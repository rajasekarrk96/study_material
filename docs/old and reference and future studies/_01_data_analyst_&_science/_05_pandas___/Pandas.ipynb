***Understanding Series and DataFrames*** is fundamental to working with Pandas, as these are the core data structures that Pandas uses for data manipulation and analysis. Let's explore these concepts in detail:

**1. Series:**
A Series is a one-dimensional array-like object in Pandas. It can hold various data types, including integers, floats, strings, and more. A Series consists of two primary components: the data and the index.

- **Data**: The data can be thought of as the values or observations in the Series. It's similar to a NumPy array but with added functionality.
- **Index**: The index is a label or identifier associated with each element in the Series. It's like a row label or key for data retrieval. The index can be explicitly defined or, by default, is created as a sequence of integers starting from 0.

**Creating a Series:**
You can create a Series using the `pd.Series()` constructor by passing a list or array of data as its argument.

**Missing Data Handling**: It offers robust tools for handling missing data, including methods for filling, dropping, or interpolating missing values.

```python
import pandas as pd
import numpy as np

data = {'A': [1, 2, np.nan, 4, 5],
        'B': [np.nan, 2, 3, np.nan, 5],
        'C': [1, 2, 3, 4, 5]}

df = pd.DataFrame(data)

# Detect missing data
print(df.isna())

# Drop rows with missing data
df_dropped = df.dropna()
print(df_dropped)

# Fill missing data with a specified value (e.g., 0)
df_filled = df.fillna(0)
print(df_filled)

# Interpolate missing data
df_interpolated = df.interpolate()
print(df_interpolated)
```

**Grouping and Aggregation**: With the `groupby()` method, you can group data by one or more columns and perform various aggregations, enabling advanced data analysis.



**Grouping Data:**

1. **Basic Grouping:**

   You can group data by one or more columns in a DataFrame using the `groupby()` method.


```python
import pandas as pd

data = {'Category': ['A', 'B', 'A', 'B', 'A'],
        'Value': [10, 15, 20, 25, 30]}

df = pd.DataFrame(data)

# Grouping data by the 'Category' column
grouped = df.groupby('Category')
```

```python
print(grouped)
```


2. **Grouping by Multiple Columns:**

   You can group data by multiple columns by passing a list of column names to `groupby()`.

```python
import pandas as pd

data = {'Category': ['A', 'B', 'A', 'B', 'A'],
        'Subcategory': ['X', 'Y', 'X', 'Y', 'Z'],
        'Value': [10, 15, 20, 25, 30]}

df = pd.DataFrame(data)

# Grouping data by both 'Category' and 'Subcategory'
grouped = df.groupby(['Category', 'Subcategory'])

# Aggregating with mean and sum
mean_values = grouped['Value'].mean()
sum_values = grouped['Value'].sum()
print(df.columns)
print(df)
```


**Aggregating Data:**

Once you have grouped the data, you can perform various aggregation functions on the grouped data to summarize or transform it:

- **Common Aggregation Functions:**

  - `mean()`: Calculate the mean of each group.
  - `sum()`: Calculate the sum of each group.
  - `count()`: Count the number of items in each group.
  - `min()`, `max()`: Find the minimum and maximum values in each group.
  - `median()`: Calculate the median of each group.
  - `agg()`: Apply custom aggregation functions.


```python
import pandas as pd

data = {'Category': ['A', 'B', 'A', 'B', 'A'],
        'Value': [10, 15, 20, 25, 30]}

df = pd.DataFrame(data)

# Grouping data by 'Category' and calculating the mean of 'Value' within each group
mean_values = df.groupby('Category')['Value'].mean()
mean_values
```

```python
# Grouping data by 'Category' and counting the number of observations within each group
count_values = df.groupby('Category')['Value'].count()
count_values
```

```python
# Grouping data by 'Category' and finding the minimum 'Value' within each group
min_values = df.groupby('Category')['Value'].min()
min_values
```

```python
# Grouping data by 'Category' and finding the maximum 'Value' within each group
max_values = df.groupby('Category')['Value'].max()
max_values
```

```python
# Grouping data by 'Category' and calculating the median 'Value' within each group
median_values = df.groupby('Category')['Value'].median()
median_values
```

**Merging and Joining**: High-performance merging and joining functions enable the combination of multiple datasets based on common columns or indices.




**1. Types of Joins:**

Pandas supports several types of joins, including:

- **Inner Join (default)**: Combines rows where the keys (common columns) are present in both DataFrames.
- **Outer Join**: Combines all rows from both DataFrames and fills in missing values with NaN.
- **Left Join**: Combines all rows from the left DataFrame and matching rows from the right DataFrame. Missing values in the right DataFrame are filled with NaN.
- **Right Join**: Combines all rows from the right DataFrame and matching rows from the left DataFrame. Missing values in the left DataFrame are filled with NaN.

**2. Merging DataFrames:**

The `pd.merge()` function is used to merge DataFrames based on common columns or indices.

```python
import pandas as pd

left_df = pd.DataFrame({'key': ['A', 'B', 'C'],
                        'value_left': [1, 2, 3]})

right_df = pd.DataFrame({'key': ['B', 'C', 'D'],
                         'value_right': [4, 5, 6]})
```

```python

# Inner join based on 'key' column
result_inner = pd.merge(left_df, right_df, on='key')

# Left join based on 'key' column
result_left = pd.merge(left_df, right_df, on='key', how='left')

# Right join based on 'key' column
result_right = pd.merge(left_df, right_df, on='key', how='right')

# Outer join based on 'key' column
result_outer = pd.merge(left_df, right_df, on='key', how='outer')
```

```python
print("Result of Inner Join:")
print(result_inner)
```

```python
print("\nResult of Left Join:")
print(result_left)
```

```python
print("\nResult of Right Join:")
print(result_right)
```

```python
print("\nResult of Outer Join:")
print(result_outer)
```

**3. Joining DataFrames:**

The `df.join()` method is used for joining DataFrames on their indices.

```python
import pandas as pd

left_df = pd.DataFrame({'value_left': [1, 2, 3]}, index=['A', 'B', 'C'])
right_df = pd.DataFrame({'value_right': [4, 5, 6]}, index=['B', 'C', 'D'])

# Inner join based on indices
result_inner = left_df.join(right_df, how='inner')

# Left join based on indices
result_left = left_df.join(right_df, how='left')

# Right join based on indices
result_right = left_df.join(right_df, how='right')

# Outer join based on indices
result_outer = left_df.join(right_df, how='outer')
```

# **1. Time Series Data Analysis:**

Suppose you want to perform time series analysis on the 'Year' column. First, you need to convert the 'Year' column to a datetime object and set it as the DataFrame's index. Then, you can resample and calculate statistics:

```python

# Convert 'Year' column to datetime
data['Year'] = pd.to_datetime(data['Year'], format='%Y')

# Set 'Year' as the index
data.set_index('Year', inplace=True)

# Resample data by year and calculate the mean of 'Life Expectancy'
life_expectancy_mean = data['Life Expectancy'].resample('Y').mean()
print(life_expectancy_mean)
```

# **2. Handling Large Datasets (Chunking):**

To handle large datasets, you can read and process data in chunks using pd.read_csv() with the chunksize parameter. Here's an example of reading data in chunks and performing a simple operation:

```python

chunk_size = 100  # Specify the chunk size
data_iterator = pd.read_csv('/content/Global Population Trends(2016-2022).csv', chunksize=chunk_size)

for chunk in data_iterator:
    # Process each chunk
    # For example, calculate the sum of 'Total Population' in each chunk
    total_population_sum = chunk['Total Population'].sum()
    print(f"Chunk Total Population Sum: {total_population_sum}")
```

```python
import pandas as pd

# Create a sample DataFrame
data = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva'],
    'Age': [25, 30, 35, 28, 22],
    'Salary': [50000, 60000, 75000, 55000, 48000],
    'Department': ['HR', 'IT', 'Finance', 'IT', 'Sales']
})

# Add a new column
data['Experience'] = [3, 5, 8, 4, 2]

# Perform element-wise operations
data['Salary After Raise'] = data['Salary'] * 1.1  # Increase salary by 10%

# Group by 'Department' and calculate the mean of 'Salary'
department_salary_mean = data.groupby('Department')['Salary'].mean()

# Sort data by 'Age' in descending order
sorted_data = data.sort_values(by='Age', ascending=False)

# Apply a custom function to the 'Salary' column
def bonus_function(salary):
    return salary + 5000

data['Salary After Bonus'] = data['Salary'].apply(bonus_function)

# Concatenate two DataFrames vertically
data2 = pd.DataFrame({
    'Name': ['Frank', 'Grace'],
    'Age': [24, 31],
    'Salary': [52000, 60000],
    'Department': ['Sales', 'Finance']
})

concatenated_data = pd.concat([data, data2], axis=0)

# Merge two DataFrames based on a common key column 'Department'
data3 = pd.DataFrame({
    'Department': ['HR', 'IT', 'Finance'],
    'Location': ['New York', 'San Francisco', 'Chicago']
})

merged_data = data.merge(data3, on='Department', how='left')

print("Modified DataFrame:")
print(data)

print("\nGrouped Data - Mean Salary by Department:")
print(department_salary_mean)

print("\nSorted Data by Age:")
print(sorted_data)

print("\nConcatenated DataFrames:")
print(concatenated_data)

print("\nMerged DataFrames:")
print(merged_data)
```

# **1. Time Series Analysis:**

Pandas provides robust tools for time series analysis, including date and time indexing, resampling, and calculating rolling statistics. Here's a code example that demonstrates some of these operations:



```python
import pandas as pd
import numpy as np

# Create a sample time series DataFrame
date_rng = pd.date_range(start='2023-01-01', end='2023-01-10', freq='D')
ts_data = pd.DataFrame(date_rng, columns=['date'])
ts_data['temperature'] = np.random.randint(0, 100, size=(len(date_rng)))

# Set the 'date' column as the index
ts_data.set_index('date', inplace=True)

# Resample data by week and calculate the mean temperature
weekly_mean = ts_data['temperature'].resample('W').mean()

# Calculate rolling mean over a window of 3 days
rolling_mean = ts_data['temperature'].rolling(window=3).mean()

print("Sample Time Series Data:")
print(ts_data)
print("\nWeekly Mean Temperature:")
print(weekly_mean)
print("\nRolling Mean Temperature:")
print(rolling_mean)
```

# **2. Chunking:**

Processing large datasets in chunks can be done using the pd.read_csv() function with the chunksize parameter. Here's an example of reading and processing data in chunks:

```python
import pandas as pd

# Replace 'large_data.csv' with the path to your large CSV file
chunk_size = 1000
data_iterator = pd.read_csv('/content/data_export.csv', chunksize=chunk_size)

for chunk in data_iterator:
    # Process each chunk
    # Calculate the sum of the 'Salary' column in each chunk
    chunk_sum = chunk['Salary'].sum()
    print(f"Chunk Sum: {chunk_sum}")
```

# **3. df.applymap(custom_function):**

The applymap() method applies a custom function element-wise to the entire DataFrame. Here's an example:

```python
import pandas as pd

# Create a sample DataFrame
data = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6]})

# Define a custom function to square each element
def square_element(x):
    return x ** 2

# Apply the custom function to all elements in the DataFrame
data_squared = data.applymap(square_element)

print("Original DataFrame:")
print(data)
print("\nDataFrame with Elements Squared:")
print(data_squared)
```

# **4. Categorical Data:**

You can convert columns to categorical data types to save memory and improve performance when working with limited unique values. Here's an example:

```python
import pandas as pd

# Create a sample DataFrame
data = pd.DataFrame({'Category': ['A', 'B', 'A', 'C', 'B', 'A']})

# Convert the 'Category' column to a categorical data type
data['Category'] = data['Category'].astype('category')

# Access category codes
category_codes = data['Category'].cat.codes

print("DataFrame with Categorical Data:")
print(data)
print("\nCategory Codes:")
print(category_codes)
```
