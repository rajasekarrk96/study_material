## Introduction

Broadcasting is a set of rules that NumPy follows when performing arithmetic operations on arrays with differing shapes. By expanding smaller arrays along singleton dimensions, NumPy can compute results without explicit replication, which makes vectorized code both simple and efficient.

# What is Broadcasting?

Broadcasting describes how NumPy treats arrays with different shapes during elementwise operations. When possible, NumPy automatically expands the smaller array to match the shape of the larger array.

## Broadcasting Rules

NumPy compares array shapes elementwise, starting from the trailing dimensions. Two dimensions are compatible when:

1. They are equal, or
2. One of them is 1.

If the shapes are not compatible, NumPy raises a `ValueError`.

Example: `(5, 4)` and `(4,)` are compatible because the trailing dimension 4 matches and the missing leading dimension is treated as 1.

Example: `(3, 1)` and `(1, 4)` are compatible because the singleton dimensions can be expanded to `(3, 4)`.

```python
import numpy as np
# Scalar and array broadcasting
vector = np.array([1, 2, 3])
print('vector + 5 ->', vector + 5)

# 1-D and 2-D broadcasting
matrix = np.array([[1, 2, 3], [4, 5, 6]])
print('matrix:', matrix)
print('matrix + vector:', matrix + vector)
```

## Broadcasting with singleton dimensions

A dimension with size 1 can be stretched to match the corresponding dimension in the other array. This is how arrays with shapes like `(3, 1)` and `(1, 4)` can participate in arithmetic operations.

```python
a = np.array([[10], [20], [30]])  # shape (3, 1)
b = np.array([[1, 2, 3, 4]])     # shape (1, 4)
print('a shape:', a.shape)
print('b shape:', b.shape)
print('a + b:', a + b)
```

## Shape Compatibility Examples

Use the `np.broadcast_shapes()` helper to determine whether shapes are compatible before applying operations.

```python
print('broadcast_shapes((5, 4), (4,)):', np.broadcast_shapes((5, 4), (4,)))
print('broadcast_shapes((3, 1), (1, 4)):', np.broadcast_shapes((3, 1), (1, 4)))
try:
    np.broadcast_shapes((2, 3), (3, 2))
except ValueError as error:
    print('incompatible shapes error:', error)
```

## Broadcasting Scalar Values

Scalars are treated as arrays with shape `()`, and they can broadcast to any shape. This means `array + 1` is equivalent to adding a constant to every element.

Scalar broadcasting is common when normalizing data or shifting values.

```python
data = np.arange(1, 7).reshape(2, 3)
print('original data:', data)
print('data - 2:', data - 2)
print('data * 3:', data * 3)
```

## Broadcasting for Normalization

Broadcasting makes it easy to normalize rows or columns by subtracting a mean vector or dividing by a standard deviation vector.

```python
data = np.array([[10.0, 20.0, 30.0], [15.0, 25.0, 35.0]])
row_mean = data.mean(axis=1, keepdims=True)
print('row_mean shape:', row_mean.shape)
print('normalized rows:', data - row_mean)

col_mean = data.mean(axis=0)
print('col_mean shape:', col_mean.shape)
print('normalized columns:', data - col_mean)
```

## Broadcasting with Higher Dimensions

Broadcasting works with arrays of any dimension, provided the trailing dimensions are compatible. For example, a `(3, 4, 1)` array can broadcast with a `(4, 5)` array as long as the shapes align from the end.

```python
a = np.ones((2, 3, 4, 1))
b = np.arange(4).reshape((1, 4))
print('a shape:', a.shape)
print('b shape:', b.shape)
print('a + b shape:', (a + b).shape)
```

## Broadcasting and Memory

Broadcasting does not actually replicate data in memory. Instead, NumPy creates a temporary view with virtual dimensions expanded. This saves memory, but operations still may allocate a new array for the final result.

Use `np.broadcast_to()` to explicitly generate a broadcasted view when you want to inspect the expanded shape.

```python
x = np.array([1, 2, 3])
b = np.broadcast_to(x, (3, 3))
print('broadcast_to shape:', b.shape)
print('broadcast_to array:', b)
print('shares memory:', np.shares_memory(x, b))
```

## Common Broadcasting Pitfalls

- Mismatched shapes that cannot be aligned raise `ValueError`.
- Silent broadcasting of shape `1` can hide bugs when arrays are not the expected shape.
- Broadcasting can make code less readable if the shape logic is unclear.

Always print shapes when debugging broadcasting issues.

```python
# Example of an incompatible broadcast
try:
    np.array([[1, 2, 3], [4, 5, 6]]) + np.array([1, 2])
except ValueError as error:
    print('broadcast error:', error)
```

## Real-world Example: Adding a Bias Vector

In machine learning, broadcasting is often used to add a bias vector to every sample in a batch.

```python
batch = np.random.randn(5, 10)  # 5 samples, 10 features
bias = np.arange(10) * 0.1
print('batch shape:', batch.shape)
print('bias shape:', bias.shape)
output = batch + bias
print('output shape:', output.shape)
```

## Real-world Example: Compute Pairwise Differences

Broadcasting can compute pairwise differences or distances between vectors without explicit loops.

```python
points = np.array([[1.0, 2.0], [4.0, 6.0], [7.0, 8.0]])
diffs = points[:, np.newaxis, :] - points[np.newaxis, :, :]
print('differences shape:', diffs.shape)
print('pairwise differences:', diffs)
```

# Summary

## Key Points

- Broadcasting aligns arrays with different shapes by comparing dimensions from the right.
- Dimensions are compatible if they are equal or if one is 1.
- Scalars broadcast to any shape.
- Broadcasting can be used for row-wise and column-wise normalization, bias addition, and pairwise computations.
- Use `np.broadcast_shapes()` and `np.broadcast_to()` to inspect compatibility and broadcasted shapes.

## Best Practices

- Print shapes when debugging broadcasting errors.
- Prefer explicit use of `np.newaxis` to clarify intended broadcasting.
- Avoid relying on silent shape expansion when array shapes are ambiguous.

## Common Mistakes

- Trying to broadcast incompatible shapes.
- Confusing `@` with broadcasting-based addition or multiplication.
- Assuming broadcasting always returns a view; it often creates a temporary result.

## Interview Tips

- Explain the broadcasting rules with concrete shape examples.
- Mention that broadcasting is a powerful alternative to explicit loops.
- Show how `keepdims=True` is useful for maintaining shapes in reductions.
