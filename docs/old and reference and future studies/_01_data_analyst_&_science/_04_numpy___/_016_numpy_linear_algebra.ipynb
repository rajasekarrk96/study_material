## Introduction

NumPy's `linalg` module provides essential linear algebra tools for scientific computing. It supports exact matrix operations and solvers for many applied problems.


# Matrix multiplication and dot products

Use `np.dot`, `@`, or `np.matmul` for matrix multiplication.


```python
import numpy as np
A = np.array([[1, 2], [3, 4]])
b = np.array([5, 6])
print('A @ b:', A @ b)
print('dot:', np.dot(A, b))
```

# Determinant, inverse, and trace

Compute matrix properties with `np.linalg.det`, `np.linalg.inv`, and `np.trace`.


```python
print('det A:', np.linalg.det(A))
print('inv A:
', np.linalg.inv(A))
print('trace A:', np.trace(A))
```

# Solving linear systems

Solve `Ax = b` using `np.linalg.solve`.


```python
x = np.linalg.solve(A, b)
print('solution x:', x)
print('check Ax:', A @ x)
```

# Eigenvalues and eigenvectors

Use `np.linalg.eig` for eigen decomposition of square matrices.


```python
vals, vecs = np.linalg.eig(A)
print('eigenvalues:', vals)
print('eigenvectors:
', vecs)
```

# Singular Value Decomposition (SVD)

`np.linalg.svd` decomposes a matrix into singular values and orthogonal factors.


```python
U, s, Vt = np.linalg.svd(A)
print('U:
', U)
print('singular values:', s)
print('Vt:
', Vt)
```

# Norms and matrix power

Compute vector and matrix norms, and raise matrices to powers.


```python
v = np.array([1.0, 2.0, 3.0])
print('norm:', np.linalg.norm(v))
print('matrix power A^2:
', np.linalg.matrix_power(A, 2))
```
