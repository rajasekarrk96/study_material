## Introduction

Sorting lets you order values and prepare data for binary search, ranking, and selection. NumPy offers both in-place and out-of-place sorting, as well as index-based sorting for efficient retrieval.


# np.sort() and ndarray.sort()

`np.sort` returns a sorted copy, while `ndarray.sort()` sorts the array in-place.


```python
import numpy as np
arr = np.array([4, 2, 7, 1])
print('sorted copy:', np.sort(arr))
arr.sort()
print('in-place sort:', arr)
```

# argsort()

`np.argsort` returns the indices that would sort an array. Use these indices to reorder other arrays consistently.


```python
scores = np.array([50, 20, 80, 40])
order = np.argsort(scores)
print('sort order:', order)
print('sorted scores:', scores[order])
```

# lexsort()

`np.lexsort` sorts by multiple keys, using the last key as the primary sort order.


```python
names = np.array(['c', 'a', 'b', 'a'])
scores = np.array([20, 10, 30, 25])
order = np.lexsort((scores, names))
print('lexsort order:', order)
print('sorted rows:
', np.vstack((names[order], scores[order])).T)
```

# partition() and argpartition()

`np.partition` rearranges an array so that the k smallest values appear before index k. `np.argpartition` returns the indices for that arrangement.


```python
arr = np.array([7, 2, 5, 1, 9])
print('partition at k=3:', np.partition(arr, 3))
print('argpartition indices:', np.argpartition(arr, 2))
```

# Sorting along axes

Sort 2-D arrays along rows or columns using `axis`.


```python
matrix = np.array([[3, 1, 2], [9, 5, 7]])
print('sort axis=1:
', np.sort(matrix, axis=1))
print('sort axis=0:
', np.sort(matrix, axis=0))
```

# Stable vs unstable sorts

NumPy supports stable sorting methods such as `mergesort`. Stability preserves the relative order of equal elements.


```python
arr = np.array([3, 1, 2, 1])
print('stable sort:', np.sort(arr, kind='mergesort'))
print('unstable sort:', np.sort(arr, kind='quicksort'))
```

# Sorting structured arrays

Structured arrays can be sorted by field names using `order`. This is useful for heterogenous records.


```python
dtype = [('name', 'U10'), ('age', 'i4')]
people = np.array([('Alice', 30), ('Bob', 25), ('Charlie', 35)], dtype=dtype)
print('sort by age:', np.sort(people, order='age'))
```

## Real-world Example

Rank products by price and select the top three cheapest items.


```python
prices = np.array([19.99, 5.99, 14.95, 2.49, 9.99])
order = np.argsort(prices)
print('cheapest items:', prices[order[:3]])
```

## Summary

- Use `np.sort` for sorted copies and `.sort()` for in-place sorting.
- `argsort` gives the permutation indices used for ranking.
- `lexsort` sorts by multiple keys.
- `partition` is efficient when only a subset of order is needed.
