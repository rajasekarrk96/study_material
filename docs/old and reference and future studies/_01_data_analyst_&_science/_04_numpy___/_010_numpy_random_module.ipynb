# The Generator API

Create a generator using `np.random.default_rng(seed)` for reproducible pseudo-random numbers. The generator methods return arrays of random values from many distributions.


```python
import numpy as np
rng = np.random.default_rng(12345)
print('random floats:', rng.random(5))
print('random integers:', rng.integers(1, 10, size=5))
```

## Reproducibility and Seeding

Using the same seed always reproduces the same sequence of values. This is essential for debugging and sharing experiments.


```python
rng1 = np.random.default_rng(2026)
rng2 = np.random.default_rng(2026)
print('first draw equal:', rng1.random(3))
print('second draw equal:', rng2.random(3))
```

## Continuous Distributions

Use methods such as `normal`, `uniform`, `exponential`, and `gamma` to sample continuous distributions.


```python
print('uniform [0, 1):', rng.uniform(0, 1, size=5))
print('normal mean 0, std 1:', rng.normal(0, 1, size=5))
print('exponential scale 1:', rng.exponential(scale=1.0, size=5))
```

## Discrete Distributions

The `integers`, `binomial`, `poisson`, and `multinomial` methods sample common discrete distributions.


```python
print('binomial (n=10, p=0.5):', rng.binomial(10, 0.5, size=5))
print('poisson (lam=2):', rng.poisson(2, size=5))
print('multinomial (n=10, p=[0.2,0.3,0.5]):', rng.multinomial(10, [0.2, 0.3, 0.5]))
```

## Selections and Permutations

Use `choice`, `shuffle`, and `permutation` to select or permute elements for randomized algorithms and sampling.


```python
items = np.arange(10)
print('choice with replacement:', rng.choice(items, size=5))
print('choice without replacement:', rng.choice(items, size=5, replace=False))
shuffled = items.copy()
rng.shuffle(shuffled)
print('shuffled:', shuffled)
print('permutation copy:', rng.permutation(items))
```

## Random Sampling for Training and Validation

Shuffle indices or sample subsets to create randomized training/validation splits in machine learning workflows.


```python
data = np.arange(20).reshape((10, 2))
labels = np.arange(10)
indices = rng.permutation(data.shape[0])
train_idx, test_idx = indices[:7], indices[7:]
print('train indices:', train_idx)
print('test indices:', test_idx)
print('train data:', data[train_idx])
print('test labels:', labels[test_idx])
```

## Legacy `np.random` vs `Generator`

The old `np.random` functions such as `np.random.rand` and `np.random.randint` use a global state. The modern `Generator` API is preferred because it avoids hidden state and makes seeds explicit.


```python
print('legacy rand:', np.random.rand(3))
print('legacy randint:', np.random.randint(1, 10, 3))
# Prefer RNG methods for new code
```

## Advanced Sampling Techniques

Use `choice` with probability weights, or sample from structured arrays to simulate real-world scenarios.


```python
events = ['A', 'B', 'C']
probabilities = [0.1, 0.3, 0.6]
print('weighted choices:', rng.choice(events, size=5, p=probabilities))
print('sample with replace=False:', rng.choice(events, size=3, replace=False))
```

## Monte Carlo Simulation Example

Estimate the probability of a random point falling inside a unit circle using the Generator API.


```python
rng = np.random.default_rng(1)
points = rng.uniform(-1, 1, size=(200000, 2))
inside = np.sum(np.sum(points**2, axis=1) <= 1)
pi_estimate = 4 * inside / points.shape[0]
print('Pi estimate:', pi_estimate)
```

## Bootstrap Sampling Example

Bootstrap sampling draws repeated samples from data with replacement to estimate uncertainty.


```python
data = np.array([10, 20, 30, 40, 50])
bootstrap_samples = np.array([rng.choice(data, size=data.size, replace=True) for _ in range(1000)])
print('bootstrap means:', np.mean(bootstrap_samples, axis=1)[:5])
print('bootstrap mean estimate:', np.mean(np.mean(bootstrap_samples, axis=1)))
```

## Edge Cases

- Use `replace=False` with `choice` only when the sample size does not exceed the population.
- Make sure probability weights sum to 1 when using `choice` with `p`.
- Random generators with the same seed produce repeatable results, but the sequence order depends on the methods called.


```python
try:
    rng.choice([1, 2, 3], size=5, replace=False)
except ValueError as error:
    print('replace=False error:', error)

try:
    rng.choice([1, 2, 3], size=5, p=[0.5, 0.2, 0.1])
except ValueError as error:
    print('weights error:', error)
```
