## Introduction

Saving arrays to disk is essential for long-running computation, data exchange, and reproducibility. NumPy supports binary `.npy` and `.npz` formats, as well as text-based CSV-like files.


# Saving and loading `.npy` files

Use `np.save` and `np.load` for single arrays in binary format.


```python
import numpy as np
arr = np.arange(10)
np.save('example.npy', arr)
loaded = np.load('example.npy')
print('loaded:', loaded)
```

# Saving compressed archives with `.npz`

`np.savez` and `np.savez_compressed` can store multiple arrays in one archive.


```python
a = np.arange(3)
b = np.arange(3, 6)
np.savez('archive.npz', first=a, second=b)
with np.load('archive.npz') as data:
    print(data['first'])
    print(data['second'])
```

# Text file I/O with `savetxt` and `loadtxt`

Use text formats for human-readable storage and interoperability with spreadsheets.


```python
matrix = np.array([[1.5, 2.5], [3.5, 4.5]])
np.savetxt('matrix.txt', matrix, delimiter=',', fmt='%.2f')
loaded_text = np.loadtxt('matrix.txt', delimiter=',')
print(loaded_text)
```

# Handling headers and comments

`np.savetxt` supports headers, and `np.loadtxt` can skip comments and header lines.


```python
np.savetxt('data.csv', matrix, delimiter=',', header='a,b', comments='')
loaded_csv = np.loadtxt('data.csv', delimiter=',', skiprows=1)
print(loaded_csv)
```

# File operations with structured arrays

`tq` can save structured arrays as text if you format fields carefully.


```python
dtype = [('name', 'U10'), ('score', 'f4')]
data = np.array([('Alice', 90.0), ('Bob', 85.5)], dtype=dtype)
np.savetxt('records.txt', data, fmt='%s,%.1f', header='name,score', comments='')
loaded = np.loadtxt('records.txt', delimiter=',', dtype=str, skiprows=1)
print(loaded)
```
