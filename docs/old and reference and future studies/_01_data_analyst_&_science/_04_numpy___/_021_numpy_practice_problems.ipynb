# NumPy Practice Problems (Chapter 21)

This chapter collects practice problems that reinforce key NumPy concepts from array creation to advanced operations.


## Problem Set

Solve the following problems using NumPy functions and arrays. Try to avoid explicit Python loops.


1. Create a 5x5 array with values 1 to 25 and reshape it into a 1-D array.


2. Generate a 10-element array of random integers from 0 to 100 and find the top 3 values.


3. Create a 3x3 identity matrix and compute its inverse.


4. Given `a = np.array([1,2,3])` and `b = np.array([[1],[2],[3]])`, compute the outer product.


5. Use boolean indexing to replace all negative values in a random array with 0.


6. Sort the rows of a 2-D array by the value in the second column.


7. Create a structured array for employees with `name`, `age`, and `salary`. Then filter age > 30.


8. Use `np.meshgrid` to compute a 2-D table of `x**2 + y**2` for x and y values from 0 to 4.


9. Load a matrix from a text file and compute its column means.
   *Hint: save the matrix first using `np.savetxt`.*


10. Convert a Pandas DataFrame of integers to a NumPy array and compute the elementwise square.


## Summary

This problem set covers NumPy array creation, arithmetic, indexing, sorting, structured arrays, file I/O, and Pandas interoperability. Use it to practice writing concise, vectorized NumPy code.
