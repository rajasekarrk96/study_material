# 1. `ndim`

## Definition

`ndim` is an integer attribute that reports the number of dimensions (axes) of an `ndarray`. It is also known as the rank of the array.

## Purpose

Use `ndim` to understand whether an array is one-dimensional, two-dimensional, or higher-dimensional. This is useful when writing code that depends on the number of axes, such as matrix multiplication, image processing, or tensor operations.

## Real-world Use Cases

- Distinguishing between feature vectors (`ndim=1`) and feature matrices (`ndim=2`) in machine learning.
- Verifying that image data is 3D (height, width, channels) before applying convolution.
- Ensuring that time-series data has the expected number of dimensions.

## Syntax

`array.ndim`

## Parameters

- `ndim` is an attribute, not a function; it takes no parameters.

## Return Value

- Returns: `int`
- Range: 0 for scalars, 1 for vectors, 2 for matrices, and higher for tensors.

## Notes

- A scalar created with `np.array(10)` has `ndim == 0`.
- `ndim` is computed from the array's shape: `len(array.shape)`.

## Common Errors

- Mistaking scalar arrays (`ndim=0`) for empty arrays.

## Related Functions

- `array.shape`
- `array.size`
- `array.reshape()`

```python
# Import NumPy and create arrays of different ranks
import numpy as np

scalar = np.array(42)  # 0-D array
vector = np.array([1, 2, 3])  # 1-D array
matrix = np.array([[1, 2], [3, 4]])  # 2-D array
tensor = np.zeros((2, 2, 2))  # 3-D array

print('scalar ndim:', scalar.ndim)
print('vector ndim:', vector.ndim)
print('matrix ndim:', matrix.ndim)
print('tensor ndim:', tensor.ndim)
```

# 2. `shape`

## Definition

`shape` is a tuple attribute that describes the size of an `ndarray` along each dimension.

## Purpose

Use `shape` to understand how many rows, columns, channels, or dimensions an array contains. It is essential for verifying compatibility before array operations such as matrix multiplication, concatenation, or broadcasting.

## Real-world Use Cases

- Validating the dimensions of training data and labels for a machine learning model.
- Checking image dimensions before feeding data into a convolutional neural network.
- Confirming array shapes before concatenating or stacking arrays.

## Syntax

`array.shape`

## Parameters

- `shape` is an attribute and takes no parameters.

## Return Value

- Returns: `tuple` of ints
- Example: `(3,)`, `(2, 4)`, `(100, 64, 64, 3)`

## Notes

- The product of the `shape` tuple equals `array.size`.
- Dtype and shape together determine the array's total memory usage.

## Common Errors

- Assuming the first element of `shape` is always rows; for 1-D arrays it is the length.

## Related Functions

- `array.reshape()`
- `array.transpose()`
- `np.newaxis`

```python
# Demonstrate shape for arrays of different dimensions
series = np.array([10, 20, 30])
table = np.array([[1, 2, 3], [4, 5, 6]])
image = np.zeros((64, 64, 3))  # height, width, channels

print('series shape:', series.shape)
print('table shape:', table.shape)
print('image shape:', image.shape)

# Use shape to validate compatibility before matrix multiplication
A = np.ones((2, 3))
B = np.ones((3, 4))
print('A shape:', A.shape, 'B shape:', B.shape)
```

# 3. `size`

## Definition

`size` is an integer attribute that returns the total number of elements in an `ndarray`.

## Purpose

Use `size` to understand how many values an array contains, which is useful for performance estimation, reshaping, and validating data volume.

## Real-world Use Cases

- Comparing the number of samples in feature arrays and label arrays.
- Estimating memory usage and processing cost.
- Validating that a reshaped array still contains the expected number of elements.

## Syntax

`array.size`

## Parameters

- `size` is an attribute and takes no parameters.

## Return Value

- Returns: `int`
- Example: `6` for a `(2, 3)` array, `3` for a `(3,)` vector.

## Notes

- `array.size == np.prod(array.shape)`

## Common Errors

- Confusing `size` with `shape[0]` for multi-dimensional arrays.

## Related Functions

- `array.shape`
- `array.ndim`
- `np.prod()`

```python
# Show size for arrays with different shapes
one_d = np.arange(5)
two_d = np.arange(6).reshape(2, 3)
three_d = np.zeros((2, 3, 4))

print('one_d size:', one_d.size)
print('two_d size:', two_d.size)
print('three_d size:', three_d.size)

# Check that the total element count remains constant after reshape
reshaped = two_d.reshape(3, 2)
print('reshaped size:', reshaped.size)
print('original shape:', two_d.shape, 'new shape:', reshaped.shape)
```

# 4. `dtype`

## Definition

`dtype` is an attribute that specifies the data type of the elements in the array.

## Purpose

Use `dtype` to inspect or enforce the numeric type stored in the array, such as integers, floats, booleans, or strings.

## Real-world Use Cases

- Choosing `np.float32` for neural network weights to save memory.
- Using `np.int64` for integer IDs and counts.
- Detecting unwanted type promotion when arithmetic operations are applied.

## Syntax

`array.dtype`

## Parameters

- `dtype` is an attribute and takes no parameters.

## Return Value

- Returns: `numpy.dtype` object
- Example: `dtype('int64')`, `dtype('float32')`.

## Notes

- Different dtypes affect memory usage, precision, and supported operations.
- NumPy may upcast or downcast values when converting arrays.

## Common Errors

- Assuming Python lists have a dtype before conversion.
- Using an incompatible dtype for strings or structured arrays.

## Related Functions

- `array.astype()`
- `np.dtype()`
- `np.zeros(..., dtype=...)`

```python
# Inspect dtype for arrays with different element types
int_array = np.array([1, 2, 3], dtype=np.int64)
float_array = np.array([1.0, 2.5, 3.25], dtype=np.float32)
bool_array = np.array([True, False, True])
string_array = np.array(['apple', 'banana'], dtype='<U10')

print('int_array dtype:', int_array.dtype)
print('float_array dtype:', float_array.dtype)
print('bool_array dtype:', bool_array.dtype)
print('string_array dtype:', string_array.dtype)

# Use dtype to avoid precision loss when converting data
data = np.array([1.2, 2.7, 3.9])
int_conversion = data.astype(np.int32)
print('original dtype:', data.dtype, 'converted dtype:', int_conversion.dtype)
    
```

# 5. `itemsize`

## Definition

`itemsize` is an integer attribute that returns the number of bytes used by each element in the array.

## Purpose

Use `itemsize` to understand how much memory each element occupies, which is important for choosing data types and optimizing storage.

## Real-world Use Cases

- Comparing memory usage of `np.float32` versus `np.float64`.
- Choosing a compact dtype for large datasets in data analytics.
- Diagnosing why an array consumes more memory than expected.

## Syntax

`array.itemsize`

## Parameters

- `itemsize` is an attribute and takes no parameters.

## Return Value

- Returns: `int` number of bytes per element
- Example: `8` for `float64`, `4` for `int32`.

## Notes

- Total array memory is `array.itemsize * array.size`.

## Related Functions

- `array.nbytes`
- `array.dtype`

```python
# Show itemsize for different array dtypes
int32_array = np.zeros(1, dtype=np.int32)
float64_array = np.zeros(1, dtype=np.float64)
bool_array = np.zeros(1, dtype=bool)

print('int32 itemsize:', int32_array.itemsize)
print('float64 itemsize:', float64_array.itemsize)
print('bool itemsize:', bool_array.itemsize)
```

# 6. `nbytes`

## Definition

`nbytes` is an integer attribute that returns the total number of bytes consumed by the array data.

## Purpose

Use `nbytes` to estimate the memory footprint of an array and to optimize memory usage in large-scale data processing.

## Real-world Use Cases

- Forecasting memory requirements for large datasets.
- Comparing memory usage across different dtypes.
- Monitoring memory consumption in machine learning experiments.

## Syntax

`array.nbytes`

## Return Value

- Returns: `int` total bytes
- Example: `800` for a `(100, 2, 4)` `float64` array with 100*2*4*8 bytes.

## Notes

- `nbytes == array.size * array.itemsize`

## Common Errors

- Expecting `nbytes` to include overhead from array metadata or Python objects; it only counts the raw data buffer.

```python
# Compare total memory usage for float32 and float64 arrays
float32_data = np.zeros((100, 100), dtype=np.float32)
float64_data = np.zeros((100, 100), dtype=np.float64)

print('float32 nbytes:', float32_data.nbytes)
print('float64 nbytes:', float64_data.nbytes)
print('expected ratio:', float64_data.nbytes / float32_data.nbytes)
```

# 7. `flags`

## Definition

`flags` is an object that provides boolean metadata about the memory layout and capabilities of an array, such as whether it is C-contiguous, F-contiguous, or writeable.

## Purpose

Use `flags` to understand how an array is stored and whether operations requiring certain memory layouts are safe and efficient.

## Syntax

`array.flags`

## Common flag names

- `C_CONTIGUOUS` / `C` — row-major order.
- `F_CONTIGUOUS` / `F` — column-major order.
- `OWNDATA` — whether the array owns its own memory.
- `WRITEABLE` — whether the array can be modified.
- `ALIGNED` — whether data is aligned in memory for efficient access.

## Notes

- Some NumPy operations require contiguous memory.
- `flags` helps diagnose performance or compatibility issues with C/Fortran extensions.

```python
# Examine flags for arrays with different memory layouts
contiguous = np.zeros((3, 4), order='C')
fortran = np.zeros((3, 4), order='F')
transposed = contiguous.T

print('contiguous flags:
', contiguous.flags)
print('fortran flags:
', fortran.flags)
print('transposed flags:
', transposed.flags)

# Check whether a view shares memory with the original array
view = contiguous[:2, :2]
print('view owns data:', view.flags.owndata)
print('view is writeable:', view.flags.writeable)
```

# 8. `strides`

## Definition

`strides` is a tuple of bytes that indicates how many bytes must be skipped in memory to move to the next element along each dimension.

## Purpose

Use `strides` to understand how array indexing maps to memory addresses and to diagnose non-contiguous arrays.

## Real-world Use Cases

- Understanding the memory layout of transposed or sliced arrays.
- Debugging performance issues related to non-contiguous memory access.
- Ensuring compatibility with low-level code that expects a specific stride pattern.

## Syntax

`array.strides`

## Notes

- For a C-contiguous array, `strides` typically equals `(itemsize * columns, itemsize)` for a 2-D array.
- For a transposed view, the strides are reversed.

## Common Errors

- Using an array with unexpected strides in code that assumes contiguous memory.

## Related Functions

- `np.ascontiguousarray()`
- `np.asfortranarray()`

```python
# Inspect strides for a contiguous array, a transposed view, and a slice
base = np.arange(12).reshape(3, 4)
transposed_base = base.T
sliced_base = base[:, ::2]  # every other column

print('base strides:', base.strides)
print('transposed strides:', transposed_base.strides)
print('sliced strides:', sliced_base.strides)

# Show how strides affect memory layout
print('base C-contiguous?', base.flags.c_contiguous)
print('transposed C-contiguous?', transposed_base.flags.c_contiguous)
print('sliced C-contiguous?', sliced_base.flags.c_contiguous)
```

# 9. `T`

## Definition

`T` is a transposed view of the array, which swaps dimensions in a new view without copying data when possible.

## Purpose

Use `T` to transpose matrices and higher-dimensional arrays for linear algebra, data reshaping, and broadcasting.

## Real-world Use Cases

- Transposing feature matrices from shape `(n_samples, n_features)` to `(n_features, n_samples)` when needed.
- Preparing data for matrix multiplication with weight matrices.
- Swapping axes for image and tensor manipulations.

## Syntax

`array.T`

## Notes

- `array.T` is a view, not a copy, for most arrays.
- For 1-D arrays, `T` has no effect because the transpose of a 1-D vector is still a 1-D vector.

## Common Errors

- Assuming `T` returns a copy when it returns a view.

## Related Functions

- `array.transpose()`
- `np.swapaxes()`

```python
# Transpose a 2D matrix and compare shapes
matrix = np.array([[1, 2, 3], [4, 5, 6]])
print('matrix shape:', matrix.shape)
print('matrix T shape:', matrix.T.shape)
print('matrix T:
', matrix.T)

# Transpose a 3D tensor with T and show that it is a view
tensor = np.arange(24).reshape(2, 3, 4)
tensor_transposed = tensor.T
print('tensor shape:', tensor.shape)
print('tensor T shape:', tensor_transposed.shape)
print('shares memory with original:', tensor_transposed.base is tensor or tensor_transposed.base is tensor.base)
```

## Edge Cases

### Empty arrays

Empty arrays can still have valid shape, ndim, and dtype attributes.

### Scalar arrays

A scalar array created with `np.array(5)` has `ndim=0`, `shape=()`, `size=1`.

### Non-contiguous arrays

Slicing and transposing can change strides and contiguity without altering shape or size.

### High-dimensional arrays

Properties are consistent even for tensors with many axes, such as `(4, 3, 2, 5)`.

```python
# Explore properties of empty arrays, scalar arrays, and high-dimensional arrays
empty_array = np.array([])
scalar_array = np.array(5)
high_dim_array = np.zeros((2, 3, 4, 5))

print('empty_array shape:', empty_array.shape, 'ndim:', empty_array.ndim, 'size:', empty_array.size)
print('scalar_array shape:', scalar_array.shape, 'ndim:', scalar_array.ndim, 'size:', scalar_array.size)
print('high_dim_array shape:', high_dim_array.shape, 'ndim:', high_dim_array.ndim)

# Sliced non-contiguous array
slice_array = high_dim_array[:, :, ::2, :]
print('slice_array shape:', slice_array.shape)
print('slice_array strides:', slice_array.strides)
print('slice_array contiguous:', slice_array.flags.c_contiguous)
```

1. Smaller dtypes use less memory but may lose precision; larger dtypes preserve precision but consume more memory.

2. `array.T` is a view because it reinterprets the same memory with swapped axes; this avoids copies but may yield non-contiguous memory.
3. An array is not writeable if it is created from an immutable buffer, if `WRITEABLE` flag is False, or if it is a protected view.
4. `dtype` determines whether operations produce integer truncation, floating-point rounding, or overflow/underflow.
5. `flags.aligned` is important when passing arrays to C extensions or hardware that requires aligned memory access.

# Summary

## Key Points

- `ndim` tells you how many axes an array has.
- `shape` describes the length of each axis.
- `size` counts the total number of elements.
- `dtype` defines how each element is stored and interpreted.
- `itemsize` and `nbytes` explain memory usage.
- `flags` reveal memory layout and writeability.
- `strides` describe how to step through elements in memory.
- `T` provides a transposed view for arrays.

## Cheat Sheet

- `array.ndim` — number of dimensions.
- `array.shape` — tuple of dimension sizes.
- `array.size` — total number of elements.
- `array.dtype` — element data type.
- `array.itemsize` — bytes per element.
- `array.nbytes` — total memory bytes.
- `array.flags` — memory layout metadata.
- `array.strides` — byte step sizes for each axis.
- `array.T` — transpose view.

## Memory Tricks

- Use `dtype=np.float32` instead of `np.float64` when memory is tight and precision is adequate.
- Use `np.ascontiguousarray()` to ensure C-order memory for performance-critical code.
- Avoid copying data unless necessary; inspect `flags.owndata` and `base`.

## Best Practices

- Validate `shape` and `dtype` before feeding arrays into models or functions.
- Use `shape` and `ndim` checks to avoid broadcasting errors.
- Inspect `nbytes` for large arrays to prevent out-of-memory issues.

## Common Mistakes

- Confusing `size` with `shape[0]` on multi-dimensional arrays.
- Assuming `array.T` makes a copy.
- Ignoring non-contiguous arrays when performance matters.

## Interview Tips

- Explain how `shape`, `ndim`, and `size` relate.
- Describe how `dtype` affects precision and memory.
- Mention the difference between a view and a copy when discussing `T` and `strides`.
- Use concrete examples such as image tensors or feature matrices.
