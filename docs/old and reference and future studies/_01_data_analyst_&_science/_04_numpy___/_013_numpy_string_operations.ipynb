## Introduction

NumPy supports string arrays and provides vectorized string operations through the `np.char` namespace. These functions are useful for processing labels, categories, and text data in arrays.


# Basic string methods

Use `np.char.lower`, `np.char.upper`, and `np.char.capitalize` for elementwise modifications.


```python
import numpy as np
text = np.array(['Hello', 'WORLD', 'NumPy'])
print('lower:', np.char.lower(text))
print('upper:', np.char.upper(text))
print('title:', np.char.title(text))
```

# Search and replace

Use `np.char.find`, `np.char.replace`, and `np.char.count` to detect and transform substrings.


```python
text = np.array(['apple', 'banana', 'cherry'])
print('find an:', np.char.find(text, 'an'))
print('replace a with A:', np.char.replace(text, 'a', 'A'))
print('count a:', np.char.count(text, 'a'))
```

# Split and join

`np.char.split` and `np.char.join` allow vectorized tokenization and reconstruction.


```python
phrases = np.array(['a b c', 'd e f'])
print('split:', np.char.split(phrases))
print('join:', np.char.join('-', np.array(['12', '34'])))
```

# startswith / endswith

Use `np.char.startswith` and `np.char.endswith` to filter text arrays.


```python
words = np.array(['cat', 'dog', 'cow', 'camel'])
mask = np.char.startswith(words, 'c')
print('startswith c:', mask)
print('matching words:', words[mask])
```

# Convert between bytes and Unicode

Use `np.char.decode` and `np.char.encode` for byte string arrays.


```python
byte_arr = np.array([b'abc', b'def'])
print('decoded:', np.char.decode(byte_arr, 'utf-8'))
print('encoded:', np.char.encode(np.array(['ghi', 'jkl']), 'utf-8'))
```

# Comparison with Python lists

Unlike Python list comprehensions, `np.char` operations are vectorized and apply to the entire array efficiently.


```python
names = np.array(['Alice', 'Bob', 'Carol'])
print('replace o with 0:', np.char.replace(names, 'o', '0'))
print('add prefix:', np.char.add('X_', names))
```

## Real-world Example

Normalize labels by stripping whitespace and converting to lowercase.


```python
labels = np.array(['  USA', 'Canada  ', ' Mexico '])
normalized = np.char.strip(np.char.lower(labels))
print('normalized labels:', normalized)
```
