# reshape()

## Definition

`reshape()` returns a new view of the array with a different shape but the same data whenever possible.

## Purpose

Use `reshape()` to change how an array is interpreted without altering the underlying values.

## Real-world Use Cases

- Converting a flat feature vector into a matrix or tensor.
- Reshaping image data into `(height, width, channels)` format.
- Preparing input arrays for linear algebra operations.

## Syntax

`array.reshape(newshape, order='C')`

## Parameters

| Parameter | Type | Default | Description |
| --- | --- | --- | --- |
| `newshape` | int or tuple of ints | required | Desired shape. One dimension can be `-1` to infer its size. |
| `order` | {'C', 'F', 'A'} | 'C' | Read the elements in C order, Fortran order, or preserve input order as much as possible. |

## Return Value

- Returns: `numpy.ndarray` with the requested shape.
- Shape: `newshape` after inference.
- Dtype: same as the original array.

## Notes

- One dimension may be `-1` to let NumPy infer its size.
- The total number of elements must remain unchanged.

## Warnings

- A `ValueError` is raised if the new shape is incompatible with the total number of elements.

## Common Errors

- `ValueError: cannot reshape array of size 12 into shape (5,)`
- Attempting to reshape an array with `-1` in more than one dimension.

## Related Functions

- `np.reshape()`
- `resize()`
- `ravel()`
- `flatten()`

```python
# reshape() examples
import numpy as np
input_array = np.arange(12)
print('input_array:', input_array)
reshaped_matrix = input_array.reshape((3, 4))
print('reshaped_matrix:
', reshaped_matrix)

# Use -1 to infer one dimension automatically
reshaped_inferred = input_array.reshape((2, -1))
print('reshaped_inferred shape:', reshaped_inferred.shape)
print('reshaped_inferred:
', reshaped_inferred)

# Demonstrate that reshape returns a view whenever possible
input_array[0] = 999
print('reshaped_matrix after origin change:
', reshaped_matrix)
```

# resize()

## Definition

`resize()` changes the shape and size of an array in-place when called as a method, or returns a new array when used as `np.resize()`.

## Purpose

Use `resize()` when you want to expand or shrink an array and potentially fill new entries with repeated data.

## Real-world Use Cases

- Expanding a weight vector to a new shape for demonstration.
- Shrinking a data buffer to a smaller size.
- Creating example arrays with repeated patterns.

## Syntax

`array.resize(new_shape, refcheck=True)`

`np.resize(array, new_shape)`

## Parameters

| Parameter | Type | Default | Description |
| --- | --- | --- | --- |
| `new_shape` | int or tuple of ints | required | Desired shape. |
| `refcheck` | bool | True | If False, allow resizing even if there are references to the array. Only for the method form. |

## Return Value

- `array.resize(...)` returns `None` and modifies the array in-place.
- `np.resize(array, new_shape)` returns a new array.

## Notes

- The method form changes the target array directly.
- `np.resize()` repeats the array if the new size is larger.

## Warnings

- The method form may fail if the array has active references and `refcheck=True`.
- Resizing can silently repeat data when expanding.

## Common Errors

- `ValueError: cannot resize an array that references or is referenced by another array in this way`
- Misunderstanding the difference between in-place and new-array behavior.

## Related Functions

- `np.reshape()`
- `np.full()`
- `np.zeros_like()`

```python
# resize() examples
base_array = np.arange(6)
print('original array:', base_array)
base_array.resize((3, 2))  # changes shape in-place
print('resized in-place:
', base_array)

# Use np.resize() to create a new array with repeated data
source_array = np.array([1, 2, 3])
resized_copy = np.resize(source_array, (2, 5))
print('source_array:', source_array)
print('resized_copy shape:', resized_copy.shape)
print('resized_copy:', resized_copy)
```

# flatten()

## Definition

`flatten()` returns a copy of the array collapsed into one dimension.

## Purpose

Use `flatten()` when you need a one-dimensional copy of the array for algorithms that require flat input.

## Real-world Use Cases

- Flattening image data for a fully connected neural network.
- Converting a tensor into a feature vector.
- Serializing array values for disk or network transport.

## Syntax

`array.flatten(order='C')`

## Parameters

| Parameter | Type | Default | Description |
| --- | --- | --- | --- |
| `order` | {'C', 'F', 'A'} | 'C' | Read the elements in row-major, column-major, or preserve order. |

## Return Value

- Returns: 1-D `numpy.ndarray` with copied data.
- Shape: `(array.size,)`.

## Notes

- Always returns a copy, so the result is independent of the original array.

## Warnings

- Because it returns a copy, `flatten()` can use more memory than `ravel()`.

## Related Functions

- `ravel()`
- `reshape()`
- `np.ravel()`

```python
# flatten() examples
matrix = np.arange(9).reshape((3, 3))
flat_copy = matrix.flatten()
print('matrix:
', matrix)
print('flat_copy:', flat_copy)

# Modify the flattened copy and show original is unchanged
flat_copy[0] = -1
print('modified flat_copy:', flat_copy)
print('original matrix after modification:
', matrix)
```

# ravel()

## Definition

`ravel()` returns a flattened array, often as a view of the original data.

## Purpose

Use `ravel()` when you want a one-dimensional view into the data if possible, to avoid copying.

## Real-world Use Cases

- Iterating over all elements without allocating new memory.
- Using flattened data as input to algorithms that accept 1-D arrays.
- Examining array contents in a single dimension.

## Syntax

`array.ravel(order='C')`

## Parameters

| Parameter | Type | Default | Description |
| --- | --- | --- | --- |
| `order` | {'C', 'F', 'A'} | 'C' | Read the elements in row-major, column-major, or preserve order. |

## Return Value

- Returns: 1-D `numpy.ndarray`.
- May return a view or a copy depending on memory layout.

## Notes

- If the original array is contiguous, `ravel()` returns a view.
- If not, it may return a copy.

## Warnings

- Modifying a returned view may modify the original array. Use `copy()` to avoid this.

## Related Functions

- `flatten()`
- `reshape()`

```python
# ravel() examples
matrix = np.arange(12).reshape((3, 4))
flat_view = matrix.ravel()
print('matrix:
', matrix)
print('flat_view:', flat_view)
print('shares memory:', np.shares_memory(matrix, flat_view))

# Modify the view and see the original change
flat_view[0] = -10
print('matrix after modification:
', matrix)
```

# squeeze()

## Definition

`squeeze()` removes axes of length one from an array.

## Purpose

Use `squeeze()` when you want to eliminate unnecessary singleton dimensions.

## Real-world Use Cases

- Converting shape `(1, n)` to `(n,)` after loading data.
- Removing extra dimensions after broadcasted computations.
- Preparing arrays for functions that expect fewer dimensions.

## Syntax

`array.squeeze(axis=None)`

## Parameters

| Parameter | Type | Default | Description |
| --- | --- | --- | --- |
| `axis` | int or tuple of ints or None | None | Specific axes of length one to remove. If None, all singleton axes are removed. |

## Return Value

- Returns: `numpy.ndarray` with singleton dimensions removed.
- Shape: fewer axes if any were length one.

## Notes

- `squeeze()` does not change axes with length greater than one.

## Common Errors

- `ValueError: cannot select an axis to squeeze out which has size not equal to one` when axis is not a singleton axis.

## Related Functions

- `np.expand_dims()`
- `reshape()`

```python
# squeeze() examples
tensor = np.zeros((1, 3, 1, 4))
print('tensor shape:', tensor.shape)
squeezed = tensor.squeeze()
print('squeezed shape:', squeezed.shape)

# Remove a specific axis of length one
tensor2 = np.zeros((1, 5, 1))
squeezed_axis = tensor2.squeeze(axis=0)
print('tensor2 shape:', tensor2.shape)
print('squeezed_axis shape:', squeezed_axis.shape)
```

# expand_dims()

## Definition

`expand_dims()` adds a new axis to an array at the specified position.

## Purpose

Use `expand_dims()` when you need to increase array rank for broadcasting or for functions that require extra dimensions.

## Real-world Use Cases

- Converting a 1-D feature vector into a 2-D batch with shape `(1, n)`.
- Adding a channel dimension to grayscale images.
- Preparing arrays for broadcasting in arithmetic calculations.

## Syntax

`np.expand_dims(array, axis)`

## Parameters

| Parameter | Type | Default | Description |
| --- | --- | --- | --- |
| `array` | ndarray | required | Input array. |
| `axis` | int | required | Position where the new axis is inserted. |

## Return Value

- Returns: `numpy.ndarray` with an additional dimension.
- Shape: one more axis than the input.

## Notes

- Negative axis values count from the end.

## Common Errors

- `AxisError` if the axis is out of range for the new array rank.

## Related Functions

- `np.squeeze()`
- `np.newaxis`

```python
# expand_dims() examples
vector = np.array([1, 2, 3])
expanded_first = np.expand_dims(vector, axis=0)  # shape becomes (1, 3)
expanded_last = np.expand_dims(vector, axis=1)  # shape becomes (3, 1)
print('vector shape:', vector.shape)
print('expanded_first shape:', expanded_first.shape)
print('expanded_last shape:', expanded_last.shape)
```

# transpose()

## Definition

`transpose()` permutes the axes of an array.

## Purpose

Use `transpose()` to swap dimensions for matrix operations, tensor manipulation, and broadcasting.

## Real-world Use Cases

- Switching from row-major to column-major orientation.
- Transposing feature matrices for linear algebra.
- Reordering dimensions in deep learning tensors.

## Syntax

`array.transpose(*axes)`

## Parameters

| Parameter | Type | Default | Description |
| --- | --- | --- | --- |
| `axes` | tuple of ints | None | Desired ordering of axes. If None, reverses axes. |

## Return Value

- Returns: `numpy.ndarray` with permuted axes.
- May return a view if possible.

## Notes

- `array.T` is a shorthand for `array.transpose()`.

## Common Errors

- `AxisError` if axes are repeated or out of range.

## Related Functions

- `np.swapaxes()`
- `np.moveaxis()`

```python
# transpose() examples
image = np.zeros((64, 128, 3))  # height, width, channels
transposed_image = image.transpose(2, 0, 1)  # channels first
print('image shape:', image.shape)
print('transposed_image shape:', transposed_image.shape)

# Transpose a matrix and compare with .T
matrix = np.array([[1, 2], [3, 4]])
print('matrix:
', matrix)
print('matrix.T:
', matrix.T)
print('matrix.transpose():
', matrix.transpose())
```

# swapaxes()

## Definition

`swapaxes()` exchanges two existing axes of the array.

## Purpose

Use `swapaxes()` when you need to swap only two axes instead of permuting all axes.

## Real-world Use Cases

- Rearranging a `(batch, height, width, channels)` tensor to `(batch, channels, height, width)` for deep learning frameworks.
- Exchanging the first and third dimensions of a 3-D tensor.
- Swapping axes for broadcasting compatibility.

## Syntax

`array.swapaxes(axis1, axis2)`

## Parameters

| Parameter | Type | Default | Description |
| --- | --- | --- | --- |
| `axis1` | int | required | First axis to swap. |
| `axis2` | int | required | Second axis to swap. |

## Return Value

- Returns: `numpy.ndarray` with the two axes swapped.
- Generally returns a view.

## Notes

- The axes must both exist in the array.

## Common Errors

- `AxisError` when axis indices are invalid or equal.

## Related Functions

- `np.transpose()`
- `np.moveaxis()`

```python
# swapaxes() examples
tensor = np.zeros((2, 3, 4))
swapped = tensor.swapaxes(0, 2)
print('tensor shape:', tensor.shape)
print('swapped shape:', swapped.shape)

# Swap width and channel axes in a 4-D tensor
batch_tensor = np.zeros((8, 64, 64, 3))
swapped_batch = batch_tensor.swapaxes(1, 3)
print('batch_tensor shape:', batch_tensor.shape)
print('swapped_batch shape:', swapped_batch.shape)
```

# moveaxis()

## Definition

`moveaxis()` moves specified axes to new positions while preserving the order of the remaining axes.

## Purpose

Use `moveaxis()` for flexible axis reordering in multi-dimensional arrays.

## Real-world Use Cases

- Moving a channel axis to the front for a convolutional neural network.
- Reordering a tensor from `(depth, height, width)` to `(height, width, depth)`.
- Preparing arrays for libraries that expect a specific axis order.

## Syntax

`np.moveaxis(array, source, destination)`

## Parameters

| Parameter | Type | Default | Description |
| --- | --- | --- | --- |
| `array` | ndarray | required | Input array. |
| `source` | int or sequence of ints | required | Axis or axes to move. |
| `destination` | int or sequence of ints | required | Destination positions for the moved axes. |

## Return Value

- Returns: `numpy.ndarray` with moved axes.
- The order of all other axes is preserved.

## Notes

- `source` and `destination` must have the same length when sequences are provided.

## Common Errors

- `AxisError` when source or destination axes are invalid.

## Related Functions

- `np.transpose()`
- `np.swapaxes()`

```python
# moveaxis() examples
input_tensor = np.zeros((10, 20, 30))
moved = np.moveaxis(input_tensor, 0, 2)  # move first axis to last position
print('input_tensor shape:', input_tensor.shape)
print('moved shape:', moved.shape)

# Move multiple axes simultaneously
tensor = np.zeros((2, 3, 4, 5))
moved_axes = np.moveaxis(tensor, [0, 3], [3, 0])
print('tensor shape:', tensor.shape)
print('moved_axes shape:', moved_axes.shape)
```

## View vs Copy Behavior

Understanding whether an operation returns a view or a copy is essential for writing correct, memory-efficient code.

- `reshape()` usually returns a view.
- `ravel()` may return a view or a copy.
- `flatten()` always returns a copy.
- `transpose()`, `swapaxes()`, and `moveaxis()` usually return views.

Use `np.shares_memory()` to verify whether two arrays share the same underlying buffer.

```python
# View vs copy summary examples
matrix = np.arange(12).reshape((3, 4))
reshaped = matrix.reshape((4, 3))
flat_ravel = matrix.ravel()
flat_copy = matrix.flatten()
transpose_view = matrix.T
print('shares_memory(matrix, reshaped):', np.shares_memory(matrix, reshaped))
print('shares_memory(matrix, flat_ravel):', np.shares_memory(matrix, flat_ravel))
print('shares_memory(matrix, flat_copy):', np.shares_memory(matrix, flat_copy))
print('shares_memory(matrix, transpose_view):', np.shares_memory(matrix, transpose_view))
```

## Real-world Example

Use array manipulation to reshape sensor data for time-series analysis and prepare tensors for deep learning.

```python
# Real-world tensor manipulation example
sensor_data = np.random.rand(100, 12)  # 100 time steps, 12 channels
# Prepare a batch of 10 sequences of length 10 with 12 features each
sequence_data = sensor_data[:100].reshape((10, 10, 12))
print('sensor_data shape:', sensor_data.shape)
print('sequence_data shape:', sequence_data.shape)

# Add a batch dimension for model input
batch_input = np.expand_dims(sequence_data, axis=0)
print('batch_input shape:', batch_input.shape)

# Move channels to the front for a model that expects (batch, channels, time, features)
reordered = np.moveaxis(batch_input, 3, 1) if batch_input.ndim > 3 else batch_input
print('reordered shape:', reordered.shape)
```

## Edge Cases

### Empty arrays

Most manipulation methods work with empty arrays but preserve the total size of zero.

### Singleton dimensions

`squeeze()` and `expand_dims()` are especially useful when removing or adding singleton axes.

### High-dimensional arrays

Functions such as `moveaxis()` and `transpose()` are designed for tensors with many axes.

### Non-contiguous arrays

Some operations may force a copy if the array is not contiguous.

```python
# Edge case examples
empty_array = np.array([])
print('empty_array reshape to (0, 1):', empty_array.reshape((0, 1)).shape)

singleton_array = np.array([[[42]]])
print('singleton_array shape:', singleton_array.shape)
print('squeezed shape:', singleton_array.squeeze().shape)
print('expanded shape:', np.expand_dims(singleton_array.squeeze(), axis=0).shape)

# Non-contiguous array from transpose
matrix = np.arange(12).reshape((3, 4))
transposed = matrix.T
print('transposed is contiguous:', transposed.flags.c_contiguous)
print('transposed reshape shape:', transposed.reshape((4, 3)).shape)
```

# Summary

## Key Points

- `reshape()` changes array shape while preserving elements.
- `resize()` can modify arrays in-place or create a repeated new array.
- `flatten()` returns a copy and `ravel()` returns a view when possible.
- `squeeze()` removes singleton dimensions and `expand_dims()` adds them.
- `transpose()`, `swapaxes()`, and `moveaxis()` reorder axes for tensor manipulation.

## Cheat Sheet

- `arr.reshape(shape)` — reshape.
- `arr.resize(shape)` — in-place resize.
- `np.resize(arr, shape)` — new array with repeated data.
- `arr.flatten()` — copy to 1-D.
- `arr.ravel()` — flattened view or copy.
- `arr.squeeze()` — remove size-1 axes.
- `np.expand_dims(arr, axis)` — add an axis.
- `arr.transpose(*axes)` — permute axes.
- `arr.swapaxes(i, j)` — swap two axes.
- `np.moveaxis(arr, source, destination)` — move axes.

## Memory Tricks

- Use `ravel()` instead of `flatten()` when you want to avoid copies.
- Use `reshape()` on contiguous data to get a view.
- Avoid `np.resize()` for large arrays unless repeated data is acceptable.

## Best Practices

- Validate the total number of elements before reshaping.
- Use `copy()` explicitly when you need independent data.
- Use clear axis ordering when transposing or moving axes.

## Common Mistakes

- Confusing `reshape()` with `resize()`.
- Assuming `flatten()` returns a view.
- Forgetting that `squeeze()` only removes axes of length one.
