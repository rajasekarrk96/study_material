## Introduction

NumPy excels when operations are expressed on entire arrays. This chapter explains common optimization patterns, avoiding bottlenecks and making the most of array computation.


# Vectorization vs Python loops

Vectorized operations are far faster than explicit Python loops for large arrays.


```python
import numpy as np
arr = np.arange(1000000)
# vectorized
res_vec = arr * 2
# Python loop
res_loop = []
for x in arr:
    res_loop.append(x * 2)
print('vectorized first 5:', res_vec[:5])
```

# Memory layout and ordering

Contiguous arrays are faster. Use `np.ascontiguousarray` or `np.asfortranarray` when needed.


```python
a = np.arange(12).reshape(3, 4)
b = a.T
print('a flags:', a.flags.c_contiguous)
print('b flags:', b.flags.c_contiguous)
c = np.ascontiguousarray(b)
print('c flags:', c.flags.c_contiguous)
```

# Broadcasting for efficiency

Broadcasting avoids explicit repetition of data and can compute results with minimal memory overhead.


```python
x = np.arange(3)
y = np.arange(4).reshape(4, 1)
print('broadcasted result:
', x + y)
```

# Using np.einsum

`np.einsum` expresses complex operations concisely and can be faster for many tensor contractions.


```python
A = np.arange(6).reshape(2, 3)
B = np.arange(6).reshape(3, 2)
print('einsum result:
', np.einsum('ij,jk->ik', A, B))
print('dot result:
', A @ B)
```

# Avoiding unnecessary copies

Copying arrays allocates memory. Use views when possible and avoid `np.copy` unless needed.


```python
x = np.arange(9).reshape(3, 3)
y = x[::2]
print('view shape:', y.shape)
print('is view:', y.base is x)
```
