## Introduction

These functions are vectorized and optimized in C, making them significantly faster than Python loops. This chapter focuses on the most important math tools for scientific computing, data preparation, and analytics.


# Arithmetic Operations

NumPy arrays support `+`, `-`, `*`, `/`, `**`, and `%` for elementwise arithmetic. These operations broadcast across shapes and return new arrays with the computed results.


```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print('a + b =', a + b)
print('a - b =', a - b)
print('a * b =', a * b)
print('a / b =', a / b)
print('a ** 2 =', a ** 2)
print('b % a =', b % a)
```

## Universal Functions (ufuncs)

Universal functions compute elementwise results for arrays. Common NumPy math ufuncs include `np.sqrt`, `np.exp`, `np.log`, `np.sin`, and `np.maximum`. Ufuncs support the `out`, `where`, and `dtype` parameters.


```python
x = np.array([1.0, 4.0, 9.0, 16.0])
print('sqrt:', np.sqrt(x))
print('exp:', np.exp(x))
print('log:', np.log(x))
print('sin:', np.sin(x))
print('maximum with 10:', np.maximum(x, 10))
```

## Power and Exponential Functions

Use `np.power`, `np.exp`, `np.log`, and `np.log10` for exponential and logarithmic computations. These functions are essential for growth models, scaling, and numerical transformations.


```python
values = np.array([1.0, 10.0, 100.0])
print('power:', np.power(values, 2))
print('exp:', np.exp(values))
print('log10:', np.log10(values))
print('log1p:', np.log1p(values))
```

## Trigonometric Functions

NumPy supports trigonometric functions such as `sin`, `cos`, `tan`, and their inverses. These are often used in signal processing, geometry, and physics simulations.


```python
angles = np.array([0.0, np.pi/6, np.pi/4, np.pi/2])
print('sin:', np.sin(angles))
print('cos:', np.cos(angles))
print('tan:', np.tan(angles))
print('arcsin:', np.arcsin(np.sin(angles)))
```

## Comparison and Boolean Math

Comparison operators produce boolean arrays that can be used for filtering, counting, and conditional arithmetic. Boolean arrays can also be treated as integers where `True` is 1 and `False` is 0.


```python
data = np.array([5, 15, 25, 35])
mask = data > 20
print('mask:', mask)
print('values > 20:', data[mask])
print('sum of mask:', np.sum(mask))
print('boolean arithmetic:', mask * data)
```

## Aggregation Functions

Use `np.sum`, `np.prod`, `np.cumsum`, and `np.cumprod` to aggregate values across arrays. These functions support the `axis` and `keepdims` parameters for multidimensional arrays.


```python
matrix = np.arange(1, 7).reshape((2, 3))
print('matrix:
', matrix)
print('sum all:', np.sum(matrix))
print('sum over axis 0:', np.sum(matrix, axis=0))
print('cumsum:', np.cumsum(matrix))
print('prod over axis 1:', np.prod(matrix, axis=1))
```

## Conditional Operations with `where`

`np.where` selects values from two arrays based on a condition, or returns indices when only a condition is provided.


```python
scores = np.array([70, 85, 90, 55, 60])
result = np.where(scores >= 75, 'pass', 'fail')
print('result:', result)
adjusted = np.where(scores < 60, scores + 10, scores)
print('adjusted scores:', adjusted)
```

## Rounding and Clipping

NumPy provides functions like `np.round`, `np.floor`, `np.ceil`, and `np.clip` to control numerical precision and bound values within a range.


```python
values = np.array([1.2, 2.5, 3.7, -0.4])
print('round:', np.round(values, 1))
print('floor:', np.floor(values))
print('ceil:', np.ceil(values))
print('clip between 0 and 3:', np.clip(values, 0, 3))
```

## Real-world Example

Use NumPy math operations to compute a logistic activation for a feature vector and then clip the outputs to a valid probability range.


```python
features = np.array([-3.0, -1.0, 0.0, 1.0, 3.0])
logistic = 1 / (1 + np.exp(-features))
print('logistic:', logistic)
print('clipped logistic:', np.clip(logistic, 0.0, 1.0))
```

## Edge Cases

- Beware of division by zero when using `/` or `np.log`.
- Use `np.nan_to_num` or `np.errstate` to manage invalid values.
- Some functions preserve dtype while others return floats for safer results.


```python
values = np.array([1.0, 0.0, -1.0])
with np.errstate(divide='ignore', invalid='ignore'):
    result = np.log(values)
print('log result:', result)
print('nan_to_num:', np.nan_to_num(result, nan=0.0, neginf=-1e9, posinf=1e9))
```
