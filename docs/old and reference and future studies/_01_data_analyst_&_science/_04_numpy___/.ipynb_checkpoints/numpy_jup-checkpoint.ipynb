```python
"""
numpy coding
"""
import numpy as np

numArray=np.array([[1,2,3],[4,5,6,]])

print(numArray)
```

```python

#can view array in console variable explorer and click the varibale name to view it graphical
np1=np.array([[1,3],[4,5]])

np2=np.array([[3,4],[6,7]])
```

```python
#numpy array multiplication(matrix)
np3=np1@np2

print(np3)
```

```python
#normal multiple with cell values
np4=np1*np2

print(np4)
```

```python

#numpy array multiplication(matrix)
np5=np.dot(np1,np2)

print(np5)
```

```python
#normal multiple with cell values
np6=np.multiply(np1,np2)

print(np6)
```

```python
sum1=np1+np2

print(sum1)
```

```python

sub1=np1-np2

print(sub1)
```

```python
sub2=np.subtract(np1,np2)

print(sub2)
```

```python
sum2=np.sum(np1)

print(sum2)
```
