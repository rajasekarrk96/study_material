## Introduction

Structured arrays support mixed data types in a single NumPy array. They are useful for tabular data where each row has multiple typed fields, such as name, age, and salary.


# Creating structured arrays

Define a dtype with fields and create arrays from tuples or dictionaries.


```python
import numpy as np
dtype = [('name', 'U10'), ('age', 'i4'), ('score', 'f4')]
people = np.array([('Alice', 30, 85.5), ('Bob', 25, 91.2)], dtype=dtype)
print(people)
print(people['age'])
print(people['name'])
```

# Field access and slicing

Access fields by name, and slice rows with array indexing.


```python
print('first person:', people[0])
print('scores > 90:', people[people['score'] > 90])
```

# Record arrays

Record arrays allow dot notation to access fields. Use `view(np.recarray)` to convert.


```python
rec = people.view(np.recarray)
print(rec.name)
print(rec.age)
```

# Structured array operations

Use boolean masks and arithmetic on individual fields.


```python
ages = people['age']
print('average age:', np.mean(ages))
people['score'] += 5
print(people)
```

# Sorting by fields

Sort structured arrays with the `order` parameter.


```python
sorted_by_age = np.sort(people, order='age')
print(sorted_by_age)
sorted_by_score = np.sort(people, order='score')
print(sorted_by_score)
```

# Advanced dtype definitions

Structured dtypes may include nested fields and variable-length strings.


```python
nested_dtype = [('id', 'i4'), ('info', [('height', 'f4'), ('weight', 'f4')])]
nested = np.array([(1, (1.7, 70.0)), (2, (1.8, 80.0))], dtype=nested_dtype)
print(nested['info']['height'])
```
