## Introduction

Bitwise operations work at the binary level of integer values. They are useful for low-level data processing, feature flags, and fast masking when working with integer arrays.


# Bitwise AND/OR/XOR

Use `np.bitwise_and`, `np.bitwise_or`, and `np.bitwise_xor` for elementwise binary operations.


```python
import numpy as np
a = np.array([1, 2, 3], dtype=np.int32)
b = np.array([3, 1, 0], dtype=np.int32)
print('and:', np.bitwise_and(a, b))
print('or :', np.bitwise_or(a, b))
print('xor:', np.bitwise_xor(a, b))
```

# Bitwise NOT

`np.invert` flips every bit in the integer representation. For booleans it returns logical NOT.


```python
flags = np.array([True, False, True])
print('invert booleans:', np.invert(flags))
vals = np.array([0, 1, 2], dtype=np.uint8)
print('invert integers:', np.invert(vals))
```

# Bit shifts

Use `np.left_shift` and `np.right_shift` to move bits left or right.


```python
nums = np.array([1, 2, 4], dtype=np.int32)
print('left shift by 1:', np.left_shift(nums, 1))
print('right shift by 1:', np.right_shift(nums, 1))
```

# Logical vs bitwise operators

Use `&`, `|`, `^`, and `~` only with boolean or integer arrays, and use Python `and` and `or` only for scalar booleans.


```python
mask = (np.array([1, 0, 1]) == 1) & (np.array([0, 1, 1]) == 1)
print('bitwise and mask:', mask)
```

# Masking with bitwise operations

Create a mask from integer flags to select data based on bits.


```python
flags = np.array([0b001, 0b010, 0b011, 0b100], dtype=np.int32)
mask = np.bitwise_and(flags, 0b001) != 0
print('flags:', flags)
print('mask for least significant bit:', mask)
print('selected flags:', flags[mask])
```

## Real-world Example

Use bitwise operations to encode and decode feature flags stored as integers.


```python
READ = 1
WRITE = 2
EXEC = 4
permissions = np.array([1, 3, 5, 7], dtype=np.int32)
can_write = np.bitwise_and(permissions, WRITE) != 0
print('can write:', can_write)
```

## Summary

- Bitwise operators work on integer arrays at the binary level.
- `invert` flips bits or negates booleans.
- `left_shift` and `right_shift` move bits along positions.
- Use boolean masks and bitwise logic for fast selection.
- Avoid Python `and`/`or` with arrays; use `&` and `|` instead.
