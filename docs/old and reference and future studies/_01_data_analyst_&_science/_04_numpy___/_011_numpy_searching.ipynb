## Introduction

Search operations are essential for finding values, positions, and patterns in array data. NumPy offers fast vectorized search and indexing functions that work with both 1-D and multidimensional arrays.


# np.where()

`np.where(condition, x, y)` selects values based on a boolean condition. With a single argument, it returns indices where the condition is True.


```python
import numpy as np
arr = np.array([5, 10, 15, 20, 25])
indices = np.where(arr > 15)
print('indices:', indices)
print('values > 15:', arr[indices])
result = np.where(arr % 2 == 0, arr, -arr)
print('conditional result:', result)
```

# np.nonzero() and np.flatnonzero()

`np.nonzero` returns a tuple of index arrays for nonzero elements. `np.flatnonzero` returns flat indices of nonzero elements.


```python
matrix = np.array([[0, 1, 2], [3, 0, 4]])
print('nonzero:', np.nonzero(matrix))
print('flatnonzero:', np.flatnonzero(matrix))
```

# np.argwhere()

`np.argwhere(condition)` returns the coordinates of True values in a boolean array.


```python
mask = matrix > 1
print('mask:
', mask)
print('argwhere:
', np.argwhere(mask))
```

# argmax and argmin

Use `np.argmax` and `np.argmin` to find the index of the maximum or minimum value along an axis.


```python
arr2 = np.array([[1, 5, 3], [4, 2, 6]])
print('argmax axis 0:', np.argmax(arr2, axis=0))
print('argmin axis 1:', np.argmin(arr2, axis=1))
```

# np.searchsorted()

`np.searchsorted(sorted_array, values)` returns insertion indices for maintaining sorted order.


```python
sorted_values = np.array([1, 3, 5, 7])
print('insertion indices:', np.searchsorted(sorted_values, [2, 6]))
```

# Membership testing with np.isin()

Use `np.isin` to test whether each element of one array is present in another array.


```python
items = np.array([1, 2, 3, 4, 5])
print('isin [2,4,6]:', np.isin(items, [2, 4, 6]))
print('filtered:', items[np.isin(items, [2, 4, 6])])
```

# Real-world Example

Filter temperature readings above a threshold and locate their positions.


```python
temps = np.array([15.2, 18.7, 21.5, 19.0, 22.3])
mask = temps > 20
print('hot indices:', np.where(mask))
print('hot temperatures:', temps[mask])
```
