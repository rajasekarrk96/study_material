## Introduction

# Descriptive Statistics

Use `np.mean`, `np.median`, `np.var`, and `np.std` to compute central tendency and dispersion.


```python
import numpy as np

data = np.array([10, 15, 20, 25, 30])
print('mean:', np.mean(data))
print('median:', np.median(data))
print('variance:', np.var(data))
print('std deviation:', np.std(data))
```

## Min, Max, and Range

`np.min`, `np.max`, `np.amin`, and `np.amax` compute extrema. The range is the difference between max and min.


```python
print('min:', np.min(data))
print('max:', np.max(data))
print('range:', np.max(data) - np.min(data))
```

## Percentiles and Quantiles

Use `np.percentile` and `np.quantile` to compute data cutoffs. Percentiles are especially useful for identifying medians, quartiles, and boundaries for outliers.


```python
print('25th percentile:', np.percentile(data, 25))
print('50th percentile:', np.percentile(data, 50))
print('75th percentile:', np.percentile(data, 75))
print('quantiles:', np.quantile(data, [0.25, 0.5, 0.75]))
```

## Unique Values and Counts

`np.unique` returns the sorted unique values in an array and can also return counts for each value.


```python
values = np.array([1, 2, 2, 3, 3, 3])
unique_values, counts = np.unique(values, return_counts=True)
print('unique:', unique_values)
print('counts:', counts)
```

## Histograms and Bins

`np.histogram` computes the distribution of values over specified bins. This is useful for visualizing data density.


```python
hist, bin_edges = np.histogram(data, bins=4)
print('hist:', hist)
print('bin edges:', bin_edges)
```

## Covariance and Correlation

Use `np.cov` to compute the covariance matrix and `np.corrcoef` to compute correlation coefficients between variables.


```python
X = np.array([[1, 2, 3], [2, 3, 5], [3, 5, 8]])
print('covariance matrix:
', np.cov(X, rowvar=False))
print('correlation matrix:
', np.corrcoef(X, rowvar=False))
```

## Handling Missing Values

NumPy provides `np.nanmean`, `np.nanstd`, `np.nanmedian`, and `np.nanpercentile` to ignore `NaN` values in computations.


```python
data_with_nan = np.array([1.0, np.nan, 3.0, 4.0])
print('mean ignoring NaN:', np.nanmean(data_with_nan))
print('std ignoring NaN:', np.nanstd(data_with_nan))
print('median ignoring NaN:', np.nanmedian(data_with_nan))
```

## Axis Reductions and `keepdims`

Statistics often require aggregation across a specific axis. Use `axis` to control which dimension is reduced, and `keepdims=True` to preserve the original rank for broadcasting.


```python
matrix = np.array([[10, 20, 30], [15, 25, 35]])
print('mean axis=0:', np.mean(matrix, axis=0))
print('mean axis=1 with keepdims:
', np.mean(matrix, axis=1, keepdims=True))
```

## Weighted Average

Use `np.average` with the `weights` parameter to compute weighted means. This is useful when some observations count more than others.


```python
values = np.array([1.0, 2.0, 3.0])
weights = np.array([0.1, 0.3, 0.6])
print('weighted average:', np.average(values, weights=weights))
```

## Counting and Cumulative Statistics

`np.count_nonzero`, `np.cumsum`, and `np.cumprod` help compute frequent statistics and running totals.


```python
array = np.array([2, 0, 2, 3])
print('count nonzero:', np.count_nonzero(array))
print('cumsum:', np.cumsum(array))
print('cumprod:', np.cumprod(array))
```

## Real-world Example

Summarize exam scores across students and subjects, including mean, median, percentiles, and variance.


```python
scores = np.array([[80, 90, 70], [85, 78, 92], [88, 82, 79]])
print('student means:', np.mean(scores, axis=1))
print('subject medians:', np.median(scores, axis=0))
print('75th percentile overall:', np.percentile(scores, 75))
print('score variance:', np.var(scores, axis=0))
```

## Edge Cases

- Many statistical functions return `NaN` when the input is empty.
- `np.nan*` functions ignore missing values but preserve `NaN` if all entries are missing.
- Covariance and correlation require at least two observations per variable.


```python
empty = np.array([])
print('empty mean:', np.mean(empty) if empty.size else 'empty array')
print('nanmean empty:', np.nanmean(empty) if empty.size else 'empty array')
```
