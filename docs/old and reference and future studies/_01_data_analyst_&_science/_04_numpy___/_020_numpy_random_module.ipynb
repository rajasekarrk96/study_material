## Introduction

Random numbers are essential for simulation, sampling, and testing. NumPy's modern `Generator` API provides a flexible and reproducible interface.


# Creating a random number generator

Use `np.random.default_rng()` to create a `Generator` object.


```python
import numpy as np
rng = np.random.default_rng(12345)
print('random float:', rng.random())
print('random integers:', rng.integers(1, 10, size=5))
```

# Common distributions

Generate random values from normal, binomial, and uniform distributions.


```python
print('normal:', rng.normal(loc=0.0, scale=1.0, size=5))
print('binomial:', rng.binomial(n=10, p=0.5, size=5))
print('uniform:', rng.uniform(low=0.0, high=1.0, size=5))
```

# Permutations and sampling

Use `shuffle`, `permutation`, and `choice` for sampling and random ordering.


```python
data = np.arange(10)
print('permutation:', rng.permutation(data))
rng.shuffle(data)
print('shuffled:', data)
print('choice:', rng.choice(data, size=4, replace=False))
```

# Reproducibility with seeds

Create the same generator state by using the same seed.


```python
rng1 = np.random.default_rng(42)
rng2 = np.random.default_rng(42)
print('same samples:', np.all(rng1.random(5) == rng2.random(5)))
```
