# Data Structure Revision Cheat Sheet

A concise reference for key data structure concepts, operations, and complexity.


## 1. Core Python Data Structures

| Structure | Ordered | Mutable | Duplicates | Use case |
|---|---|---|---|---|
| `list` | Yes | Yes | Yes | ordered collection, indexing |
| `tuple` | Yes | No | Yes | fixed records, immutable keys |
| `set` | No | Yes | No | unique items, membership tests |
| `dict` | No | Yes | No (keys) | key-value lookup and mapping |
| `deque` | Yes | Yes | Yes | fast insert/delete at both ends |


## 2. Linear Structures

### Stack
- LIFO: push, pop, peek, is_empty.
- Use `list.append()` / `list.pop()` or `collections.deque`.

### Queue
- FIFO: enqueue, dequeue, peek.
- Use `collections.deque` for `append` and `popleft`.

### Linked List
- Node-based sequence with `next` pointers.
- Efficient insertion/deletion when you have a node reference.
- No direct random access.


## 3. Non-linear Structures

### Tree
- Hierarchical nodes with root and children.
- Binary tree: at most 2 children.
- BST property: left < node < right.
- Traversals: inorder, preorder, postorder, level-order.

### Graph
- Nodes (vertices) connected by edges.
- Types: undirected/directed, weighted/unweighted.
- Representations: adjacency list, adjacency matrix.
- Search: DFS, BFS.

### Heap
- Binary heap is a complete tree.
- Min-heap: smallest element at root.
- Use `heapq` for priority queue behavior.

### Trie
- Prefix tree for words.
- Efficient prefix search and autocomplete.

### BFS on a graph
```python
from collections import deque

def bfs(graph, start):
    queue = deque([start])
    visited = {start}
    order = []
    while queue:
        node = queue.popleft()
        order.append(node)
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)
    return order
```

### DFS template
```python
def dfs(graph, node, visited=None):
    if visited is None:
        visited = set()
    visited.add(node)
    for neighbor in graph[node]:
        if neighbor not in visited:
            dfs(graph, neighbor, visited)
    return visited
```
