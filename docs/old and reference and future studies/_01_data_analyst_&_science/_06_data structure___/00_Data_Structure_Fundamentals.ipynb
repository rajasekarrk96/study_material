## 1. What is a Data Structure?

A data structure is a way to store, organize, and manage data so it can be accessed and modified efficiently. Data structures are essential for writing fast, maintainable programs and solving complex problems in data science, algorithms, and software engineering.

**Key ideas:**
- **Storage:** how data is placed in memory or storage.
- **Organization:** how data elements relate to each other.
- **Operations:** insert, delete, search, update, traverse, sort.
- **Efficiency:** time complexity and space complexity for operations.

**Why this matters:**
- Data structure choice affects the performance of software.
- In data science, the right structure helps manage data pipelines, feature stores, and algorithm inputs.
- In teaching, data structures give students a foundation for problem solving and code design.

## 2. Classification of Data Structures

Data structures can be classified by shape and behavior. The main categories are linear and non-linear.

### 2.1 Linear Data Structures
- Elements are arranged in sequence.
- Each element has a unique predecessor and successor, except the first and last.
- Examples: `list`, `tuple`, `array`, `linked list`, `stack`, `queue`.

### 2.2 Non-linear Data Structures
- Elements are connected in a hierarchical or graph-based way.
- There is no single sequence order.
- Examples: `tree`, `graph`, `heap`, `matrix`.

### 2.3 Static vs Dynamic
- **Static structures**: fixed size after creation (conceptually like arrays or tuples).
- **Dynamic structures**: size can grow or shrink during execution (like Python lists, sets, dictionaries).

## 3. Abstract Data Types (ADT)

An Abstract Data Type defines the operations on a structure without defining how those operations are implemented.

### Examples of ADTs
- `Stack`: push, pop, peek, is_empty.
- `Queue`: enqueue, dequeue, peek, is_empty.
- `Map` / `Dictionary`: put, get, remove, contains.
- `Set`: add, remove, contains, union, intersection.

**Important:** The same ADT can be implemented differently. For example, a stack can use a list, linked list, or deque internally.

## 4. Python Built-in Data Structures

Python provides powerful built-in data structures for common use cases. Each structure has a clear purpose and set of operations.

### 4.1 `list`
- Ordered sequence of elements.
- Mutable and allows duplicates.
- Fast random access by index.

### 4.2 `tuple`
- Ordered and immutable.
- Can be used as dictionary keys if elements are hashable.

### 4.3 `set`
- Unordered collection of unique elements.
- Supports membership tests and set algebra.

### 4.4 `dict`
- Unordered mapping of keys to values.
- Fast key-based lookup.

### 4.5 `deque` from `collections`
- Double-ended queue with fast insertion and deletion at both ends.
- Useful for implementing stacks and queues efficiently.

```python
# Python built-in data structure examples
numbers = [10, 20, 30]  # list
coordinates = (1, 2, 3)  # tuple
unique_words = {'python', 'data', 'science'}  # set
student = {'name': 'Alex', 'age': 25, 'city': 'Delhi'}  # dict

print('list:', numbers)
print('tuple:', coordinates)
print('set:', unique_words)
print('dict:', student)

from collections import deque
dq = deque(['first', 'second'])
dq.append('third')
dq.appendleft('zero')
print('deque:', dq)
```

## 5. Data Structure Operations

Common operations for data structures include:
- **Create**: build a structure from data.
- **Read**: access elements or values.
- **Insert**: add new elements.
- **Delete**: remove elements.
- **Search**: find elements by value or key.
- **Traverse**: visit each element in order.
- **Sort**: order elements by a rule.
- **Update**: change values in place.

Different structures optimize different operations, so choose based on the use case.
