## 1. Frequency Counting with Dictionaries

Dictionaries are ideal for counting occurrences and building fast lookup tables.

### Example: word frequency in text


```python
text = 'data science data analysis data science'
counts = {}
for word in text.split():
    counts[word] = counts.get(word, 0) + 1
print('Word counts:', counts)
# Use-case: feature extraction, bag-of-words, token counting
```

## 2. Set Operations for Deduplication and Membership

Sets are efficient for unique value extraction and membership tests.

### Example: remove duplicate entries from a dataset


```python
emails = ['a@example.com', 'b@example.com', 'a@example.com', 'c@example.com']
unique_emails = set(emails)
print('Unique emails:', unique_emails)
# Use-case: remove duplicate user IDs or values
```

## 3. Lists and Tuples for Ordered Data

Lists and tuples store ordered sequences. Use lists for mutable data and tuples for immutable records.

### Example: sequence of measurements and coordinate pairs

## 4. Matrix Applications in Image and Grid Data

Matrices represent 2D data like images, heatmaps, and grid-based simulations.

### Example: transpose and row sums


```python
image = [[0, 1, 2], [3, 4, 5], [6, 7, 8]]
transpose = [[image[r][c] for r in range(len(image))] for c in range(len(image[0]))]
row_sums = [sum(row) for row in image]
print('Transpose:')
for row in transpose: print(row)
print('Row sums:', row_sums)
# Use-case: image processing, table operations
```

## 5. Priority Queues for Scheduling

Heaps are useful when tasks need to be processed based on priority.

### Example: schedule jobs by priority


```python
import heapq
jobs = [(2, 'low priority task'), (1, 'high priority task'), (3, 'medium priority task')]
heapq.heapify(jobs)
while jobs:
    priority, task = heapq.heappop(jobs)
    print('Processing', task, 'with priority', priority)
# Use-case: scheduling, event simulation, server queues
```

## 6. Graph Applications

Graphs model relationships in social networks, recommendations, maps, and supply chains.

### Example: BFS for shortest unweighted path


```python
from collections import deque
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D'],
    'C': ['A', 'D'],
    'D': ['B', 'C', 'E'],
    'E': ['D']
}
def shortest_path(graph, start, goal):
    queue = deque([(start, [start])])
    visited = {start}
    while queue:
        node, path = queue.popleft()
        if node == goal:
            return path
        for neighbor in graph[node]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append((neighbor, path + [neighbor]))
    return None

print('Shortest path from A to E:', shortest_path(graph, 'A', 'E'))
# Use-case: recommendation systems, route planning
```

## 7. Tree Applications

Trees capture hierarchical data such as file systems, organization charts, XML/JSON parsing, and decision trees.

### Example: simple decision tree logic


```python
class DecisionNode:
    def __init__(self, question=None, yes=None, no=None, result=None):
        self.question = question
        self.yes = yes
        self.no = no
        self.result = result

def evaluate(node, answer):
    if node.result is not None:
        return node.result
    if answer(node.question):
        return evaluate(node.yes, answer)
    return evaluate(node.no, answer)

root = DecisionNode('is_raining', yes=DecisionNode(result='Take umbrella'), no=DecisionNode(result='No umbrella needed'))
print(evaluate(root, lambda q: q == 'is_raining'))
# Use-case: decision logic, rule-based systems
```

## 8. Hash Table Applications

Hash tables are used for caching results, counting features, and fast membership tests.

### Example: caching function results with a dictionary


```python
cache = {}
def fib(n):
    if n in cache:
        return cache[n]
    if n < 2:
        return n
    cache[n] = fib(n - 1) + fib(n - 2)
    return cache[n]

print('Fibonacci 10:', fib(10))
print('Cache keys:', cache.keys())
# Use-case: memoization, caching expensive computations
```
