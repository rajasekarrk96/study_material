## 1. Matrix Basics

A matrix is a 2D grid of values organized into rows and columns. In Python, matrices are most commonly represented as nested lists.

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9],
]
print('Matrix:')
for row in matrix:
    print(row)
print('Rows:', len(matrix), 'Cols:', len(matrix[0]))
```

### 1.1 Accessing elements
- `matrix[row][col]` returns the element at that position.
- Row and column indexes start at 0.

### 1.2 Traversal
- Row-major: visit elements row by row.
- Column-major: visit elements column by column.

### 1.3 Common operations
- Sum of rows or columns.
- Transpose.
- Search in sorted matrix.
- Rotate or flip matrix.

```python
def traverse_row_major(matrix):
    for row in matrix:
        for value in row:
            print(value, end=' ')
        print()

def traverse_col_major(matrix):
    rows = len(matrix)
    cols = len(matrix[0]) if rows else 0
    for c in range(cols):
        for r in range(rows):
            print(matrix[r][c], end=' ')
        print()

print('Row-major traversal:')
traverse_row_major(matrix)
print('Column-major traversal:')
traverse_col_major(matrix)
```

## 2. Matrix Transformations

### 2.1 Transpose
Transposing a matrix flips rows and columns.

```python
def transpose(matrix):
    return [[matrix[r][c] for r in range(len(matrix))] for c in range(len(matrix[0]))]

print('Transpose:')
for row in transpose(matrix):
    print(row)
```

### 2.2 Matrix multiplication
- Multiply an `m x n` matrix by an `n x p` matrix to get an `m x p` result.
- Each output element is the dot product of one row and one column.

### 2.3 Search in sorted matrix
- If rows and columns are sorted, start from top-right and move left/down to find a target in O(m + n).

```python
def multiply_matrices(a, b):
    rows_a, cols_a = len(a), len(a[0])
    rows_b, cols_b = len(b), len(b[0])
    if cols_a != rows_b:
        raise ValueError('Incompatible matrix sizes')
    result = [[0] * cols_b for _ in range(rows_a)]
    for i in range(rows_a):
        for j in range(cols_b):
            for k in range(cols_a):
                result[i][j] += a[i][k] * b[k][j]
    return result

a = [[1, 2], [3, 4]]
b = [[5, 6], [7, 8]]
print('Product:')
for row in multiply_matrices(a, b):
    print(row)
```

```python
def search_sorted_matrix(matrix, target):
    if not matrix or not matrix[0]:
        return None
    rows, cols = len(matrix), len(matrix[0])
    r, c = 0, cols - 1
    while r < rows and c >= 0:
        if matrix[r][c] == target:
            return (r, c)
        elif matrix[r][c] > target:
            c -= 1
        else:
            r += 1
    return None

sorted_matrix = [
    [1, 4, 7, 11],
    [2, 5, 8, 12],
    [3, 6, 9, 16],
    [10, 13, 14, 17],
]
print('Search 9:', search_sorted_matrix(sorted_matrix, 9))
print('Search 15:', search_sorted_matrix(sorted_matrix, 15))
```

## 3. Nested Lists

Nested lists let you build hierarchical or matrix-like data in plain Python. Use them for adjacency lists, grids, and JSON-like structures.

### Example: nested dictionary and list combination


```python
school = {
    'classA': ['Alice', 'Bob'],
    'classB': ['Charlie', 'Dana'],
    'classC': ['Esha', 'Farhan']
}
for class_name, students in school.items():
    print(class_name, '->', students)

# Access nested element
print('Second student in classB:', school['classB'][1])
```

## 4. Sparse Matrices

A sparse matrix has many zero values. Store it efficiently using a dictionary or list of tuples rather than a full nested list.

### Example representation


```python
sparse = {(0, 1): 5, (2, 3): 10}
print('Sparse matrix entries:')
for position, value in sparse.items():
    print(position, '->', value)

# Convert sparse to dense 4x4
rows, cols = 4, 4
dense = [[sparse.get((r, c), 0) for c in range(cols)] for r in range(rows)]
for row in dense:
    print(row)
```
