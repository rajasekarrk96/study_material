## 1. Even Squares from a List

**Problem:** Given a list of integers, return a new list that contains the square of each even number only.

**Example:** `[1, 2, 3, 4]` -> `[4, 16]`


```python
def even_squares(numbers):
    return [x * x for x in numbers if x % 2 == 0]

print(even_squares([1, 2, 3, 4]))  # [4, 16]
print(even_squares([10, 11, 12]))  # [100, 144]
```

## 2. Remove Duplicates and Preserve Order

**Problem:** Given a list, return a new list with duplicate values removed while preserving the original order.

**Example:** `[3, 1, 2, 3, 2]` -> `[3, 1, 2]`


```python
def remove_duplicates_preserve_order(items):
    seen = set()
    result = []
    for item in items:
        if item not in seen:
            seen.add(item)
            result.append(item)
    return result

print(remove_duplicates_preserve_order([3, 1, 2, 3, 2]))  # [3, 1, 2]
print(remove_duplicates_preserve_order(["a", "b", "a", "c"]))  # ["a", "b", "c"]
```

## 3. Word Frequency with a Dictionary

**Problem:** Count the frequency of each word in a sentence and return a dictionary of counts.

**Example:** `"hello world hello"` -> `{"hello": 2, "world": 1}`


```python
def word_frequency(sentence):
    words = sentence.lower().split()
    counts = {}
    for word in words:
        counts[word] = counts.get(word, 0) + 1
    return counts

print(word_frequency("hello world hello"))  # {"hello": 2, "world": 1}
print(word_frequency("Data science is data science"))
```

## 4. Breadth-First Search (BFS) on a Graph

**Problem:** Implement BFS traversal on an adjacency-list graph and return the visit order.

**Example:** `graph = {'A': ['B', 'C'], 'B': ['A', 'D'], 'C': ['A'], 'D': ['B']}`


```python
from collections import deque

def bfs(graph, start):
    visited = {start}
    queue = deque([start])
    order = []
    while queue:
        node = queue.popleft()
        order.append(node)
        for neighbor in graph.get(node, []):
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)
    return order

graph = {
    "A": ["B", "C"],
    "B": ["A", "D"],
    "C": ["A"],
    "D": ["B"],
}
print(bfs(graph, "A"))  # ["A", "B", "C", "D"]
```

## 5. Parentheses Balance with a Stack

**Problem:** Use a stack to validate whether the parentheses in an expression are balanced.

**Example:** `"(a + b) * (c + d)"` -> `True`; `"(a + b * (c + d)"` -> `False`


```python
def is_balanced_parentheses(expr):
    stack = []
    pairs = {
        ")": "(",
        "]": "[",
        "}": "{",
    }
    for char in expr:
        if char in pairs.values():
            stack.append(char)
        elif char in pairs:
            if not stack or stack[-1] != pairs[char]:
                return False
            stack.pop()
    return len(stack) == 0

print(is_balanced_parentheses("(a + b) * (c + d)"))  # True
print(is_balanced_parentheses("(a + b * (c + d)"))  # False
```

## 6. Search in a Sorted Matrix

**Problem:** Given a matrix with rows and columns sorted in ascending order, search for a target value in O(m+n) time.

**Example:** `target = 9` in `[[1,4,7],[2,5,8],[3,6,9]]`


```python
def search_sorted_matrix(matrix, target):
    if not matrix or not matrix[0]:
        return None
    rows, cols = len(matrix), len(matrix[0])
    r, c = 0, cols - 1
    while r < rows and c >= 0:
        if matrix[r][c] == target:
            return (r, c)
        elif matrix[r][c] > target:
            c -= 1
        else:
            r += 1
    return None

matrix = [
    [1, 4, 7],
    [2, 5, 8],
    [3, 6, 9],
]
print(search_sorted_matrix(matrix, 9))  # (2, 2)
print(search_sorted_matrix(matrix, 10))  # None
```
