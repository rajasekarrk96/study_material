## 1. What is a Data Structure?

A data structure is a way to store and organize data so it can be used efficiently. In Python, data structures can be built-in types like `list`, `tuple`, `set`, and `dict`, or custom structures like linked lists, stacks, queues, trees, and graphs.

Key concepts:
- **Data**: values or objects that represent information.
- **Structure**: the organization of data in memory.
- **Operations**: insertion, deletion, traversal, search, update, and sorting.
- **Complexity**: time and space cost for operations.

### 1.1 Classification of Data Structures

- **Linear data structures**: elements arranged in sequential order. Example: `list`, `tuple`, `array`, `linked list`, `stack`, `queue`.
- **Non-linear data structures**: elements connected in complex relationships. Example: `tree`, `graph`, `heap`, `matrix`.
- **Static data structures**: fixed size and layout. Example: arrays in low-level languages, but in Python `tuple` is conceptually static.
- **Dynamic data structures**: can change size. Example: `list`, `dict`, `set`.

### 1.2 Abstract Data Types (ADT)

An ADT defines a set of operations without specifying implementation. For example: `Stack`, `Queue`, `Map`, `Set`, `List`. The implementation can use arrays, linked lists, or other containers.

## 2. Python Built-in Data Structures

Python provides several built-in data structures with rich syntax and behavior. These are the most commonly used for data analysis and algorithm design.

### 2.1 List

A `list` is an ordered, mutable collection of items. It can store elements of different types.

```python
# List examples and operations
fruits = ['apple', 'banana', 'cherry']
print('Original:', fruits)

# Access by index
print('First:', fruits[0])
print('Last:', fruits[-1])

# Modify
fruits[1] = 'blueberry'
print('Modified:', fruits)

# Add and remove
fruits.append('date')
fruits.insert(1, 'avocado')
print('After append and insert:', fruits)

fruits.remove('apple')
print('After remove apple:', fruits)
popped = fruits.pop()
print('Popped:', popped, 'Remaining:', fruits)

# Slicing and copying
slice_example = fruits[1:3]
copy_example = fruits.copy()
print('Slice:', slice_example)
print('Copy:', copy_example)
```

#### Important list operations
- `len(list)`
- `list.append(x)`
- `list.extend(iterable)`
- `list.insert(i, x)`
- `list.remove(x)`
- `list.pop([i])`
- `list.clear()`
- `list.index(x)`
- `list.count(x)`
- `list.sort()`
- `list.reverse()`
- `sorted(list)`
- `list.copy()`

Lists are ideal for ordered collections when you need fast indexing and iteration.

### 2.2 Tuple

A `tuple` is an ordered, immutable collection. Use tuples when data should not change and when you want hashable container elements.

```python
# Tuple examples
coordinates = (10, 20)
print('Coordinates:', coordinates)

# Access and unpack
x, y = coordinates
print('x =', x, 'y =', y)

# Tuple operations
print('Length:', len(coordinates))
print('Count of 10:', coordinates.count(10))
print('Index of 20:', coordinates.index(20))

# Nested tuple with mixed types
mixed = (1, 'python', (2, 3))
print('Mixed:', mixed)
```

#### Tuple notes
- Tuples are immutable. Elements cannot be changed after creation.
- Tuples can be used as dictionary keys or set elements if they contain only hashable types.
- Use tuples when you need fixed collections or when returning multiple values from a function.

### 2.3 Set

A `set` is an unordered collection of unique items. Sets are useful for membership tests, duplicates elimination, and mathematical set operations.

```python
# Set examples
fruits = {'apple', 'banana', 'cherry'}
print('Fruits:', fruits)

# Add and remove
fruits.add('date')
fruits.discard('banana')
print('After add and discard:', fruits)

# Membership
print('cherry in fruits?', 'cherry' in fruits)

# Set operations
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}
print('Union:', a | b)
print('Intersection:', a & b)
print('Difference:', a - b)
print('Symmetric difference:', a ^ b)
```

#### Set notes
- Sets are unordered and unindexed.
- Duplicate items are automatically removed.
- Sets are mutable but elements must be hashable.
- Use sets for fast membership tests and operations on unique items.

### 2.4 Dictionary

A `dict` is an unordered collection of key-value pairs. Dictionaries are efficient for lookup by key.

```python
# Dictionary examples
student = {'name': 'Alice', 'age': 23, 'grades': [85, 90, 95]}
print('Student:', student)

# Access values
print('Name:', student['name'])
print('Age:', student.get('age'))
print('Grades:', student.get('grades', []))

# Update and add
student['age'] = 24
student['major'] = 'Data Science'
print('Updated:', student)

# Keys, values, items
print('Keys:', list(student.keys()))
print('Values:', list(student.values()))
print('Items:', list(student.items()))

# Remove entries
student.pop('major')
print('After pop:', student)
student.clear()
print('After clear:', student)
```

#### Dictionary notes
- Keys must be unique and hashable.
- Values can be any type.
- Dictionaries are ideal for mapping relationships, fast lookups, and counting.
- `dict` preserves insertion order in Python 3.7+.

### 2.5 Comparison of built-in structures

- `list`: ordered, mutable, allows duplicates, indexable.
- `tuple`: ordered, immutable, allows duplicates, indexable.
- `set`: unordered, mutable, unique elements, no indexing.
- `dict`: unordered mapping, mutable, unique keys, values indexed by key.

Choose based on whether order, mutability, uniqueness, or key lookup matters.

## 3. Sequence Operations and Iteration

Python sequences support looping, comprehensions, slicing, and built-in functions. These patterns are useful across data structures.

```python
numbers = [1, 2, 3, 4, 5]

# Iteration
for number in numbers:
    print(number, end=' ')
print()

# List comprehension
squares = [n * n for n in numbers]
print('Squares:', squares)

# Filtering with comprehension
even = [n for n in numbers if n % 2 == 0]
print('Even numbers:', even)

# Generator expression
even_gen = (n for n in numbers if n % 2 == 0)
print('Even generator:', list(even_gen))
```

## 4. Common Data Structure Patterns

This section covers classic ADTs and their Python implementations.

### 4.1 Stack

A stack is a LIFO (Last-In, First-Out) structure. Use append/pop on a list or `deque` for efficient operations.

```python
from collections import deque

# Stack with list
stack = []
stack.append(1)
stack.append(2)
stack.append(3)
print('Stack list:', stack)

top = stack.pop()
print('Popped from list stack:', top, 'Remaining:', stack)

# Stack with deque
stack_deque = deque()
stack_deque.append(1)
stack_deque.append(2)
stack_deque.append(3)
print('Stack deque:', stack_deque)

top_deque = stack_deque.pop()
print('Popped from deque stack:', top_deque, 'Remaining:', stack_deque)
```

#### Stack notes
- Use when you need backtracking, undo history, or expression evaluation.
- `push` is `append`, `pop` is `pop()`.
- `deque` is faster for large stacks.

### 4.2 Queue

A queue is a FIFO (First-In, First-Out) structure. Use `deque` for efficient queue operations.

```python
from collections import deque

queue = deque()
queue.append('a')
queue.append('b')
queue.append('c')
print('Queue:', queue)

first = queue.popleft()
print('Dequeued:', first, 'Remaining:', queue)

second = queue.popleft()
print('Dequeued:', second, 'Remaining:', queue)
```

#### Queue notes
- Use when processing tasks in arrival order and scheduling jobs.
- `append` adds, `popleft` removes from the front.
- Avoid list `pop(0)` for performance.

### 4.3 Linked List

A linked list stores elements as nodes that point to the next node. Each node contains data and a reference to the next node.

```python
class ListNode:
    def __init__(self, value=None, next=None):
        self.value = value
        self.next = next

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, value):
        new_node = ListNode(value)
        if not self.head:
            self.head = new_node
            return
        current = self.head
        while current.next:
            current = current.next
        current.next = new_node

    def traverse(self):
        values = []
        current = self.head
        while current:
            values.append(current.value)
            current = current.next
        return values

    def search(self, value):
        current = self.head
        while current:
            if current.value == value:
                return True
            current = current.next
        return False

# Create and use linked list
linked = LinkedList()
linked.append(10)
linked.append(20)
linked.append(30)
print('Linked list values:', linked.traverse())
print('Search 20:', linked.search(20))
print('Search 40:', linked.search(40))
```

#### Linked list notes
- Offers flexible insertions and deletions.
- No direct indexing; traversal is required.
- Useful for building stacks, queues, and adjacency lists.

### 4.4 Matrix and 2D Lists

A matrix is a two-dimensional structure represented by nested lists. It is useful for grids, image data, and numerical tables.

```python
# Matrix traversal and search examples
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

def traverse_row_major(matrix):
    for row in matrix:
        for value in row:
            print(value, end=' ')
        print()

print('Row-major traversal:')
traverse_row_major(matrix)

def get_matrix_dimensions(matrix):
    rows = len(matrix)
    cols = len(matrix[0]) if rows > 0 else 0
    return rows, cols

print('Dimensions:', get_matrix_dimensions(matrix))
```

#### Matrix notes
- Use nested lists for matrices.
- Access elements with `matrix[row][col]`.
- Common operations: traversal, transpose, row/column sum, and matrix multiplication.

### 4.5 Tree (Binary Tree)

A tree is a hierarchical structure with nodes and child relationships. A binary tree has at most two children per node.

```python
class TreeNode:
    def __init__(self, value):
        self.value = value
        self.left = None
        self.right = None

def inorder_traversal(node):
    if node is None:
        return []
    return inorder_traversal(node.left) + [node.value] + inorder_traversal(node.right)

def preorder_traversal(node):
    if node is None:
        return []
    return [node.value] + preorder_traversal(node.left) + preorder_traversal(node.right)

def postorder_traversal(node):
    if node is None:
        return []
    return postorder_traversal(node.left) + postorder_traversal(node.right) + [node.value]

# Build tree
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)

print('Inorder:', inorder_traversal(root))
print('Preorder:', preorder_traversal(root))
print('Postorder:', postorder_traversal(root))
```

#### Tree notes
- Use trees for hierarchical and sorted data.
- Common traversals: inorder, preorder, postorder, breadth-first.
- Binary trees are the foundation of binary search trees, heaps, and expression parsing.

### 4.6 Graphs

A graph is a collection of nodes (vertices) connected by edges. Graphs can be directed or undirected, weighted or unweighted.

```python
# Graph using adjacency list
graph = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()
    visited.add(start)
    for neighbor in graph[start]:
        if neighbor not in visited:
            dfs(graph, neighbor, visited)
    return visited

def bfs(graph, start):
    visited = {start}
    queue = [start]
    order = []
    while queue:
        vertex = queue.pop(0)
        order.append(vertex)
        for neighbor in graph[vertex]:
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)
    return order

print('DFS from A:', dfs(graph, 'A'))
print('BFS from A:', bfs(graph, 'A'))
```

#### Graph notes
- Use graphs for networks, relationships, and routing problems.
- Representations: adjacency list, adjacency matrix, edge list.
- Common algorithms: DFS, BFS, shortest path, cycle detection.

## 5. Complexity and Performance

Understanding time and space complexity is crucial when choosing a data structure.

### 5.1 Common operation complexity

- `list` indexing: O(1)
- `list` append: O(1) amortized
- `list` insert/remove at front: O(n)
- `dict` lookup/insert/delete: O(1) average
- `set` add/remove/membership: O(1) average
- `tuple` indexing: O(1)
- `deque` append/pop both ends: O(1)

For nested structures, complexity depends on traversal depth and size of inner collections.

### 5.2 Choosing the right structure

- Use `list` for ordered collection with frequent random access.
- Use `tuple` for fixed collections and immutable records.
- Use `set` for unique elements and fast membership checks.
- Use `dict` for key-value mappings and frequency counts.
- Use `deque` for efficient FIFO/LIFO operations.
- Use linked lists for frequent insertions/deletions in the middle when random access is not required.
- Use trees or graphs for hierarchical or networked data.
