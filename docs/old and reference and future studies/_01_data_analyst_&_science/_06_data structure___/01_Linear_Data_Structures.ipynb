# Linear Data Structures

This notebook covers the main linear data structures in Python with definitions, implementation details, examples, and study notes.

## 1. Arrays and Lists

### Definition
A list is a dynamic array in Python. It stores items in order and allows fast access by index.

### Properties
- Ordered sequence.
- Mutable: elements can be changed after creation.
- Allows duplicates.
- Supports slicing, indexing, and concatenation.

### Common operations
- `append()` adds to the end.
- `insert()` adds at a specific position.
- `pop()` removes by index.
- `remove()` deletes by value.
- `sort()` orders elements.

### Example


```python
numbers = [5, 2, 9, 1, 5]
print('Original list:', numbers)
numbers.append(10)
print('After append:', numbers)
numbers.insert(2, 7)
print('After insert(2,7):', numbers)
numbers.remove(5)
print('After remove(5):', numbers)
last_item = numbers.pop()
print('Popped item:', last_item)
print('Sorted list:', sorted(numbers))
print('Sliced list [1:4]:', numbers[1:4])
```

### Notes on Python lists
- Lists are built on a dynamic array; append is amortized O(1).
- Insertion or removal at the beginning is O(n).
- Use lists when order matters and you need indexing.

## 2. Tuples

### Definition
A tuple is an ordered, immutable sequence. Use tuples for fixed collections and records.

### Properties
- Ordered and indexable.
- Immutable: cannot change elements after creation.
- Can contain mixed types.
- Can be used as dictionary keys if all elements are hashable.

### Example


```python
point = (10, 20)
print('Point tuple:', point)
x, y = point
print('x:', x, 'y:', y)
record = ('Raja', 28, 'Python')
print('Record:', record)
print('Index of 28:', record.index(28))
print('Count of 10 in point:', point.count(10))
```

## 3. Sets

### Definition
A set is an unordered collection of unique items. It supports fast membership checks and set operations.

### Properties
- Unordered and unindexed.
- No duplicate elements.
- Supports union, intersection, difference, symmetric difference.

### Example


```python
tools = {'hammer', 'wrench', 'screwdriver', 'hammer'}
print('Set of tools:', tools)
tools.add('pliers')
print('After add:', tools)
tools.discard('wrench')
print('After discard:', tools)
a = {1, 2, 3}
b = {2, 3, 4}
print('Union:', a | b)
print('Intersection:', a & b)
print('Difference:', a - b)
print('Symmetric difference:', a ^ b)
```

## 4. Dictionaries

### Definition
A dictionary maps keys to values. It is ideal for associative lookup, counting, and storing structured records.

### Properties
- Unordered (in older Python versions) but preserves insertion order in Python 3.7+.
- Keys must be unique and hashable.
- Values can be any type.

### Example


```python
student = {'name': 'Asha', 'age': 24, 'grade': 'A'}
print('Student dict:', student)
print('Name:', student['name'])
print('Age with get:', student.get('age'))
student['grade'] = 'A+'
student['city'] = 'Mumbai'
print('Updated dict:', student)
print('Keys:', list(student.keys()))
print('Values:', list(student.values()))
print('Items:', list(student.items()))
```

## 5. Linked Lists

### Definition
A linked list is a linear data structure where each element is a node that contains data and a reference to the next node.

### Types
- Singly linked list: nodes point only to the next node.
- Doubly linked list: nodes point to both next and previous nodes.
- Circular linked list: last node points back to first node.

### Advantages
- Dynamic size.
- Efficient insertion and deletion at arbitrary positions when you have a node reference.

### Disadvantages
- No fast random access by index.
- Higher memory overhead for pointers.

### Example: Singly Linked List implementation


```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class SinglyLinkedList:
    def __init__(self):
        self.head = None

    def insert_at_end(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            return
        current = self.head
        while current.next:
            current = current.next
        current.next = new_node

    def traverse(self):
        current = self.head
        values = []
        while current:
            values.append(current.data)
            current = current.next
        return values

    def search(self, target):
        current = self.head
        while current:
            if current.data == target:
                return current
            current = current.next
        return None

linked = SinglyLinkedList()
linked.insert_at_end(10)
linked.insert_at_end(20)
linked.insert_at_end(30)
print('Linked list values:', linked.traverse())
print('Search for 20:', linked.search(20).data if linked.search(20) else None)
```

## 6. Stack

### Definition
A stack is a linear data structure that follows Last-In-First-Out (LIFO). Elements are added and removed from the top.

### Primary operations
- `push(item)` adds an item to the top.
- `pop()` removes the top item.
- `peek()` returns the top item without removing it.
- `is_empty()` checks whether the stack is empty.

### Example using Python list


```python
stack = []
stack.append('first')
stack.append('second')
stack.append('third')
print('Stack content:', stack)
print('Top element:', stack[-1])
print('Pop element:', stack.pop())
print('After pop:', stack)
```

## 7. Queue

### Definition
A queue is a linear data structure that follows First-In-First-Out (FIFO). Elements are added at the rear and removed from the front.

### Primary operations
- `enqueue(item)` adds to the rear.
- `dequeue()` removes from the front.
- `peek()` returns the front item.

### Example using `deque` for efficiency


```python
from collections import deque
queue = deque()
queue.append('first')
queue.append('second')
queue.append('third')
print('Queue content:', queue)
print('Dequeue:', queue.popleft())
print('After dequeue:', queue)
```

## 8. Notes and Best Practices

- Use Python lists for general ordered data with frequent indexing.
- Use tuples for fixed records or keys in dictionaries.
- Use sets when uniqueness and membership speed are most important.
- Use dictionaries for key-value lookup and counts.
- Use linked lists when insertion/deletion in the middle is frequent and random access is not required.
- Use a `deque` for real queues and fast double-ended access.
