## 1. Complexity Basics

### Big-O notation
Big-O describes the upper bound on the runtime or space use of an algorithm as input size grows. It expresses the worst-case behavior in terms of `n`.

### Common complexity classes
- `O(1)`: constant time. operation cost remains fixed.
- `O(log n)`: logarithmic time. often binary search or balanced tree operations.
- `O(n)`: linear time. cost grows in proportion to input size.
- `O(n log n)`: linearithmic time. common for efficient sorting.
- `O(n^2)`: quadratic time. nested loops over input.
- `O(2^n)`: exponential time. recursive combinations and power sets.

### Space complexity
- Measures additional memory used by the algorithm.
- `O(1)` means no extra memory besides inputs.
- `O(n)` means memory grows with input size.

## 2. Complexity of Common Python Data Structures

| Structure | Operation | Time Complexity | Notes |
|---|---|---|---|
| `list` | index access | O(1) | constant-time random access |
| `list` | append | O(1) amortized | resizing occasionally costs O(n) |
| `list` | insert/delete at front | O(n) | shifts elements |
| `tuple` | access | O(1) | immutable sequence |
| `set` | add/remove | O(1) average | hash-based |
| `dict` | get/set | O(1) average | hash-based |
| `deque` | append/pop ends | O(1) | double-ended queue |
| linked list | insert/delete at front | O(1) | if node reference is available |
| linked list | access by index | O(n) | sequential traversal |
| binary tree | search/insert | O(h) | h = tree height |
| graph BFS/DFS | traverse all nodes | O(V + E) | vertices + edges |

### Note
Actual performance may vary with implementation, Python version, and underlying hardware.

## 3. Choosing the Right Data Structure

### Decision rules
- If you need fast index access, choose `list` or `tuple`.
- If you need unique values, choose `set`.
- If you need mapping from key to value, choose `dict`.
- If you need FIFO order, choose `queue` (`deque`).
- If you need LIFO order, choose `stack`.
- If you need dynamic insertion or deletion, consider linked lists.
- If you need hierarchical or networked data, choose trees or graphs.

### Data science note
- Use `pandas` and `numpy` when working with large tabular or numerical data, but understand Python structures for preprocessing and custom algorithms.

```python
# Exercise 1: Count unique words
def count_unique_words(sentence):
    words = sentence.lower().split()
    return len(set(words))

sentence = 'Data science is data science and Python is powerful'
print('Unique words:', count_unique_words(sentence))

# Exercise 2: Reverse a linked list
class Node:
    def __init__(self, value):
        self.value = value
        self.next = None

def reverse_list(head):
    prev, current = None, head
    while current:
        nxt = current.next
        current.next = prev
        prev = current
        current = nxt
    return prev

head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
reversed_head = reverse_list(head)
values = []
current = reversed_head
while current:
    values.append(current.value)
    current = current.next
print('Reversed linked list:', values)

# Exercise 3: Matrix diagonal sum
def diagonal_sum(matrix):
    n = len(matrix)
    primary = sum(matrix[i][i] for i in range(n))
    secondary = sum(matrix[i][n-1-i] for i in range(n))
    if n % 2 == 1:
        return primary + secondary - matrix[n//2][n//2]
    return primary + secondary

matrix = [[1,2,3],[4,5,6],[7,8,9]]
print('Diagonal sum:', diagonal_sum(matrix))

# Exercise 4: BFS traversal
def bfs(graph, start):
    visited = set([start])
    queue = [start]
    order = []
    while queue:
        node = queue.pop(0)
        order.append(node)
        for neighbor in graph.get(node, []):
            if neighbor not in visited:
                visited.add(neighbor)
                queue.append(neighbor)
    return order

graph = {'A': ['B', 'C'], 'B': ['D'], 'C': ['E'], 'D': [], 'E': []}
print('BFS order:', bfs(graph, 'A'))

# Exercise 5: Balanced parentheses checker
def is_balanced(expression):
    stack = []
    pairs = {'(': ')', '[': ']', '{': '}'}
    for ch in expression:
        if ch in pairs:
            stack.append(ch)
        elif ch in pairs.values():
            if not stack or pairs[stack.pop()] != ch:
                return False
    return not stack

print('Balanced example:', is_balanced('({[]})'))
print('Unbalanced example:', is_balanced('([)]'))
```
