### Q14. Update Salary By ID

**Approach:** Solve the problem with clear PL/SQL-style procedural thinking first, then verify it with direct test cases.


```python
def update_plsql_salary_by_id(rows: list[dict], employee_id: int, new_salary: int) -> list[dict]:
    # Step 1: Build a new list so the input is not mutated.
    updated_rows = []

    # Step 2: Copy each row and update only the matching one.
    for row in rows:
        new_row = row.copy()
        if new_row["id"] == employee_id:
            new_row["salary"] = new_salary
        updated_rows.append(new_row)

    # Step 3: Return the updated rows.
    return updated_rows


# Dry run:
# id 1 matches -> salary becomes 55000

# Time complexity: O(n)

input_data = ([{"id": 1, "salary": 50000}], 1, 55000)
expected_output = [{"id": 1, "salary": 55000}]
print(update_plsql_salary_by_id(*input_data))  # should match expected_output
assert update_plsql_salary_by_id(*input_data) == expected_output
```

### Q19. Detect Missing Required Fields

**Approach:** Solve the problem with clear PL/SQL-style procedural thinking first, then verify it with direct test cases.


```python
def detect_plsql_missing_required_fields(rows: list[dict], required_fields: list[str]) -> list[int]:
    # Step 1: Collect indices of rows missing at least one required field.
    invalid_indices = []

    # Step 2: Check each row against the required field list.
    for index, row in enumerate(rows):
        for field in required_fields:
            if field not in row:
                invalid_indices.append(index)
                break

    # Step 3: Return the invalid row indices.
    return invalid_indices


# Dry run:
# row 0 missing salary
# row 1 missing name

# Time complexity: O(rows * required_fields)

input_data = ([{"name": "Asha"}, {"salary": 50000}], ["name", "salary"])
expected_output = [0, 1]
print(detect_plsql_missing_required_fields(*input_data))  # should match expected_output
assert detect_plsql_missing_required_fields(*input_data) == expected_output
```

### Q21. Find Rows Above Average Salary

**Approach:** Solve the problem with clear PL/SQL-style procedural thinking first, then verify it with direct test cases.


```python
def find_plsql_rows_above_average_salary(rows: list[dict]) -> list[dict]:
    # Step 1: Return early for empty input.
    if not rows:
        return []

    # Step 2: Compute the average salary.
    average_salary = sum(row["salary"] for row in rows) / len(rows)

    # Step 3: Keep only rows whose salary is above average.
    return [row for row in rows if row["salary"] > average_salary]


# Dry run:
# average = 60000
# only 70000 stays

# Time complexity: O(n)

input_data = [{"salary": 50000}, {"salary": 70000}]
expected_output = [{"salary": 70000}]
print(find_plsql_rows_above_average_salary(input_data))  # should match expected_output
assert find_plsql_rows_above_average_salary(input_data) == expected_output
assert find_plsql_rows_above_average_salary([]) == []
```

### Q22. Rank Departments By Bonus

**Approach:** Solve the problem with clear PL/SQL-style procedural thinking first, then verify it with direct test cases.


```python
def rank_departments_by_bonus(rows: list[dict]) -> list[str]:
    # Step 1: Aggregate bonus totals by department.
    totals = {}
    for row in rows:
        department = row["department"]
        totals[department] = totals.get(department, 0) + row["bonus"]

    # Step 2: Sort by descending total bonus, then by department name.
    ordered = sorted(totals.items(), key=lambda item: (-item[1], item[0]))

    # Step 3: Return only the department names.
    return [department for department, _ in ordered]


# Dry run:
# HR 2000 before IT 1000

# Time complexity: O(n log n)

input_data = [{"department": "IT", "bonus": 1000}, {"department": "HR", "bonus": 2000}]
expected_output = ["HR", "IT"]
print(rank_departments_by_bonus(input_data))  # should match expected_output
assert rank_departments_by_bonus(input_data) == expected_output
```

```python
def collapse_consecutive_procedure_names(procedure_names: list[str]) -> list[str]:
    # Step 1: Return early for empty input.
    if not procedure_names:
        return []

    # Step 2: Start with the first procedure name.
    collapsed = [procedure_names[0]]

    # Step 3: Keep only changes in consecutive names.
    for name in procedure_names[1:]:
        if name != collapsed[-1]:
            collapsed.append(name)

    # Step 4: Return the collapsed names.
    return collapsed


# Dry run:
# load once, then send

# Time complexity: O(n)

input_data = ["load", "load", "send"]
expected_output = ["load", "send"]
print(collapse_consecutive_procedure_names(input_data))  # should match expected_output
assert collapse_consecutive_procedure_names(input_data) == expected_output
```

### Q27. Validate Unique IDs

**Approach:** Solve the problem with clear PL/SQL-style procedural thinking first, then verify it with direct test cases.


```python
def validate_plsql_unique_ids(rows: list[dict]) -> list[int]:
    # Step 1: Count how many times each id appears.
    counts = {}
    for row in rows:
        row_id = row["id"]
        counts[row_id] = counts.get(row_id, 0) + 1

    # Step 2: Keep ids that appear more than once.
    duplicates = [row_id for row_id, count in counts.items() if count > 1]

    # Step 3: Return the duplicates sorted.
    return sorted(duplicates)


# Dry run:
# id 1 appears twice -> [1]

# Time complexity: O(n log n)

input_data = [{"id": 1}, {"id": 1}, {"id": 2}]
expected_output = [1]
print(validate_plsql_unique_ids(input_data))  # should match expected_output
assert validate_plsql_unique_ids(input_data) == expected_output
```

### Q35. Employee Bonus Index

**Approach:** Solve the problem with clear PL/SQL-style procedural thinking first, then verify it with direct test cases.


```python
def employee_bonus_index(rows: list[dict]) -> dict[str, int]:
    # Step 1: Prepare the output dictionary.
    index = {}

    # Step 2: Map each employee name to its bonus.
    for row in rows:
        index[row["name"]] = row["bonus"]

    # Step 3: Return the bonus index.
    return index


# Dry run:
# Asha -> 1000

# Time complexity: O(n)

input_data = [{"name": "Asha", "bonus": 1000}]
expected_output = {"Asha": 1000}
print(employee_bonus_index(input_data))  # should match expected_output
assert employee_bonus_index(input_data) == expected_output
```
