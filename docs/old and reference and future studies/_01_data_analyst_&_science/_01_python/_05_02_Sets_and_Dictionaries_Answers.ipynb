## Q1. Remove Duplicate Numbers
**Step-by-step logic:** Convert the input list into a set.

**Dry run:** `[1, 2, 2, 3, 4, 4]` becomes `{1, 2, 3, 4}`.

**Time complexity:** `O(n)`

```python
def unique_numbers(numbers):
    # Convert the list to a set so duplicate values are removed automatically
    return set(numbers)

input_data = [1, 2, 2, 3, 4, 4]
expected_output = {1, 2, 3, 4}
print(unique_numbers(input_data))  # should match expected
print(expected_output)
```

**Dry run:** `['cat', 'dog', 'cat', 'bird']` becomes `{'cat', 'dog', 'bird'}` and its length is `3`.

**Step-by-step logic:** Copy the set and use `add()`.

```python
def add_city(cities, city):
    # Create a copy so the original set is not changed outside the function unexpectedly
    updated_cities = set(cities)

    # Add the new city to the set
    updated_cities.add(city)

    # Return the updated set
    return updated_cities

input_data = {'Delhi', 'Pune'}
expected_output = {'Delhi', 'Pune', 'Mumbai'}
print(add_city(input_data, 'Mumbai'))  # should match expected
print(expected_output)
```

## Q13. Word Frequency Counter
**Step-by-step logic:** Split the sentence into words and count each word using a dictionary.

**Dry run:** Every time a word appears, its count increases by 1.

**Time complexity:** `O(n)`

```python
def word_frequency(sentence):
    # Split the sentence into a list of words
    words = sentence.split()

    # Create an empty dictionary for counting
    counts = {}

    # Visit each word in the list
    for word in words:
        # Increase the old count if present, otherwise start from 0 and add 1
        counts[word] = counts.get(word, 0) + 1

    # Return the frequency dictionary
    return counts

input_data = 'red blue red green red'
expected_output = {'red': 3, 'blue': 1, 'green': 1}
print(word_frequency(input_data))  # should match expected
print(expected_output)
```

## Q14. Merge Two Dictionaries
**Step-by-step logic:** Copy the first dictionary, then update it using the second dictionary.

**Dry run:** Common keys get replaced by values from the second dictionary.

**Time complexity:** `O(n + m)`

```python
def merge_dicts(dict_a, dict_b):
    # Create a copy of the first dictionary
    merged = dict(dict_a)

    # Update the copy with the second dictionary so repeated keys are overwritten
    merged.update(dict_b)

    # Return the merged dictionary
    return merged

input_a = {'a': 1, 'b': 2}
input_b = {'b': 5, 'c': 3}
expected_output = {'a': 1, 'b': 5, 'c': 3}
print(merge_dicts(input_a, input_b))  # should match expected
print(expected_output)
```

## Q16. Group Numbers by Even and Odd
**Step-by-step logic:** Traverse the list once and place each number into the correct dictionary entry.

**Dry run:** Even numbers go into one list and odd numbers go into the other.

**Time complexity:** `O(n)`

```python
def group_even_odd(numbers):
    # Prepare the output dictionary with two empty lists
    result = {'even': [], 'odd': []}

    # Process each number in the input list
    for number in numbers:
        # Check whether the number is divisible by 2
        if number % 2 == 0:
            # Store the number in the even list
            result['even'].append(number)
        else:
            # Store the number in the odd list
            result['odd'].append(number)

    # Return the grouped dictionary
    return result

input_data = [1, 2, 3, 4, 5]
expected_output = {'even': [2, 4], 'odd': [1, 3, 5]}
print(group_even_odd(input_data))  # should match expected
print(expected_output)
```

## Q19. Invert a Dictionary

**Time complexity:** `O(n)`

## Q20. Count Character Frequency
**Step-by-step logic:** Ignore spaces and count every other character.

**Dry run:** Each letter increases its own counter in the dictionary.

**Time complexity:** `O(n)`

```python
def character_frequency(text):
    # Create an empty dictionary to store character counts
    counts = {}

    # Read one character at a time from the string
    for character in text:
        # Skip spaces because the problem says to ignore them
        if character == ' ':
            continue

        # Increase the count of the current character
        counts[character] = counts.get(character, 0) + 1

    # Return the frequency dictionary
    return counts

input_data = 'data science'
expected_output = {'d': 1, 'a': 2, 't': 1, 's': 1, 'c': 2, 'i': 1, 'e': 2, 'n': 1}
print(character_frequency(input_data))  # should match expected
print(expected_output)
```

## Q21. Find the First Repeating Element
**Step-by-step logic:** Use a set to remember what has already been seen.

**Dry run:** `5` is seen, then `3`, then `4`, then `3` repeats first.

**Time complexity:** `O(n)`

```python
def first_repeating(numbers):
    # Create an empty set to store elements we have already seen
    seen = set()

    # Traverse the list from left to right
    for number in numbers:
        # If the number is already in seen, it is the first repeating element
        if number in seen:
            return number

        # Otherwise add it to seen and continue
        seen.add(number)

    # If no value repeats, return None
    return None

input_data = [5, 3, 4, 3, 5, 6]
expected_output = 3
print(first_repeating(input_data))  # should match expected
print(expected_output)
```

## Q22. Find Duplicate Elements
**Step-by-step logic:** Use one set for seen values and one set for duplicates.

**Dry run:** When a value appears again, add it to the duplicates set.

**Time complexity:** `O(n)`

```python
def find_duplicates(numbers):
    # Create a set for values that have already been seen once
    seen = set()

    # Create a set for values that repeat
    duplicates = set()

    # Read each number from the input list
    for number in numbers:
        # If the number was already seen, record it as a duplicate
        if number in seen:
            duplicates.add(number)
        else:
            # Otherwise store it in the seen set
            seen.add(number)

    # Return all duplicate values
    return duplicates

input_data = [1, 2, 3, 2, 4, 1, 5]
expected_output = {1, 2}
print(find_duplicates(input_data))  # should match expected
print(expected_output)
```

## Q24. Create a Nested Marks Report
**Step-by-step logic:** For each student, sum the values inside the inner marks dictionary.

**Dry run:** Student `A` total is `80 + 90 = 170`, and student `B` total is `70 + 60 = 130`.

**Time complexity:** `O(total subject entries)`

```python
def total_marks(report):
    # Create an empty dictionary to store total marks for each student
    totals = {}

    # Loop through every student and that student's subject-mark dictionary
    for student, marks in report.items():
        # Sum all marks for the current student and store the total
        totals[student] = sum(marks.values())

    # Return the totals dictionary
    return totals

input_data = {'A': {'math': 80, 'science': 90}, 'B': {'math': 70, 'science': 60}}
expected_output = {'A': 170, 'B': 130}
print(total_marks(input_data))  # should match expected
print(expected_output)
```

## Q26. Build a Reverse Index
**Step-by-step logic:** For each student-course pair, add the student name under that course in a new dictionary.

**Dry run:** `Python` collects `Amit` and `Riya`, while `SQL` collects `Neha`.

**Time complexity:** `O(n)`

```python
def reverse_index(student_courses):
    # Create an empty dictionary where each course maps to a set of students
    course_students = {}

    # Process one student-course pair at a time
    for student, course in student_courses.items():
        # If the course is not yet a key, create an empty set for it
        if course not in course_students:
            course_students[course] = set()

        # Add the current student to the set for the current course
        course_students[course].add(student)

    # Return the reverse mapping dictionary
    return course_students

input_data = {'Amit': 'Python', 'Neha': 'SQL', 'Riya': 'Python'}
expected_output = {'Python': {'Amit', 'Riya'}, 'SQL': {'Neha'}}
print(reverse_index(input_data))  # should match expected
print(expected_output)
```

## Q27. Symmetric Difference of Two Lists
**Step-by-step logic:** Convert both lists to sets and use the symmetric difference operator.

**Dry run:** Remove common elements and keep elements present in exactly one side.

**Time complexity:** `O(n + m)`

```python
def symmetric_difference_lists(list_a, list_b):
    # Convert both lists into sets and return elements that appear in exactly one set
    return set(list_a) ^ set(list_b)

input_a = [1, 2, 3, 4]
input_b = [3, 4, 5, 6]
expected_output = {1, 2, 5, 6}
print(symmetric_difference_lists(input_a, input_b))  # should match expected
print(expected_output)
```

## Q28. Most Frequent Word
**Step-by-step logic:** Count each word, then return the key whose value is maximum.

**Dry run:** Word `a` reaches count `3`, which is higher than the others.

**Time complexity:** `O(n)`

```python
def most_frequent_word(sentence):
    # Split the sentence into words
    words = sentence.split()

    # Create an empty dictionary for word counts
    counts = {}

    # Count each word in the sentence
    for word in words:
        counts[word] = counts.get(word, 0) + 1

    # Return the word whose count is highest
    return max(counts, key=counts.get)

input_data = 'a b a c b a'
expected_output = 'a'
print(most_frequent_word(input_data))  # should match expected
print(expected_output)
```

## Q29. Dictionary Difference by Keys

## Q33. Course Enrollment System
**Step-by-step logic:** Check each student to see whether the target course exists in that student's course set.

**Dry run:** Students `A` and `C` contain `Python` in their course lists.

**Time complexity:** `O(n)`

```python
def students_in_course(enrollment, target_course):
    # Create an empty set to store names of matching students
    result = set()

    # Visit each student and that student's course set
    for student, courses in enrollment.items():
        # Add the student if the target course is present
        if target_course in courses:
            result.add(student)

    # Return the set of matching students
    return result

input_data = {'A': {'Python', 'SQL'}, 'B': {'Excel'}, 'C': {'Python'}}
expected_output = {'A', 'C'}
print(students_in_course(input_data, 'Python'))  # should match expected
print(expected_output)
```

## Q35. Login Tracking Dashboard
**Step-by-step logic:** Use a set for unique users and a dictionary for counts at the same time.

**Dry run:** Each login adds the username to the set and increases its counter in the dictionary.

**Time complexity:** `O(n)`

```python
def login_dashboard(logins):
    # Create an empty set to track unique usernames
    unique_users = set()

    # Create an empty dictionary to track how many times each user logged in
    login_counts = {}

    # Process each login event in order
    for username in logins:
        # Add the username to the set of unique users
        unique_users.add(username)

        # Increase the count for the current username
        login_counts[username] = login_counts.get(username, 0) + 1

    # Return both the unique users set and the login count dictionary
    return unique_users, login_counts

input_data = ['raj', 'neha', 'raj', 'aman', 'raj']
expected_output = ({'raj', 'neha', 'aman'}, {'raj': 3, 'neha': 1, 'aman': 1})
print(login_dashboard(input_data))  # should match expected
print(expected_output)
```
