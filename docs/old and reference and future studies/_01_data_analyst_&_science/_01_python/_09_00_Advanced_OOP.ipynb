# Basic Syntax Explanation

## Main ideas
- `class Child(Parent):` creates inheritance.
- Method overriding lets a child class replace parent behavior.
- `super()` calls the parent version of a method.
- Dunder methods such as `__str__()`, `__len__()`, and `__add__()` customize object behavior.
- `@classmethod` works with the class itself.
- `@staticmethod` is related to the class but does not need class or object state.

## Operators involved
- Dot operator `.` for method and attribute access.
- Arithmetic and comparison operators through dunder methods like `+`, `==`, `<`.


```python
# Define a parent class
class Animal:
    # Store the common animal name
    def __init__(self, name):
        # Save the name on the current object
        self.name = name


# Define a child class that inherits from Animal
class Dog(Animal):
    # Add a method specific to Dog
    def speak(self):
        # Return the dog sound
        return f"{self.name} says Woof"


# Create a Dog object
dog = Dog("Bruno")

# Print the inherited and child behavior
print(dog.name)
print(dog.speak())
```

# Concept 1: Inheritance

Inheritance allows one class to reuse and extend another class.

## Why useful
- Reduces code duplication
- Groups related types together
- Makes systems easier to expand

## Common mistake
Creating inheritance when classes do not have a true parent-child relationship.


```python
# Create a parent class
class Vehicle:
    # Initialize the shared brand attribute
    def __init__(self, brand):
        # Save the brand for the current object
        self.brand = brand


# Create a child class
class Bike(Vehicle):
    # Add a child-specific method
    def show_type(self):
        # Print a message using inherited data
        print(f"{self.brand} is a bike")


# Create a child object
bike = Bike("Yamaha")

# Use inherited data through the child object
bike.show_type()
```

# Concept 2: Using `super()`

`super()` lets a child class call methods from its parent class.

## Why important
- Avoids rewriting parent initialization code
- Keeps inheritance chains clean
- Makes maintenance easier

## Edge case
If you forget `super()` in a child constructor, some parent attributes may never be created.


```python
# Define a parent class
class Person:
    # Set name when the object is created
    def __init__(self, name):
        # Store the name
        self.name = name


# Define a child class with extra data
class Teacher(Person):
    # Initialize both inherited and new attributes
    def __init__(self, name, subject):
        # Call the parent constructor to set name
        super().__init__(name)

        # Store the new subject attribute
        self.subject = subject


# Create a teacher object
teacher = Teacher("Anita", "Math")

# Print both attributes
print(teacher.name)
print(teacher.subject)
```

# Concept 3: Method Overriding

Method overriding happens when a child class defines a method with the same name as the parent class method.

## Why useful
Different child classes can behave differently while keeping the same method name.

## Real-world example
Different user roles may implement their own dashboard display.


```python
# Define a parent class
class Notification:
    # Define a general send method
    def send(self):
        # Return the generic message
        return "Sending a notification"


# Define a child class that overrides the method
class EmailNotification(Notification):
    # Replace the parent behavior
    def send(self):
        # Return the email-specific message
        return "Sending an email notification"


# Create objects from both classes
general_message = Notification()
email_message = EmailNotification()

# Print both method results
print(general_message.send())
print(email_message.send())
```

# Concept 4: Polymorphism

Polymorphism means different objects can respond to the same method name in their own way.

## Why it matters
You can write code that works with many related object types through one common interface.

## Common mistake
Confusing polymorphism with inheritance only. Polymorphism is about common behavior, not just class hierarchy.


```python
# Define a parent class
class Shape:
    # Create a common interface method
    def area(self):
        # Raise an error if not implemented in child classes
        raise NotImplementedError("Child class must implement area")


# Define one child class
class Rectangle(Shape):
    # Initialize rectangle dimensions
    def __init__(self, width, height):
        # Store width
        self.width = width

        # Store height
        self.height = height

    # Implement area for rectangle
    def area(self):
        # Return width multiplied by height
        return self.width * self.height


# Define another child class
class Circle(Shape):
    # Initialize circle radius
    def __init__(self, radius):
        # Store the radius
        self.radius = radius

    # Implement area for circle
    def area(self):
        # Return the area formula
        return 3.14159 * self.radius * self.radius


# Store different objects in one list
shapes = [Rectangle(4, 5), Circle(3)]

# Use the same method name on both objects
for shape in shapes:
    print(shape.area())
```

# Concept 5: Dunder Methods

Dunder methods are special methods surrounded by double underscores.

## Common examples
- `__str__()` for readable printing
- `__len__()` for length behavior
- `__add__()` for `+`
- `__eq__()` for equality comparison

## Why useful
They let custom objects behave like built-in Python objects.


```python
# Define a custom class with a dunder method
class Playlist:
    # Store the songs list
    def __init__(self, songs):
        # Save the songs in the object
        self.songs = songs

    # Customize len() behavior for this class
    def __len__(self):
        # Return how many songs are in the playlist
        return len(self.songs)


# Create one playlist object
playlist = Playlist(["Song A", "Song B", "Song C"])

# Call len on the custom object
print(len(playlist))
```

# Concept 6: `__str__()` and Friendly Printing

`__str__()` decides what should be shown when `print(object)` is used.

## Why useful
- Cleaner notebook outputs
- Easier debugging
- Better user-facing messages


```python
# Define a class with a custom string representation
class StudentProfile:
    # Initialize object data
    def __init__(self, name, marks):
        # Save the name
        self.name = name

        # Save the marks
        self.marks = marks

    # Define friendly text for print()
    def __str__(self):
        # Return a readable object summary
        return f"Student(name={self.name}, marks={self.marks})"


# Create one student profile object
profile = StudentProfile("Asha", 91)

# Print the custom string output
print(profile)
```

# Concept 7: Operator Overloading with `__add__()`

Operator overloading allows objects to define what operators like `+` should do.

## Why useful
It makes custom objects expressive and natural to use.

## Common mistake
Returning the wrong type from a dunder method can confuse later code.


```python
# Define a vector class
class Vector:
    # Initialize x and y values
    def __init__(self, x, y):
        # Save x
        self.x = x

        # Save y
        self.y = y

    # Define how + should work for vectors
    def __add__(self, other):
        # Return a new Vector object with added coordinates
        return Vector(self.x + other.x, self.y + other.y)

    # Define readable printing for vectors
    def __str__(self):
        # Return a friendly text representation
        return f"Vector({self.x}, {self.y})"


# Create two vectors
vector_one = Vector(2, 3)
vector_two = Vector(5, 7)

# Add the vectors and print the result
print(vector_one + vector_two)
```

# Concept 8: Equality Comparison with `__eq__()`

`__eq__()` defines how `==` should compare two objects.

## Why important
By default, two objects are compared by identity, not by their data.

## Real-world example
Two user objects may represent the same user ID and should be treated as equal.


```python
# Define a class for books
class Book:
    # Initialize title and author
    def __init__(self, title, author):
        # Save the title
        self.title = title

        # Save the author
        self.author = author

    # Define custom equality logic
    def __eq__(self, other):
        # Compare the data fields instead of object identity
        return self.title == other.title and self.author == other.author


# Create two different objects with the same data
book_one = Book("Python Basics", "Riya")
book_two = Book("Python Basics", "Riya")

# Compare the objects
print(book_one == book_two)
```

# Concept 9: Encapsulation

Encapsulation means keeping object details controlled inside the class.

## Python style
- `_name` suggests protected internal use
- `__name` triggers name mangling for stronger hiding

## Why useful
- Prevents accidental misuse
- Protects sensitive state
- Keeps class APIs cleaner


```python
# Define a class with an internal attribute
class Wallet:
    # Initialize the wallet amount
    def __init__(self, amount):
        # Save amount as a protected-style attribute
        self._amount = amount

    # Provide a safe method to read the amount
    def get_amount(self):
        # Return the internal amount
        return self._amount


# Create a wallet object
wallet = Wallet(2500)

# Access the internal value using the public method
print(wallet.get_amount())

print(wallet._amount)  # This is possible but not recommended
```

# Concept 10: Name Mangling with Double Underscore

Double underscore attributes are renamed internally by Python to reduce accidental direct access.

## Important note
This is not full security. It is mainly a protection against accidental misuse.


```python
# Define a class with a private-style attribute
class SecureAccount:
    # Initialize the object with a PIN
    def __init__(self, pin):
        # Save the PIN using double underscore name mangling
        self.__pin = pin

    # Provide a method to verify a PIN
    def check_pin(self, entered_pin):
        # Compare the entered value with the stored PIN
        return self.__pin == entered_pin


# Create an account object
secure_account = SecureAccount(1234)

# Verify the PIN through the method
print(secure_account.check_pin(1234))
print(secure_account.check_pin(1111))
#print(secure_account.__pin)  # This will raise an AttributeError
```

# Concept 11: `@classmethod`

A class method works with the class itself instead of a specific object.

## Uses
- Alternative constructors
- Shared setup logic
- Class-level utilities

## Operator or syntax involved
Use the `@classmethod` decorator and `cls` as the first parameter.


```python
# Define a class with an alternative constructor
class Employee:
    # Initialize normal object data
    def __init__(self, name, salary):
        # Save the employee name
        self.name = name

        # Save the salary
        self.salary = salary

    # Create an alternative constructor from text input
    @classmethod
    def from_string(cls, text):#cls is keyword for class method, it is used to refer to the class itself, and it is a convention to name it cls or can be any name like c, one, etc.
        # Split the string into parts
        name, salary = text.split("-")

        # Create and return a new object using the class itself
        return cls(name, int(salary))


# Create an object using the class method
employee = Employee.from_string("Aman-50000")

# Print the generated object data
print(employee.name)
print(employee.salary)
```

# Concept 12: `@staticmethod`

A static method belongs logically to the class but does not need `self` or `cls`.

## When to use
Use it for helper logic that is related to the class concept but does not need object state.


```python
# Define a utility-style class
class Temperature:
    # Create a static method for conversion
    @staticmethod
    def celsius_to_fahrenheit(celsius):
        # Apply the conversion formula
        return (celsius * 9 / 5) + 32


# Call the static method directly from the class
print(Temperature.celsius_to_fahrenheit(25))
```

# Concept 13: Abstract Behavior with Base Classes

Sometimes a parent class should define the expected interface, while child classes provide the actual implementation.

## Why useful
- Enforces consistent method names
- Helps design clean systems
- Supports polymorphism well


```python
# Define a base payment class
class Payment:
    # Define the interface method
    def pay(self, amount):
        # Raise an error if child classes do not implement it
        raise NotImplementedError("Child classes must implement pay")


# Define one payment method
class CardPayment(Payment):
    # Implement the required behavior
    def pay(self, amount):
        # Return a card payment message
        return f"Paid {amount} by card"


# Define another payment method
class CashPayment(Payment):
    # Implement the required behavior
    def pay(self, amount):
        # Return a cash payment message
        return f"Paid {amount} by cash"


# Create a list of different payment objects
payments = [CardPayment(), CashPayment()]

# Use the same interface on both objects
for payment in payments:
    print(payment.pay(500))
```

# Concept 14: Multiple Inheritance Awareness

Python supports multiple inheritance, but beginners should use it carefully.

## Why careful?
- It can make code harder to follow
- Method resolution can become confusing

## Best practice
Use multiple inheritance only when the design is genuinely clear and necessary.


```python
# Define the first parent class
class Camera:
    # Add camera behavior
    def click_photo(self):
        # Return a camera action message
        return "Photo captured"


# Define the second parent class
class Phone:
    # Add phone behavior
    def call(self):
        # Return a phone action message
        return "Calling..."


# Define a child with multiple inheritance
class SmartPhone(Camera, Phone):
    # No extra code is needed for this demonstration
    pass


# Create a smartphone object
smartphone = SmartPhone()

# Use behavior from both parent classes
print(smartphone.click_photo())
print(smartphone.call())
```

# Concept 15: Common Mistakes and Best Practices

## Common mistakes
- Using inheritance where composition would be better
- Forgetting `super()`
- Overusing dunder methods without a clear reason
- Mixing too many responsibilities into one class
- Confusing class methods and static methods

## Best practices
- Keep parent classes general and child classes specific
- Override methods only when behavior truly changes
- Use dunder methods to make objects natural and readable
- Prefer simple clear class design


```python
# Define a clean base class
class User:
    # Initialize shared user data
    def __init__(self, username):
        # Save the username
        self.username = username

    # Return the role label
    def role(self):
        # Return the default role
        return "User"


# Define a more specific child class
class Admin(User):
    # Override the role behavior
    def role(self):
        # Return the admin role
        return "Admin"


# Create one admin object
admin_user = Admin("root_user")

# Print the overridden behavior
print(admin_user.username)
print(admin_user.role())
```
