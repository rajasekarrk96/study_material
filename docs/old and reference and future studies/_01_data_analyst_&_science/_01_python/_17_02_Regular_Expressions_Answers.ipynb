## Q1. Find First Digits

**Problem:** Use `re.search` to find the first group of digits in a string and return it.

**Step-by-step logic:**
- Read the required input shape: `text: str`.
- Choose the most suitable regex function or pattern.
- Use raw strings for clarity and safer backslashes.
- Handle the stated constraint carefully: Return None when no digits are present.
- Return the final result in the required format `str | None`.


````python
def find_first_digits(text: str) -> str | None:
    # Import the regex module.
    import re

    # Search for the first digit group anywhere in the text.
    match = re.search(r"\d+", text)

    # Return the matched text when found, otherwise None.
    return match.group() if match else None

# Time Complexity: depends on the text length and how many regex passes or matches are performed.
# Space Complexity: depends on match lists, replacement output, or captured groups created above.
````

## Q11. Extract Email User Domain

**Problem:** Use capturing groups to extract the username and domain from a simple `.com` email and return them as a tuple.

**Step-by-step logic:**
- Read the required input shape: `email: str`.
- Choose the most suitable regex function or pattern.
- Use raw strings for clarity and safer backslashes.
- Handle the stated constraint carefully: Return None if the pattern does not match.
- Return the final result in the required format `tuple[str, str] | None`.


````python
def extract_email_user_domain(email: str) -> tuple[str, str] | None:
    # Import the regex module.
    import re

    # Match a simple .com email with two capture groups.
    match = re.fullmatch(r"(\w+)@(\w+)\.com", email)

    # Return both groups as a tuple when matched.
    return (match.group(1), match.group(2)) if match else None

# Time Complexity: depends on the text length and how many regex passes or matches are performed.
# Space Complexity: depends on match lists, replacement output, or captured groups created above.
````

## Q12. Named Group Date

**Problem:** Use named groups to extract year, month, and day from a date string `YYYY-MM-DD` and return them in a dictionary.

**Step-by-step logic:**
- Read the required input shape: `text: str`.
- Choose the most suitable regex function or pattern.
- Use raw strings for clarity and safer backslashes.
- Handle the stated constraint carefully: Return None if the date format does not match.
- Return the final result in the required format `dict[str, str] | None`.


````python
def named_group_date(text: str) -> dict[str, str] | None:
    # Import the regex module.
    import re

    # Match a full date and capture each part by name.
    match = re.fullmatch(r"(?P<year>\d{4})-(?P<month>\d{2})-(?P<day>\d{2})", text)

    # Return the named groups dictionary when matched.
    return match.groupdict() if match else None

# Time Complexity: depends on the text length and how many regex passes or matches are performed.
# Space Complexity: depends on match lists, replacement output, or captured groups created above.
````
