# Introduction

- **What is it?** Concurrency is about handling multiple tasks during overlapping periods of time. Asyncio is Python's framework for cooperative asynchronous programming.
- **Why it is used?** Some tasks spend time waiting, such as network requests, file I/O, timers, and message queues. Concurrency helps keep programs responsive and efficient during those waits.
- **Real-world relevance:** Chat apps, dashboards, crawlers, APIs, task runners, background workers, and data pipelines often depend on concurrency or async patterns.


# Basic Syntax Explanation

Important modern async syntax:

```python
import asyncio

async def work():
    await asyncio.sleep(1)
    return "done"

result = asyncio.run(work())
```

Core ideas:
- `async def` defines a coroutine function
- `await` pauses until another awaitable is ready
- `asyncio.run(...)` runs the event loop for a top-level coroutine
- threads and queues are useful for some non-async coordination patterns


````python
# Basic Example with detailed comments

# Import asyncio for asynchronous programming.
import asyncio

# Define a coroutine function with async def.
async def say_hello():
    # Pause without blocking the whole program flow.
    await asyncio.sleep(0)

    # Return a short result string.
    return "hello"


# Run the coroutine from normal top-level code.
print(asyncio.run(say_hello()))
````

# 1. Concurrency vs Parallelism

Concurrency means managing multiple tasks with overlapping progress. Parallelism means tasks literally run at the same time on different CPU resources.

Why useful:
- helps choose the right tool
- avoids unrealistic performance assumptions


````python
# Store short comparison labels.
concurrency = "overlapping progress"
parallelism = "same-time execution"

# Print both ideas.
print(concurrency)
print(parallelism)
````

# 2. Blocking vs Non-Blocking Work

Blocking work waits and prevents progress in the current flow. Non-blocking designs let other tasks make progress while one task waits.

Examples of blocking:
- sleeping
- waiting for network responses
- waiting for disk or database I/O


````python
# Import time for a blocking sleep example.
import time

# Block the current thread briefly.
time.sleep(0.1)

# Print after the blocking wait ends.
print("Done")
````

# 3. When Threads Help

Threads are often useful for I/O-bound work, not for speeding up heavy pure Python CPU loops.

Good thread examples:
- waiting on files
- waiting on sockets
- waiting on APIs
- producer-consumer patterns


````python
# Import the threading module.
import threading

# Define a short worker function.
def worker():
    # Print a worker message.
    print("Thread worker ran")


# Create and start a thread.
thread = threading.Thread(target=worker)
thread.start()
thread.join()
````

# 4. Thread Basics

Thread flow usually includes:

1. define a worker function
2. create a Thread
3. `start()`
4. `join()` when you need to wait for completion


````python
# Import threading for a simple demo.
import threading

# Define a worker that prints a number.
def print_number(number):
    # Print the received value.
    print(number)


# Start one thread with arguments.
thread = threading.Thread(target=print_number, args=(5,))
thread.start()
thread.join()
````

# 5. Shared State and Race Conditions

When multiple threads touch the same mutable state, race conditions can happen.

Why important:
- results may become inconsistent
- bugs may appear only sometimes

Safe tools:
- `Lock`
- queues
- avoiding shared mutable state when possible


````python
# Import threading for a lock demonstration.
import threading

# Create a shared counter and a lock.
counter = 0
lock = threading.Lock()

# Define a safe increment function.
def safe_increment():
    global counter
    with lock:
        counter += 1


# Call the safe increment once.
safe_increment()
print(counter)
````

# 6. Queues for Thread Communication

`queue.Queue` is a very useful thread-safe coordination tool.

Why useful:
- producer-consumer pattern
- safer than manual shared lists
- built-in blocking behavior


````python
# Import Queue from queue.
from queue import Queue

# Create a queue and add items.
tasks = Queue()
tasks.put("task1")
tasks.put("task2")

# Read one item safely.
print(tasks.get())
````

# 7. ThreadPoolExecutor

`concurrent.futures.ThreadPoolExecutor` gives a higher-level thread API.

Why useful:
- less boilerplate
- easier mapping of work
- future-based results


````python
# Import ThreadPoolExecutor.
from concurrent.futures import ThreadPoolExecutor

# Define a small worker function.
def square(number):
    # Return the squared value.
    return number * number


# Run work in a thread pool.
with ThreadPoolExecutor(max_workers=2) as executor:
    print(list(executor.map(square, [1, 2, 3])))
````

# 8. When Processes Help

Common tool:
- `ProcessPoolExecutor`

Why useful:
- CPU-heavy calculations

````python
# Import ProcessPoolExecutor for the concept example.
from concurrent.futures import ProcessPoolExecutor

# Define a CPU-style function.
def cube(number):
    # Return the cube value.
    return number ** 3


# Show the executor class name without launching heavy work.
print(ProcessPoolExecutor.__name__)
````

# 9. What Is Asyncio?

Asyncio is Python's framework for cooperative asynchronous I/O.

Why useful:
- many concurrent waiting tasks
- network-heavy apps
- large numbers of sockets or timers

Important idea:
- tasks cooperate by `await`ing, rather than preemptive thread switching


````python
# Import asyncio for coroutine work.
import asyncio

# Define a tiny coroutine.
async def ping():
    # Yield control briefly.
    await asyncio.sleep(0)
    return "pong"


# Run and print the coroutine result.
print(asyncio.run(ping()))
````

# 10. Coroutines

A coroutine is a function defined with `async def`.

Why useful:
- represents async work clearly
- can pause with `await`
- integrates with the event loop


````python
# Import asyncio.
import asyncio

# Define a coroutine function.
async def greet(name):
    # Pause cooperatively.
    await asyncio.sleep(0)
    return f"Hello, {name}"


# Run the coroutine and print the result.
print(asyncio.run(greet("Asha")))
````

# 11. The `await` Keyword

`await` pauses the current coroutine until another awaitable completes.

Why useful:
- lets other tasks run
- avoids blocking the whole loop


````python
# Import asyncio.
import asyncio

# Define one coroutine that waits for another.
async def outer():
    # Await an async sleep call.
    await asyncio.sleep(0)
    return "done"


# Run the outer coroutine.
print(asyncio.run(outer()))
````

# 12. `asyncio.run()`

`asyncio.run()` is the common top-level entry point for running a coroutine from regular synchronous code.

Why useful:
- starts and closes the event loop cleanly
- beginner-friendly async entry point


````python
# Import asyncio.
import asyncio

# Define a simple coroutine.
async def work():
    # Return a small result.
    return 123


# Run it from normal code.
print(asyncio.run(work()))
````

# 13. `asyncio.sleep()`

`asyncio.sleep()` is different from `time.sleep()`.

Important difference:
- `time.sleep()` blocks the current thread
- `asyncio.sleep()` yields control back to the event loop


````python
# Import asyncio.
import asyncio

# Define a coroutine that pauses cooperatively.
async def pause():
    await asyncio.sleep(0.01)
    return "awake"


# Run and print the result.
print(asyncio.run(pause()))
````

# 14. Running Multiple Tasks with `gather`

`asyncio.gather()` runs multiple awaitables and collects their results.

Why useful:
- concurrent waits
- cleaner task grouping
- one result list


````python
# Import asyncio.
import asyncio

# Define two coroutines.
async def one():
    await asyncio.sleep(0)
    return 1

async def two():
    await asyncio.sleep(0)
    return 2


# Define a parent coroutine using gather.
async def main():
    return await asyncio.gather(one(), two())


# Run and print all gathered results.
print(asyncio.run(main()))
````

# 15. Creating Tasks

Tasks schedule coroutines to run concurrently in the event loop.

Common function:
- `asyncio.create_task()`

Why useful:
- start work before waiting for final results


````python
# Import asyncio.
import asyncio

# Define a small coroutine.
async def job(number):
    await asyncio.sleep(0)
    return number * 10


# Create tasks inside a main coroutine.
async def main():
    task1 = asyncio.create_task(job(1))
    task2 = asyncio.create_task(job(2))
    return await asyncio.gather(task1, task2)


# Run and print task results.
print(asyncio.run(main()))
````

# 16. Async Queues

`asyncio.Queue` is useful for async producer-consumer patterns.

Why useful:
- safe coordination between coroutines
- backpressure-aware workflows


````python
# Import asyncio.
import asyncio

# Define an async queue example.
async def queue_demo():
    queue = asyncio.Queue()
    await queue.put("task")
    return await queue.get()


# Run and print the queue result.
print(asyncio.run(queue_demo()))
````

# 17. Timeout Handling

Concurrency code often needs timeouts.

Useful async helper:
- `asyncio.wait_for(...)`

Why useful:
- prevents hanging tasks
- improves resilience


````python
# Import asyncio.
import asyncio

# Define a coroutine with a short pause.
async def slow():
    await asyncio.sleep(0.01)
    return "done"


# Run it with a timeout.
async def main():
    return await asyncio.wait_for(slow(), timeout=1)


# Print the result.
print(asyncio.run(main()))
````

# 18. Cancellation Basics

Async tasks can be cancelled.

Why useful:
- shutting down cleanly
- stopping outdated work
- freeing resources


````python
# Import asyncio.
import asyncio

# Define a long-running coroutine.
async def worker():
    try:
        await asyncio.sleep(1)
    except asyncio.CancelledError:
        return "cancelled"


# Create and cancel the task.
async def main():
    task = asyncio.create_task(worker())
    task.cancel()
    try:
        return await task
    except asyncio.CancelledError:
        return "cancelled"


# Print the cancellation result.
print(asyncio.run(main()))
````

# 22. Short-Hand and Optimized Patterns

Helpful patterns:

- keep shared state minimal
- prefer queues over manual list sharing
- use `gather` for grouped awaits
- use executors for blocking work
- separate CPU-bound and I/O-bound designs


````python
# Import asyncio.
import asyncio

# Define a tiny coroutine.
async def double(number):
    await asyncio.sleep(0)
    return number * 2


# Map a few numbers with gather.
async def main():
    return await asyncio.gather(*(double(n) for n in [1, 2, 3]))


# Print the gathered results.
print(asyncio.run(main()))
````

# 23. Real-World Example: Worker Queue

A common concurrency pattern is producer-consumer processing.

Why useful:
- task pipelines
- background jobs
- batched work dispatch


````python
# Import Queue from queue.
from queue import Queue

# Create a work queue.
queue = Queue()
queue.put("extract")
queue.put("clean")
queue.put("load")

# Process all queued tasks in order.
processed = []
while not queue.empty():
    processed.append(queue.get())

# Print the processed order.
print(processed)
````

# 24. Real-World Example: Async Batch Runner

Asyncio shines when many tasks mostly wait.

Goal:
- create coroutines
- run them concurrently
- gather results cleanly


````python
# Import asyncio.
import asyncio

# Define a coroutine representing one async API-like wait.
async def fetch_value(number):
    await asyncio.sleep(0)
    return number * 100


# Run several async tasks together.
async def main():
    tasks = [fetch_value(1), fetch_value(2), fetch_value(3)]
    return await asyncio.gather(*tasks)


# Print the final gathered values.
print(asyncio.run(main()))
````
