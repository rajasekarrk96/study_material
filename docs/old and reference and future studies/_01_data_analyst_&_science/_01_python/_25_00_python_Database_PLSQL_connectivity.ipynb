# Introduction

- **What is it?** PL/SQL is Oracle's procedural extension to SQL. It allows SQL statements to be combined with variables, conditions, loops, procedures, functions, and exception handling.
- **Why it is used?** It helps move business logic close to the database, automate repetitive tasks, validate data, and manage complex transactional workflows.
- **Real-world relevance:** PL/SQL is widely used in Oracle-based enterprise systems such as banking applications, ERP systems, payroll systems, inventory tools, reporting pipelines, and internal workflow automation.


# Basic Syntax Explanation

```sql
DECLARE
    v_name VARCHAR2(100);
BEGIN
    v_name := 'Asha';
    DBMS_OUTPUT.PUT_LINE(v_name);
END;
/
```

Main idea:
- `DECLARE` is used for variables and constants.
- `BEGIN ... END` contains executable statements.
- PL/SQL combines SQL with procedural logic such as assignment, conditions, loops, and exception handling.


````python
# Basic Example with detailed comments

# Simulate the idea of a PL/SQL block using ordinary Python variables and output.

# Declare a variable-like value.
employee_name = "Asha"

# Declare another value that could be used in logic.
salary = 50000

# Simulate executable block behavior with simple processing.
if salary >= 50000:
    message = f"{employee_name} is in the high salary group."
else:
    message = f"{employee_name} is not in the high salary group."

# Print the final message, similar to DBMS_OUTPUT.PUT_LINE.
print(message)
````

# 2. Anonymous Blocks

An anonymous block is a one-time PL/SQL block that is not stored permanently in the database.


````python
plsql_block = '''
DECLARE
    v_message VARCHAR2(100);
BEGIN
    v_message := 'Hello from PL/SQL';
END;
/
'''

print(plsql_block)
````

# 3. DECLARE Section

The `DECLARE` section is used for variables, constants, cursors, and local types.


````python
declared_values = {
    "v_name": "Asha",
    "v_salary": 50000,
    "c_bonus_rate": 0.10,
}

print(declared_values)
````

# 4. Variables and Assignment

PL/SQL variables hold values, and assignment uses `:=` rather than `=`.


````python
employee_name = "Asha"
salary = 50000

# Update a variable value similarly to procedural assignment logic.
salary = salary + 2000

print(employee_name, salary)
````

# 5. Common Data Types

PL/SQL commonly uses `NUMBER`, `VARCHAR2`, `DATE`, `BOOLEAN`, and `%TYPE` references.


````python
type_examples = {
    "employee_id": "NUMBER",
    "employee_name": "VARCHAR2(100)",
    "joining_date": "DATE",
    "is_active": "BOOLEAN",
}

for name, data_type in type_examples.items():
    print(name, "->", data_type)
````

# 6. Constants

A constant stores a value that should not change during the block.


````python
BONUS_RATE = 0.10
salary = 50000
bonus = salary * BONUS_RATE

print(bonus)
````

# 7. BEGIN END Execution Block

The `BEGIN ... END` section contains the executable statements.


````python
steps = [
    "read data",
    "apply logic",
    "produce result",
]

for step in steps:
    print(step)
````

# 8. DBMS_OUTPUT.PUT_LINE Idea

This built-in package is commonly used for displaying debug or informational output from PL/SQL blocks.


````python
messages = ["Block started", "Variable updated", "Block completed"]

for message in messages:
    print(message)
````

# 9. IF THEN ELSE

Conditional logic allows the block to branch based on values or query results.


````python
score = 82

if score >= 75:
    result = "Pass"
else:
    result = "Fail"

print(result)
````

# 10. ELSIF Chains

Use `ELSIF` when more than two conditions are needed.


````python
score = 88

if score >= 90:
    grade = "A"
elif score >= 75:
    grade = "B"
elif score >= 60:
    grade = "C"
else:
    grade = "D"

print(grade)
````

# 11. Loop Basics

PL/SQL supports simple loops for repeating work.


````python
values = [10, 20, 30]

for value in values:
    print(value)
````

# 12. FOR Loops

A `FOR` loop is common when the number of iterations is known.


````python
total = 0

for number in range(1, 6):
    total += number

print(total)
````

# 13. WHILE Loops

A `WHILE` loop repeats while a condition remains true.


````python
counter = 1

while counter <= 3:
    print(counter)
    counter += 1
````

# 14. Procedures

A procedure is a stored PL/SQL program unit used to perform actions.


````python
procedure_text = '''
CREATE OR REPLACE PROCEDURE give_bonus(p_employee_id NUMBER) AS
BEGIN
    NULL;
END;
/
'''

print(procedure_text)
````

# 15. Functions

A function returns a value and can often be used inside SQL or other PL/SQL code.


````python
def calculate_bonus(salary, rate):
    # Return a computed bonus amount.
    return salary * rate


print(calculate_bonus(50000, 0.10))
````

# 16. Parameters IN OUT and IN OUT

PL/SQL procedures can accept input parameters, output parameters, and parameters that do both.


````python
parameter_roles = {
    "IN": "input only",
    "OUT": "output only",
    "IN OUT": "input and output",
}

for role, meaning in parameter_roles.items():
    print(role, "->", meaning)
````

# 17. SELECT INTO Idea

`SELECT ... INTO` reads query results directly into PL/SQL variables.


````python
employee = {"name": "Asha", "salary": 50000}

# Simulate selecting column values into variables.
employee_name = employee["name"]
employee_salary = employee["salary"]

print(employee_name, employee_salary)
````

# 18. Records

A record groups related fields together, similar to one row of structured data.


````python
employee_record = {
    "employee_id": 1,
    "employee_name": "Asha",
    "salary": 50000,
}

print(employee_record)
````

# 19. Explicit Cursors

Cursors are used to process query result rows one at a time in procedural logic.


````python
rows = [
    {"employee_name": "Asha", "salary": 50000},
    {"employee_name": "Ravi", "salary": 42000},
]

# Simulate reading rows one by one from a cursor.
for row in rows:
    print(row["employee_name"], row["salary"])
````

# 20. Cursor FOR Loops

Cursor FOR loops reduce boilerplate because open, fetch, and close steps are handled automatically.


````python
rows = [
    {"name": "Asha"},
    {"name": "Ravi"},
    {"name": "Meera"},
]

for row in rows:
    print(row["name"])
````

# 21. Exceptions

PL/SQL has an `EXCEPTION` section for handling runtime problems gracefully.


````python
try:
    value = 10 / 0
except ZeroDivisionError:
    print("Handled division error safely.")
````

# 22. Common Built-in Exceptions

Typical exceptions include `NO_DATA_FOUND`, `TOO_MANY_ROWS`, and `ZERO_DIVIDE`.


````python
exception_names = [
    "NO_DATA_FOUND",
    "TOO_MANY_ROWS",
    "ZERO_DIVIDE",
]

for exception_name in exception_names:
    print(exception_name)
````

# 23. User-defined Exceptions

Custom exceptions allow business-specific error handling.


````python
class LowSalaryError(Exception):
    pass


salary = 1000

try:
    if salary < 5000:
        raise LowSalaryError("Salary is below the minimum business rule.")
except LowSalaryError as error:
    print(error)
````

# 24. Transactions COMMIT and ROLLBACK

PL/SQL blocks often control transactions so related changes stay consistent.


````python
balance_a = 1000
balance_b = 500
amount = 200

# Simulate a successful transfer.
balance_a -= amount
balance_b += amount

print(balance_a, balance_b)
````

# 25. savepoints and Partial Rollback Idea

A savepoint allows some work to be undone without discarding the whole transaction.


````python
original_total = 100
after_first_step = original_total + 20
after_second_step = after_first_step - 5

# Simulate rolling back only the second step.
rolled_back_total = after_first_step

print(rolled_back_total)
````

# 26. Triggers

Triggers automatically run in response to table events such as insert, update, or delete.


````python
trigger_text = '''
CREATE OR REPLACE TRIGGER trg_before_insert_employee
BEFORE INSERT ON employees
FOR EACH ROW
BEGIN
    NULL;
END;
/
'''

print(trigger_text)
````

# 27. Packages

# 29. Bind Variables Idea

Bind variables pass values separately from SQL text and improve efficiency and safety.


````python
employee_id = 10
query_template = "SELECT salary FROM employees WHERE employee_id = :employee_id"

print(query_template)
print({"employee_id": employee_id})
````

# 30. Edge Cases and Common Mistakes

Common mistakes include assuming one row when many rows may exist, forgetting exception handling, or writing procedural logic when simple SQL would be enough.


````python
salaries = [50000, 42000, 61000]

# Safe pattern: check whether the collection is empty before averaging.
if salaries:
    average_salary = sum(salaries) / len(salaries)
else:
    average_salary = 0

print(average_salary)
````

# 31. Real-world Example Payroll Logic

PL/SQL is often used for payroll calculations and rule-based updates.


````python
employees = [
    {"name": "Asha", "salary": 50000},
    {"name": "Ravi", "salary": 42000},
]

bonuses = {}
for employee in employees:
    bonuses[employee["name"]] = employee["salary"] * 0.10

print(bonuses)
````

# 32. Real-world Example Data Validation

Before committing, PL/SQL can validate values and reject invalid transactions.


````python
orders = [
    {"order_id": 1, "amount": 200},
    {"order_id": 2, "amount": -50},
]

invalid_orders = []
for order in orders:
    if order["amount"] < 0:
        invalid_orders.append(order["order_id"])

print(invalid_orders)
````

# 34. Mini Practical Example

A typical PL/SQL workflow is read -> validate -> process -> handle exception -> commit or rollback.


````python
accounts = {"A": 1000, "B": 500}
amount = 300

try:
    if accounts["A"] < amount:
        raise ValueError("Insufficient balance")

    accounts["A"] -= amount
    accounts["B"] += amount
except ValueError as error:
    print("Transfer failed:", error)
else:
    print("Transfer succeeded:", accounts)
````
