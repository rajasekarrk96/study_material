```python
# Define a SafeBankAccount class
class SafeBankAccount:
    # Initialize the balance
    def __init__(self, balance):
        # Store the starting balance
        self.balance = balance

    # Withdraw only when enough balance is available
    def withdraw(self, amount):
        # Check whether the account has enough money
        if amount <= self.balance:
            self.balance -= amount
            return 'Withdrawal successful'
        return 'Insufficient balance'

# Create one safe account object
account = SafeBankAccount(500)

# Test case 1
expected_output = 'Insufficient balance'
print(account.withdraw(700))  # should match expected
print(expected_output)
```

### Q20. Alternate Constructor

Dry run: split `'Aman-50000'` into `Aman` and `50000`, convert salary to integer, and return the new object. Time complexity is `O(n)` in the string length.


```python
# Define an Employee class with a class method constructor
class Employee:
    # Initialize name and salary
    def __init__(self, name, salary):
        # Store the employee name
        self.name = name
        # Store the employee salary
        self.salary = salary

    # Define an alternate constructor
    @classmethod
    def from_string(cls, text):
        # Split the incoming text into name and salary
        name, salary = text.split('-')
        # Return a new Employee object
        return cls(name, int(salary))

# Create one employee object from string
employee = Employee.from_string('Aman-50000')

# Test case 1
expected_output = ('Aman', 50000)
print((employee.name, employee.salary))  # should match expected
print(expected_output)
```

### Q26. Descriptor-Like Validation via Property

Dry run: setting `85` passes validation and is stored, while values below `0` or above `100` would raise an error. Time complexity is `O(1)`.


```python
# Define a Percentage class with a validating property
class Percentage:
    # Initialize through the property setter
    def __init__(self, value):
        # Set the percentage value through validation
        self.value = value

    # Read the stored percentage value
    @property
    def value(self):
        # Return the internal stored value
        return self._value

    # Validate and set the percentage value
    @value.setter
    def value(self, new_value):
        # Reject invalid percentages
        if not 0 <= new_value <= 100:
            raise ValueError('Percentage must be between 0 and 100')
        # Store the validated value
        self._value = new_value

# Create one valid percentage object
percentage = Percentage(85)

# Test case 1
expected_output = 85
print(percentage.value)  # should match expected
print(expected_output)
```

### Q27. Bank Account Statement

Dry run: start with `1000`, deposit `100` to reach `1100`, withdraw `50` to reach `1050`, and count two transactions. Time complexity is `O(1)` per operation.


```python
# Define a StatementAccount class
class StatementAccount:
    # Initialize the account balance and transaction count
    def __init__(self, balance):
        # Store the starting balance
        self.balance = balance
        # Start with zero transactions
        self.transaction_count = 0

    # Deposit money and count the transaction
    def deposit(self, amount):
        self.balance += amount
        self.transaction_count += 1

    # Withdraw money and count the transaction
    def withdraw(self, amount):
        self.balance -= amount
        self.transaction_count += 1

    # Return a final statement dictionary
    def statement(self):
        return {'balance': self.balance, 'count': self.transaction_count}

# Create one account and apply two operations
account = StatementAccount(1000)
account.deposit(100)
account.withdraw(50)

# Test case 1
expected_output = {'balance': 1050, 'count': 2}
print(account.statement())  # should match expected
print(expected_output)
```

### Q31. Student Report Card

Dry run: average of `[88, 76, 91, 84]` is `84.75`, which falls in the `B` grade range. Time complexity is `O(n)`.


```python
# Define a StudentReport class
class StudentReport:
    # Initialize the report with name and marks list
    def __init__(self, name, marks):
        # Store the student name
        self.name = name
        # Store the marks list
        self.marks = marks

    # Calculate the average mark
    def average(self):
        # Return the arithmetic mean of the marks
        return sum(self.marks) / len(self.marks)

    # Calculate the letter grade
    def grade(self):
        # Use the average to decide the grade
        avg = self.average()
        if avg >= 90:
            return 'A'
        if avg >= 75:
            return 'B'
        if avg >= 60:
            return 'C'
        return 'D'

    # Return a report dictionary
    def report(self):
        return {'name': self.name, 'average': self.average(), 'grade': self.grade()}

# Create one report object
report = StudentReport('Asha', [88, 76, 91, 84])

# Test case 1
expected_output = {'name': 'Asha', 'average': 84.75, 'grade': 'B'}
print(report.report())  # should match expected
print(expected_output)
```

### Q32. Shopping Cart Object

Dry run: `300 * 2 = 600` for the book and `20 * 5 = 100` for pens, so total bill is `700`. Time complexity is `O(n)` for `n` items.


```python
# Define a ShoppingCart class
class ShoppingCart:
    # Initialize an empty list of items
    def __init__(self):
        # Store all cart items in one list
        self.items = []

    # Add one item to the cart
    def add_item(self, name, price, quantity):
        # Store the item as a dictionary in the cart list
        self.items.append({'name': name, 'price': price, 'quantity': quantity})

    # Calculate the total bill
    def total_bill(self):
        # Sum price multiplied by quantity for every item
        return sum(item['price'] * item['quantity'] for item in self.items)

# Create one cart object
cart = ShoppingCart()
cart.add_item('Book', 300, 2)
cart.add_item('Pen', 20, 5)

# Test case 1
expected_output = 700
print(cart.total_bill())  # should match expected
print(expected_output)
```

### Q34. ATM Simulation

```python
# Define an ATMAccount class
class ATMAccount:
    # Initialize balance and transaction count
    def __init__(self, balance):
        # Store the starting balance
        self.balance = balance
        # Start with zero transactions
        self.transactions = 0

    # Deposit money into the account
    def deposit(self, amount):
        # Update the balance and transaction count
        self.balance += amount
        self.transactions += 1

    # Withdraw money from the account
    def withdraw(self, amount):
        # Update the balance and transaction count
        self.balance -= amount
        self.transactions += 1

    # Return a mini statement dictionary
    def mini_statement(self):
        return {'balance': self.balance, 'transactions': self.transactions}

# Create one ATM account object
account = ATMAccount(1000)
account.deposit(500)
account.withdraw(200)

# Test case 1
expected_output = {'balance': 1300, 'transactions': 2}
print(account.mini_statement())  # should match expected
print(expected_output)
```

### Q35. Employee Payroll Object

Dry run: gross salary `50000` with `10%` tax means tax amount `5000`, so net salary becomes `45000.0`. Time complexity is `O(1)`.


```python
# Define an EmployeePayroll class
class EmployeePayroll:
    # Initialize name, gross salary, and tax percentage
    def __init__(self, name, gross_salary, tax_percent):
        # Store the employee name
        self.name = name
        # Store the gross salary
        self.gross_salary = gross_salary
        # Store the tax percentage
        self.tax_percent = tax_percent

    # Calculate net salary after tax
    def net_salary(self):
        # Compute the tax amount from the gross salary
        tax_amount = self.gross_salary * (self.tax_percent / 100)
        # Return the salary left after subtracting tax
        return self.gross_salary - tax_amount

# Create one payroll object
payroll = EmployeePayroll('Ravi', 50000, 10)

# Test case 1
expected_output = 45000.0
print(payroll.net_salary())  # should match expected
print(expected_output)
```
