# Basic Syntax Explanation

Modern Python often looks like this:

```python
name = "Asha"
message = f"Hello, {name}"

from dataclasses import dataclass

@dataclass
class Student:
    name: str
    score: int

if (count := len(message)) > 5:
    print(count)
```

Important modern features:
- `f"..."` for f-strings
- `@dataclass` for data classes
- `:=` for assignment expressions
- `str | None` style union types
- `match ... case` for structural pattern matching


````python
# Basic Example with detailed comments

# Store a person's name in a variable.
name = "Asha"

# Use an f-string to build a readable message.
message = f"Hello, {name}"

# Print the formatted message.
print(message)

# Use the walrus operator to assign and test in one step.
if (length := len(message)) > 5:
    # Print the saved length value.
    print(length)
````

````python
# ==========================================================
# PYTHON WALRUS OPERATOR (:=) — COMPLETE NOTES + EXAMPLES
# ==========================================================
# The walrus operator := is called the "assignment expression".
#
# It allows you to:
#   ✔ Assign a value to a variable
#   ✔ AND use that value in the same expression
#
# Introduced in Python 3.8

print("\n========== 1. BASIC IDEA ==========\n")

# Normal way (2 steps)
message = "Hello, Asha"
length = len(message)
if length > 5:
    print("Normal way length:", length)

# Walrus way (1 step)
if (length := len(message)) > 5:
    print("Walrus length:", length)

# Explanation:
# (length := len(message))
# → first assigns length = len(message)
# → then returns that value for comparison
````

````python
print("\n========== 2. HOW IT WORKS INTERNALLY ==========\n")

# Step-by-step mental model:
# (x := value)
# ↓
# 1. Assign x = value
# 2. Return value

value = (x := 10)
print("x =", x)        # 10
print("value =", value)  # 10 (returned result)
````

````python
print("\n========== 3. AVOID REPEATED COMPUTATION ==========\n")

# Without walrus → repeated calculation
text = "PythonProgramming"

if len(text) > 10:
    print("Length:", len(text))   # len() called twice ❌

# With walrus → efficient
if (l := len(text)) > 10:
    
    print("Length (optimized):", l)   # len() called once ✔
````

````python
print("\n========== 4. USE IN WHILE LOOP ==========\n")

# Very common real-world usage

# Simulating input using list
data = ["raj", "kumar", "exit"]

i = 0
while (user := data[i]) != "exit":
    print("Processing:", user)
    i += 1

# Explanation:
# user := data[i]
# → assign and check in same line
````

````python
print("\n========== 5. USE IN LIST COMPREHENSION ==========\n")

numbers = [1, 2, 3, 4, 5]

# Store square AND filter at same time
squares = [sq for n in numbers if (sq := n*n) > 5]

print("Squares > 5:", squares)

# Explanation:
# sq := n*n → assign square
# condition → check > 5
# result → reuse sq
````

````python
print("\n========== 6. COMMON MISTAKE ==========\n")

# ❌ Wrong (syntax error)
# if length := len("hello") > 3:
#     pass

# ✔ Correct (use parentheses)
if (length := len("hello")) > 3:
    print("Correct usage, length =", length)
````

# 15. Positional-Only Parameters

Modern Python supports positional-only parameters using `/`.

Example:
```python
def divide(a, b, /):
    ...
```

Why useful:
- clearer API design
- prevents unsupported keyword usage


````python
# ==========================================================
# POSITIONAL-ONLY PARAMETERS ( / ) — COMPLETE EXPLANATION
# ==========================================================
# Syntax:
# def function(a, b, /):
#     ...
#
# The "/" means:
# → All parameters BEFORE "/" must be passed ONLY by position
# → You CANNOT use keyword arguments for them

print("\n========== 1. BASIC FUNCTION ==========\n")

# Define function with positional-only parameters
def divide(a, b, /):
    # a and b must be passed positionally
    return a / b


print("Correct call (positional):", divide(10, 2))  # ✔ Works

try:
    divide(a=10, b=2)   # ❌ This will fail
except TypeError as e:
    print("Error:", e)


print("\n========== 2. WHY '/' IS USED ==========\n")

# Normally, Python allows both:
def normal_divide(a, b):
    return a / b

print("Keyword allowed:", normal_divide(a=10, b=2))  # ✔ Works


# But with "/" → keyword usage is NOT allowed
print("\nTrying keyword with positional-only parameters:")




print("\n========== 3. WHAT '/' ACTUALLY DOES ==========\n")

# Think of "/" as a boundary:
#
# def func(a, b, /, c, d)
#         ↑ ↑
#         these are positional-only
#
# After "/" → parameters behave normally


def example(a, b, /, c, d):
    return a + b + c + d

# Valid calls:
print("Mixed call:", example(1, 2, c=3, d=4))  # ✔ allowed

# Invalid:
try:
    example(a=1, b=2, c=3, d=4)  # ❌ a and b cannot be keywords
except TypeError as e:
    print("Error:", e)


print("\n========== 4. WHY PYTHON INTRODUCED THIS ==========\n")

# ✔ Avoid ambiguity
# ✔ Improve API design
# ✔ Protect parameter names from users

# Example:
def power(base, exp, /):
    return base ** exp

# User cannot misuse names like:
# power(base=2, exp=3) ❌


print("\n========== 5. REAL-WORLD USE CASE ==========\n")

# Many built-in functions use positional-only parameters
# Example: len(), pow(), etc.

print("len('abc') =", len("abc"))  # ✔ works

try:
    len(obj="abc")  # ❌ not allowed (positional-only)
except TypeError as e:
    print("Built-in error:", e)


print("\n========== 6. SUMMARY ==========\n")

# "/" means:
# → parameters BEFORE it = positional-only
#
# Key Rules:
# ✔ Must pass by position
# ❌ Cannot use keyword arguments
#
# Mental Model:
# def func(a, b, /)
# call → func(1, 2) ✔
# call → func(a=1, b=2) ❌

print("\nPOSITIONAL-ONLY PARAMETERS UNDERSTOOD ✔")
````

# 16. Keyword-Only Parameters

Keyword-only parameters use `*` in the signature.

Example:
```python
def build_name(*, first, last):
    ...
```

Why useful:
- clearer calls
- fewer argument-order mistakes


````python
# Define a function with keyword-only parameters.
def build_name(*, first, last):
    # Return the combined full name.
    return f"{first} {last}"


# Call the function with explicit keyword arguments.
print(build_name(first="Asha", last="Sharma"))
````

# 17. Structural Pattern Matching

`match ... case` is a major modern Python feature.

Why useful:
- cleaner branching
- readable alternatives to long if-elif chains
- useful for commands and structured data

Important note:
- requires newer Python versions


````python
# Store a command value.
command = "start"

# Match the command against known cases.
match command:
    case "start":
        print("Starting")
    case "stop":
        print("Stopping")
    case _:
        print("Unknown command")
````

# 18. Pattern Matching with Tuples

Pattern matching can also unpack structured values.

Why useful:
- readable dispatch logic
- command parsing
- event handling


````python
# Store a tuple-like command.
event = ("move", 10, 20)

# Match the tuple structure.
match event:
    case ("move", x, y):
        print(f"Move to {x}, {y}")
    case _:
        print("Unknown event")
````

# 19. `zip(strict=True)`

Modern Python can enforce matching lengths with `zip(strict=True)`.

Why useful:
- catches silent bugs
- safer paired iteration

Common old issue:
- normal zip stops at the shortest iterable silently


````python
# Use zip with strict mode on equal-length lists.
for name, score in zip(["Asha", "Ravi"], [95, 88], strict=True):
    # Print each paired record.
    print(name, score)
````

# 22. Edge Cases

Important edge cases:

- empty strings in formatting
- empty lists in unpacking
- pattern matching fall-through to `_`
- bad assumptions about types despite hints
- mutable defaults in dataclasses if handled carelessly


````python
# Import field for safe dataclass defaults.
from dataclasses import dataclass, field


# Use field(default_factory=...) for mutable defaults.
@dataclass
class Bag:
    # Create a fresh list for every instance.
    items: list[str] = field(default_factory=list)


# Print a safe default instance.
print(Bag())
````

# 24. Real-World Mini Practical Example

This example combines multiple new Python features.

Goal:
- use a dataclass
- use type hints
- use pathlib
- use an f-string
- use the dictionary merge operator


````python
# Import modern tools.
from dataclasses import dataclass
from pathlib import Path


# Define a typed dataclass for a report item.
@dataclass
class Report:
    # Store the report title.
    title: str

    # Store the item count.
    count: int


# Create a report object.
report = Report("Python Notes", 3)

# Build a modern dictionary using merge syntax.
base = {"status": "ready"}
details = {"title": report.title, "count": report.count}
payload = base | details

# Build a path with pathlib.
file_path = Path("reports") / "summary.txt"

# Print a modern formatted message.
print(f"Save {payload} to {file_path}")
````
