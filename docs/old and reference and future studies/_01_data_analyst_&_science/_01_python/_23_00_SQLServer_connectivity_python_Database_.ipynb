# Basic Syntax Explanation

## Standard workflow
1. Install `pyodbc`
2. Import `pyodbc`
3. Build a SQL Server connection string
4. Open a connection with `pyodbc.connect(...)`
5. Create a cursor with `connection.cursor()`
6. Execute SQL using `execute()` or `executemany()`
7. Call `commit()` after write operations
8. Fetch rows after `SELECT`
9. Close the cursor and connection

## Operators and patterns involved
- Python assignment: `=`
- SQL comparison: `=`, `>`, `<`, `>=`, `<=`, `!=`
- SQL logic: `AND`, `OR`, `NOT`
- SQL pattern matching: `LIKE`
- SQL grouping and sorting: `GROUP BY`, `ORDER BY`
- `pyodbc` positional placeholders: `?`

## Safety rule
Never merge raw user input directly into SQL text. Use parameterized queries with `?` placeholders.


```python
# Import pyodbc so Python can talk to SQL Server through ODBC.
import pyodbc

try:
    # Build the SQL Server connection string.
    # Replace these values with real server details before running.
    connection_string = (
        'DRIVER={ODBC Driver 17 for SQL Server};'
        'SERVER=localhost;'
        'DATABASE=school_db;'
        'UID=sa;'
        'PWD=your_password;'
    )

    # Open the connection to SQL Server.
    connection = pyodbc.connect(connection_string)

    # Create a cursor so SQL statements can be executed.
    cursor = connection.cursor()

    # Run a small query that returns the current database name.
    cursor.execute('SELECT DB_NAME()')

    # Read one row from the result.
    active_database = cursor.fetchone()

    # Print the result so the connection can be verified.
    print('Connected to SQL Server successfully')
    print('Active database:', active_database)

    # Close the cursor when query work is done.
    cursor.close()
except pyodbc.Error as error:
    # Print the exact ODBC or SQL Server error.
    print('SQL Server error:', error)
finally:
    # Close the connection only if it was created.
    if 'connection' in locals():
        connection.close()
        print('Connection closed')
```

## Concept 3: Connection Strings


A SQL Server connection string tells Python how to reach the server.

### Important parts
- `DRIVER`
- `SERVER`
- `DATABASE`
- `UID`
- `PWD`

### Variations
- SQL Server authentication
- Windows trusted connection

### Common mistakes
- Typing the wrong driver name
- Forgetting the database name
- Exposing passwords in shared code


```python
# Build a SQL Server authentication connection string.
sql_auth_connection_string = (
    'DRIVER={ODBC Driver 17 for SQL Server};'
    'SERVER=localhost;'
    'DATABASE=school_db;'
    'UID=sa;'
    'PWD=secret;'
)

# Build a Windows trusted connection string.
trusted_connection_string = (
    'DRIVER={ODBC Driver 17 for SQL Server};'
    'SERVER=localhost;'
    'DATABASE=school_db;'
    'Trusted_Connection=yes;'
)

print(sql_auth_connection_string)
print(trusted_connection_string)
```

## Concept 4: Establishing a Connection


A connection object represents the live session between Python and SQL Server.

### Why it matters
- No query can run without a connection
- Connections should be opened carefully and closed reliably

### Common mistakes
- Reopening many connections unnecessarily
- Not handling connection failures
- Leaving the connection open after work is done


```python
import pyodbc

try:
    # Prepare the connection string.
    connection_string = (
        'DRIVER={ODBC Driver 17 for SQL Server};'
        'SERVER=localhost;'
        'DATABASE=school_db;'
        'UID=sa;'
        'PWD=your_password;'
    )

    # Open the SQL Server connection.
    connection = pyodbc.connect(connection_string)
    print('Connection established')
except pyodbc.Error as error:
    print('Connection failed:', error)
finally:
    if 'connection' in locals():
        connection.close()
        print('Connection closed')
```

## Concept 5: Connection Object vs Cursor Object


The connection manages the database session, while the cursor sends SQL commands and fetches results.

### Connection object
- Opens and closes the database session
- Commits or rolls back transactions
- Creates cursor objects

### Cursor object
- Executes SQL commands
- Reads result rows
- Exposes metadata like `description`

### Common mistake
- Trying to fetch rows directly from the connection instead of the cursor


```python
import pyodbc

# Show the standard connection and cursor creation pattern.
connection_string = (
    'DRIVER={ODBC Driver 17 for SQL Server};'
    'SERVER=localhost;'
    'DATABASE=school_db;'
    'UID=sa;'
    'PWD=your_password;'
)

connection = pyodbc.connect(connection_string)
cursor = connection.cursor()

print('Connection type:', type(connection).__name__)
print('Cursor type:', type(cursor).__name__)

cursor.close()
connection.close()
```

## Concept 7: Creating Tables and Choosing SQL Server Data Types


Table structure should match the type of data being stored.

### Common SQL Server types
- `INT`
- `VARCHAR(n)`
- `NVARCHAR(n)`
- `DATE`
- `DECIMAL(p,s)`
- `BIT`

### SQL Server identity columns
SQL Server commonly uses `IDENTITY(1,1)` for auto-incrementing integer keys.

```python
create_table_query = '''
CREATE TABLE students (
    id INT IDENTITY(1,1) PRIMARY KEY,
    name VARCHAR(100) NOT NULL,
    email VARCHAR(150) UNIQUE NOT NULL,
    age INT,
    course VARCHAR(100) NOT NULL
)
'''

print(create_table_query)
```

## Concept 8: Inserting Data Safely


`INSERT` adds new rows into a table.

### Safe approach
Use parameterized queries with `?` placeholders in `pyodbc`.

### Why it matters
- Prevents SQL injection
- Keeps values separate from SQL structure
- Makes code cleaner and safer

### Common mistakes
- Using f-strings inside SQL
- Forgetting `commit()`
- Sending values in the wrong order


```python
insert_query = 'INSERT INTO students (name, email, age, course) VALUES (?, ?, ?, ?)'

# Store the data values separately in the same order as the placeholders.
student_values = ('Aman', 'aman@example.com', 20, 'Python')

print(insert_query)
print(student_values)
```

## Concept 9: Reading Data with `SELECT`


`SELECT` reads rows from SQL Server tables.

### Variations
- Read all columns
- Read selected columns only
- Filter with `WHERE`
- Sort with `ORDER BY`
- Restrict count with `TOP`

### Edge cases
- Query may return zero rows
- Result order is not guaranteed without explicit sorting


```python
select_query = '''
SELECT TOP 5 name, course
FROM students
WHERE age >= ?
ORDER BY name ASC
'''

filter_values = (18,)

print(select_query)
print(filter_values)
```

## Concept 10: Fetch Methods


After a `SELECT`, rows can be fetched in different ways.

### Variations
- `fetchone()`
- `fetchmany(size)`
- `fetchall()`

### Common mistakes
- Calling fetch methods before running a query
- Assuming `fetchall()` can be repeated without re-running the query


```python
fetch_methods = {
    'fetchone()': 'Returns one row or None',
    'fetchmany(3)': 'Returns the next 3 rows',
    'fetchall()': 'Returns all remaining rows'
}

for method, meaning in fetch_methods.items():
    print(method, '->', meaning)
```

## Concept 11: Updating Rows


`UPDATE` changes existing data.

### Important rule
Always use a `WHERE` clause unless every row is supposed to change.

### Common mistakes

```python
update_query = 'UPDATE students SET course = ? WHERE email = ?'
update_values = ('Advanced Python', 'aman@example.com')

print(update_query)
print(update_values)
```

## Concept 13: Transactions, Commit, and Rollback


A transaction groups related updates into one logical unit.

### Why it matters
- Prevents partial updates
- Keeps data consistent
- Allows recovery with `rollback()`

### Real-world example
Money transfer between two accounts should either complete fully or fail fully.


```python
withdraw_query = 'UPDATE accounts SET balance = balance - ? WHERE account_id = ?'
deposit_query = 'UPDATE accounts SET balance = balance + ? WHERE account_id = ?'

withdraw_values = (500, 101)
deposit_values = (500, 202)

print(withdraw_query)
print(withdraw_values)
print(deposit_query)
print(deposit_values)
```

## Concept 14: Error Handling


Database operations can fail because of connection issues, bad credentials, permission problems, syntax errors, or constraint violations.

### Good practice
- Use `try` / `except` / `finally`
- Catch `pyodbc.Error`
- Roll back writes when a transaction fails
- Close the cursor and connection


```python
import pyodbc

try:
    connection = pyodbc.connect(
        'DRIVER={ODBC Driver 17 for SQL Server};'
        'SERVER=localhost;'
        'DATABASE=school_db;'
        'UID=sa;'
        'PWD=wrong_password;'
    )
except pyodbc.Error as error:
    print('Handled SQL Server error:', error)
finally:
    if 'connection' in locals():
        connection.close()
```

## Concept 15: Bulk Insert with `executemany()`


When many rows need to be inserted, `executemany()` is cleaner than calling `execute()` repeatedly.

### Benefits
- Easier to read
- Cleaner import logic
- Better for batch inserts


```python
bulk_insert_query = 'INSERT INTO students (name, email, age, course) VALUES (?, ?, ?, ?)'

bulk_values = [
    ('Aman', 'aman@example.com', 20, 'Python'),
    ('Bina', 'bina@example.com', 21, 'SQL'),
    ('Charu', 'charu@example.com', 22, 'SQL Server')
]

print(bulk_insert_query)
for row in bulk_values:
    print(row)
```

## Concept 16: Filtering, Sorting, Grouping, and Aggregation

## Concept 22: Mini Practical Example - Student Management with SQL Server


This mini practical combines connection setup, table creation, insert, select, update, and final display.

It is close to the kind of small project workflow beginners build when learning SQL Server connectivity.


```python
import pyodbc

try:
    connection = pyodbc.connect(
        'DRIVER={ODBC Driver 17 for SQL Server};'
        'SERVER=localhost;'
        'DATABASE=school_db;'
        'UID=sa;'
        'PWD=your_password;'
    )
    cursor = connection.cursor()

    # Create the demo table if it does not already exist.
    cursor.execute(
        '''
        IF OBJECT_ID('student_demo', 'U') IS NULL
        CREATE TABLE student_demo (
            id INT IDENTITY(1,1) PRIMARY KEY,
            name VARCHAR(100) NOT NULL,
            course VARCHAR(100) NOT NULL,
            marks INT NOT NULL
        )
        '''
    )

    # Insert sample rows.
    cursor.executemany(
        'INSERT INTO student_demo (name, course, marks) VALUES (?, ?, ?)',
        [
            ('Aman', 'Python', 88),
            ('Bina', 'Python', 94),
            ('Charu', 'SQL Server', 81)
        ]
    )
    connection.commit()

    # Read Python learners.
    cursor.execute(
        'SELECT name, marks FROM student_demo WHERE course = ? ORDER BY marks DESC',
        ('Python',)
    )
    print('Python learners:', cursor.fetchall())

    # Update one learner.
    cursor.execute(
        'UPDATE student_demo SET marks = marks + ? WHERE name = ?',
        (5, 'Aman')
    )
    connection.commit()

    # Show the final rows.
    cursor.execute('SELECT name, course, marks FROM student_demo ORDER BY name')
    print('Final rows:', cursor.fetchall())

    cursor.close()
except pyodbc.Error as error:
    print('SQL Server practical error:', error)
finally:
    if 'connection' in locals():
        connection.close()
```
