# Introduction

## What is it?
Functional programming is a style of writing code where functions are treated as important building blocks. Instead of changing data step by step in many places, functional programming prefers transforming data through clear reusable functions.

## Why it is used?
- It makes logic easier to reuse.
- It can reduce side effects and hidden bugs.
- It helps create transformation pipelines for data.
- It is useful in data analysis, validation, filtering, and reporting tasks.

## Real-world relevance
- Cleaning lists of raw text values.
- Transforming CSV data into useful structures.
- Filtering API responses.
- Building small predictable helper functions for testing.


# Basic Syntax Explanation

## Main ideas
- Functions in Python are first-class objects, so they can be stored in variables and passed to other functions.
- `lambda` creates a small anonymous function.
- `map(function, iterable)` applies a function to each item.
- `filter(function, iterable)` keeps items for which the function returns `True`.
- `reduce(function, iterable)` combines items into one result.
- Comprehensions are often a readable alternative to `map()` and `filter()`.

## Operators involved
- `lambda x: ...` for inline function expressions.
- `+`, `-`, `*`, `/`, `%`, `**` inside transformation functions.
- Comparison operators like `>`, `<`, `==`.
- Logical operators like `and`, `or`, `not`.


```python
# Store a function in a variable to show that functions are first-class objects
def double(number):
    # Return the given number multiplied by two
    return number * 2


# Assign the function itself to another variable without calling it
operation = double

# Call the function through the new variable name
result = operation(8)

# Print the final result
print(result)
```

# Concept 1: First-Class Functions

In Python, functions are first-class objects. That means a function can be:
- stored in a variable,
- passed as an argument,
- returned from another function.

## Why this matters
This idea is the base of functional programming in Python.

## Common mistake
Writing `my_function()` when you want to pass the function itself. `my_function()` calls it immediately, while `my_function` passes the function object.


```python
# ================= FUNCTION AS ARGUMENT (DETAILED) =================

# Function 1: Converts text to uppercase
def to_upper_text(text):
    """
    Converts input string to uppercase

    Parameter:
    text (str) : input text

    Returns:
    str : uppercase version of text
    """
    return text.upper()


# Function 2: Higher-order function
def apply_to_value(value, func):
    """
    Applies a given function to a value

    Parameters:
    value : any data (string here)
    func  : function to apply on value

    Returns:
    Result of func(value)
    """

    # func is a FUNCTION OBJECT (reference), not executed yet
    return func(value)


# ================= EXECUTION FLOW =================

# Call:
# apply_to_value("python", to_upper_text)

# Step 1:
# value = "python"
# func = reference to function to_upper_text

# Step 2:
# Inside function → func(value)
# → to_upper_text("python")

# Step 3:
# "python".upper() → "PYTHON"

# Step 4:
# Return "PYTHON"

print("Result:", apply_to_value("python", to_upper_text))   # OUTPUT: PYTHON


# ================= IMPORTANT CONCEPT =================

# ✅ CORRECT → pass function reference
# apply_to_value("python", to_upper_text)

# ❌ WRONG → calling function immediately
# apply_to_value("python", to_upper_text())
# ERROR → missing argument


# ================= USING LAMBDA =================

# Inline function (anonymous)
print("Lambda:", apply_to_value("python", lambda x: x.capitalize()))  # Python


# ================= MULTIPLE FUNCTIONS =================

def reverse_text(text):
    return text[::-1]

def add_exclamation(text):
    return text + "!!!"

functions = [to_upper_text, reverse_text, add_exclamation]

for f in functions:
    print(f"{f.__name__} →", apply_to_value("python", f))


# ================= REAL-WORLD USE =================

# Example: Dynamic text processing pipeline
def process_text(text, operations):
    for op in operations:
        text = op(text)
    return text

pipeline = [str.strip, str.upper, add_exclamation]

print("Pipeline Result:", process_text("  python  ", pipeline))


# ================= SUMMARY =================
# 1. Functions are FIRST-CLASS objects in Python
# 2. You can pass functions as arguments
# 3. func(value) → dynamically executes passed function
# 4. Used in:
#    - Callbacks
#    - Data processing pipelines
#    - Flask routes & middleware
#    - AI preprocessing steps
```

# Concept 2: Pure Functions

A pure function gives the same output for the same input and does not change outside state.

## Why pure functions are useful
- Easier to test
- Easier to debug
- Easier to reuse

## Edge case
If a function depends on global variables, current time, random values, or external files, it may not be pure.


```python
# Define a pure function because the result depends only on the input values
def add_tax(amount, tax_rate):
    # Return the final amount after adding tax
    return amount + (amount * tax_rate)


# Call the function twice with the same input
first_result = add_tax(100, 0.18)
second_result = add_tax(100, 0.18)

# Print both results to show they are the same
print(first_result)
print(second_result)
```

# Concept 3: Immutability Idea

Functional programming prefers not mutating original data when possible. Python lists are mutable, but we can still follow an immutable style by returning new values instead of changing old ones.

## Why this helps
- Original data remains safe
- Bugs from accidental changes become less common
- Code becomes easier to reason about

## Common mistake
Changing the original list when the user expects a new transformed list.


```python
# Create an original list
numbers = [1, 2, 3, 4]

# Build a new list instead of changing the original list in place
doubled_numbers = [number * 2 for number in numbers]

# Print both lists to compare them
print(numbers)
print(doubled_numbers)
```

# Concept 4: Lambda Functions

`lambda` creates a short anonymous function in one line.

## Syntax
`lambda parameters: expression`

## Best use
- Small temporary functions
- Sorting keys
- Quick transformations inside `map()` or `filter()`

## Common mistake
Using `lambda` for long complicated logic makes code hard to read.


```python
# Create a lambda function that squares a number
square = lambda value: value * value

# Call the lambda function
print(square(5))

# Use lambda with sorted to sort by string length
words = ["data", "ai", "python", "ml"]
sorted_words = sorted(words, key=lambda word: len(word))

# Print the sorted output
print(sorted_words)
```

# Concept 5: `map()` for Transformation

`map()` applies a function to every item of an iterable and returns a map object.

## Important note
In Python 3, `map()` does not return a list directly. It returns an iterator, so you often convert it using `list()`.

## Common mistake
Forgetting to wrap `map()` with `list()` when you want to display or reuse all values immediately.


```python
# Create a list of numbers
numbers = [1, 2, 3, 4]

# Use map to square each number
squared_map = map(lambda number: number * number, numbers)

# Convert the map object into a real list
squared_list = list(squared_map)

# Print the transformed list
print(squared_list)
```

# Concept 6: `filter()` for Selection

`filter()` keeps only the items for which the function returns `True`.

## Why useful
- Removes unwanted values
- Creates clean subsets
- Helps build validation pipelines

## Edge case
If the condition is always false, the result becomes an empty iterable.


```python
# Create a list of numbers
numbers = [1, 2, 3, 4, 5, 6]

# Use filter to keep only even numbers
even_numbers = filter(lambda number: number % 2 == 0, numbers)

# Convert the filtered result into a list
even_list = list(even_numbers)

# Print the selected values
print(even_list)
```

# Concept 7: `reduce()` for Accumulation

`reduce()` repeatedly combines values until one result remains.

## Import needed
`reduce()` is in the `functools` module.

## Real use
- Total product
- Combined score
- Building cumulative values

## Common mistake
Using `reduce()` when `sum()` or another simpler built-in would be more readable.


```python
# ================= REDUCE FUNCTION - FULL EXPLANATION =================

from functools import reduce

# Input list
numbers = [1, 2, 3, 4]

# reduce() applies a function cumulatively to the list
# Syntax: reduce(function, iterable)

# Lambda function:
# lambda left, right: left * right
# → takes two values and returns their product

product = reduce(lambda left, right: left * right, numbers)

# ---------------- HOW REDUCE WORKS INTERNALLY ----------------
# Step 1: left = 1, right = 2 → 1 * 2 = 2
# Step 2: left = 2, right = 3 → 2 * 3 = 6
# Step 3: left = 6, right = 4 → 6 * 4 = 24
# Final result = 24

# Equivalent normal loop (for understanding):
# result = numbers[0]
# for n in numbers[1:]:
#     result = result * n

# Print final output
print("Final Product:", product)   # Output: 24


# ---------------- EXTRA EXAMPLES ----------------

# Sum using reduce
sum_result = reduce(lambda a, b: a + b, numbers)
print("Sum:", sum_result)   # 10

# Maximum value using reduce
max_result = reduce(lambda a, b: a if a > b else b, numbers)
print("Max:", max_result)   # 4


# ---------------- KEY POINTS ----------------
# 1. reduce() reduces a list → single value
# 2. It always works with 2 elements at a time
# 3. Lambda defines the operation
# 4. Common use cases: sum, product, max, custom aggregation
```

```python
# ================= REDUCE - PRODUCT EXPLANATION =================

from functools import reduce

# Input list
numbers = [5, 10, 3, 4]

# reduce() applies the lambda function step-by-step across the list
# lambda left, right: left * right → multiplies two values

product = reduce(lambda left, right: left * right, numbers)

# ---------------- INTERNAL WORKING ----------------
# Step 1: left = 5,  right = 10 → 5 * 10 = 50
# Step 2: left = 50, right = 3  → 50 * 3 = 150
# Step 3: left = 150, right = 4 → 150 * 4 = 600
# Final result = 600

# Equivalent loop (for understanding):
# result = numbers[0]   # 5
# for n in numbers[1:]: # 10, 3, 4
#     result = result * n

# Print final output
print("Final Product:", product)   # Output: 600


# ---------------- KEY IDEA ----------------
# reduce() keeps combining elements:
# (((5 * 10) * 3) * 4) = 600
```

# Concept 8: Comprehensions vs `map()` and `filter()`

List comprehensions are often more readable in Python than `map()` and `filter()`.

## Example comparison
- `map()` is concise for direct function application.
- Comprehensions are often easier to read when conditions are involved.

## Best practice
Use the form that makes the code clearest to a beginner or teammate.


```python
# Create a list of numbers
numbers = [1, 2, 3, 4, 5]

# Use map to square values
map_result = list(map(lambda number: number * number, numbers))

# Use a list comprehension to do the same work
comprehension_result = [number * number for number in numbers]

# Print both results to compare them
print(map_result)
print(comprehension_result)
```

# Concept 9: `sorted()` with Key Functions

Functional programming often uses functions as sorting rules.

## Why important
Sorting with a key function is a very common interview and real-world task.

## Advanced usage
You can sort by:
- string length,
- a tuple element,
- lowercase version of text,
- custom computed values.


```python
# Create a list of student records as tuples
students = [("Asha", 82), ("Ravi", 95), ("Meena", 88)]

# Sort the tuples by the score value stored at index 1
sorted_students = sorted(students, key=lambda student: student[1], reverse=True)

# Print the sorted records
print(sorted_students)
```

# Concept 10: `zip()` for Combining Sequences

`zip()` combines elements from multiple iterables position by position.

## Why useful
- Pairs related data
- Helps build dictionaries
- Supports parallel processing of small sequences

## Edge case
`zip()` stops when the shortest iterable ends.


```python
# Create two related lists
names = ["Asha", "Ravi", "Meena"]
scores = [82, 95, 88]

# Combine both lists item by item
combined_data = list(zip(names, scores))

# Print the paired result
print(combined_data)
```

# Concept 11: `any()` and `all()`

These functions help evaluate many boolean conditions.

## Meaning
- `any()` returns `True` if at least one item is true.
- `all()` returns `True` only if every item is true.

## Real-world relevance
- Validation checks
- Permission rules
- Pass/fail conditions


```python
# Create a list of scores
scores = [45, 60, 75, 90]

# Check whether any score is above 80
has_high_score = any(score > 80 for score in scores)

# Check whether all scores are at least 40
all_passed = all(score >= 40 for score in scores)

# Print both boolean results
print(has_high_score)
print(all_passed)
```

# Concept 12: Higher-Order Functions

A higher-order function either takes another function as input or returns a function.

## Why it matters
This is a major functional programming idea and appears in decorators, callbacks, and reusable pipelines.


```python
# ================= CLOSURE / FUNCTION RETURNING FUNCTION =================

# Outer function
def make_multiplier(multiplier):
    """
    Returns a function that multiplies any value by 'multiplier'

    multiplier → stored value (captured by closure)
    """

    # Inner function
    def multiply(value):
        """
        This function REMEMBERS 'multiplier' from outer scope
        even after make_multiplier() has finished execution
        """
        return value * multiplier   # uses outer variable

    # Return the function itself (not calling it)
    return multiply


# ---------------- FUNCTION CREATION ----------------

# Call outer function
# multiplier = 3
# It returns the inner function 'multiply'
triple = make_multiplier(3)

# Now:
# triple is actually → multiply function with multiplier = 3 stored inside

# ---------------- EXECUTION FLOW ----------------

# triple(10)
# → multiply(10)
# → 10 * 3 = 30

print("Result:", triple(10))   # Output: 30


# ---------------- IMPORTANT CONCEPT: CLOSURE ----------------
# Closure = function + remembered environment

# Even though make_multiplier() finished,
# 'multiplier' is still remembered inside 'multiply'


# ---------------- MULTIPLE INSTANCES ----------------

double = make_multiplier(2)
quadruple = make_multiplier(4)

print("Double:", double(5))        # 10
print("Quadruple:", quadruple(5))  # 20


# ---------------- EQUIVALENT (WITHOUT CLOSURE) ----------------
# Normally you'd pass multiplier every time:
# def multiply(value, multiplier):
#     return value * multiplier


# ---------------- KEY POINTS ----------------
# 1. Functions are FIRST-CLASS objects
# 2. Functions can RETURN other functions
# 3. Inner function "captures" outer variables → CLOSURE
# 4. Each call creates a separate memory (state)

# Real-world usage:
# - Custom configurations
# - AI preprocessing pipelines
# - Flask decorators / middleware
```

# Concept 13: Data Cleaning Pipeline

Functional programming is very useful for step-by-step data cleaning.

## Pipeline idea
- strip spaces,
- convert to lowercase,
- remove empty values,
- keep only valid items.

## Real-world example
Cleaning raw survey answers or imported CSV text data.


```python
# Create a list with messy text values
raw_values = ["  Python  ", "", "DATA", " Ai ", "  "]

# Strip spaces from every value
stripped_values = map(lambda value: value.strip(), raw_values)

# Convert every value to lowercase
lower_values = map(lambda value: value.lower(), stripped_values)

# Keep only non-empty values
clean_values = list(filter(lambda value: value != "", lower_values))

# Print the cleaned list
print(clean_values)
```

# Concept 15: Short-Hand and Optimized Methods

Python already provides useful built-ins that often replace more complex logic.

## Examples
- `sum()` instead of `reduce()` for addition
- comprehensions instead of simple `map()` or `filter()`
- generator expressions for memory-efficient processing

## Best practice
Choose the clearest tool, not just the most clever one.


```python
# Create a list of numbers
numbers = [10, 20, 30, 40]

# Use sum as the simplest way to add values
total = sum(numbers)

# Use a generator expression to count even values without creating a full extra list
even_count = sum(1 for number in numbers if number % 2 == 0)

# Print the optimized results
print(total)
print(even_count)
```

# Concept 16: Advanced Usage with Generator Expressions

Generator expressions produce values lazily, one at a time.

## Why advanced users like them
- Better memory usage for large data
- Great for streaming computations
- Common in analytics code


```python
# Create a generator expression for squared values
squared_generator = (number * number for number in range(1, 6))

# Loop through the generator and print values one by one
for value in squared_generator:
    # Print each generated value
    print(value)
```

# Concept 17: Mini Practical Example

Let us build a small practical workflow that:
- filters valid marks,
- converts them to grades,
- summarizes the result.

This combines multiple functional ideas in one place.


```python
# Create a list of raw marks with one invalid value
marks = [95, 82, 67, 48, -5, 110]

# Keep only valid marks between 0 and 100
valid_marks = list(filter(lambda mark: 0 <= mark <= 100, marks))

# Convert valid marks to letter grades
grades = list(
    map(
        lambda mark: "A" if mark >= 90 else "B" if mark >= 75 else "C" if mark >= 60 else "D",
        valid_marks,
    )
)

# Count how many students passed with grade A or B
high_grades = list(filter(lambda grade: grade in {"A", "B"}, grades))

# Print each stage of the mini pipeline
print(valid_marks)
print(grades)
print(high_grades)
```
