# 2. Nested Functions

A nested function is a function defined inside another function.

Why useful:
- keeps helper logic close to where it is used
- hides implementation details
- prepares the ground for closures and decorators

Common mistake:
- trying to call the inner function directly from outside when it is not returned


````python
# Define an outer function.
def outer():
    # Define an inner helper function.
    def inner():
        # Print a message to show the inner function is active.
        print("Inner helper is running")

    # Call the helper from inside the outer function.
    inner()


# Call the outer function.
outer()
````

# 3. Closures

A closure is a function that remembers values from the scope where it was created, even after the outer function finishes.

Why important:
- builds function factories
- useful for callbacks
- useful for configuration-based behavior

Edge case:
- closures capture references, so mutable outer values can still change later


````python
# Define a function factory.
def make_multiplier(multiplier):
    # Define an inner function that remembers multiplier.
    def multiply(number):
        # Return the multiplication result using the remembered value.
        print(f"Multiplying {number} by {multiplier}")
        return number * multiplier

    # Return the inner function.
    return multiply


# Build a custom doubling function.
double = make_multiplier(2)

# Build a custom tripling function.
triple = make_multiplier(3)

# Print results from both closures.
print(double(10))
print(triple(10))
````

# 4. The `nonlocal` Keyword

`nonlocal` allows an inner function to modify a variable from the nearest enclosing function scope.

Why useful:
- stateful counters
- memoized closures
- custom trackers

Common mistake:
- assigning to an outer variable without `nonlocal`, which creates a new local variable and causes confusion


````python
# Define a counter factory.
def make_counter():
    # Store state in the outer scope.
    count = 0

    # Define the inner function.
    def increment():
        # Tell Python to reuse the outer count variable.
        nonlocal count

        # Update the outer state.
        count += 1

        # Return the updated count.
        return count

    # Return the inner function.
    return increment


# Create a reusable counter.
counter = make_counter()

# Print several counter values.
print(counter())
print(counter())
print(counter())
````

# 5. LEGB Scope Rule
PYTHON LEGB RULE (NAME RESOLUTION) - COMPLETE NOTES
 ==========================================================
 LEGB = Local → Enclosing → Global → Built-in

 When Python sees a variable/function name, it searches in this order:
 1) Local      → Inside current function
 2) Enclosing  → Outer (nested) function
 3) Global     → Module (file-level)
 4) Built-in   → Python predefined names (print, len, etc.)

 Python STOPS at the FIRST match found.



````python

print("\n========== 1. LOCAL SCOPE ==========")
def local_example():
    x = 10  # Local variable (exists only inside this function)
    print("Local x =", x)

local_example()

# print(x)  # ❌ ERROR: x is not defined outside the function

````

````python
print("\n========== 2. ENCLOSING SCOPE ==========")
def outer():
    x = 20  # Enclosing variable (belongs to outer function)

    def inner():
        # Python cannot find x in local → checks enclosing → finds here
        print("Enclosing x =", x)

    inner()

outer()
````

````python
print("\n========== 3. GLOBAL SCOPE ==========")
x = 30  # Global variable (accessible anywhere in this file)

def global_example():
    # No local x → no enclosing → Python finds global x
    print("Global x =", x)

global_example()
````

````python
print("\n========== 4. BUILT-IN SCOPE ==========")
# Built-in names are predefined in Python (like print, len, range)
text = "Hello"
print("Length using built-in len() =", len(text))
````

````python

print("\n========== 5. COMPLETE LEGB DEMO ==========")
x = "global"

def outer():
    x = "enclosing"

    def inner():
        x = "local"
        print("LEGB result =", x)  # Stops at LOCAL

    inner()

outer()
# Output → local
````

````python
print("\n========== 6. SHADOWING (IMPORTANT) ==========")
# Shadowing means a variable in inner scope hides outer scope variable

x = "global"

def test_shadow():
    x = "local"  # This hides global x inside function
    print("Inside function:", x)

test_shadow()
print("Outside function:", x)  # Global remains unchanged
````

````python
print("\n========== 7. GLOBAL KEYWORD ==========")
# Used to MODIFY a global variable inside a function

x = 100

def modify_global():
    global x   # tells Python: use global x, not create local
    x = 200

modify_global()
print("Modified global x =", x)
````

````python
print("\n========== 8. NONLOCAL KEYWORD ==========")
# Used to modify enclosing (outer function) variable

def outer_nonlocal():
    x = 10

    def inner():
        nonlocal x  # refers to outer function's x
        x = 50

    inner()
    print("Nonlocal modified x =", x)

outer_nonlocal()
````

````python
print("\n========== 9. BUILT-IN OVERRIDING (DANGEROUS) ==========")
# You can override built-in names (not recommended)

len = 999   # ❌ overwriting built-in function name

# print(len("Hello"))  # ❌ ERROR: 'int' object is not callable

# Fix:
del len
print("Restored built-in len =", len("Hello"))
````

# 12. Custom Context Manager with a Class

A custom context manager class implements:

- `__enter__`
- `__exit__`

Advanced use:
- timers
- transactions
- custom resource pools


````python
# Define a custom context manager class.
class SimpleManager:
    # Run when entering the with block.
    def __enter__(self):
        # Print a setup message.
        print("Entering the block")

        # Return a value for the as variable.
        return {"status": "ready"}

    # Run when leaving the with block.
    def __exit__(self, exc_type, exc_value, traceback):
        # Print a cleanup message.
        print("Exiting the block")


# Use the custom context manager.
with SimpleManager() as resource:
    # Print the resource returned by __enter__.
    print(resource)
# x=SimpleManager()
# print(x)
````

# 13. Context Managers with `contextlib.contextmanager`

`contextlib.contextmanager` lets us create context managers with a generator-like function.

Why useful:
- less boilerplate than a full class
- good for simple setup and cleanup flows


````python
# Import the helper decorator.
from contextlib import contextmanager


# Build a context manager from a function.
@contextmanager
def managed_message():
    # Create a resource value.
    value = "temporary resource"

    # Yield control to the with block.
    yield value

    # Run cleanup code after the block ends.
    print("Cleanup complete")


# Use the context manager.
with managed_message() as item:
    # Print the yielded value.
    print(item)
````

# 20. Shallow Copy vs Deep Copy

A shallow copy copies the outer container only. A deep copy recursively copies nested mutable objects too.

Common copy methods:
- `list.copy()`
- slicing
- `copy.deepcopy()`

Edge case:
- shallow copies still share inner lists or dictionaries


````python
# Import the copy module.
import copy

# Create a nested list.
nested = [[1, 2], [3, 4]]

# Make a shallow copy.
shallow_copy = nested.copy()

# Make a deep copy.
deep_copy = copy.deepcopy(nested)

# Change the inner list in the original structure.
nested[0].append(99)

# Print all versions to compare behavior.
print(nested)
print(shallow_copy)
print(deep_copy)
````

# 21. Properties

`@property` lets a method behave like a normal attribute.

Why useful:
- validation
- computed values
- read-only fields

Short-hand advantage:
- cleaner than explicit getter methods


````python
# Define a class with a property.
class Temperature:
    # Store Celsius at object creation time.
    def __init__(self, celsius):
        # Save the value in a private-style attribute.
        self._celsius = celsius

    # Expose a computed read-only property.
    @property
    def fahrenheit(self):
        # Convert Celsius to Fahrenheit.
        return (self._celsius * 9 / 5) + 32


# Create a Temperature object.
today = Temperature(25)

# Access the property like a normal attribute.
print(today.fahrenheit)
````

# 22. Property Setters for Validation

A property setter lets us validate values during assignment.

Why useful:
- keeps object state valid
- avoids repeating validation code
- gives a clean public API


````python
# Define a class with getter and setter properties.
class BankAccount:
    # Initialize with a starting balance.
    def __init__(self, balance):
        # Assign through the property setter for validation.
        self.balance = balance

    # Define the readable property.
    @property
    def balance(self):
        # Return the stored private value.
        return self._balance

    # Define the writable property.
    @balance.setter
    def balance(self, value):
        # Reject negative balances.
        if value < 0:
            raise ValueError("Balance cannot be negative")

        # Store the validated value.
        self._balance = value


# Create a valid account.
account = BankAccount(500)

# Print the balance.
print(account.balance)
````

# 23. Descriptors

A descriptor is an object that controls attribute access through special methods such as `__get__` and `__set__`.

Why advanced code uses descriptors:
- reusable validation
- framework internals
- ORM field handling
- model attributes


````python
# Define a descriptor class.
class PositiveNumber:
    # Learn the attribute name when the owner class is created.
    def __set_name__(self, owner, name):
        # Save a private storage name.
        self.private_name = "_" + name

    # Control reading the attribute.
    def __get__(self, instance, owner):
        # Return the stored value from the instance.
        return getattr(instance, self.private_name)

    # Control writing the attribute.
    def __set__(self, instance, value):
        # Validate the incoming value.
        if value <= 0:
            raise ValueError("Value must be greater than zero")

        # Save the validated value on the instance.
        setattr(instance, self.private_name, value)


# Use the descriptor in another class.
class Product:
    # Connect the descriptor to the price attribute.
    price = PositiveNumber()

    # Initialize the object through descriptor validation.
    def __init__(self, price):
        # Set the price using the descriptor.
        self.price = price


# Create a product and print the validated price.
book = Product(250)
print(book.price)
````

# 24. Callable Objects

An object becomes callable when its class defines `__call__`.

Why useful:
- stateful function-like behavior
- customizable processors
- cleaner APIs in some designs


````python
# Define a callable class.
class Adder:
    # Save an increment value.
    def __init__(self, increment):
        # Store the increment for later use.
        self.increment = increment

    # Make object instances callable like functions.
    def __call__(self, value):
        # Return the processed value.
        return value + self.increment


# Create a callable object.
add_five = Adder(5)

# Call the object like a function.
print(add_five(10))
````

# 26. Dataclasses as a Practical Modern Pattern

`dataclasses` reduce boilerplate when creating classes mainly used to store data.

Why useful:
- automatic `__init__`
- cleaner class definitions

````python
# Import the dataclass decorator.
from dataclasses import dataclass


# Build a data-focused class with less boilerplate.
@dataclass
class Student:
    # Store the student name.
    name: str

    # Store the score.
    score: int


# Create a Student object.
student = Student("Asha", 92)

# Print the generated representation.
print(student)
````

# 28. Real-World Mini Practical Example

This example combines closures, decorators, context managers, and generators.

Goal:
- count how many records were processed
- log each processing step
- safely write results to a file
- lazily transform input values


````python
# Import wraps to preserve metadata.
from functools import wraps


# Create a decorator for simple logging.
def log_call(function):
    # Preserve the original function metadata.
    @wraps(function)
    def wrapper(*args, **kwargs):
        # Print the function name before running it.
        print(f"Processing with {function.__name__}")

        # Return the original function result.
        return function(*args, **kwargs)

    # Return the wrapper function.
    return wrapper


# Build a stateful counter with a closure.
def make_counter():
    # Store how many items were processed.
    count = 0

    # Define the inner function that updates the state.
    def increment():
        # Reuse the outer variable.
        nonlocal count

        # Increase the count.
        count += 1

        # Return the updated count.
        return count

    # Return the inner function.
    return increment


# Decorate a transformation function.
@log_call
def normalize_name(name):
    # Return a cleaned title-cased version.
    return name.strip().title()


# Create a lazy generator of cleaned names.
raw_names = ["  asha", "ravi  ", "  meera  "]
clean_name_generator = (normalize_name(name) for name in raw_names)

# Create the stateful counter.
count_processed = make_counter()

# Open a file safely for writing results.
with open("advanced_names_report.txt", "w", encoding="utf-8") as handle:
    # Loop through the generator one value at a time.
    for clean_name in clean_name_generator:
        # Update the processed count.
        current_count = count_processed()

        # Write a numbered line to the file.
        handle.write(f"{current_count}. {clean_name}\n")


# Read the result file and print its contents.
with open("advanced_names_report.txt", "r", encoding="utf-8") as handle:
    # Display the generated report.
    print(handle.read())
````
