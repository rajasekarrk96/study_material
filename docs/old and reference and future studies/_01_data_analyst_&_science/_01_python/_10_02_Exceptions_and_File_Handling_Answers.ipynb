### Q1. Convert String to Integer Safely

Dry run: converting `'42'` succeeds and returns `42`; converting bad text would return `'Invalid'`. Time complexity is `O(n)` in text length.


```python
# Define a function that converts text to an integer safely
def safe_int(text: str):
    # Try converting the incoming text into an integer
    try:
        return int(text)
    except ValueError:
        # Return a friendly fallback when conversion fails
        return 'Invalid'

# Test case 1
input_value = '42'
expected_output = 42
print(safe_int(input_value))  # should match expected
print(expected_output)
```

### Q21. Retry Conversion with Fallback

Dry run: `'10'` becomes `10`, `'x'` fails so fallback `0` is used, and `'20'` becomes `20`. Time complexity is `O(n)`.


```python
# Define a function that converts strings with a fallback value
def convert_with_fallback(values: list[str], fallback: int) -> list[int]:
    # Create a list for converted values
    result = []

    # Convert each string one by one
    for value in values:
        try:
            result.append(int(value))
        except ValueError:
            result.append(fallback)

    # Return the completed converted list
    return result

# Test case 1
expected_output = [10, 0, 20]
print(convert_with_fallback(['10', 'x', '20'], 0))  # should match expected
print(expected_output)
```

### Q23. Custom Context Manager

Dry run: entering the context writes `Entering`, leaving it writes `Exiting`, and reading the file returns both messages. Time complexity is `O(1)` per special method plus file write size.


```python
# Define a custom context manager class
class FileSession:
    # Initialize with the log file name
    def __init__(self, file_name):
        self.file_name = file_name

    # Run when entering the with block
    def __enter__(self):
        with open(self.file_name, 'a', encoding='utf-8') as file:
            file.write('Entering\n')
        return self

    # Run when leaving the with block
    def __exit__(self, exc_type, exc_value, traceback):
        with open(self.file_name, 'a', encoding='utf-8') as file:
            file.write('Exiting\n')

# Clear sample log file
with open('session.txt', 'w', encoding='utf-8') as file:
    file.write('')

# Use the custom context manager
with FileSession('session.txt'):
    pass

# Test case 1
with open('session.txt', 'r', encoding='utf-8') as file:
    print(file.read().rstrip())  # should include Entering and Exiting
```

### Q25. Batch File Merger

Dry run: read the content of each input file in order, concatenate it, write the result into the output file, and return it. Time complexity is `O(n)` in total merged content size.


```python
# Define a function that merges multiple text files
def merge_files(input_files: list[str], output_file: str) -> str:
    # Start with an empty merged string
    merged = ''

    # Read each input file and add its content
    for file_name in input_files:
        with open(file_name, 'r', encoding='utf-8') as file:
            merged += file.read()

    # Write the merged content into the output file
    with open(output_file, 'w', encoding='utf-8') as file:
        file.write(merged)

    # Return the merged text
    return merged

# Prepare sample files
with open('a.txt', 'w', encoding='utf-8') as file:
    file.write('A')
with open('b.txt', 'w', encoding='utf-8') as file:
    file.write('B')

# Test case 1
expected_output = 'AB'
print(merge_files(['a.txt', 'b.txt'], 'out.txt'))  # should match expected
print(expected_output)
```

### Q34. Log Parsing Tool

Dry run: count lines containing `ERROR`, write the summary line into the output file, and return the summary text. Time complexity is `O(n)` in number of log lines.


```python
# Define a function that parses a log file and writes a summary
def log_parsing_tool(log_file: str, summary_file: str) -> str:
    # Count error lines in the log file
    with open(log_file, 'r', encoding='utf-8') as file:
        error_count = sum(1 for line in file if 'ERROR' in line)

    # Build the summary line
    summary = f'ERROR count: {error_count}'

    # Write the summary into the output file
    with open(summary_file, 'w', encoding='utf-8') as file:
        file.write(summary)

    # Return the summary text
    return summary

# Prepare sample log file
with open('app.log', 'w', encoding='utf-8') as file:
    file.write('INFO start\nERROR issue one\nERROR issue two\nERROR issue three')

# Test case 1
expected_output = 'ERROR count: 3'
print(log_parsing_tool('app.log', 'error_summary.txt'))  # should match expected
print(expected_output)
```
