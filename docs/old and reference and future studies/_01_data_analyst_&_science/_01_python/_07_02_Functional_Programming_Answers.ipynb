Step-by-step logic: apply a doubling lambda to each list element with `map`, then convert the map object to a list. Dry run: `[1, 2, 3]` becomes `[2, 4, 6]`. Time complexity is `O(n)`.


```python
# Define a function that doubles all numbers
def double_all(numbers: list[int]) -> list[int]:
    # Apply a lambda to every number and convert the map result to a list
    return list(map(lambda number: number * 2, numbers))

# Test case 1
input_values = [1, 2, 3]
expected_output = [2, 4, 6]
print(double_all(input_values))  # should match expected
print(expected_output)
```

### Q10. Safe Maximum

Step-by-step logic: return `None` for an empty list; otherwise find the largest value. Dry run: `[3, 9, 1]` returns `9`. Time complexity is `O(n)`.

### Q11. Frequency Map

Step-by-step logic: loop through each item and count how many times it appears using a dictionary. Dry run: `['a', 'b', 'a']` becomes `{'a': 2, 'b': 1}`. Time complexity is `O(n)`.


```python
# Define a function that builds a frequency map
def frequency_map(items: list[str]) -> dict[str, int]:
    # Start with an empty dictionary
    counts = {}

    # Update the count for each item
    for item in items:
        counts[item] = counts.get(item, 0) + 1

    # Return the completed frequency map
    return counts

# Test case 1
input_values = ['a', 'b', 'a']
expected_output = {'a': 2, 'b': 1}
print(frequency_map(input_values))  # should match expected
print(expected_output)
```

### Q15. Flatten One Level

Step-by-step logic: loop through each small list, then through each value inside it, and collect all values into one result list. Time complexity is `O(n)` over all values.


```python
# Define a function that flattens one level of nesting
def flatten_once(groups: list[list[int]]) -> list[int]:
    # Create a result list for flattened values
    flat = []

    # Visit each inner list
    for group in groups:
        # Add every value from the inner list to the flat result
        for value in group:
            flat.append(value)

    # Return the flat list
    return flat

# Test case 1
input_values = [[1, 2], [3], [4, 5]]
expected_output = [1, 2, 3, 4, 5]
print(flatten_once(input_values))  # should match expected
print(expected_output)
```

### Q18. Normalize Text

Step-by-step logic: strip spaces, lowercase values, and skip cleaned values that become empty. Time complexity is `O(n)`.


```python
# Define a function that normalizes text values
def normalize_text(values: list[str]) -> list[str]:
    # Create a result list for cleaned values
    cleaned = []

    # Clean each incoming value
    for value in values:
        # Strip spaces and lowercase the text
        normalized = value.strip().lower()

        # Keep only non-empty cleaned values
        if normalized:
            cleaned.append(normalized)

    # Return the cleaned list
    return cleaned

# Test case 1
input_values = [' Python ', '', ' DATA ']
expected_output = ['python', 'data']
print(normalize_text(input_values))  # should match expected
print(expected_output)
```

### Q19. Group by First Letter

Step-by-step logic: use the first character as the dictionary key and collect matching words under that key. Time complexity is `O(n)`.


```python
# Define a function that groups words by first letter
def group_by_first_letter(words: list[str]) -> dict[str, list[str]]:
    # Create an empty dictionary for groups
    groups = {}

    # Process each word one by one
    for word in words:
        # Use the first letter as the group key
        key = word[0]

        # Create the list if this key has not appeared before
        groups.setdefault(key, []).append(word)

    # Return the grouped dictionary
    return groups

# Test case 1
input_values = ['apple', 'ant', 'banana']
expected_output = {'a': ['apple', 'ant'], 'b': ['banana']}
print(group_by_first_letter(input_values))  # should match expected
print(expected_output)
```

### Q21. Top K Frequent Words

Step-by-step logic: count frequencies, sort by highest frequency and alphabetical order, then take the first `k` words. Time complexity is `O(n log n)` because of sorting.


```python
# Define a function that returns the top k frequent words
def top_k_frequent(words: list[str], k: int) -> list[str]:
    # Count word frequencies in a dictionary
    counts = {}
    for word in words:
        counts[word] = counts.get(word, 0) + 1

    # Sort by descending frequency and then alphabetically
    ordered = sorted(counts.items(), key=lambda item: (-item[1], item[0]))

    # Return only the first k words from the ordered pairs
    return [word for word, _ in ordered[:k]]

# Test case 1
input_values = ['b', 'a', 'b', 'c', 'a', 'a']
expected_output = ['a', 'b']
print(top_k_frequent(input_values, 2))  # should match expected
print(expected_output)
```

### Q22. Compose Two Functions

Step-by-step logic: return a new function that first applies `g`, then applies `f` to that result. Dry run: with `square` after `double`, input `3` becomes `6` then `36`. Time complexity is `O(1)` per call beyond the two function calls.


```python
# Define a function that composes two single-argument functions
def compose(f, g):
    # Return a new function that applies g first and then f
    return lambda value: f(g(value))

# Create two sample functions
def double(number):
    # Return the number doubled
    return number * 2

def square(number):
    # Return the square of the number
    return number * number

# Build the composed function
composed = compose(square, double)

# Test case 1
expected_output = 36
print(composed(3))  # should match expected
print(expected_output)
```

### Q23. Closure Counter

Step-by-step logic: create an outer variable, use `nonlocal` inside an inner function, and return the inner function. Each call updates stored state. Time complexity is `O(1)` per call.


```python
# Define a closure-based counter factory
def make_counter():
    # Store the count in the outer scope
    count = 0

    # Define the inner function that updates outer state
    def increment():
        # Allow updates to the outer count variable
        nonlocal count
        count += 1
        return count

    # Return the inner function
    return increment

# Build one counter
counter = make_counter()

# Test cases
print(counter())  # should be 1
print(counter())  # should be 2
print(counter())  # should be 3
```

### Q24. Manual Reduce

Step-by-step logic: start from the first value and combine it with each later value one by one. Time complexity is `O(n)`.


```python
# Define a manual reduce function
def manual_reduce(combine, values: list[int]) -> int:
    # Start with the first value as the accumulator
    result = values[0]

    # Combine the accumulator with each remaining value
    for value in values[1:]:
        result = combine(result, value)

    # Return the final reduced result
    return result

# Test case 1
expected_output = 10
print(manual_reduce(lambda a, b: a + b, [1, 2, 3, 4]))  # should match expected
print(expected_output)
```

### Q25. Unique Stable Filter

Step-by-step logic: keep a `seen` set and add each new item only once while preserving first appearance order. Time complexity is `O(n)`.


```python
# Define a function that keeps unique items in stable order
def unique_stable(items: list[str]) -> list[str]:
    # Track seen items in a set
    seen = set()

    # Store unique items in the result list
    result = []

    # Visit each item in order
    for item in items:
        # Add it only if it has not been seen before
        if item not in seen:
            seen.add(item)
            result.append(item)

    # Return the stable unique list
    return result

# Test case 1
input_values = ['a', 'b', 'a', 'c', 'b']
expected_output = ['a', 'b', 'c']
print(unique_stable(input_values))  # should match expected
print(expected_output)
```

### Q26. Score Pipeline

Step-by-step logic: filter records by score range, sort the filtered records by score descending, then return only the names. Time complexity is `O(n log n)` because of sorting.


```python
# Define a function that builds a score pipeline
def score_pipeline(records: list[dict]) -> list[str]:
    # Keep only records with scores in the required range
    filtered = [record for record in records if 60 <= record['score'] <= 90]

    # Sort the filtered records by score descending
    ordered = sorted(filtered, key=lambda record: record['score'], reverse=True)

    # Return only the names from the ordered records
    return [record['name'] for record in ordered]

# Test case 1
records = [{'name': 'A', 'score': 95}, {'name': 'B', 'score': 88}, {'name': 'C', 'score': 60}]
expected_output = ['B', 'C']
print(score_pipeline(records))  # should match expected
print(expected_output)
```

### Q27. Partition Even and Odd

Step-by-step logic: loop through numbers and place each one into the even list or odd list. Time complexity is `O(n)`.


```python
# Define a function that partitions numbers into even and odd
def partition_even_odd(numbers: list[int]) -> tuple[list[int], list[int]]:
    # Create separate lists for even and odd values
    evens = []
    odds = []

    # Visit each number and place it into the correct list
    for number in numbers:
        if number % 2 == 0:
            evens.append(number)
        else:
            odds.append(number)

    # Return both lists as a tuple
    return evens, odds

# Test case 1
expected_output = ([2, 4], [1, 3])
print(partition_even_odd([1, 2, 3, 4]))  # should match expected
print(expected_output)
```

### Q28. Longest Clean Word

Step-by-step logic: normalize words, remove empties, and return the longest cleaned word or `None` when nothing valid remains. Time complexity is `O(n)`.


```python
# Define a function that returns the longest cleaned word
def longest_clean_word(words: list[str]) -> str | None:
    # Normalize and keep only non-empty cleaned words
    cleaned = [word.strip().lower() for word in words if word.strip()]

    # Return None if no cleaned word exists
    if not cleaned:
        return None

    # Return the longest cleaned word
    return max(cleaned, key=len)

# Test case 1
input_values = ['  Python ', '', ' ai ']
expected_output = 'python'
print(longest_clean_word(input_values))  # should match expected
print(expected_output)
```

### Q29. Memoized Fibonacci

Step-by-step logic: store already computed Fibonacci values in a dictionary so repeated subproblems are not solved again. Time complexity is `O(n)`.


```python
# Define a memoized Fibonacci function
def memoized_fibonacci(n: int) -> int:
    # Store base results in a cache
    cache = {0: 0, 1: 1}

    # Define a helper function that reuses the cache
    def helper(value: int) -> int:
        if value in cache:
            return cache[value]
        cache[value] = helper(value - 1) + helper(value - 2)
        return cache[value]

    # Return the nth Fibonacci number
    return helper(n)

# Test case 1
expected_output = 55
print(memoized_fibonacci(10))  # should match expected
print(expected_output)
```

### Q32. Sales Summary

Step-by-step logic: compute the total, compute the average from that total, then count how many values are above average. Time complexity is `O(n)`.


```python
# Define a function that returns sales summary values
def sales_summary(amounts: list[int]) -> tuple[int, float, int]:
    # Compute the total of all sales amounts
    total = sum(amounts)

    # Compute the average sales amount
    average = total / len(amounts)

    # Count how many amounts are above average
    count_above = sum(1 for amount in amounts if amount > average)

    # Return total, average, and above-average count
    return total, average, count_above

# Test case 1
input_values = [100, 150, 250]
expected_output = (500, 500 / 3, 1)
print(sales_summary(input_values))  # should match expected
print(expected_output)
```
