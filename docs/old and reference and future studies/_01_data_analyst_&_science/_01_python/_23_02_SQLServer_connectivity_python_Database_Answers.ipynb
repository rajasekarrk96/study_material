## Q22. Rollback Pattern on Error


```python
# Problem Title: Rollback Pattern on Error
# Step-by-step logic:
# 1. Read the SQL Server or pyodbc requirement.
# 2. Build the safe query, helper, or connection pattern.
# 3. Print the requested output clearly.
# Dry run:
# The script creates the SQL Server pattern requested and prints the expected result shape for review.
# Time complexity:
# These notebook answers mostly run in O(1) time because they build and print fixed patterns.

import pyodbc

try:
    connection = pyodbc.connect(
        'DRIVER={ODBC Driver 17 for SQL Server};'
        'SERVER=localhost;'
        'DATABASE=school_db;'
        'UID=sa;'
        'PWD=your_password;'
    )
    cursor = connection.cursor()
    cursor.execute('UPDATE accounts SET balance = balance - ? WHERE account_id = ?', (500, 101))
    raise pyodbc.Error('Simulated failure')
    connection.commit()
except pyodbc.Error:
    if 'connection' in locals():
        connection.rollback()
    print('rollback()')
finally:
    if 'cursor' in locals():
        cursor.close()
    if 'connection' in locals():
        connection.close()

# Test case:
expected_output = '`rollback()`'
print('Expected output:')
print(expected_output)
```
