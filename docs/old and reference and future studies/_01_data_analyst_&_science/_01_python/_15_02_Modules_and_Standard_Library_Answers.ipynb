## Q11. Defaultdict Group

**Problem:** Use `defaultdict(list)` to group words by their first letter and return the grouped dictionary.

**Step-by-step logic:**
- Read the required input shape: `words: list[str]`.
- Choose the most suitable standard library module or import style.
- Use built-in tools instead of reinventing common functionality.
- Handle the stated constraint carefully: Words are non-empty strings.
- Return the final answer in the required format `dict[str, list[str]]`.


````python
def defaultdict_group(words: list[str]) -> dict[str, list[str]]:
    # Import defaultdict from collections.
    from collections import defaultdict

    # Create a defaultdict with list as the default factory.
    groups = defaultdict(list)

    # Group words by their first character.
    for word in words:
        groups[word[0]].append(word)

    # Return a normal dictionary for display.
    return dict(groups)

# Time Complexity: depends on the number of elements processed or file/text size handled above.
# Space Complexity: depends on extra containers, parsed data, or intermediate results created above.
````

````python
def date_add_days(date_text: str, days: int) -> str:
    # Import datetime and timedelta.
    from datetime import datetime, timedelta

    # Parse the input date string into a date object.
    start_date = datetime.strptime(date_text, "%Y-%m-%d").date()

    # Add the requested number of days.
    new_date = start_date + timedelta(days=days)

    # Return the new date as a string.
    return str(new_date)

# Time Complexity: depends on the number of elements processed or file/text size handled above.
# Space Complexity: depends on extra containers, parsed data, or intermediate results created above.
````

````python
def csv_read_names(csv_text: str) -> list[str]:
    # Import csv and StringIO for in-memory CSV reading.
    import csv
    from io import StringIO

    # Open the CSV text like a file.
    handle = StringIO(csv_text)

    # Build a reader for dictionary rows.
    reader = csv.DictReader(handle)

    # Return the names from the name column.
    return [row["name"] for row in reader]

# Time Complexity: depends on the number of elements processed or file/text size handled above.
# Space Complexity: depends on extra containers, parsed data, or intermediate results created above.
````

## Q35. Queue Processing With Deque

**Problem:** Use `deque` to process tasks from left to right and return the order in which tasks were processed.

**Step-by-step logic:**
- Read the required input shape: `tasks: list[str]`.
- Choose the most suitable standard library module or import style.
- Use built-in tools instead of reinventing common functionality.
- Handle the stated constraint carefully: The list may be empty.
- Return the final answer in the required format `list[str]`.


````python
def queue_processing_with_deque(tasks: list[str]) -> list[str]:
    # Import deque from collections.
    from collections import deque

    # Create a deque for left-to-right processing.
    queue = deque(tasks)

    # Store processed task order here.
    processed = []

    # Process tasks until the queue is empty.
    while queue:
        processed.append(queue.popleft())

    # Return the processed order.
    return processed

# Time Complexity: depends on the number of elements processed or file/text size handled above.
# Space Complexity: depends on extra containers, parsed data, or intermediate results created above.
````
