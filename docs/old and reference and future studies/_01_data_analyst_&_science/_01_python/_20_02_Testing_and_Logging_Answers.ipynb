### Q11. Count Log Levels

**Approach:** Solve the problem with clear beginner-friendly logic first, then verify it with direct test cases.


```python
def count_log_levels(levels: list[str]) -> dict[str, int]:
    # Step 1: Prepare an empty dictionary for counts.
    counts = {}

    # Step 2: Loop through each raw level string.
    for level in levels:
        # Step 3: Normalize spaces and letter case.
        normalized = level.strip().upper()

        # Step 4: Increase the count for that normalized level.
        counts[normalized] = counts.get(normalized, 0) + 1

    # Step 5: Return the completed frequency dictionary.
    return counts


# Dry run:
# 'info' -> INFO
# 'ERROR' -> ERROR
# ' info ' -> INFO

# Time complexity: O(n)

input_data = ["info", "ERROR", " info "]
expected_output = {"INFO": 2, "ERROR": 1}
print(count_log_levels(input_data))  # should match expected_output
assert count_log_levels(["info", "ERROR", " info "]) == expected_output
assert count_log_levels([]) == {}
```

### Q16. Merge Test Reports

**Approach:** Solve the problem with clear beginner-friendly logic first, then verify it with direct test cases.


```python
def merge_test_reports(report_a: dict[str, int], report_b: dict[str, int]) -> dict[str, int]:
    # Step 1: Start with a copy so the original first report is not changed.
    merged = report_a.copy()

    # Step 2: Visit every key-value pair in the second report.
    for key, value in report_b.items():
        # Step 3: Add the incoming value to any existing count.
        merged[key] = merged.get(key, 0) + value

    # Step 4: Return the combined dictionary.
    return merged


# Dry run:
# passed starts at 2
# add passed 1 -> 3
# add failed 3 -> 3

# Time complexity: O(n + m)

input_data = ({"passed": 2}, {"passed": 1, "failed": 3})
expected_output = {"passed": 3, "failed": 3}
print(merge_test_reports(*input_data))  # should match expected_output
assert merge_test_reports({"passed": 2}, {"passed": 1, "failed": 3}) == expected_output
assert merge_test_reports({}, {"failed": 1}) == {"failed": 1}
```

```python
def detect_duplicate_test_names(names: list[str]) -> list[str]:
    # Step 1: Count how many times each test name appears.
    counts = {}
    for name in names:
        counts[name] = counts.get(name, 0) + 1

    # Step 2: Keep only names with counts greater than one.
    duplicates = [name for name, count in counts.items() if count > 1]

    # Step 3: Return them in sorted order.
    return sorted(duplicates)


# Dry run:
# test_a count 2
# test_b count 1
# duplicates -> ['test_a']

# Time complexity: O(n log n) because of sorting

input_data = ["test_a", "test_b", "test_a"]
expected_output = ["test_a"]
print(detect_duplicate_test_names(input_data))  # should match expected_output
assert detect_duplicate_test_names(input_data) == expected_output
assert detect_duplicate_test_names(["one", "two"]) == []
```

### Q21. Group Logs By Level

**Approach:** Solve the problem with clear beginner-friendly logic first, then verify it with direct test cases.


```python
def group_logs_by_level(logs: list[dict]) -> dict[str, list[str]]:
    # Step 1: Prepare an empty dictionary for grouped messages.
    grouped = {}

    # Step 2: Visit each log row.
    for log in logs:
        # Step 3: Normalize the level and read the message.
        level = str(log.get("level", "")).strip().upper()
        message = str(log.get("message", ""))

        # Step 4: Create the list for a new level if needed.
        if level not in grouped:
            grouped[level] = []

        # Step 5: Append the message to the appropriate level group.
        grouped[level].append(message)

    # Step 6: Return the grouped dictionary.
    return grouped


# Dry run:
# INFO gets ['start']
# ERROR gets ['fail']

# Time complexity: O(n)

input_data = [{"level": "info", "message": "start"}, {"level": "ERROR", "message": "fail"}]
expected_output = {"INFO": ["start"], "ERROR": ["fail"]}
print(group_logs_by_level(input_data))  # should match expected_output
assert group_logs_by_level(input_data) == expected_output
assert group_logs_by_level([]) == {}
```

### Q22. Find Longest Failure Streak

**Approach:** Solve the problem with clear beginner-friendly logic first, then verify it with direct test cases.


```python
def find_longest_failure_streak(results: list[bool]) -> int:
    # Step 1: Track the current streak and the best streak seen so far.
    current_streak = 0
    longest_streak = 0

    # Step 2: Process every result.
    for result in results:
        # Step 3: Increase the streak for failures.
        if result is False:
            current_streak += 1
            longest_streak = max(longest_streak, current_streak)
        else:
            # Step 4: Reset the streak when a pass appears.
            current_streak = 0

    # Step 5: Return the longest failure run.
    return longest_streak


# Dry run:
# True -> streak 0
# False -> streak 1
# False -> streak 2
# True -> reset

# Time complexity: O(n)

input_data = [True, False, False, True, False]
expected_output = 2
print(find_longest_failure_streak(input_data))  # should match expected_output
assert find_longest_failure_streak(input_data) == 2
assert find_longest_failure_streak([]) == 0
```

**Approach:** Solve the problem with clear beginner-friendly logic first, then verify it with direct test cases.


```python
def collapse_duplicate_consecutive_logs(messages: list[str]) -> list[str]:
    # Step 1: Return early for empty input.
    if not messages:
        return []

    # Step 2: Start the result with the first message.
    collapsed = [messages[0]]

    # Step 3: Visit the remaining messages.
    for message in messages[1:]:
        # Step 4: Append only when the current message differs from the previous kept value.
        if message != collapsed[-1]:
            collapsed.append(message)

    # Step 5: Return the compressed list.
    return collapsed


# Dry run:
# Start with A
# next A same -> skip
# next B different -> keep

# Time complexity: O(n)

input_data = ["A", "A", "B", "B", "C"]
expected_output = ["A", "B", "C"]
print(collapse_duplicate_consecutive_logs(input_data))  # should match expected_output
assert collapse_duplicate_consecutive_logs(input_data) == expected_output
assert collapse_duplicate_consecutive_logs([]) == []
```

**Approach:** Solve the problem with clear beginner-friendly logic first, then verify it with direct test cases.


```python
def validate_monotonic_timestamps(timestamps: list[int]) -> bool:
    # Step 1: Compare each timestamp with the one before it.
    for index in range(1, len(timestamps)):
        # Step 2: If the current value is smaller, the sequence decreases.
        if timestamps[index] < timestamps[index - 1]:
            return False

    # Step 3: If no decrease was found, the sequence is monotonic.
    return True


# Dry run:
# 10 <= 10 <= 12 <= 15
# No decreases -> True

# Time complexity: O(n)

input_data = [10, 10, 12, 15]
expected_output = True
print(validate_monotonic_timestamps(input_data))  # should match expected_output
assert validate_monotonic_timestamps([10, 10, 12, 15]) is True
assert validate_monotonic_timestamps([10, 9]) is False
```

### Q30. Build Failure Matrix

**Approach:** Solve the problem with clear beginner-friendly logic first, then verify it with direct test cases.


```python
def build_failure_matrix(batches: list[list[bool]]) -> list[list[int]]:
    # Step 1: Prepare the outer matrix list.
    matrix = []

    # Step 2: Process one batch at a time.
    for batch in batches:
        # Step 3: Convert each boolean to 0 for pass and 1 for failure.
        converted_row = []
        for result in batch:
            if result is False:
                converted_row.append(1)
            else:
                converted_row.append(0)

        # Step 4: Add the converted row to the matrix.
        matrix.append(converted_row)

    # Step 5: Return the final matrix.
    return matrix


# Dry run:
# [True, False] -> [0, 1]
# [False, False] -> [1, 1]

# Time complexity: O(total number of boolean values)

input_data = [[True, False], [False, False]]
expected_output = [[0, 1], [1, 1]]
print(build_failure_matrix(input_data))  # should match expected_output
assert build_failure_matrix(input_data) == expected_output
assert build_failure_matrix([]) == []
```

```python
def api_health_summary(status_codes: list[int]) -> dict[str, int]:
    # Step 1: Prepare counters for response categories.
    summary = {"success": 0, "client_error": 0, "server_error": 0}

    # Step 2: Inspect every status code.
    for code in status_codes:
        # Step 3: Classify by numeric range.
        if 200 <= code <= 299:
            summary["success"] += 1
        elif 400 <= code <= 499:
            summary["client_error"] += 1
        elif 500 <= code <= 599:
            summary["server_error"] += 1

    # Step 4: Return the categorized counts.
    return summary


# Dry run:
# 200 success
# 201 success
# 404 client_error
# 500 server_error

# Time complexity: O(n)

input_data = [200, 201, 404, 500]
expected_output = {"success": 2, "client_error": 1, "server_error": 1}
print(api_health_summary(input_data))  # should match expected_output
assert api_health_summary(input_data) == expected_output
assert api_health_summary([]) == {"success": 0, "client_error": 0, "server_error": 0}
```

### Q32. Login Audit Report

**Approach:** Solve the problem with clear beginner-friendly logic first, then verify it with direct test cases.


```python
def login_audit_report(events: list[dict]) -> dict[str, int]:
    # Step 1: Start counters for successful and failed logins.
    report = {"success": 0, "failed": 0}

    # Step 2: Process each login event.
    for event in events:
        # Step 3: Update the appropriate bucket based on the success flag.
        if event.get("success") is True:
            report["success"] += 1
        else:
            report["failed"] += 1

    # Step 4: Return the audit report.
    return report


# Dry run:
# True -> success 1
# False -> failed 1

# Time complexity: O(n)

input_data = [{"success": True}, {"success": False}]
expected_output = {"success": 1, "failed": 1}
print(login_audit_report(input_data))  # should match expected_output
assert login_audit_report(input_data) == expected_output
assert login_audit_report([]) == {"success": 0, "failed": 0}
```

### Q33. ETL Validation Errors

**Approach:** Solve the problem with clear beginner-friendly logic first, then verify it with direct test cases.


```python
def etl_validation_errors(rows: list[dict]) -> list[int]:
    # Step 1: Store the positions of invalid rows.
    invalid_indices = []

    # Step 2: Enumerate rows so both index and content are available.
    for index, row in enumerate(rows):
        # Step 3: Mark the row if either required field is missing.
        if "id" not in row or "name" not in row:
            invalid_indices.append(index)

    # Step 4: Return the invalid row positions.
    return invalid_indices


# Dry run:
# row 0 has id and name -> valid
# row 1 misses name -> collect index 1

# Time complexity: O(n)

input_data = [{"id": 1, "name": "A"}, {"id": 2}]
expected_output = [1]
print(etl_validation_errors(input_data))  # should match expected_output
assert etl_validation_errors(input_data) == expected_output
assert etl_validation_errors([]) == []
```

### Q34. Payment Retry Analysis

**Approach:** Solve the problem with clear beginner-friendly logic first, then verify it with direct test cases.


```python
def payment_retry_analysis(transactions: list[list[bool]]) -> int:
    # Step 1: Count how many transactions eventually succeed.
    successful_transactions = 0

    # Step 2: Process one transaction attempt history at a time.
    for attempts in transactions:
        # Step 3: any(attempts) is True if at least one attempt succeeded.
        if any(attempts):
            successful_transactions += 1

    # Step 4: Return the final success count.
    return successful_transactions


# Dry run:
# [False, True] -> success
# [False, False] -> no success
# [True] -> success

# Time complexity: O(total number of attempts)

input_data = [[False, True], [False, False], [True]]
expected_output = 2
print(payment_retry_analysis(input_data))  # should match expected_output
assert payment_retry_analysis(input_data) == expected_output
assert payment_retry_analysis([]) == 0
```

### Q35. Production Error Dashboard

**Approach:** Solve the problem with clear beginner-friendly logic first, then verify it with direct test cases.


```python
def production_error_dashboard(logs: list[dict]) -> dict[str, int]:
    # Step 1: Prepare error counters per module.
    counts = {}

    # Step 2: Visit every log row.
    for log in logs:
        # Step 3: Normalize the level before comparison.
        level = str(log.get("level", "")).strip().upper()

        # Step 4: Count only ERROR rows.
        if level == "ERROR":
            module = log.get("module")
            counts[module] = counts.get(module, 0) + 1

    # Step 5: Return the module-to-error-count dictionary.
    return counts


# Dry run:
# auth ERROR -> auth 1
# auth INFO -> ignored

# Time complexity: O(n)

input_data = [{"module": "auth", "level": "ERROR"}, {"module": "auth", "level": "INFO"}]
expected_output = {"auth": 1}
print(production_error_dashboard(input_data))  # should match expected_output
assert production_error_dashboard(input_data) == expected_output
assert production_error_dashboard([]) == {}
```
