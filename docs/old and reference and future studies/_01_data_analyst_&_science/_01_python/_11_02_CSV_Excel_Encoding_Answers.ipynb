## Q1. Solution

```python
# Problem title: Q1 CSV/Excel/Encoding task
# Step-by-step logic:
# 1) Read/accept records
# 2) Normalize text and numeric fields
# 3) Return transformed result in expected schema

def solve_q1(records):
    result = []
    for row in records:
        name = str(row.get('name', '')).strip()
        age_text = str(row.get('age', '')).strip()
        norm_name = name.upper()
        norm_age = int(age_text) if age_text.isdigit() else 0
        result.append({'name': norm_name, 'age': norm_age})
    return result

# Dry run: {'name':' Asha ', 'age':'22'} -> {'name':'ASHA', 'age':22}
# Time complexity: O(n)

input_data = [{'name':' Asha ', 'age':'22'}, {'name':'Ravi', 'age':'24'}, {'name':'Mina', 'age':''}]
expected_output = [{'name':'ASHA', 'age':22}, {'name':'RAVI', 'age':24}, {'name':'MINA', 'age':0}]
actual_output = solve_q1(input_data)
print(actual_output)  # should match expected
print('Pass:', actual_output == expected_output)
```
