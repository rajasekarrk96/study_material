# Introduction

## What is it?
A class is a blueprint for creating objects. An object is a real instance created from that class. Classes group data and behavior together.

## Why it is used?
- To organize related data and logic.
- To model real-world entities.
- To make bigger programs easier to understand.
- To reuse logic by creating many objects from one class.

## Real-world relevance
- A `Student` object can store name, marks, and grade methods.
- A `BankAccount` object can store balance and support deposit or withdrawal.
- A `CartItem` object can keep price, quantity, and total calculations together.


# Basic Syntax Explanation

## Main syntax
- Use `class ClassName:` to define a class.
- Use `__init__()` to initialize object data.
- Use `self` to refer to the current object.
- Use dot notation like `object.attribute` and `object.method()`.

## Important ideas
- `self` is not a keyword, but it is the standard Python name.
- Instance attributes belong to each object separately.
- Methods are functions defined inside a class.

## Operators involved
- Assignment operator `=` for attributes.
- Dot operator `.` for accessing attributes and methods.


```python
# Define a simple class named Student
class Student:
    # Define the constructor that runs when a new object is created
    def __init__(self):
        # Store the incoming name inside the current object
        print("okay, I am running")


# Create an object from the Student class
student_one = Student()

# Print the object's stored attribute
print(student_one)
```

# Concept 1: Creating a Class

A class is a template. It describes what attributes and methods its objects will have.

## Why this matters
Without classes, related data and functions often stay scattered in many places.

## Common mistake
Beginners sometimes forget that a class definition does not create an object by itself. You still need to instantiate it.


```python
# Define a class that represents a book
class Book:
    # Use pass because this first example only shows the class structure
    pass


# Create an object from the Book class
book_one = Book()

# Print the type of the created object
print(type(book_one))
```

# Concept 2: The Constructor `__init__()`

The constructor initializes object data at creation time.

## Why useful
- Sets starting values
- Makes objects ready to use immediately
- Reduces setup mistakes

## Edge case
If you define required parameters in `__init__()`, you must provide them while creating the object.


```python
# Define a class with a constructor
class Laptop:
    # Accept brand and ram while creating the object
    def __init__(self, brand, ram):
        # Store brand inside the object
        self.brand = brand

        # Store ram inside the object
        self.ram = ram


# Create one laptop object
laptop_one = Laptop("Lenovo", 16)

# Print the initialized values
print(laptop_one.brand)
print(laptop_one.ram)
print(type(laptop_one.brand),type(laptop_one))
```

# Concept 3: Understanding `self`

`self` refers to the current object. It helps Python know which object's data should be used.

## Why important
When multiple objects exist, `self` keeps their data separate.

## Common mistake
Forgetting to include `self` as the first parameter in instance methods.


```python
# Define a class that uses self in a method
class Employee:
    # Initialize the object with a name
    def __init__(self, name):
        # Save the name inside this specific object
        self.name = name

    # Define an instance method
    def show_name(self):
        # Use self to access this object's name
        print(self.name)


# Create an employee object
employee_one = Employee("Ravi")

# Call the instance method
employee_one.show_name()
```

# Concept 4: Instance Attributes

Instance attributes belong to individual objects. Different objects can store different values in the same attribute name.

## Real-world meaning
Two students can both have a `name` attribute, but with different values.


```python
# Define a class with two instance attributes
class Car:
    # Initialize the object with brand and color
    def __init__(self, brand, color):
        # Store the brand for this object
        self.brand = brand

        # Store the color for this object
        self.color = color


# Create two different objects
car_one = Car("Toyota", "White")
car_two = Car("Honda", "Black")

# Print each object's separate data
print(car_one.brand, car_one.color)
print(car_two.brand, car_two.color)
```

# Concept 5: Methods

Methods are functions defined inside a class. They describe what an object can do.

## Why useful
- Keep behavior close to related data
- Make code easier to read and reuse

## Common mistake
Trying to access object attributes inside a method without using `self`.


```python
# Define a class with a method
class Fan:
    # Initialize the object with a speed value
    def __init__(self, speed):
        # Store the starting speed
        self.speed = speed

    # Define a method to increase speed
    def increase_speed(self):
        # Add one to the current speed
        self.speed += 1


# Create a fan object
fan_one = Fan(2)

# Change the object state by calling the method
fan_one.increase_speed()

# Print the updated state
print(fan_one.speed)
```

# Concept 6: Object State Changes

Objects can change over time through methods.

## Why important
This is one of the biggest advantages of objects. They can remember data and update it.

## Real-world example
A bank account balance changes after deposits and withdrawals.


```python
# Define a bank account class
class BankAccount:
    # Set the starting balance
    def __init__(self, balance):
        # Store the balance in the object
        self.balance = balance

    # Add money to the balance
    def deposit(self, amount):
        # Increase the balance
        self.balance += amount

    # Subtract money from the balance
    def withdraw(self, amount):
        # Decrease the balance
        self.balance -= amount


# Create one bank account object
account = BankAccount(1000)

# Update the account state
account.deposit(500)
account.withdraw(200)

# Print the final balance
print(account.balance)
```

# Concept 7: Class Attributes

Class attributes belong to the class itself and are shared by all objects unless overridden.

## Difference from instance attributes
- Instance attribute: specific to one object
- Class attribute: shared default value across objects

## Common mistake
Changing a class attribute when you actually wanted different values per object.


```python
# Define a class with a class attribute
class MobilePhone:
    # Create one shared class attribute
    category = "Electronics"

    # Initialize each phone object with its own brand
    def __init__(self, brand):
        # Store the brand separately for each object
        self.brand = brand


# Create two objects
phone_one = MobilePhone("Samsung")
phone_two = MobilePhone("Apple")

# Print both instance and class attributes
print(phone_one.brand, phone_one.category)
print(phone_two.brand, phone_two.category)
```

# Concept 8: Class Attributes vs Instance Attributes

It is important to know where data should live.

## Good rule
- Use class attributes for shared constant-like values.
- Use instance attributes for object-specific data.

## Edge case
Assigning `object.attribute = value` can create an instance attribute with the same name as a class attribute.


```python
# Define a class with a class attribute
class School:
    # Create a shared school name
    school_name = "Green Valley School"

    # Initialize each student object with its own name
    def __init__(self, student_name):
        # Save the student name
        self.student_name = student_name


# Create an object
student = School("Meena")

# Override school_name only for this object
student.school_name = "Riverdale School"

# Print the instance-level value
print(student.school_name)

# Print the class-level shared value
print(School.school_name)
print(student)
```

# Concept 9: String Representation with `__str__()`

`__str__()` lets you control how an object appears when printed.

## Why useful
- Better readability
- Better debugging
- Friendlier output


```python
# Define a class with a custom string representation
class Course:
    # Initialize the object with title and duration
    def __init__(self, title, duration):
        # Store the title
        self.title = title

        # Store the duration
        self.duration = duration

    # Return a friendly string when the object is printed
    def __str__(self):
        # Build a readable description of the object
        return f"{self.title} - {self.duration} hours"


# Create one course object
course = Course("Python", 40)

# Print the object directly
print(course)
```

# Concept 10: Default Values in Objects

Constructors can use default values for optional attributes.

## Why helpful
- Makes object creation more flexible
- Reduces repeated arguments

## Example
A product can default to quantity `1` unless a different quantity is passed.


```python
# Define a class with a default constructor argument
class Product:
    # Give quantity a default value
    def __init__(self, name, quantity=1):
        # Store the product name
        self.name = name

        # Store the quantity
        self.quantity = quantity


# Create one product with default quantity
product_one = Product("Notebook")

# Create another product with custom quantity
product_two = Product("Pen", 5)

# Print both objects' data
print(product_one.name, product_one.quantity)
print(product_two.name, product_two.quantity)
```

# Concept 11: Validation Inside Methods

Objects can protect their data by validating values before changing them.

## Why important
Prevents invalid state such as negative balance, negative stock, or impossible marks.

## Real-world example
A withdrawal method should stop the balance from becoming negative.


```python
# Define a safer bank account class
class SafeBankAccount:
    # Initialize the account with a balance
    def __init__(self, balance):
        # Store the starting balance
        self.balance = balance

    # Withdraw money only when it is allowed
    def withdraw(self, amount):
        # Check whether the balance is enough
        if amount <= self.balance:
            # Reduce the balance only when valid
            self.balance -= amount
            return "Withdrawal successful"

        # Return a message if withdrawal is not allowed
        return "Insufficient balance"


# Create an account object
safe_account = SafeBankAccount(500)

# Test a valid withdrawal
print(safe_account.withdraw(200))

# Test an invalid withdrawal
print(safe_account.withdraw(400))

# Print the final balance
print(safe_account.balance)
```

# Concept 12: Mini Real-World Example

Let us combine attributes and methods into one useful object.

## Example goal
A shopping cart item should:
- store item name,
- store price,
- store quantity,
- calculate total cost.


```python
# Define a cart item class
class CartItem:
    # Initialize the object with item data
    def __init__(self, name, price, quantity):
        # Store the item name
        self.name = name

        # Store the item price
        self.price = price

        # Store the quantity
        self.quantity = quantity

    # Define a method to calculate total price
    def total_cost(self):
        # Return price multiplied by quantity
        return self.price * self.quantity


# Create one cart item object
item = CartItem("Keyboard", 1500, 2)

# Print the calculated total cost
print(item.total_cost())
```

# Concept 13: Common Mistakes and Edge Cases

## Common mistakes
- Forgetting `self`
- Calling methods on the class instead of the object
- Using a class attribute when an instance attribute is needed
- Forgetting required constructor arguments
- Allowing invalid object state

## Edge cases
- Empty strings as names
- Negative values for prices, marks, or balances
- Creating objects without all required data


```python
# Define a class that validates marks
class ExamResult:
    # Initialize the object with marks
    def __init__(self, marks):
        # Save the marks value
        self.marks = marks

    # Check whether marks are valid
    def is_valid(self):
        # Return True only for marks between 0 and 100
        return 0 <= self.marks <= 100


# Create a few test objects for edge cases
results = [ExamResult(95), ExamResult(-5), ExamResult(120)]

# Check the validity of each object
for result in results:
    print(result.marks, result.is_valid())
```

# Concept 14: Short-Hand and Good Style

## Good style tips
- Use clear class names like `Student`, `BankAccount`, `Product`
- Use clear method names like `deposit`, `withdraw`, `calculate_total`
- Keep each class focused on one responsibility
- Put validation where state changes happen

## Optimized thinking
OOP is not about making everything a class. Use a class when data and related actions naturally belong together.


```python
# Define a clean and focused student class
class StudentRecord:
    # Initialize object data
    def __init__(self, name, marks):
        # Save the student name
        self.name = name

        # Save the student marks
        self.marks = marks

    # Return the grade for this student
    def grade(self):
        # Decide the grade based on marks
        if self.marks >= 90:
            return "A"
        if self.marks >= 75:
            return "B"
        if self.marks >= 60:
            return "C"
        return "D"


# Create one student record object
record = StudentRecord("Anita", 84)

# Print the object's computed grade
print(record.name, record.grade())
```
