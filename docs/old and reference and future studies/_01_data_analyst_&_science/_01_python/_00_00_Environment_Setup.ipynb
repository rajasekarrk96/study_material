## 9. Virtual Environments

A virtual environment keeps packages separate for each project.

Why it is important:

- Avoid package conflicts.
- Keep projects clean.
- Make projects easier to share.
- Use different package versions in different projects.

### Create And Activate Virtual Environment

Create:

```bash
python -m venv .venv
```

Activate on Windows PowerShell:

```bash
.venv\Scripts\Activate.ps1
```

Activate on Windows Command Prompt:

```bash
.venv\Scripts\activate.bat
```

Deactivate:

```bash
deactivate
```

## 10. requirements.txt

A `requirements.txt` file stores project dependencies.

Create `requirements.txt`:

```bash
pip freeze > requirements.txt
```

Install packages from `requirements.txt`:

```bash
pip install -r requirements.txt
```

````python
# Example package list for data analyst learning
packages = [
    "numpy",
    "pandas",
    "matplotlib",
    "requests",
    "jupyter",
]

for package in packages:
    print(package)
````

## 11. Basic Python Project Folder Structure

```text

    src/
    outputs/
    requirements.txt
    README.md
```

Folder meaning:

- `data/` stores datasets.
- `notebooks/` stores Jupyter notebooks.
- `src/` stores reusable Python scripts.
- `outputs/` stores generated files, charts, and reports.
- `requirements.txt` stores dependencies.
- `README.md` explains the project.

## 13. Jupyter Notebook Workflow

A notebook contains cells. For data analyst and data science work, notebooks are useful because they can keep code, output, explanation, and observations in one place.

Main cell types:

- Code cell: runs Python code.
- Markdown cell: writes headings, notes, explanations, observations, and conclusions.

Common notebook actions:

- Run one cell.
- Run all cells.
- Restart kernel.
- Restart and run all.
- Clear output.
- Save notebook.

Good notebook habits:

- Use headings.
- Explain important code.
- Keep output clean.
- Run notebook from top to bottom before final saving.
- Do not leave broken cells.

````python
# Code cell example
x = 10
y = 20
result = x + y

result
````
