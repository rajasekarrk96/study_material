# Introduction

- To break big programs into smaller understandable parts.

# Basic Syntax Explanation

## Function syntax
- Use `def` to define a function.
- Write the function name after `def`.
- Put parameters inside parentheses.
- End the first line with a colon `:`.
- Indent the function body.
- Use `return` when you want to send a value back.

## Module syntax
- A module is simply a `.py` file.
- Use `import module_name` to use code from another file.
- Use `from module_name import item_name` to import a specific item.
- Use `if __name__ == "__main__":` to run code only when a file is executed directly.


```python
# Define a simple function named greet with one parameter called name
def greet(name):
    # Return a formatted greeting string to the caller
    return f"Hello, {name}!"


# Call the function and store the returned string in a variable
message = greet("Asha")

# Print the result so we can see what the function returned
print(message)
```

# Concept 1: Defining and Calling Functions

Functions start with `def`. A function does not run when it is defined. It runs only when it is called.

## Important ideas
- Function definition creates the reusable block.
- Function call executes the block.
- A good function usually has one clear responsibility.

## Common mistake
- Beginners often define a function and forget to call it.


```python
# Define a function named show_course
def show_course():
    # Print a message when the function is called
    print("Python Functions and Modules")


# Call the function for the first time
show_course()

# Call the same function again to show reusability
show_course()
```

# Concept 2: Parameters and Arguments

Parameters are variable names written in the function definition. Arguments are actual values passed during the function call.

## Types of arguments
- Positional arguments
- Keyword arguments

## Why this matters
It gives flexibility and makes functions reusable with different inputs.


```python
# Define a function with two parameters
def add_numbers(first_number, second_number):
    # Add the two incoming values and return the result
    return first_number + second_number


# Pass values by position
sum_one = add_numbers(10, 20)

# Pass values by keyword for readability
sum_two = add_numbers(second_number=5, first_number=7)

# Print both results
print(sum_one)
print(sum_two)
```

# Concept 3: Return Values

`return` sends a value back to the caller. If no `return` is written, Python returns `None`.

## Why return is important
- Returned values can be stored.
- Returned values can be reused in later calculations.
- Functions become more useful than functions that only print.

## Edge case
A `print()` inside a function shows output, but it does not automatically return that output.


```python
# Define a function that returns a multiplication result
def multiply(a, b):
    # Send the calculated result back to the caller
    return a * b


# Define a function with no explicit return statement
def just_print(text):
    # Print the incoming text only
    print(text)


# Store the returned multiplication result
product = multiply(6, 4)

# Print the returned result
print(product)

# Store the output of the second function
result = just_print("This is printed, not returned.")

# Print the stored value to show that Python returned None
print(result)
```

# Concept 4: Local Scope and Global Scope

Variables created inside a function are local to that function. Variables created outside are global.

## Why scope matters
- Prevents accidental variable conflicts.
- Keeps code predictable.
- Helps avoid bugs in large programs.

## Common mistake
Trying to use a local variable outside the function.


```python
# Create a global variable outside any function
course_name = "Python Mastery"


# Define a function to demonstrate local scope
def describe_course():
    # Create a local variable available only inside this function
    level = "Beginner to Advanced"

    # Access both the global and local variables inside the function
    print(course_name)
    print(level)


# Call the function so the local variable exists during execution
describe_course()

# Access the global variable outside the function
print(course_name)

# Do not try to print level here because it exists only inside the function
```

# Concept 5: Default Parameters

Default parameters let a function work even when the caller does not provide every value.

## Benefits
- Reduces repeated arguments.
- Makes function calls shorter.
- Helps create optional behavior.

## Common mistake
Using mutable objects like lists or dictionaries as default values can cause confusing bugs.


```python
# Define a function with a default parameter
def welcome(name, greeting="Hello"):
    # Build and return a greeting using the provided or default greeting
    return f"{greeting}, {name}!"


# Use the default greeting
message_one = welcome("Ravi")

# Override the default greeting
message_two = welcome("Ravi", "Good morning")

# Print both outputs
print(message_one)
print(message_two)
```

# Concept 6: Mutable Default Argument Edge Case

This is one of the most important function edge cases for beginners and interview preparation.

## Problem
If a list is used as a default parameter, that same list is reused across function calls.

## Correct pattern
Use `None` as the default and create a new list inside the function.


```python
# Incorrect example shown for learning purposes
def bad_append(item, bucket=[]):
    # Append the item to the same reused default list
    bucket.append(item)
    return bucket


# Correct approach using None
def safe_append(item, bucket=None):
    # Create a fresh list only when the caller did not pass one
    if bucket is None:
        bucket = []

    # Append the new item to the correct list
    bucket.append(item)
    return bucket


# Show the surprising behavior of the bad function
print(bad_append("a"))
print(bad_append("b"))

# Show the safe behavior of the corrected function
print(safe_append("a"))
print(safe_append("b"))
```

# Concept 7: Arbitrary Positional Arguments with `*args`

`*args` collects extra positional arguments into a tuple.

## When useful
- When the number of inputs can vary.
- When building helper utilities like sum, logging, or formatting tools.

## Operator involved
The `*` operator is used for packing positional arguments.


```python
# Define a function that can accept many positional numbers
def find_total(*numbers):
    # Start the running total at zero
    total = 0

    # Loop through every received value
    for number in numbers:
        # Add each number to the running total
        total += number

    # Return the final total
    return total


# Call the function with different counts of values
print(find_total(5, 10, 15))
print(find_total(1, 2, 3, 4, 5))
```

# Concept 8: Arbitrary Keyword Arguments with `**kwargs`

`**kwargs` collects extra keyword arguments into a dictionary.

## When useful
- Flexible configuration functions
- Logging utilities
- Metadata collection

## Operator involved
The `**` operator is used for packing keyword arguments.


```python
# Define a function that accepts any number of keyword arguments
def show_profile(**details):
    # Loop through the dictionary of received keyword arguments
    for key, value in details.items():
        # Print each key and value pair
        print(f"{key}: {value}")


# Pass multiple named values to the function
show_profile(name="Anita", role="Analyst", city="Chennai")
```

# Concept 9: Recursion

Recursion means a function calls itself.

## When useful
- Tree-like problems
- Divide-and-conquer logic
- Mathematical definitions such as factorial

## Important rule
There must be a base case, otherwise recursion continues forever and raises an error.


```python
# Define a recursive factorial function
def factorial(number):
    # Base case stops recursion when the number is 0 or 1
    if number <= 1:
        return 1

    # Recursive case multiplies the number by the factorial of the smaller value
    return number * factorial(number - 1)


# Call the recursive function and print the result
print(factorial(5))
```

# Concept 10: Lambda Functions

Lambda functions are small anonymous functions written in one line.

## Syntax idea
`lambda parameters: expression`

## Best use
- Short throwaway functions
- Sorting keys
- Small mapping/filtering operations

## Common mistake
Using lambda for complex multi-step logic makes code harder to read.


```python
# Create a small lambda function that squares a number
square = lambda value: value * value

# Use the lambda function directly
print(square(6))

# Use lambda as a sorting key for tuples
students = [("Riya", 88), ("Aman", 92), ("Sara", 85)]

# Sort students by marks in descending order
sorted_students = sorted(students, key=lambda item: item[1], reverse=True)

# Print the sorted result
print(sorted_students)
```

# Concept 11: Docstrings and Type Hints

Docstrings explain what a function does. Type hints show the expected types of parameters and return values.

## Why these help
- Improve readability
- Support editors and static tools
- Make teamwork easier

## Note
Type hints do not enforce types at runtime by themselves.


```python
# ================= FUNCTION WITH DOCSTRING =================

def calculate_discount(price: float, percent: float) -> float:
    """
    Return the discounted price after subtracting the given percentage.

    Parameters:
    price (float)   : Original price of the product
    percent (float) : Discount percentage

    Returns:
    float : Final price after applying discount
    """
    discount_amount = price * (percent / 100)
    return price - discount_amount


# ================= HOW TO VIEW DOCSTRING =================

# 1) Using __doc__ attribute (direct access)
print("Using __doc__:\n", calculate_discount.__doc__)

# 2) Using help() function (BEST & DETAILED)
print("\nUsing help():")
help(calculate_discount)

# 3) Using dir() to explore attributes
print("\nUsing dir():")
print(dir(calculate_discount))


# ================= CALL FUNCTION =================
print("\nFunction Output:")
print(calculate_discount(1000.0, 15.0))
```

# Concept 12: Higher-Order Functions

A higher-order function either takes another function as input or returns a function.

## Why it matters
- Encourages reusable logic
- Useful in callbacks, decorators, and data processing
- Common in advanced Python patterns


```python
# ================= HIGHER-ORDER FUNCTION (DETAILED) =================

# A function that takes another function as an argument
def apply_operation(value, operation):
    """
    apply_operation is a Higher-Order Function because:
    - It accepts another function (operation) as input
    - It calls that function inside

    Parameters:
    value      : input data (e.g., number)
    operation  : a function that will be applied to value

    Returns:
    Result of operation(value)
    """

    # IMPORTANT:
    # operation is NOT a value → it is a FUNCTION OBJECT (reference to function)
    # So we call it using operation(value)
    return operation(value)


# Normal function to pass as argument
def double(number):
    """
    Doubles the given number

    Parameter:
    number : numeric value

    Returns:
    number * 2
    """
    return number * 2


# ================= EXECUTION FLOW =================

# Step 1:
# apply_operation(12, double)
# value = 12
# operation = reference to function 'double'

# Step 2:
# Inside apply_operation → operation(value)
# → double(12)

# Step 3:
# double(12) returns 24

# Step 4:
# apply_operation returns 24

print("Result:", apply_operation(12, double))   # Output: 24


# ================= IMPORTANT CONCEPT =================

# ❌ WRONG (common mistake)
# print(apply_operation(12, double()))
# double() runs immediately → ERROR (missing argument)

# ✅ CORRECT
# Pass function WITHOUT parentheses → just reference
# apply_operation(12, double)


# ================= USING LAMBDA (ANONYMOUS FUNCTION) =================

# Lambda = small inline function
result_lambda = apply_operation(10, lambda x: x + 5)
print("Lambda Result:", result_lambda)   # Output: 15


# ================= MULTIPLE OPERATIONS =================

def square(n):
    return n * n

def cube(n):
    return n * n * n

print("Square:", apply_operation(4, square))  # 16
print("Cube:", apply_operation(3, cube))      # 27


# ================= REAL-WORLD THINKING =================

# You can dynamically change behavior
operations = [double, square, cube]

for op in operations:
    print(f"Applying {op.__name__} on 5 →", apply_operation(5, op))


# ================= SUMMARY =================
# 1. Functions are FIRST-CLASS OBJECTS in Python
# 2. You can:
#    - Pass functions as arguments
#    - Store functions in variables
#    - Return functions from other functions
# 3. apply_operation is a HIGHER-ORDER FUNCTION
# 4. operation(value) → calls the passed function dynamically
```

# Concept 13: Modules and Importing

A module is a Python file containing reusable code.

## Import styles
- `import math`
- `from math import sqrt`
- `import math as m`

## Real-world relevance
Projects split utilities, database code, and business logic into separate modules.


```python
# Import the built-in math module
import math

# Use a function from the imported module with dot notation
print(math.sqrt(81))

# Use a constant from the same module
print(math.pi)
```

# Concept 14: Selective Import and Aliasing

Selective import brings only what you need. Aliasing provides a shorter name.

## Common mistake
`from module import *` can pollute the namespace and make code harder to understand.


```python
# Import only the needed function and give it a short alias
from math import factorial as math_factorial

# Call the aliased function directly
print(math_factorial(6))
```

# Concept 15: The `__name__` Guard

When a Python file runs directly, `__name__` becomes `"__main__"`. When it is imported, `__name__` becomes the module name.

## Why used
- Lets a file act both as an importable module and a runnable script.
- Prevents test or demo code from running during import.


```python
# Check whether this file is being executed directly
if __name__ == "__main__":
    # This block runs only when the file is executed as the main program
    print("This file is running directly.")
else:
    # This block runs only when the file is imported from another file
    print("This file was imported.")
```

# Concept 17: Common Mistakes and Best Practices

## Common mistakes
- Forgetting `return`
- Using mutable default arguments
- Shadowing built-in names like `list`, `sum`, or `max`
- Importing everything with `*`
- Writing one huge function that does too many jobs

## Best practices
- Keep functions focused
- Use descriptive names
- Write docstrings for reusable code
- Group related functions into modules
- Test small functions with sample inputs


```python
# Avoid naming a variable `list` because that hides Python's built-in list type
numbers = [1, 2, 3]

# Use a clear function name that describes the job
def get_average(values):
    # Return zero safely when the list is empty to avoid division by zero
    if not values:
        return 0

    # Return the arithmetic mean of the values
    return sum(values) / len(values)


# Print the average of the example list
print(get_average(numbers))

# Print the safe result for an empty list edge case
print(get_average([]))
```

# Concept 18: Real-World and Mini Practical Examples

Functions and modules become powerful when applied to repeated tasks.

## Mini practical examples
- Salary calculator function
- Student grade function
- Utility module for text cleaning
- Report generator helpers

## Advanced usage direction
In larger systems, functions are combined with modules, packages, tests, and decorators for maintainable software design.


```python
# Define a reusable function to assign a grade based on marks
def assign_grade(score):
    # Return grade A for scores 90 and above
    if score >= 90:
        return "A"

    # Return grade B for scores from 75 to 89
    if score >= 75:
        return "B"

    # Return grade C for scores from 60 to 74
    if score >= 60:
        return "C"

    # Return grade D for lower scores
    return "D"


# Create a sample list of student scores
scores = [95, 82, 67, 48]

# Loop through each score and print its grade
for score in scores:
    print(score, assign_grade(score))
```
