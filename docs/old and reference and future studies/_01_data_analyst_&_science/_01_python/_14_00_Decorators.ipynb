# Introduction

- **What is it?** A decorator is a function that takes another function, adds some behavior, and returns a new function.
- **Why it is used?** Decorators help us reuse cross-cutting behavior such as logging, input validation, timing, caching, authentication, and permission checks without rewriting the same code in many places.
- **Real-world relevance:** Web frameworks use decorators for routes and permissions, testing libraries use decorators for fixtures and markers, and production code uses them for logging, retries, and caching.


# Basic Syntax Explanation

Basic decorator pattern:

```python
def decorator(function):
    def wrapper(*args, **kwargs):
        return function(*args, **kwargs)
    return wrapper

@decorator
def greet(name):
    return f"Hello, {name}"
```

Important parts:
- the original function is received as an argument
- a wrapper function usually runs before or after the original function
- `*args` and `**kwargs` let the wrapper accept any argument pattern
- `@decorator_name` is short for reassigning the function manually


````python
# Basic Example with detailed comments

# Define a simple decorator function.
def logger(function):
    # Define a wrapper that accepts any positional and keyword arguments.
    def wrapper(*args, **kwargs):
        # Print a message before the original function runs.
        print(f"Calling {function.__name__}")

        # Run the original function and store its result.
        result = function(*args, **kwargs)

        # Print a message after the original function finishes.
        print(f"Finished {function.__name__}")

        # Return the original result to the caller.
        return result

    # Return the new wrapped function.
    return wrapper


# Apply the decorator to a normal function.
@logger
def greet(name):
    # Return a greeting string.
    return f"Hello, {name}"


# Call the decorated function.
print(greet("Asha"))
````

# 3. Manual Decoration Without `@`

Before using `@decorator_name`, it helps to understand the manual version:

```python
greet = logger(greet)
```

This means the original function name now refers to the wrapped version.


````python
# Define a simple decorator.
def logger(function):
    # Define the wrapper.
    def wrapper(*args, **kwargs):
        print("Before function call")
        result = function(*args, **kwargs)
        print("After function call")
        return result

    # Return the wrapped function.
    return wrapper


# Define the original function.
def greet(name):
    # Return a greeting.
    return f"Hello, {name}"


# Decorate the function manually.
greet = logger(greet)

# Call the wrapped function.
print(greet("Ravi"))
````

# 4. Basic Decorator Structure

A standard decorator has:

1. an outer decorator function
2. an inner wrapper function
3. a call to the original function
4. a returned wrapper

This structure appears again and again in real code.


````python
# Define a decorator with the standard pattern.
def simple_decorator(function):
    # Define the wrapper around the original function.
    def wrapper():
        # Run extra behavior before the function.
        print("Start")

        # Call the original function.
        function()

        # Run extra behavior after the function.
        print("End")

    # Return the wrapper.
    return wrapper


# Decorate a no-argument function.
@simple_decorator
def show_message():
    # Print a message from the original function.
    print("Inside the function")


# Call the decorated function.
show_message()
````

# 5. Using `*args` and `**kwargs`

Wrappers should usually accept any arguments so the decorator works with many different functions.

Why useful:
- flexible decorator design
- supports unknown function signatures
- common best practice


````python
# Define a flexible decorator.
def logger(function):
    # Accept any arguments in the wrapper.
    def wrapper(*args, **kwargs):
        # Print the incoming arguments.
        print("Args:", args, "Kwargs:", kwargs)

        # Call the original function with the same arguments.
        return function(*args, **kwargs)

    # Return the wrapper.
    return wrapper


# Decorate a function with multiple parameters.
@logger
def introduce(name, city="Pune"):
    # Return a sentence using both parameters.
    return f"{name} lives in {city}"


# Call the decorated function.
print(introduce("Asha", city="Mumbai"))
````

# 6. Returning Results Correctly

A common decorator mistake is forgetting to return the original function result from the wrapper.

If the wrapper does not return the result, the caller may unexpectedly receive `None`.


````python
# Define a decorator that returns the original result correctly.
def logger(function):
    # Define the wrapper.
    def wrapper(*args, **kwargs):
        # Print a message before the call.
        print("Running function")

        # Save the original result.
        result = function(*args, **kwargs)

        # Return the saved result.
        return result

    # Return the wrapper.
    return wrapper


# Decorate a function that returns a value.
@logger
def square(number):
    # Return the square of the number.
    return number * number


# Print the decorated result.
print(square(5))
````

# 7. `functools.wraps`

Without `functools.wraps`, a decorator can hide the original function's name and docstring.

Why important:
- debugging
- introspection
- documentation
- testing
- framework integration


````python
# Import wraps from functools.
from functools import wraps


# Define a decorator that preserves metadata.
def logger(function):
    # Copy the original function metadata onto the wrapper.
    @wraps(function)
    def wrapper(*args, **kwargs):
        # Print the original function name.
        print(f"Running {function.__name__}")
        return function(*args, **kwargs)

    # Return the wrapper.
    return wrapper


# Decorate a function with a docstring.
@logger
def multiply(a, b):
    # Return the product of both numbers.
    return a * b


# Print preserved metadata.
print(multiply.__name__)
print(multiply(4, 5))
````

# 8. Logging Decorators

Logging decorators add visibility to function calls.

Real-world use:
- debug API functions
- trace batch jobs
- monitor utilities


````python
# Define a logging decorator.
def log_call(function):
    # Define the wrapper.
    def wrapper(*args, **kwargs):
        # Print the function name before execution.
        print(f"Calling {function.__name__}")

        # Run and return the original function.
        return function(*args, **kwargs)

    # Return the wrapper.
    return wrapper


# Decorate a function that processes data.
@log_call
def process_value(value):
    # Return a transformed value.
    return value * 10


# Print the decorated result.
print(process_value(7))
````

# 9. Validation Decorators

Validation decorators check inputs before the original function runs.

Why useful:
- reusable rules
- safer APIs
- cleaner business functions


````python
# Define a decorator that requires positive input.
def require_positive(function):
    # Define the wrapper.
    def wrapper(number):
        # Reject invalid input before calling the function.
        if number <= 0:
            raise ValueError("Number must be positive")

        # Call the original function for valid input.
        return function(number)

    # Return the wrapper.
    return wrapper


# Decorate a function that depends on positive input.
@require_positive
def square_root_like(number):
    # Return a placeholder computation.
    return number ** 0.5


# Call the decorated function.
print(square_root_like(9))
````

# 10. Timing Decorators

Timing decorators measure how long a function takes.

Real-world use:
- performance checks
- optimization work
- profiling experiments


````python
# Import the time module.
import time


# Define a timing decorator.
def timer(function):
    # Define the wrapper.
    def wrapper(*args, **kwargs):
        # Record the start time.
        start = time.time()

        # Run the original function.
        result = function(*args, **kwargs)

        # Record the end time.
        end = time.time()

        # Print the elapsed time.
        print("Elapsed:", round(end - start, 6), "seconds")

        # Return the original result.
        return result

    # Return the wrapper.
    return wrapper


# Decorate a sample function.
@timer
def compute_sum():
    # Return the sum of a short range.
    return sum(range(10000))


# Call the decorated function.
print(compute_sum())
````

# 11. Call Counting Decorators

A decorator can keep state by closing over a variable or dictionary.

Why useful:
- count usage
- rate limiting
- monitoring


````python
# Define a decorator that counts calls.
def count_calls(function):
    # Store call count in the outer scope.
    count = 0

    # Define the wrapper.
    def wrapper(*args, **kwargs):
        # Reuse the outer count variable.
        nonlocal count

        # Increase the count on every call.
        count += 1

        # Print the current call count.
        print("Call number:", count)

        # Run the original function.
        return function(*args, **kwargs)

    # Return the wrapper.
    return wrapper


# Decorate a simple function.
@count_calls
def greet(name):
    # Return a greeting string.
    return f"Hello, {name}"


# Call the decorated function multiple times.
print(greet("Asha"))
print(greet("Ravi"))
````

# 12. Parameterized Decorators

A parameterized decorator is a decorator that accepts its own arguments.

Structure:
- outer factory function receives decorator settings
- inner decorator receives the function
- wrapper runs the final logic


````python
# Define a decorator factory.
def repeat(times):
    # Define the real decorator.
    def decorator(function):
        # Define the wrapper.
        def wrapper(*args, **kwargs):
            # Store the last result from repeated calls.
            result = None
            for _ in range(times):
                result = function(*args, **kwargs)
            return result

        # Return the wrapper.
        return wrapper

    # Return the actual decorator.
    return decorator


# Apply a parameterized decorator.
@repeat(3)
def say_hi():
    # Print a short message.
    print("Hi")


# Call the decorated function.
say_hi()
````

# 13. Multiple Decorators

A function can have more than one decorator.

Important rule:
- decorators are applied from the bottom upward
- execution order can affect the result greatly


````python
# Define the first decorator.
def add_brackets(function):
    # Wrap the result in brackets.
    def wrapper(*args, **kwargs):
        return "[" + function(*args, **kwargs) + "]"
    return wrapper


# Define the second decorator.
def add_exclamation(function):
    # Add an exclamation mark to the result.
    def wrapper(*args, **kwargs):
        return function(*args, **kwargs) + "!"
    return wrapper


# Apply two decorators.
@add_brackets
@add_exclamation
def greet(name):
    # Return a greeting message.
    return f"Hello, {name}"


# Print the fully decorated result.
print(greet("Asha"))
````

````python
# ==========================================================
# DECORATORS DRY RUN (STEP-BY-STEP EXECUTION TRACE)
# ==========================================================
# Goal: Understand how multiple decorators work internally
# and how execution flows during a function call.

# ----------------------------------------------------------
# STEP 1: DEFINE FIRST DECORATOR
# ----------------------------------------------------------
def add_brackets(function):
    # 'function' = whatever function is passed to this decorator

    def wrapper(*args, **kwargs):
        # This wrapper will:
        # 1. Call the original function
        # 2. Add brackets [] around its result

        result = function(*args, **kwargs)   # call inner function
        return "[" + result + "]"            # modify result

    return wrapper   # decorator returns wrapper function


# ----------------------------------------------------------
# STEP 2: DEFINE SECOND DECORATOR
# ----------------------------------------------------------
def add_exclamation(function):
    # 'function' = whatever function is passed to this decorator

    def wrapper(*args, **kwargs):
        # This wrapper will:
        # 1. Call the original function
        # 2. Add "!" at the end

        result = function(*args, **kwargs)   # call inner function
        return result + "!"                  # modify result

    return wrapper   # return wrapper


# ----------------------------------------------------------
# STEP 3: APPLY DECORATORS
# ----------------------------------------------------------
# IMPORTANT:
# Decorators are applied BOTTOM → UP
#
# So this:
#   @add_brackets
#   @add_exclamation
#   def greet():
#
# becomes internally:
#   greet = add_brackets(add_exclamation(greet))

@add_brackets
@add_exclamation
def greet(name):
    return f"Hello, {name}"


# ----------------------------------------------------------
# STEP 4: INTERNAL STRUCTURE AFTER DECORATION
# ----------------------------------------------------------
# After decoration, greet is NOT the original function anymore
#
# Structure becomes:
#
# greet → wrapper₂ (from add_brackets)
#          ↓
#          wrapper₁ (from add_exclamation)
#             ↓
#             original greet function
#
# So calling greet() actually triggers wrapper₂


# ----------------------------------------------------------
# STEP 5: EXECUTION (DRY RUN)
# ----------------------------------------------------------
print("\n========== DRY RUN START ==========\n")

# Call function
print("Calling greet('Asha')...\n")

# Step-by-step flow:
# greet("Asha")
# → wrapper₂("Asha")   (add_brackets)
# → wrapper₁("Asha")   (add_exclamation)
# → original greet("Asha")

result = greet("Asha")

# ----------------------------------------------------------
# STEP 6: TRACE THE FLOW (MENTAL EXECUTION)
# ----------------------------------------------------------
# 1. wrapper₂ starts:
#       return "[" + wrapper₁("Asha") + "]"
#
# 2. wrapper₁ starts:
#       return original_greet("Asha") + "!"
#
# 3. original greet:
#       "Hello, Asha"
#
# 4. wrapper₁ result:
#       "Hello, Asha!"
#
# 5. wrapper₂ result:
#       "[Hello, Asha!]"


# ----------------------------------------------------------
# STEP 7: FINAL OUTPUT
# ----------------------------------------------------------
print("Final Output:", result)


# ==========================================================
# SUMMARY (VERY IMPORTANT)
# ==========================================================
# ✔ Decoration order  → Bottom to Top
#       add_exclamation → then add_brackets
#
# ✔ Execution order   → Top to Bottom
#       add_brackets → add_exclamation → original function
#
# ✔ Final Flow:
#       greet → brackets → exclamation → original
#
# ✔ Final Output:
#       [Hello, Asha!]
#
# ==========================================================
print("\n========== DRY RUN COMPLETE ==========")
````

# 14. Decorators for Access Control

Decorators can check permissions before allowing a function to run.

Real-world use:
- admin-only actions
- login checks
- role-based features


````python
# Define a decorator that checks admin access.
def require_admin(function):
    # Define the wrapper.
    def wrapper(user_role, *args, **kwargs):
        # Block access if the role is not admin.
        if user_role != "admin":
            return "Access denied"

        # Otherwise run the protected function.
        return function(user_role, *args, **kwargs)

    # Return the wrapper.
    return wrapper


# Protect an admin-only action.
@require_admin
def delete_report(user_role, report_name):
    # Return a success message.
    return f"{report_name} deleted"


# Print both outcomes.
print(delete_report("admin", "sales.csv"))
print(delete_report("guest", "sales.csv"))
````

# 15. Decorators and Exceptions

A decorator can catch exceptions or add custom error reporting.

Why useful:
- consistent error handling
- safer scripts
- user-friendly APIs


````python
# Define a decorator that catches errors.
def safe_run(function):
    # Define the wrapper.
    def wrapper(*args, **kwargs):
        try:
            # Try to run the original function.
            return function(*args, **kwargs)
        except Exception as error:
            # Return a readable error message.
            return f"Error: {error}"

    # Return the wrapper.
    return wrapper


# Decorate a risky function.
@safe_run
def divide(a, b):
    # Return the division result.
    return a / b


# Show both a valid case and an error case.
print(divide(10, 2))
print(divide(10, 0))
````

# 16. Caching with Decorators

Decorators can cache expensive repeated results.

Why useful:
- faster repeated calls
- recursive problems
- optimization of pure functions


````python
# Define a simple memoization decorator.
def memoize(function):
    # Store results in a dictionary cache.
    cache = {}

    # Define the wrapper.
    def wrapper(number):
        # Return the cached result if available.
        if number in cache:
            return cache[number]

        # Compute and store the missing result.
        cache[number] = function(number)

        # Return the stored result.
        return cache[number]

    # Return the wrapper.
    return wrapper


# Decorate a function with repeated work.
@memoize
def square(number):
    # Print when the real computation happens.
    print("Computing square for", number)
    return number * number


# Call the function twice with the same input.
print(square(5))
print(square(5))
````

# 17. Built-in Decorators

Python includes useful built-in decorators such as:

- `@property`
- `@staticmethod`
- `@classmethod`

These are also decorators, even though they feel different from custom wrappers.


````python
# Define a class with built-in decorators.
class Example:
    # Create a class-level value.
    label = "Python"

    # Define a static method.
    @staticmethod
    def add(a, b):
        # Return the sum of two numbers.
        return a + b

    # Define a class method.
    @classmethod
    def describe(cls):
        # Return a message using the class attribute.
        return f"Label is {cls.label}"


# Call both decorated methods.
print(Example.add(2, 3))
print(Example.describe())
````

# 18. Decorators on Methods

Decorators can wrap instance methods too.

Important detail:
- `self` still flows through the wrapper like any other argument


````python
# Define a decorator for methods.
def announce(function):
    # Define the wrapper.
    def wrapper(*args, **kwargs):
        # Print the method name.
        print("Running method:", function.__name__)
        return function(*args, **kwargs)
    return wrapper


# Define a class with a decorated method.
class Calculator:
    # Decorate the instance method.
    @announce
    def double(self, value):
        # Return twice the input value.
        return value * 2


# Create an instance and call the method.
calculator = Calculator()
print(calculator.double(9))
````

# 19. Common Mistakes

Common mistakes with decorators:

- forgetting to return the wrapper
- forgetting to return the original result
- not accepting `*args` and `**kwargs`
- losing metadata by skipping `wraps`
- misunderstanding multiple decorator order
- storing mutable shared state carelessly


````python
# Define a correct decorator.
def uppercase_result(function):
    # Define the wrapper.
    def wrapper(*args, **kwargs):
        # Return the uppercased original result.
        return function(*args, **kwargs).upper()
    return wrapper


# Decorate a function returning text.
@uppercase_result
def get_message():
    # Return a lowercase string.
    return "hello"


# Print the transformed result.
print(get_message())
````

# 20. Edge Cases

Important edge cases:

- decorating functions that return `None`
- decorating recursive functions
- using decorators with keyword-only arguments
- wrapping functions that raise exceptions
- applying decorators to methods and class methods


````python
# Define a decorator that handles functions returning None.
def logger(function):
    # Define the wrapper.
    def wrapper(*args, **kwargs):
        print("Calling function")
        result = function(*args, **kwargs)
        print("Returned:", result)
        return result
    return wrapper


# Decorate a function that returns None.
@logger
def save_data():
    # Print a side-effect message only.
    print("Data saved")


# Call the decorated function.
save_data()
````

# 22. Real-World Mini Practical Example

This example combines validation, logging, and timing in a small realistic workflow.

Goal:
- validate input
- log function usage
- measure execution time
- keep the original business logic small


````python
# Import time for timing.
import time


# Define a validation decorator.
def require_non_empty(function):
    # Define the wrapper.
    def wrapper(name):
        # Reject blank names after stripping spaces.
        if not name.strip():
            raise ValueError("Name cannot be empty")
        return function(name)
    return wrapper


# Define a logging decorator.
def log_call(function):
    # Define the wrapper.
    def wrapper(*args, **kwargs):
        print(f"Calling {function.__name__}")
        return function(*args, **kwargs)
    return wrapper


# Define a timing decorator.
def timer(function):
    # Define the wrapper.
    def wrapper(*args, **kwargs):
        start = time.time()
        result = function(*args, **kwargs)
        end = time.time()
        print("Elapsed:", round(end - start, 6), "seconds")
        return result
    return wrapper


# Apply all three decorators.
@timer
@log_call
@require_non_empty
def build_username(name):
    # Return a cleaned username.
    return name.strip().lower().replace(" ", "_")


# Call the decorated function.
print(build_username("Asha Sharma"))
````
