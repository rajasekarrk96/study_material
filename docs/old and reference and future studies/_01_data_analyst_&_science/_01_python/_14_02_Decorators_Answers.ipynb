## Q1. Simple Logger Decorator

**Problem:** Write a decorator that prints 'start' before a function call and 'end' after it. Apply it to a function returning a greeting and return the decorated result.

**Step-by-step logic:**
- Read the required input shape: `name: str`.

- Preserve or modify the original function behavior carefully.
- Handle validation, exceptions, and decorator order when needed.
- Return the final output in the required format `str`.
- Respect the constraint: The input name is a non-empty string.


````python
def simple_logger_decorator(name: str) -> str:
    # Define a decorator that prints before and after execution.
    def logger(function):
        # Create the wrapper for the original function.
        def wrapper(*args, **kwargs):
            # Print a message before the call.
            print("start")

            # Run the original function.
            result = function(*args, **kwargs)

            # Print a message after the call.
            print("end")

            # Return the original result.
            return result

        # Return the wrapped version.
        return wrapper

    # Apply the decorator to a greeting function.
    @logger
    def greet(student_name):
        # Return a greeting string.
        return f"Hello, {student_name}"

    # Return the decorated result.
    return greet(name)

# Time Complexity: depends on the number of calls, wrapper work, or loops used above.
# Space Complexity: depends on extra closure state, cache dictionaries, logs, or wrapper-created values.
````

````python
def username_sanitizer_decorators(name: str) -> str:
    # Define a decorator that rejects blank names.
    def require_non_empty(function):
        # Define the wrapper.
        def wrapper(value):
            # Reject blank input after stripping spaces.
            if not value.strip():
                raise ValueError("Blank name")

            # Run the original function.
            return function(value)

        # Return the wrapper.
        return wrapper

    # Define a decorator that lowercases the result.
    def lowercase_result(function):
        # Define the wrapper.
        def wrapper(*args, **kwargs):
            # Lowercase the original returned username.
            return function(*args, **kwargs).lower()

        # Return the wrapper.
        return wrapper

    # Apply both decorators.
    @lowercase_result
    @require_non_empty
    def build_username(value):
        # Build a username by stripping and replacing spaces.
        return value.strip().replace(" ", "_")

    # Return either the sanitized username or invalid marker.
    try:
        return build_username(name)
    except ValueError:
        return "Invalid"

# Time Complexity: depends on the number of calls, wrapper work, or loops used above.
# Space Complexity: depends on extra closure state, cache dictionaries, logs, or wrapper-created values.
````
