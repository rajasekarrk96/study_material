## Q15. Join Students and Marks

```python
# Problem Title: Join Students and Marks\n# Step-by-step logic, dry run notes, and test case are included below.\nimport sqlite3

connection = sqlite3.connect(':memory:')
cursor = connection.cursor()
cursor.execute('CREATE TABLE students (student_id INTEGER, name TEXT)')
cursor.execute('CREATE TABLE marks (student_id INTEGER, subject TEXT, score INTEGER)')
cursor.executemany('INSERT INTO students VALUES (?, ?)', [(1, 'Asha'), (2, 'Ravi')])
cursor.executemany('INSERT INTO marks VALUES (?, ?, ?)', [(1, 'Python', 92), (2, 'SQL', 88)])
connection.commit()
cursor.execute('''
    SELECT students.name, marks.subject, marks.score
    FROM students JOIN marks ON students.student_id = marks.student_id
    ORDER BY students.name
''')
print(cursor.fetchall())
cursor.close()
connection.close()

expected_output = [('Asha', 'Python', 92), ('Ravi', 'SQL', 88)]
print('Expected:', expected_output)
```

## Q18. Reusable Helper Function

```python
# Problem Title: Reusable Helper Function\n# Step-by-step logic, dry run notes, and test case are included below.\nimport sqlite3

def get_ready_connection():
    connection = sqlite3.connect(':memory:')
    cursor = connection.cursor()
    cursor.execute('CREATE TABLE IF NOT EXISTS students (name TEXT)')
    connection.commit()
    cursor.close()
    return connection

connection = get_ready_connection()
print('Database ready')
connection.close()

expected_output = 'Database ready'
print('Expected:', expected_output)
```

## Q19. Constraint Check Example

```python
# Problem Title: Constraint Check Example\n# Step-by-step logic, dry run notes, and test case are included below.\nimport sqlite3

connection = sqlite3.connect(':memory:')
cursor = connection.cursor()
cursor.execute('''
    CREATE TABLE employees (
        id INTEGER PRIMARY KEY,
        name TEXT NOT NULL,
        age INTEGER CHECK(age >= 18)
    )
''')
try:
    cursor.execute('INSERT INTO employees (name, age) VALUES (?, ?)', ('Neo', 16))
    connection.commit()
except sqlite3.IntegrityError:
    connection.rollback()
    print('Constraint prevented insert')
cursor.close()
connection.close()

expected_output = 'Constraint prevented insert'
print('Expected:', expected_output)
```

## Q22. SQLite Upsert

```python
# Problem Title: SQLite Upsert\n# Step-by-step logic, dry run notes, and test case are included below.\nimport sqlite3

connection = sqlite3.connect(':memory:')
cursor = connection.cursor()
cursor.execute('CREATE TABLE products (name TEXT PRIMARY KEY, price INTEGER)')
cursor.execute('INSERT INTO products VALUES (?, ?)', ('Pen', 15))
cursor.execute('''
    INSERT INTO products (name, price)
    VALUES (?, ?)
    ON CONFLICT(name) DO UPDATE SET price = excluded.price
''', ('Pen', 18))
connection.commit()
cursor.execute('SELECT name, price FROM products WHERE name = ?', ('Pen',))
print(cursor.fetchone())
cursor.close()
connection.close()

expected_output = ('Pen', 18)
print('Expected:', expected_output)
```

## Q23. Transfer Transaction

```python
# Problem Title: Transfer Transaction\n# Step-by-step logic, dry run notes, and test case are included below.\nimport sqlite3

connection = sqlite3.connect(':memory:')
cursor = connection.cursor()
cursor.execute('CREATE TABLE accounts (name TEXT PRIMARY KEY, balance INTEGER)')
cursor.executemany('INSERT INTO accounts VALUES (?, ?)', [('Asha', 1000), ('Bharat', 500)])
connection.commit()
try:
    cursor.execute('UPDATE accounts SET balance = balance - ? WHERE name = ?', (200, 'Asha'))
    cursor.execute('UPDATE accounts SET balance = balance + ? WHERE name = ?', (200, 'Bharat'))
    connection.commit()
except sqlite3.Error:
    connection.rollback()

cursor.execute('SELECT name, balance FROM accounts ORDER BY name')
print(cursor.fetchall())
cursor.close()
connection.close()

expected_output = [('Asha', 800), ('Bharat', 700)]
print('Expected:', expected_output)
```
