# Introduction

## What is it?
A **set** is an unordered collection of unique items.

A **dictionary** is a collection of key-value pairs where each key maps to one value.

## Why is it used?
- Sets are used when you care about uniqueness and fast membership testing.
- Dictionaries are used when you want to connect one piece of information to another.
- Both are core Python data structures used in almost every real program.

## Real-world relevance
- Removing duplicate email addresses from a signup list
- Storing student marks by roll number
- Counting how often words appear in a document
- Tracking unique users and their activity counts

# Basic Syntax Explanation

## Set syntax
- Create a set using curly braces like `{1, 2, 3}`
- Create an empty set using `set()`
- Remember that `{}` creates an empty dictionary, not an empty set

## Dictionary syntax
- Create a dictionary using curly braces and key-value pairs like `{'name': 'Asha', 'age': 21}`
- Create an empty dictionary using `{}` or `dict()`

## Important rules
- Set elements must be hashable
- Dictionary keys must be hashable
- Dictionary values can be almost any Python object

```python
# Create a set with duplicate values
numbers = {1, 2, 2, 3, 4, 4}

# Print the set to show that duplicates are removed automatically
print('Unique numbers in set:', numbers)

# Create a dictionary with key-value pairs
student = {'name': 'Ravi', 'age': 20, 'course': 'Python'}

# Print the dictionary to show its structure
print('Student dictionary:', student)

# Access a dictionary value using a key
print('Student name:', student['name'])
```

# Concept 1: Understanding Sets

A set stores only unique values.

## Basic concepts
- Duplicate items are removed automatically
- Sets are unordered, so positions are not fixed like lists
- Sets are mutable, which means you can add and remove elements
- You usually use sets for membership tests and duplicate removal

## Real-world relevance
If a website records the same visitor ID many times, a set can quickly tell you how many unique visitors there were.

```python
# Create a set from values that contain duplicates
fruits = {'apple', 'banana', 'apple', 'orange', 'banana'}

# Print the set to confirm duplicates do not remain
print('Unique fruits:', fruits)

# Check whether a specific fruit exists in the set
print('Is banana present?', 'banana' in fruits)

# Check a fruit that is not present
print('Is mango present?', 'mango' in fruits)
```

# Concept 2: Creating Sets in Different Ways

## All variations
- Set literal: `{10, 20, 30}`
- `set()` from a list, tuple, or string
- Set comprehension
- `frozenset()` for immutable sets

## Edge case
Calling `set('hello')` creates a set of unique characters, not a set containing the full word.

```python
# Create a set directly using curly braces
set_a = {10, 20, 30}

# Create a set from a list using set()
set_b = set([20, 30, 40, 40])

# Create an empty set correctly using set()
empty_set = set()

# Create a set from a string to get unique characters
letters = set('hello')

# Print each result
print('set_a:', set_a)
print('set_b:', set_b)
print('empty_set:', empty_set)
print('letters:', letters)
```

# Concept 3: Set Methods

## Important methods
- `add()` adds one element
- `update()` adds multiple elements
- `remove()` removes an element and raises an error if missing
- `discard()` removes an element safely
- `pop()` removes an arbitrary element
- `clear()` removes everything

## Common mistake
Using `remove()` for a value that is not present causes a `KeyError`.

```python
# Start with a basic set
colors = {'red', 'blue'}

# Add a single new value
colors.add('green')

# Add multiple values at once
colors.update(['yellow', 'black'])

# Remove a value that is definitely present
colors.remove('red')

# discard() does nothing if the value is absent
colors.discard('purple')

# Print the current state of the set
print('Colors after add and remove operations:', colors)

# pop() removes one arbitrary element from the set
removed_color = colors.pop()

# Print the removed value and remaining set
print('Removed color:', removed_color)
print('Remaining colors:', colors)
```

# Concept 4: Set Operators

## Operators involved
- Union: `|`
- Intersection: `&`
- Difference: `-`
- Symmetric difference: `^`

## Why they matter
These operators make comparison between groups of items short and readable.

```python
# Create two sets to compare
group_a = {1, 2, 3, 4}
group_b = {3, 4, 5, 6}

# Union combines all unique values from both sets
print('Union:', group_a | group_b)

# Intersection keeps only common values
print('Intersection:', group_a & group_b)

# Difference keeps values that are only in the first set
print('Difference A-B:', group_a - group_b)

# Symmetric difference keeps values that are in exactly one set
print('Symmetric difference:', group_a ^ group_b)
```

# Concept 5: Set Relationships

## Relation methods
- `issubset()` checks whether one set is fully contained in another
- `issuperset()` checks whether one set fully contains another
- `isdisjoint()` checks whether two sets have no common items

## Mini practical example
You can use subset checking to verify whether a user has all required permissions.

```python
# Create a complete skill set for a role
required_skills = {'python', 'sql', 'excel'}

# Create a candidate skill set
candidate_skills = {'python', 'sql'}

# Create another unrelated skill set
other_skills = {'java', 'c++'}

# Check if candidate skills are a subset of required skills
print('Subset check:', candidate_skills.issubset(required_skills))

# Check if required skills contain candidate skills
print('Superset check:', required_skills.issuperset(candidate_skills))

# Check whether two sets share no common values
print('Disjoint check:', required_skills.isdisjoint(other_skills))
```

# Concept 6: Frozen Sets

A `frozenset` is an immutable version of a set.

## Why it exists
- Normal sets cannot be dictionary keys because they are mutable
- `frozenset` can be used as a dictionary key or placed inside another set

## Advanced usage
Use `frozenset` when you want a group of values to stay fixed.

```python
# Create an immutable set
frozen_numbers = frozenset([1, 2, 3])

# Print the frozenset
print('Frozen set:', frozen_numbers)

# Use the frozenset as a dictionary key
lookup = {frozen_numbers: 'fixed group'}

# Access the dictionary value using the frozenset key
print('Value for frozenset key:', lookup[frozen_numbers])
```

# Concept 7: Set Comprehension

## Short-hand or optimized method
A set comprehension lets you create a set in one clean expression.

## Syntax
`{expression for item in iterable if condition}`

## When to use
- Filtering values
- Transforming values
- Keeping only unique transformed results

```python
# Create a list with repeated values
numbers = [1, 2, 2, 3, 4, 4, 5, 6]

# Build a set of squares of even numbers only
even_squares = {number ** 2 for number in numbers if number % 2 == 0}

# Print the result of the set comprehension
print('Even squares:', even_squares)
```

# Concept 8: Understanding Dictionaries

A dictionary stores data in **key-value** form.

## Basic concepts
- Keys are unique
- Values can repeat
- Keys must be hashable
- Dictionaries are mutable
- Dictionaries are excellent for lookup and mapping problems

## Real-world relevance
Use a dictionary to store product IDs and prices, usernames and passwords, or subject names and marks.

```python
# Create a dictionary containing product data
product = {
    'id': 101,
    'name': 'Keyboard',
    'price': 799,
    'in_stock': True
}

# Print the full dictionary
print('Product dictionary:', product)

# Access values using dictionary keys
print('Product name:', product['name'])
print('Product price:', product['price'])
```

# Concept 9: Creating Dictionaries in Different Ways

## All variations
- Dictionary literal with braces
- `dict()` with keyword arguments
- `dict()` from a list of tuples
- Dictionary comprehension

## Edge case
If you repeat a key, the later value replaces the earlier one.

```python
# Create a dictionary using literal syntax
person_1 = {'name': 'Amit', 'city': 'Delhi'}

# Create a dictionary using dict() with keyword arguments
person_2 = dict(name='Sara', city='Pune')

# Create a dictionary from a list of tuples
person_3 = dict([('name', 'Isha'), ('city', 'Jaipur')])

# Repeat a key to show overwrite behavior
duplicate_key_demo = {'x': 10, 'x': 20}

# Print all results
print('person_1:', person_1)
print('person_2:', person_2)
print('person_3:', person_3)
print('duplicate_key_demo:', duplicate_key_demo)
```

# Concept 10: Accessing and Updating Dictionary Data

## Common operations
- Access with `dictionary[key]`
- Safe access with `get()`
- Update with assignment
- Update many values with `update()`

## Common mistake
Trying to access a missing key with square brackets raises `KeyError`.

```python
# Create a sample employee dictionary
employee = {'name': 'Kiran', 'department': 'Sales', 'salary': 45000}

# Access an existing key directly
print('Employee name:', employee['name'])

# Access a possibly missing key safely with a default value
print('Bonus amount:', employee.get('bonus', 0))

# Update an existing value
employee['salary'] = 50000

# Add a new key-value pair
employee['city'] = 'Mumbai'

# Update multiple entries together
employee.update({'department': 'Marketing', 'active': True})

# Print the updated dictionary
print('Updated employee:', employee)
```

# Concept 11: Dictionary Methods

## Important methods
- `keys()` gives all keys
- `values()` gives all values
- `items()` gives all key-value pairs
- `pop()` removes a key and returns its value
- `popitem()` removes the last inserted key-value pair
- `clear()` removes everything
- `setdefault()` gets a value or creates a default entry

```python
# Create a dictionary for method demonstration
scores = {'math': 90, 'science': 85, 'english': 88}

# View the list of keys
print('Keys:', list(scores.keys()))

# View the list of values
print('Values:', list(scores.values()))

# View key-value pairs
print('Items:', list(scores.items()))

# Remove one key and capture its value
removed_score = scores.pop('science')
print('Removed score:', removed_score)

# Add a default key only if it does not exist already
scores.setdefault('history', 75)

# Print the final dictionary
print('Updated scores:', scores)
```

# Concept 12: Iterating Through Dictionaries

## Variations
- Loop through keys
- Loop through values
- Loop through keys and values together using `items()`

## Best practice
If you need both key and value, use `items()` because it is clear and efficient.

```python
# Create a dictionary of student marks
student_marks = {'Asha': 91, 'Rohan': 87, 'Meera': 95}

# Loop through keys only
for name in student_marks:
    # Print each student name
    print('Student name:', name)

# Loop through values only
for mark in student_marks.values():
    # Print each student mark
    print('Student mark:', mark)

# Loop through keys and values together
for name, mark in student_marks.items():
    # Print a readable sentence for each pair
    print(name, 'scored', mark)
```

# Concept 13: Dictionary Comprehension

## Short-hand or optimized method
Dictionary comprehension creates dictionaries in a compact and readable way.

## Syntax
`{key_expression: value_expression for item in iterable if condition}`

## Mini practical use
Create quick lookup tables, filtered maps, or transformed records.

```python
# Create a list of numbers
numbers = [1, 2, 3, 4, 5]

# Create a dictionary where each number maps to its square
square_map = {number: number ** 2 for number in numbers}

# Create another dictionary for even numbers only
even_square_map = {number: number ** 2 for number in numbers if number % 2 == 0}

# Print both dictionaries
print('Square map:', square_map)
print('Even square map:', even_square_map)
```

# Concept 14: Nested Structures

## Advanced usage
Dictionaries often contain other dictionaries, lists, or sets as values.

## Real-world example
A student record may store name, marks, and enrolled subjects in one nested structure.

## Important note
Sets can contain immutable values, but a normal set cannot directly contain another normal set.

```python
# Create a nested dictionary for a student record
student_record = {
    'name': 'Diya',
    'marks': {'math': 92, 'science': 89},
    'subjects': {'Math', 'Science', 'English'}
}

# Access a nested dictionary value
print('Math marks:', student_record['marks']['math'])

# Access a nested set value
print('Subjects:', student_record['subjects'])
```

# Concept 15: Edge Cases and Common Mistakes

## Set edge cases
- `{}` is an empty dictionary, not an empty set
- Lists cannot be set elements because lists are mutable
- Set order should not be relied upon

## Dictionary edge cases
- Missing keys raise `KeyError` when accessed with square brackets
- Lists cannot be used as dictionary keys
- Repeated keys overwrite older values

## Why this matters
Many beginner bugs in Python come from these exact issues.

```python
# Create {} and inspect its type
empty_structure = {}

# Print the type to show that {} is a dictionary
print('Type of {}:', type(empty_structure))

# Use get() to avoid KeyError on a missing key
user = {'username': 'admin'}
print('Password value:', user.get('password', 'Not Set'))

# Show that duplicate dictionary keys keep only the last value
sample = {'a': 1, 'a': 99}
print('Duplicate key result:', sample)
```

# Concept 16: Real-World Example Using Sets

Imagine an event registration system where the same person may register more than once.

A set helps us find unique attendees quickly.

```python
# Create a list of registration names with repeats
registrations = ['Anu', 'Rahul', 'Anu', 'Vikas', 'Rahul', 'Pooja']

# Convert the list to a set to remove duplicates
unique_attendees = set(registrations)

# Print the unique attendee names
print('Unique attendees:', unique_attendees)

# Print the number of unique attendees
print('Total unique attendees:', len(unique_attendees))
```

# Concept 17: Real-World Example Using Dictionaries

Suppose you want to count how many times each word appears in a sentence.

A dictionary is perfect because the word becomes the key and the count becomes the value.

```python
# Store a sentence in a string variable
sentence = 'python is easy and python is powerful'

# Split the sentence into individual words
words = sentence.split()

# Create an empty dictionary for counting words
word_count = {}

# Traverse each word in the list
for word in words:
    # Increase the old count if the word exists, otherwise start from 0 and add 1
    word_count[word] = word_count.get(word, 0) + 1

# Print the completed frequency dictionary
print('Word counts:', word_count)
```

# Concept 18: Mini Practical Example Combining Sets and Dictionaries

In many real applications, sets and dictionaries work together.

## Example idea
- Use a set to track unique users
- Use a dictionary to track how many times each user logged in

```python
# Store repeated login events in a list
logins = ['amit', 'neha', 'amit', 'amit', 'rohan', 'neha']

# Create an empty set for unique users
unique_users = set()

# Create an empty dictionary for login counts
login_counts = {}

# Process each login event one by one
for username in logins:
    # Add the username to the unique users set
    unique_users.add(username)

    # Update the user's login count in the dictionary
    login_counts[username] = login_counts.get(username, 0) + 1

# Print the set of unique users
print('Unique users:', unique_users)

# Print the dictionary of login counts
print('Login counts:', login_counts)
```

# Concept 19: Practical Thinking and Performance

## Why these structures are popular
- Set membership checking is usually fast
- Dictionary key lookup is usually fast
- Both structures make code expressive and clean

## Decision guide
- Use a list when order matters and duplicates matter
- Use a set when uniqueness matters
- Use a dictionary when mapping or counting matters

```python
# Create a list and set with similar values
numbers_list = [10, 20, 30, 40, 50]
numbers_set = {10, 20, 30, 40, 50}

# Check membership in the list
print('30 in list:', 30 in numbers_list)

# Check membership in the set
print('30 in set:', 30 in numbers_set)

# Create a simple lookup dictionary
price_map = {'pen': 10, 'book': 50, 'bag': 500}

# Access a value directly by its key
print('Price of book:', price_map['book'])
```

# Summary

## Sets
- Store unique values
- Support union, intersection, difference, and membership testing
- Are excellent for removing duplicates and comparing groups

## Dictionaries
- Store key-value mappings
- Are excellent for lookup, counting, grouping, and structured records
- Support clear and powerful real-world data modeling

## Final idea
When solving a Python problem, ask yourself:
- Do I need uniqueness? Use a set.
- Do I need mapping or counting? Use a dictionary.
