# Introduction

- **Why it is used?** Modules help organize large programs, reduce repetition, improve readability, and support reuse across files and projects.
- **Real-world relevance:** Real applications use modules for utilities, database access, APIs, file handling, date logic, data formats, command-line scripts, and automation. The standard library provides many ready-made tools so we do not need to write everything from scratch.

# 3. Basic Import Syntax

The most common import style is:

```python
import module_name
```

Then we call module content using dot notation:

```python
module_name.function_name()
```


````python
# Import the random module.
import random

# Choose one random value from a list.
print(random.choice(["python", "sql", "excel"]))
````

# 4. Selective Import

Another style is:

```python
from module_name import item_name
```

Why useful:
- shorter calls
- direct access to a specific function or class

Common caution:
- too many direct imports can reduce clarity


````python
# Import only sqrt from the math module.
from math import sqrt

# Call the imported function directly.
print(sqrt(81))
````

# 5. Import Aliases

We can shorten long module names with aliases.

Example:
```python
import datetime as dt
```

Why useful:
- shorter code
- clearer repeated usage
- common convention in projects


````python
# Import datetime with a shorter alias.
import datetime as dt

# Print today's date using the alias.
print(dt.date.today())
````

# 6. Why `from module import *` Is Usually Avoided

Wildcard imports can make code harder to read and may cause naming conflicts.

Common issues:
- unclear origin of names
- overwritten names
- lower readability

Best practice:
- prefer explicit imports


````python
# Use an explicit import instead of wildcard import.
from math import ceil, floor

# Print both values clearly.
print(ceil(3.2))
print(floor(3.8))
````

# 9. The `datetime` Module

`datetime` helps with dates, times, timestamps, and formatting.

Common tools:
- `date`
- `datetime`
- `timedelta`
- `strftime`
- `strptime`

Real-world use:
- reports

````python
# Import date and timedelta.
from datetime import date, timedelta

# Get today's date.
today = date.today()

# Add seven days to today's date.
next_week = today + timedelta(days=7)

# Print both values.
print(today)
print(next_week)
````

# 10. Formatting and Parsing Dates

Date formatting converts a date or datetime object to a string.

Parsing converts a string to a datetime object.

Why useful:
- user input
- reports
- API data
- logs


````python
# Import datetime for formatting and parsing.
from datetime import datetime

# Parse a date-time string.
value = datetime.strptime("2026-04-27 15:30", "%Y-%m-%d %H:%M")

# Format it in another style.
print(value.strftime("%d/%m/%Y %I:%M %p"))
````

# 14. The `json` Module

`json` converts between Python objects and JSON text.

Common tools:
- `json.dumps`
- `json.loads`
- `json.dump`
- `json.load`

Real-world use:
- APIs
- config files
- data exchange


````python
# Import the json module.
import json

# Create a Python dictionary.
student = {"name": "Asha", "score": 95}

# Convert the dictionary to a JSON string.
json_text = json.dumps(student)

# Convert the JSON string back to a dictionary.
restored = json.loads(json_text)

# Print both results.
print(json_text)
print(restored)
````

# 15. The `collections` Module

`collections` provides specialized container types.

Very common tools:
- `Counter`
- `defaultdict`
- `deque`
- `namedtuple`

Why useful:
- cleaner code
- fewer manual checks
- better fit for common patterns


````python
# Import Counter and defaultdict.
from collections import Counter, defaultdict

# Count word frequencies quickly.
counts = Counter(["python", "sql", "python", "excel"])

# Build grouped values automatically.
groups = defaultdict(list)
groups["passed"].append("Asha")
groups["passed"].append("Ravi")

# Print both structures.
print(counts)
print(dict(groups))
````

# 25. Real-World Mini Practical Example

This example combines several standard library tools.

Goal:
- build a path
- create JSON report data
- add a timestamp
- count labels
- save output as formatted JSON


````python
# Import required standard library tools.
from pathlib import Path
from datetime import datetime
from collections import Counter
import json

# Create source records.
records = ["python", "sql", "python", "excel"]

# Count label frequencies.
counts = Counter(records)

# Build a report dictionary with a timestamp.
report = {
    "created_at": datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
    "counts": dict(counts),
}

# Create a local output path.
output_path = Path("standard_library_report.json")

# Save the report as formatted JSON.
output_path.write_text(json.dumps(report, indent=2), encoding="utf-8")

# Read the saved file back and print it.
print(output_path.read_text(encoding="utf-8"))
````
