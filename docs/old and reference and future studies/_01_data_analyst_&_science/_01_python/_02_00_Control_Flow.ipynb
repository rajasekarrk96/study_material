# Introduction

- **What is it?** Control flow is the order in which Python executes statements.
- **Why it is used?** Programs need decisions and repetition. Control flow lets code react to input, validate data, search collections, retry operations, and stop when a goal is reached.
- **Real-world relevance:** Login checks, payment validation, grading systems, retrying API calls, menu-driven apps, fraud rules, file processing, and interview algorithms all depend on control flow.

- **Boolean thinking:** `True`, `False`, truthy values, falsy values

# Basic Syntax Explanation

```python
if condition:
    # runs when condition is True
elif another_condition:
    # runs when first condition is False and this one is True
else:
    # runs when all previous conditions are False

for item in iterable:
    # repeats once for every item

while condition:
    # repeats while condition stays True
```

Python uses **indentation** to define blocks. Usually, use 4 spaces.


````python
# Basic Example with detailed comments

# Store marks in a variable so the program can make a decision.
marks = 78

# Check the highest category first.
if marks >= 90:
    # This line runs only when marks is 90 or above.
    grade = "A"
elif marks >= 70:
    # This line runs when marks is below 90 but at least 70.
    grade = "B"
elif marks >= 50:
    # This line runs when marks is below 70 but at least 50.
    grade = "C"
else:
    # This line runs when all earlier conditions are False.
    grade = "Fail"

# Print the final decision.
print("Marks:", marks)
print("Grade:", grade)
````

# 1. Boolean Conditions

A condition is any expression Python can treat as `True` or `False`.


````python
# A comparison creates a Boolean value.
age = 20

# Store the result of the comparison.
is_adult = age >= 18

# Print both value and type to understand the result.
print(is_adult)
print(type(is_adult))
````

# 2. Comparison Operators

`==`, `!=`, `>`, `<`, `>=`, and `<=` compare values and return Booleans.


````python
# Create sample values.
a = 10
b = 3

# Each expression below returns True or False.
print("a == b:", a == b)
print("a != b:", a != b)
print("a > b:", a > b)
print("a <= b:", a <= b)
````

# 3. Logical Operators

`and` requires all conditions, `or` requires at least one, and `not` reverses a Boolean.


````python
# Define three facts about a user.
age = 21
has_id = True
is_banned = False

# User is allowed only when all required checks pass.
can_enter = age >= 18 and has_id and not is_banned

# Print the decision.
print("Can enter:", can_enter)
````

# 4. Membership Operators

`in` and `not in` check whether a value exists inside a collection.


````python
# Store allowed roles in a list.
allowed_roles = ["admin", "manager", "analyst"]
user_role = "analyst"

# Check membership before granting access.
if user_role in allowed_roles:
    print("Access granted")
else:
    print("Access denied")
````

# 5. Identity Operators

`is` checks object identity. Use it mainly with `None`, not for normal value comparison.


````python
# None means no value is available.
result = None

# Correct way to check None.
if result is None:
    print("No result yet")

# For normal values, use == instead of is.
print(1000 == int("1000"))
````

# 6. if Statement

`if` runs a block only when the condition is True.


````python
# Define temperature.
temperature = 35

# Run this block only for hot weather.
if temperature > 30:
    print("Drink water")

# This line is outside the if block, so it always runs.
print("Weather check completed")
````

# 7. if else Statement

`else` gives a fallback path when the `if` condition is False.


````python
# Define a number.
number = 7

# Decide even or odd.
if number % 2 == 0:
    print("Even")
else:
    print("Odd")
````

# 8. if elif else Chain

`elif` is used when there are multiple exclusive choices.


````python
# Define order total.
amount = 1250

# Apply discount based on amount range.
if amount >= 5000:
    discount = 0.20
elif amount >= 1000:
    discount = 0.10
else:
    discount = 0.0

print("Discount:", discount)
````

# 9. Nested Conditions

A nested condition is an `if` inside another `if`. Use carefully to avoid messy code.


````python
# Define user state.
is_logged_in = True
is_admin = False

# First check login.
if is_logged_in:
    # This condition runs only when user is logged in.
    if is_admin:
        print("Open admin dashboard")
    else:
        print("Open user dashboard")
else:
    print("Show login page")
````

# 10. Guard Clauses

Guard clauses return early when input is invalid, making code easier to read.


````python
def divide(a, b):
    # Stop early if denominator is zero.
    if b == 0:
        return "Cannot divide by zero"

    # Main logic runs only after validation passes.
    return a / b


print(divide(10, 2))
print(divide(10, 0))
````

# 11. Short-hand if

A conditional expression writes simple if-else logic in one line.


````python
# Define age.
age = 16

# One-line decision: value_if_true if condition else value_if_false.
status = "adult" if age >= 18 else "minor"

print(status)
````

# 12. match case

`match case` handles structured choices clearly in Python 3.10+.


````python
# Define command from a menu.
command = "start"

# Match the command against possible cases.
match command:
    case "start":
        print("Starting process")
    case "stop":
        print("Stopping process")
    case "pause":
        print("Pausing process")
    case _:
        print("Unknown command")
````

# 13. for Loop

`for` repeats over each item in an iterable such as list, tuple, string, range, or dictionary.


````python
# Store names in a list.
names = ["Asha", "Ravi", "Meera"]

# Loop once for each name.
for name in names:
    # Current item is available in the variable name.
    print("Hello", name)
````

# 14. range

`range(start, stop, step)` generates numbers lazily and stops before `stop`.


````python
# Print odd numbers from 1 to 9.
for number in range(1, 10, 2):
    print(number)

# Convert range to list only when you need to see all values.
print(list(range(5)))
````

# 15. enumerate

`enumerate` gives both index and value while looping.


````python
# Store tasks in order.
tasks = ["read", "practice", "revise"]

# start=1 makes numbering human-friendly.
for position, task in enumerate(tasks, start=1):
    print(position, task)
````

# 16. Looping Dictionaries

Use `.items()` when you need both key and value.


````python
# Store marks by subject.
marks = {"math": 90, "python": 95, "sql": 85}

# Loop over key-value pairs.
for subject, score in marks.items():
    print(subject, "->", score)
````

# 17. while Loop

`while` repeats as long as its condition remains True.


````python
# Start counter.
count = 1

# Continue until count becomes greater than 5.
while count <= 5:
    print(count)
    # Update counter to avoid infinite loop.
    count += 1
````

# 18. break

`break` exits the nearest loop immediately.


````python
# Search for first number divisible by 7.
numbers = [4, 8, 13, 21, 28]

for number in numbers:
    if number % 7 == 0:
        print("First match:", number)
        break
````

# 19. continue

`continue` skips the rest of the current loop iteration.


````python
# Print only positive values.
values = [3, -1, 0, 7, -5]

for value in values:
    if value <= 0:
        continue
    print("Positive:", value)
````

# 20. pass

`pass` is a placeholder when syntax requires a block but logic is not ready yet.


````python
# Create a function skeleton.
def future_feature():
    # pass prevents IndentationError.
    pass

print("Function created successfully")
````

# 21. Loop else

A loop `else` block runs only if the loop did not end by `break`.


````python
# Search for a target.
target = 99
numbers = [10, 20, 30]

for number in numbers:
    if number == target:
        print("Found")
        break
else:
    print("Target not found")
````

# 22. Truthy and Falsy Values

Empty containers, zero, `None`, and `False` are falsy. Most other values are truthy.


````python
# Try changing this list.
items = []

# Empty list is treated as False.
if items:
    print("List has data")
else:
    print("List is empty")
````

# 23. Common Mistakes

Common mistakes include using `=` instead of `==`, wrong indentation, missing updates in while loops, and comparing with `is` incorrectly.


````python
# Correct comparison uses ==.
password = "python123"

if password == "python123":
    print("Correct password")

# Correct while loop updates the counter.
i = 0
while i < 3:
    print(i)
    i += 1
````

# 24. Edge Cases

Always test empty input, boundary values, single items, duplicate values, and invalid-looking data.


````python
def first_even(numbers):
    # Empty list should not crash.
    for number in numbers:
        if number % 2 == 0:
            return number
    return None


print(first_even([]))
print(first_even([1, 3, 5]))
print(first_even([1, 4, 6]))
````

# 25. Optimized any and all

`any()` checks whether at least one condition is True. `all()` checks whether every condition is True.


````python
# Store user answers.
scores = [80, 75, 90]

# all returns True only if every score passes.
all_passed = all(score >= 50 for score in scores)

# any returns True if at least one score is excellent.
has_excellent = any(score >= 90 for score in scores)

print(all_passed)
print(has_excellent)
````

# 26. Real-world Validation Example

Control flow is often used to validate form input before saving data.


````python
def validate_username(username):
    # Username must not be empty after removing spaces.
    if not username.strip():
        return "Username required"

    # Username must be long enough.
    if len(username) < 3:
        return "Username too short"

    # Username must contain only letters, digits, or underscore.
    if not username.replace("_", "").isalnum():
        return "Invalid characters"

    return "Valid username"


print(validate_username("raja_123"))
````

# 27. Mini Practical: ATM Menu

Menu-based programs use conditions to choose actions.


````python
# Store balance and selected option.
balance = 1000
option = "withdraw"
amount = 300

if option == "check":
    print("Balance:", balance)
elif option == "withdraw":
    if amount <= balance:
        balance -= amount
        print("Withdrawn:", amount)
        print("New balance:", balance)
    else:
        print("Insufficient balance")
else:
    print("Invalid option")
````
