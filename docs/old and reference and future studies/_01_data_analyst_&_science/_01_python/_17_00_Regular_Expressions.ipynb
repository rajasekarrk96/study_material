# Introduction

- **What is it?** A regular expression is a pattern used to search, validate, extract, replace, or split text.
- **Why it is used?** Regex helps solve many text-processing tasks quickly, such as checking email formats, extracting numbers, finding keywords, and cleaning logs.
- **Real-world relevance:** Data cleaning, log parsing, CSV cleanup, validation forms, scraping support, API payload cleanup, and text mining often rely on regular expressions.


# Basic Syntax Explanation

Core regex ideas:

```python
import re
match = re.search(r"\d+", "Room 12")
```

Important symbols:
- `.` any character except newline in basic mode
- `\d` digit
- `\w` word character
- `\s` whitespace
- `*`, `+`, `?`, `{m,n}` quantifiers
- `^` start anchor
- `$` end anchor
- `(...)` capturing group
- `|` alternation
- `[...]` character class

Important best practice:
- use raw strings like `r"\d+"` for regex patterns


````python
# Basic Example with detailed comments

# Import the regular expression module.
import re

# Store a sample string with numbers in it.
text = "Room 12, Floor 5"

# Search for the first group of digits.
match = re.search(r"\d+", text)

# Print the full matched text if a match exists.
print(match.group())

# Find all groups of digits in the string.
all_numbers = re.findall(r"\d+", text)

# Print every matched number.
print(all_numbers)
````

# 1. Why Use Regular Expressions

Regex is useful when plain string methods become too repetitive or too limited.

Good use cases:
- validation
- extraction
- replacements
- tokenization
- pattern-based cleanup


````python
# Import re for regex operations.
import re

# Find all words starting with 'p'.
print(re.findall(r"p\w*", "python pandas sql powerbi"))
````

# 2. Raw Strings

Regex patterns often use backslashes, so raw strings are strongly recommended.

Why useful:
- avoids double escaping
- makes patterns easier to read

Example:
- better: `r"\d+"`
- harder: `"\\d+"`


````python
# Import re for the demo.
import re

# Use a raw string pattern to find digits.
print(re.findall(r"\d+", "A1 B22 C333"))
````

# 3. `re.search()`

`search()` looks anywhere in the string and returns the first match object.

Why useful:
- quick presence checks
- first extraction target


````python
# Import re for searching text.
import re

# Search for the first number in the string.
match = re.search(r"\d+", "Invoice 450 due")

# Print the matched text.
print(match.group())
````

# 4. `re.match()`

`match()` checks only from the start of the string.

Why useful:
- prefix validation
- checking how a string begins


````python
# Import re for matching text.
import re

# Try matching digits only at the start.
print(re.match(r"\d+", "123abc").group())
print(re.match(r"\d+", "abc123"))
````

# 5. `re.fullmatch()`

`fullmatch()` requires the entire string to match the pattern.

Why useful:
- validation tasks
- checking complete formats


````python
# Import re for full-string validation.
import re

# Validate whether the full text is digits only.
print(bool(re.fullmatch(r"\d+", "12345")))
print(bool(re.fullmatch(r"\d+", "123a5")))
````

# 6. `re.findall()`

`findall()` returns all matches as a list.

Why useful:
- extraction
- quick reporting
- data cleaning


````python
# Import re for extracting all matches.
import re

# Find all email-like names ending in .com.
print(re.findall(r"\b\w+@\w+\.com\b", "a@test.com b@data.com c@test.org"))
````

# 7. `re.finditer()`

`finditer()` returns match objects one by one.

Why useful:
- large text processing
- access to match positions
- richer inspection than `findall()`


````python
# Import re for iterator-based matching.
import re

# Iterate through all digit matches with positions.
for match in re.finditer(r"\d+", "A1 B22 C333"):
    print(match.group(), match.start(), match.end())
````

# 8. Character Classes

Character classes use square brackets to match one character from a set.

Examples:
- `[abc]`
- `[A-Z]`
- `[0-9]`
- `[^aeiou]`

Why useful:
- flexible validation and extraction


````python
# Import re for character-class matching.
import re

# Find all lowercase vowels.
print(re.findall(r"[aeiou]", "education"))
````

# 9. Predefined Character Classes

Useful predefined shortcuts:

- `\d` digits
- `\D` non-digits
- `\w` word characters
- `\W` non-word characters
- `\s` whitespace
- `\S` non-whitespace


````python
# Import re for shortcut class matching.
import re

# Find words and digit groups separately.
print(re.findall(r"\w+", "Room 12, Floor 5"))
print(re.findall(r"\d+", "Room 12, Floor 5"))
````

# 10. Quantifiers

Quantifiers control how many times a pattern may repeat.

Common quantifiers:
- `*` zero or more
- `+` one or more
- `?` zero or one
- `{m}`
- `{m,n}`


````python
# Import re for quantifier examples.
import re

# Match one or more digits.
print(re.findall(r"\d+", "a1 b22 c333"))

# Match exactly two digits.
print(re.findall(r"\d{2}", "a1 b22 c333"))
````

# 11. Anchors

Anchors match positions rather than characters.

Important anchors:
- `^` start of string
- `$` end of string

Why useful:
- format validation
- line-aware checks


````python
# Import re for anchor examples.
import re

# Check whether the text starts with Python.
print(bool(re.search(r"^Python", "Python basics")))

# Check whether the text ends with .txt.
print(bool(re.search(r"\.txt$", "notes.txt")))
````

# 12. Alternation

Alternation with `|` means "or".

Why useful:
- multiple valid keywords
- multiple formats
- compact branching in text patterns


````python
# Import re for alternation.
import re

# Match either cat or dog.
print(re.findall(r"cat|dog", "cat bird dog fish cat"))
````

# 13. Capturing Groups

Parentheses create capturing groups.

Why useful:
- extract subparts
- reuse parts in substitutions
- structured parsing


````python
# Import re for grouped extraction.
import re

# Capture username and domain separately.
match = re.search(r"(\w+)@(\w+)\.com", "asha@test.com")

# Print both captured groups.
print(match.group(1))
print(match.group(2))
````

# 14. Named Groups

Named groups make patterns easier to read.

Syntax:
```python
(?P<name>pattern)
```


````python
# Import re for named-group matching.
import re

# Capture parts of the email with names.
match = re.search(r"(?P<user>\w+)@(?P<domain>\w+)\.com", "asha@test.com")

# Print values using group names.
print(match.group("user"))
print(match.group("domain"))
````

# 15. Greedy vs Non-Greedy

Regex is greedy by default, meaning it tries to match as much as possible.

Non-greedy versions use `?` after the quantifier.

Examples:
- greedy: `.*`
- non-greedy: `.*?`


````python
# Import re for greedy matching examples.
import re

# Compare greedy and non-greedy matching.
text = "<b>one</b><b>two</b>"
print(re.findall(r"<b>.*</b>", text))
print(re.findall(r"<b>.*?</b>", text))
````

# 16. `re.sub()`

`sub()` replaces matched patterns with new text.

Why useful:
- cleaning
- masking
- normalization


````python
# Import re for substitutions.
import re

# Replace all digit groups with X.
print(re.sub(r"\d+", "X", "Order 123 shipped on 2026-04-27"))
````

# 17. `re.split()`

`split()` breaks text by a regex pattern.

Why useful:
- multi-separator splitting
- flexible tokenization


````python
# Import re for regex splitting.
import re

# Split by commas or semicolons.
print(re.split(r"[,;]\s*", "python, sql; excel, powerbi"))
````

# 18. Regex Flags

Common flags:

- `re.IGNORECASE`
- `re.MULTILINE`
- `re.DOTALL`

Why useful:
- case-insensitive search
- line-based anchors
- dot matching across newlines


````python
# Import re for flag-based matching.
import re

# Find python in any case.
print(re.findall(r"python", "Python PYTHON python", flags=re.IGNORECASE))
````

# 19. Escaping Special Characters

Some characters are special in regex and must be escaped when matched literally.

Examples:
- `.`
- `+`
- `*`
- `?`
- `(` and `)`


````python
# Import re for literal special-character matching.
import re

# Match literal dots in a file name.
print(re.findall(r"\.", "report.v1.txt"))
````

# 20. Email-Like Pattern Example

Regex is often used for simple email-like matching.

Important note:
- full real email validation is more complex than beginner patterns
- keep training patterns practical but honest


````python
# Import re for email extraction.
import re

# Extract simple email-like values.
text = "Contact asha@test.com and ravi@data.org"
print(re.findall(r"\b[\w.-]+@[\w.-]+\.\w+\b", text))
````

# 21. Phone Number Pattern Example

Regex can help detect specific phone formats.

Why useful:
- data cleaning
- parsing user input
- form validation


````python
# Import re for phone matching.
import re

# Match a simple 10-digit style number with dashes.
print(bool(re.fullmatch(r"\d{3}-\d{3}-\d{4}", "123-456-7890")))
````

# 22. Common Mistakes

Common mistakes:

- forgetting raw strings
- using `match()` when `search()` is needed
- writing patterns that are too broad
- forgetting greediness
- assuming regex validates every real-world format perfectly


````python
# Import re for a quick comparison.
import re

# search finds digits anywhere, match checks only the start.
print(re.search(r"\d+", "abc123").group())
print(re.match(r"\d+", "abc123"))
````

# 23. Edge Cases

Important edge cases:

- no match found
- empty strings
- multiple overlapping-like regions
- unexpected whitespace
- patterns that overmatch or undermatch


````python
# Import re for a no-match-safe example.
import re

# Search for digits in a string without digits.
match = re.search(r"\d+", "no numbers here")

# Check safely before reading a group.
print(match is None)
````

# 24. Short-Hand and Optimized Patterns

Useful regex practices:

- compile reused patterns
- use raw strings
- prefer clear patterns over extremely clever ones
- use named groups for readability
- test patterns on edge cases


````python
# Import re for compiled patterns.
import re

# Compile a reusable pattern once.
pattern = re.compile(r"\b\w+@\w+\.com\b")

# Reuse it on text.
print(pattern.findall("a@test.com b@data.com c@test.org"))
````

# 25. Real-World Mini Practical Example

This example combines validation, extraction, and substitution.

Goal:
- extract emails
- mask numbers
- clean repeated separators


````python
# Import re for multi-step text cleanup.
import re

# Create sample messy text.
text = "Email asha@test.com, call 9876543210;; visit office 42."

# Extract simple email-like values.
emails = re.findall(r"\b[\w.-]+@[\w.-]+\.\w+\b", text)

# Mask long digit groups.
masked = re.sub(r"\d{4,}", "####", text)

# Split by one or more semicolons.
parts = re.split(r";+", masked)

# Print all three outputs.
print(emails)
print(masked)
print(parts)
````
