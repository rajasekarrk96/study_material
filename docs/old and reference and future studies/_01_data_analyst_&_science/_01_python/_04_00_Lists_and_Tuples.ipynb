# Introduction

- **What is it?** A list is an ordered mutable collection. A tuple is an ordered immutable collection.
- **Why it is used?** Lists are used when data changes. Tuples are used when grouped data should stay fixed.
- **Real-world relevance:** Lists store rows, scores, tasks, search results, and cart items. Tuples store coordinates, database rows, fixed pairs, and function return values.


# Concept Map / Topic Breakdown

Lists and Tuples expands into these connected areas:

- **Core fundamentals:** ordered collections, indexing, slicing, iteration
- **List behavior:** mutable, dynamic size, append, extend, insert, remove, pop, sort
- **Tuple behavior:** immutable, fixed records, unpacking, hashability when contents are hashable
- **Operators involved:** indexing `[]`, slicing `:`, concatenation `+`, repetition `*`, membership `in`
- **Built-in functions:** `len()`, `sum()`, `min()`, `max()`, `sorted()`, `reversed()`, `enumerate()`, `zip()`
- **Copying and memory:** assignment aliasing, shallow copy, nested list reference behavior
- **Pythonic shortcuts:** comprehensions, tuple unpacking, star unpacking, `zip`, `enumerate`
- **Edge cases:** empty lists, one-item tuples, negative indexes, out-of-range indexes, duplicates
- **Performance:** list append average O(1), insert/pop at front O(n), sort O(n log n), tuple lookup O(1)
- **Interview patterns:** two pointers, sliding window, stack with list, matrix transpose, flattening
- **Real-world use:** marks, rows, coordinates, carts, sales records, rankings, task lists


# Basic Syntax Explanation

```python
numbers = [10, 20, 30]       # list
point = (10, 20)             # tuple
single = (10,)               # one-item tuple needs comma
```

Important:
- Lists use square brackets `[]`.
- Tuples use parentheses `()`.
- Both support indexing, slicing, iteration, `in`, `len()`, and unpacking.
- Only lists support item assignment and mutating methods like `append()`.


````python
# Basic Example with detailed comments

# Create a list because marks may change.
marks = [80, 85, 90]

# Create a tuple because coordinates usually stay fixed.
point = (10, 20)

# Update the first mark because lists are mutable.
marks[0] = 82

# Unpack tuple values into two variables.
x, y = point

# Print final values.
print("Marks:", marks)
print("Point x:", x)
print("Point y:", y)
````

# 1. Creating Lists

Lists can be created with literal brackets or `list()`.


````python
# Create list using brackets.
numbers = [1, 2, 3]

# Create list from a range.
range_list = list(range(1, 6))

print(numbers)
print(range_list)
````

# 2. Creating Tuples

Tuples can be created with commas or `tuple()`.


````python
# Create tuple using parentheses.
point = (10, 20)

# Comma is what makes a one-item tuple.
single = (10,)

print(point)
print(type(single))
````

# 3. Indexing

Indexing reads one item. First index is 0.


````python
items = ["python", "sql", "excel"]

# Read first item.
print(items[0])

# Read last item.
print(items[-1])
````

# 4. Slicing

Slicing returns part of a list or tuple.


````python
nums = [0, 1, 2, 3, 4, 5]

# Read values from index 1 to 3.
print(nums[1:4])

# Reverse a list copy.
print(nums[::-1])
````

# 5. Mutability

Lists are mutable. Tuples are immutable.


````python
names = ["Asha", "Ravi"]

# This works because list is mutable.
names[1] = "Meera"

point = (10, 20)
# point[0] = 99  # This would fail because tuple is immutable.

print(names)
print(point)
````

# 6. append and extend

`append` adds one item. `extend` adds many items from an iterable.


````python
nums = [1, 2]

# Adds the whole value as one item.
nums.append(3)

# Adds each item from the list.
nums.extend([4, 5])

print(nums)
````

# 7. insert, remove, pop

Use these methods to add or remove list items.


````python
tasks = ["read", "revise"]

# Insert at index 1.
tasks.insert(1, "practice")

# Remove by value.
tasks.remove("read")

# Remove and return last item.
last = tasks.pop()

print(tasks)
print("Popped:", last)
````

# 8. Sorting Lists

`sort()` changes the list. `sorted()` returns a new list.


````python
scores = [70, 95, 80]

# Create a new sorted list.
sorted_scores = sorted(scores)

# Original list remains unchanged.
print("Original:", scores)
print("Sorted:", sorted_scores)
````

# 9. Tuple Unpacking

Unpacking assigns tuple/list values to variables.


````python
student = ("Asha", 91)

# Unpack into name and score.
name, score = student

print(name)
print(score)
````

# 10. Star Unpacking

Star unpacking catches remaining values.


````python
values = [10, 20, 30, 40]

# first gets first value, middle gets remaining except last.
first, *middle, last = values

print(first)
print(middle)
print(last)
````

# 11. Nested Lists

Nested lists store table-like data.


````python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
]

# Read row 1, column 2.
print(matrix[1][2])
````

# 12. List Comprehensions

Comprehensions create lists from loops.


````python
nums = [1, 2, 3, 4]

# Square every number.
squares = [n * n for n in nums]

# Keep only even numbers.
evens = [n for n in nums if n % 2 == 0]

print(squares)
print(evens)
````

# 13. Copying Lists

Assignment shares the same list. Use copy or slicing for a separate list.


````python
original = [1, 2, 3]
alias = original
copy_one = original.copy()

alias.append(4)
copy_one.append(99)

print("Original:", original)
print("Copy:", copy_one)
````

# 14. Operators with Lists and Tuples

`+` concatenates and `*` repeats.


````python
a = [1, 2]
b = [3, 4]

print(a + b)
print(a * 3)
print((1, 2) + (3, 4))
````

# 15. Membership

`in` checks whether a value exists.


````python
skills = ["python", "sql", "excel"]

if "python" in skills:
    print("Python found")
else:
    print("Python missing")
````

# 16. enumerate with Lists

`enumerate` gives index and value together.


````python
names = ["Asha", "Ravi", "Meera"]

for index, name in enumerate(names, start=1):
    print(index, name)
````

# 17. zip with Lists and Tuples

`zip` pairs values from multiple iterables.


````python
names = ["Asha", "Ravi"]
scores = [90, 85]

pairs = list(zip(names, scores))

print(pairs)
````

# 18. min, max, sum

Built-in functions make numeric list work simple.


````python
marks = [80, 95, 70]

print("Min:", min(marks))
print("Max:", max(marks))
print("Sum:", sum(marks))
print("Average:", sum(marks) / len(marks))
````

# 19. Common Mistakes

Avoid modifying a list while looping over it and remember tuple comma edge cases.


````python
nums = [1, 2, 3, 4]

# Correct: create a new list instead of removing during loop.
evens = []
for num in nums:
    if num % 2 == 0:
        evens.append(num)

print(evens)
````

# 20. Edge Cases

Always test empty lists, one-item lists, duplicate values, and negative indexes.


````python
def safe_first(items):
    # Empty list has no first item.
    if not items:
        return None
    return items[0]

print(safe_first([]))
print(safe_first([10]))
````

# 21. Advanced Sorting

Sort complex rows using `key`.


````python
students = [("Asha", 90), ("Ravi", 75), ("Meera", 95)]

# Sort by score descending.
ranked = sorted(students, key=lambda row: row[1], reverse=True)

print(ranked)
````

# 22. Real-world Example

Lists and tuples often model records and reports.


````python
sales = [
    ("Jan", 12000),
    ("Feb", 15000),
    ("Mar", 9000),
]

best_month = max(sales, key=lambda row: row[1])

print("Best month:", best_month)
````
