**Problem:** Remove duplicates while preserving order.

**Approach explanation:**

- Handle edge cases from constraints: Values are hashable.
- Use brute force for simple scans; use optimized patterns like slicing, sets, stacks, or sliding windows when useful.
- Keep original inputs unchanged when the question asks for a new list.


````python
def remove_duplicates(items: list[int]) -> list[int]:
    seen = set()
    result = []
    for item in items:
        if item not in seen:
            seen.add(item)
            result.append(item)
    return result

# Time Complexity: based on loops, slicing, sorting, or nested traversal used above.
# Space Complexity: based on extra lists/sets/stacks created above.
````
