### Q13. Factorial Recursion

Dry run: factorial(5) becomes 5 * 4 * 3 * 2 * 1. Base case stops at 1. Time complexity is O(n).


```python
# Define a recursive factorial function
def factorial_recursive(n):
    # Return 1 for the base case
    if n <= 1:
        return 1

    # Multiply the current number by the factorial of the previous number
    return n * factorial_recursive(n - 1)


# Test case 1
expected_output = 120
print(factorial_recursive(5))  # should match expected
print(expected_output)
```

### Q14. Fibonacci Recursion

Dry run: fib(7) expands into smaller Fibonacci calls until base cases 0 and 1 are reached.


```python
# Define a recursive Fibonacci function
def fibonacci_recursive(n):
    # Return n directly for base cases 0 and 1
    if n <= 1:
        return n

    # Return the sum of the previous two Fibonacci numbers
    return fibonacci_recursive(n - 1) + fibonacci_recursive(n - 2)


# Test case 1
expected_output = 13
print(fibonacci_recursive(7))  # should match expected
print(expected_output)
```

### Q21. Memoized Fibonacci

Dry run: cached values prevent recalculating the same Fibonacci subproblems many times. Time complexity becomes O(n).


```python
# Define an efficient Fibonacci function using memoization
def memoized_fibonacci(n):
    # Create a cache with known base values
    cache = {0: 0, 1: 1}

    # Define an inner helper function that reuses the cache
    def helper(value):
        # Return the cached result if it already exists
        if value in cache:
            return cache[value]

        # Compute, store, and return the missing value
        cache[value] = helper(value - 1) + helper(value - 2)
        return cache[value]

    # Start the recursive helper
    return helper(n)


# Test case 1
expected_output = 55
print(memoized_fibonacci(10))  # should match expected
print(expected_output)
```

### Q23. Closure Counter

Dry run: each call updates `count` inside the closure, so the state is remembered between calls.


```python
# Define a function that creates and returns a counter
def make_counter():
    # Create the outer state variable
    count = 0

    # Define the inner function that updates outer state
    def increment():
        nonlocal count
        count += 1
        return count

    # Return the inner function
    return increment


# Build one counter instance
counter = make_counter()

# Test cases
print(counter())  # should be 1
print(counter())  # should be 2
print(counter())  # should be 3
```

### Q26. Recursive Directory Planner

Dry run: move through nested lists and dictionaries, counting only string file names. Time complexity is O(n) over the total nested items.


```python
# Define a recursive function to count file names in nested structures
def count_files(structure):
    # Count one file when the current item is a string
    if isinstance(structure, str):
        return 1

    # Count contents of a list by summing recursive results
    if isinstance(structure, list):
        return sum(count_files(item) for item in structure)

    # Count contents of a dictionary by visiting every value
    if isinstance(structure, dict):
        return sum(count_files(value) for value in structure.values())

    # Return zero for unsupported item types
    return 0


# Test case 1
sample_structure = {"root": ["a", {"sub": ["b", "c"]}]}
expected_output = 3
print(count_files(sample_structure))  # should match expected
print(expected_output)
```

### Q32. Student Report Module

```python
# Define a helper to calculate total marks
def total_marks(marks):
    # Return the sum of all marks
    return sum(marks)


# Define a helper to calculate percentage
def percentage(marks):
    # Return the average mark value as percentage
    return total_marks(marks) / len(marks)


# Define a helper to assign a grade
def grade(marks):
    # Calculate the percentage once for decision making
    percent = percentage(marks)

    # Return grade based on the calculated percentage
    if percent >= 85:
        return "A"
    if percent >= 70:
        return "B"
    if percent >= 60:
        return "C"
    return "D"


# Test case 1
marks_list = [88, 76, 91, 84]
print(total_marks(marks_list))  # should match expected total
print(percentage(marks_list))  # should match expected percentage
print(grade(marks_list))  # should match expected grade
```

### Q35. Text Analytics Helper

Dry run: clean the text first, split into words, count them, and build a frequency dictionary to find the most common word.


```python
# Import punctuation support for text cleaning
import string


# Define a helper to clean text
def clean_sentence(text):
    # Remove punctuation and convert all characters to lowercase
    translator = str.maketrans("", "", string.punctuation)
    return text.translate(translator).lower()


# Define a helper to count words
def word_count(text):
    # Split the cleaned text and return the number of words
    return len(clean_sentence(text).split())


# Define a helper to find the most frequent word
def most_frequent_word(text):
    # Create a list of cleaned words
    words = clean_sentence(text).split()

    # Build a frequency dictionary
    counts = {}
    for word in words:
        counts[word] = counts.get(word, 0) + 1

    # Return the word with the highest frequency
    return max(counts, key=counts.get)


# Test case 1
sample_text = "Python is great, and Python is easy"
print(clean_sentence(sample_text))  # should match expected cleaned text
print(word_count(sample_text))  # should match expected word count
print(most_frequent_word(sample_text))  # should match expected frequent word
```
