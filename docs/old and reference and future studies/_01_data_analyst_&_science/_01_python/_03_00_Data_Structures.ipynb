# Introduction

- **What is it?** A data structure is a way to organize data so programs can store, find, update, and process it efficiently.
- **Why it is used?** Different problems need different access patterns. Lists are good for ordered data, dictionaries are good for fast lookup, sets are good for uniqueness, and queues are good for first-in-first-out processing.
- **Real-world relevance:** Shopping carts, contact books, database rows, API JSON, caching, task queues, search suggestions, dashboards, and interview algorithms all depend on correct data structure choices.

# Basic Syntax Explanation

```python
numbers = [10, 20, 30]              # list: ordered and mutable
point = (10, 20)                    # tuple: ordered and immutable
student = {"name": "Asha"}          # dict: key-value mapping
unique_ids = {101, 102, 103}        # set: unique unordered values
```

Main idea:
- Use a **list** when order and modification matter.
- Use a **tuple** when fixed grouped data matters.
- Use a **dict** when lookup by key matters.
- Use a **set** when uniqueness and membership checks matter.


````python
# Basic Example with detailed comments

# Create a list to store ordered marks.
marks = [80, 90, 75]

# Create a tuple to store fixed coordinates.
point = (10, 20)

# Create a dictionary to store related named information.
student = {"name": "Asha", "marks": marks}

# Create a set to store unique subjects.
subjects = {"python", "sql", "python"}

# Print each structure to compare behavior.
print("List:", marks)
print("Tuple:", point)
print("Dictionary:", student)
print("Set removes duplicates:", subjects)
````

# 1. Data Structure Selection

Choose based on access pattern: order, lookup, uniqueness, or fixed grouping.


````python
# Ordered values usually fit a list.
ordered_scores = [88, 92, 75]

# Fast key lookup usually fits a dictionary.
profile = {"id": 101, "name": "Ravi"}

# Unique membership usually fits a set.
visited_pages = {"home", "about", "home"}

print(ordered_scores)
print(profile["name"])
print(visited_pages)
````

# 2. List Basics

A list is ordered, mutable, and allows duplicate values.


````python
# Create a list.
numbers = [10, 20, 30]

# Append adds one value at the end.
numbers.append(40)

# Modify by index because lists are mutable.
numbers[1] = 25

print(numbers)
````

# 3. List Indexing

Indexing starts at 0. Negative indexes count from the end.


````python
items = ["a", "b", "c", "d"]

# First element has index 0.
print(items[0])

# Last element can be read using -1.
print(items[-1])
````

# 4. List Slicing

Slicing creates a new list from part of another list.


````python
numbers = [0, 1, 2, 3, 4, 5]

# Get index 1 up to but not including index 4.
print(numbers[1:4])

# Reverse using a negative step.
print(numbers[::-1])
````

# 5. List Methods

Common methods include append, extend, insert, pop, remove, sort, and reverse.


````python
names = ["Asha", "Ravi"]

# append adds one item.
names.append("Meera")

# insert adds at a specific position.
names.insert(1, "John")

# pop removes and returns an item.
removed = names.pop()

print(names)
print("Removed:", removed)
````

# 6. List Comprehension

Comprehensions build lists in a compact and readable way.


````python
# ================= LIST COMPREHENSION - FULL DEMO =================

numbers = [1, 2, 3, 4, 5]

# ---------------------------------------------------------------
# 1) BASIC LIST COMPREHENSION (Transformation)
# Square each number
# Syntax: [expression for item in iterable]
# ---------------------------------------------------------------
squares = [n * n for n in numbers]

# Equivalent normal loop:
# squares = []
# for n in numbers:
#     squares.append(n * n)


# ---------------------------------------------------------------
# 2) LIST COMPREHENSION WITH IF (Filtering)
# Keep only even numbers
# Syntax: [expression for item in iterable if condition]
# ---------------------------------------------------------------
evens = [n for n in numbers if n % 2 == 0]

# Equivalent normal loop:
# evens = []
# for n in numbers:
#     if n % 2 == 0:
#         evens.append(n)


# ---------------------------------------------------------------
# 3) LIST COMPREHENSION WITH IF-ELSE (Decision making)
# Label numbers as Even or Odd
# Syntax: [true_value if condition else false_value for item in iterable]
# ---------------------------------------------------------------
labels = ["Even" if n % 2 == 0 else "Odd" for n in numbers]

# Equivalent normal loop:
# labels = []
# for n in numbers:
#     if n % 2 == 0:
#         labels.append("Even")
#     else:
#         labels.append("Odd")


# ---------------------------------------------------------------
# 4) COMBINED (Filter + If-Else)
# Condition 1: take only numbers > 2
# Condition 2: if even → square, else → keep same
# ---------------------------------------------------------------
result = [n*n if n % 2 == 0 else n for n in numbers if n > 2]

# Equivalent normal loop:
# result = []
# for n in numbers:
#     if n > 2:
#         if n % 2 == 0:
#             result.append(n*n)
#         else:
#             result.append(n)


# ---------------------------------------------------------------
# 5) MULTIPLE CONDITIONS
# Even AND greater than 2
# ---------------------------------------------------------------
multi_cond = [n for n in numbers if n % 2 == 0 if n > 2]

# Equivalent:
# multi_cond = []
# for n in numbers:
#     if n % 2 == 0 and n > 2:
#         multi_cond.append(n)


# ---------------------------------------------------------------
# OUTPUTS
# ---------------------------------------------------------------
print("Numbers      :", numbers)
print("Squares      :", squares)
print("Evens        :", evens)
print("Labels       :", labels)
print("Result       :", result)
print("Multi Cond   :", multi_cond)
````

# 7. Tuple Basics

A tuple is ordered and immutable. Use it for fixed records.


````python
# Coordinates should usually not change.
point = (10, 20)

# Tuple unpacking assigns values clearly.
x, y = point

print(x)
print(y)
````

# 8. Tuple Edge Cases

A one-value tuple needs a comma.


````python
# This is an integer, not a tuple.
not_tuple = (10)

# This is a one-item tuple.
one_item_tuple = (10,)

print(type(not_tuple))
print(type(one_item_tuple))
````

# 9. Dictionary Basics

A dictionary stores key-value pairs and supports fast lookup by key.


````python
student = {"name": "Asha", "score": 91}

# Read using key.
print(student["name"])

# Add a new key-value pair.
student["city"] = "Pune"

print(student)
````

# 10. Safe Dictionary Access

`get()` avoids KeyError when a key may be missing.


````python
student = {"name": "Asha"}

# Direct access would fail for missing key.
city = student.get("city", "Unknown")

print(city)
````

# 11. Dictionary Iteration

Use keys, values, or items depending on what you need.


````python
marks = {"python": 90, "sql": 85}

# items gives key and value together.
for subject, score in marks.items():
    print(subject, score)
````

# 12. Dictionary Comprehension

Dictionary comprehensions build mappings from loops.


````python
numbers = [1, 2, 3, 4]

# Map each number to its square.
squares = {n: n * n for n in numbers}

print(squares)
````

# 13. Set Basics

A set stores unique values and has fast membership checks.


````python
ids = [101, 102, 101, 103]

# Convert list to set to remove duplicates.
unique_ids = set(ids)

print(unique_ids)
print(102 in unique_ids)
````

# 14. Set Operations

Sets support union, intersection, difference, and symmetric difference.


````python
python_students = {"Asha", "Ravi", "Meera"}
sql_students = {"Ravi", "John"}

print("Union:", python_students | sql_students)
print("Intersection:", python_students & sql_students)
print("Difference:", python_students - sql_students)
````

# 15. String as Sequence

Strings behave like immutable sequences of characters.


````python
text = "python"

# Indexing works like a list.
print(text[0])

# Slicing works too.
print(text[::-1])
````

# 16. Mutability

Lists, dictionaries, and sets are mutable. Tuples and strings are immutable.


````python
# ================= MUTABLE vs IMMUTABLE TYPES =================

# ---------------------------------------------------------------
# THEORY:
# Mutable   → Can be changed after creation (same memory location)
# Immutable → Cannot be changed after creation (new object created)
# ---------------------------------------------------------------


# ===================== MUTABLE TYPES =====================

# 1) LIST (Mutable)
print("\n--- LIST (Mutable) ---")
my_list = [1, 2, 3]
print("Original list:", my_list, "| id:", id(my_list))

# Modify element
my_list[0] = 100
print("After modification:", my_list, "| id:", id(my_list))  # same id

# Add element
my_list.append(4)
print("After append:", my_list, "| id:", id(my_list))  # still same id


# 2) DICTIONARY (Mutable)
print("\n--- DICTIONARY (Mutable) ---")
my_dict = {"a": 1, "b": 2}
print("Original dict:", my_dict, "| id:", id(my_dict))

# Modify value
my_dict["a"] = 100
print("After change:", my_dict, "| id:", id(my_dict))  # same id

# Add new key
my_dict["c"] = 3
print("After adding key:", my_dict, "| id:", id(my_dict))


# 3) SET (Mutable)
print("\n--- SET (Mutable) ---")
my_set = {1, 2, 3}
print("Original set:", my_set, "| id:", id(my_set))

# Add element
my_set.add(4)
print("After add:", my_set, "| id:", id(my_set))  # same id

# Remove element
my_set.remove(2)
print("After remove:", my_set, "| id:", id(my_set))


# ===================== IMMUTABLE TYPES =====================

# 4) TUPLE (Immutable)
print("\n--- TUPLE (Immutable) ---")
my_tuple = (1, 2, 3)
print("Original tuple:", my_tuple, "| id:", id(my_tuple))

# Try modifying (will cause error)
# my_tuple[0] = 100   # ❌ TypeError

# Instead, new tuple is created
new_tuple = my_tuple + (4,)
print("New tuple:", new_tuple, "| id:", id(new_tuple))
print("Old tuple still:", my_tuple, "| id:", id(my_tuple))


# 5) STRING (Immutable)
print("\n--- STRING (Immutable) ---")
my_string = "hello"
print("Original string:", my_string, "| id:", id(my_string))

# Try modifying (will cause error)
# my_string[0] = 'H'   # ❌ TypeError

# New string is created
new_string = my_string.upper()
print("New string:", new_string, "| id:", id(new_string))
print("Old string still:", my_string, "| id:", id(my_string))


# ===================== IMPORTANT INTERVIEW POINT =====================

# Mutable inside immutable (tuple containing list)
print("\n--- SPECIAL CASE ---")
mixed = (1, 2, [3, 4])
print("Original:", mixed, "| id:", id(mixed))

# Modify inner list
mixed[2].append(5)
print("After modifying inner list:", mixed, "| id:", id(mixed))

# NOTE:
# Tuple is immutable, but it can contain mutable objects inside it.


# ===================== FINAL SUMMARY =====================
print("\n--- SUMMARY ---")
print("Mutable   :", "List, Dictionary, Set")
print("Immutable :", "Tuple, String")
````

# 17. Aliasing and Copying

Assignment shares the same object. Use copy when you need a separate list.


````python
original = [1, 2, 3]
alias = original
copy_list = original.copy()

# This changes original too because alias points to same list.
alias.append(4)

# This changes only copy_list.
copy_list.append(99)

print("Original:", original)
print("Copy:", copy_list)
````

# 18. Nested Data Structures

Real data often combines lists and dictionaries.


````python
students = [
    {"name": "Asha", "score": 90},
    {"name": "Ravi", "score": 75},
]

for student in students:
    print(student["name"], student["score"])
````
