# Introduction

- **What is it?** Testing is the process of checking whether code behaves as expected. Logging is the process of recording what a program is doing while it runs.
- **Why it is used?** Testing helps catch bugs early, prevents regressions, and makes refactoring safer. Logging helps us debug issues, monitor behavior, and understand failures in real applications.
- **Real-world relevance:** Companies use tests in CI pipelines before deployment. Logs are used in web apps, APIs, background jobs, ETL pipelines, ML workflows, and production monitoring systems.


# Basic Syntax Explanation

```python
assert add(2, 3) == 5

import unittest

class TestMath(unittest.TestCase):
    def test_add(self):
        self.assertEqual(add(2, 3), 5)

import logging
logging.basicConfig(level=logging.INFO)
logging.info("Program started")
```

Main idea:
- Use **`assert`** for quick checks and learning.
- Use **`unittest`** for organized automated test suites.
- Use **`logging`** instead of `print()` when you need structured runtime information.


````python
# Basic Example with detailed comments

import logging


def add(a, b):
    # Return the sum of the two input numbers.
    return a + b


# Check that the function gives the expected answer.
assert add(2, 3) == 5

# Configure logging so INFO messages are visible.
logging.basicConfig(level=logging.INFO, format="%(levelname)s:%(message)s")

# Write a log message that describes what the program just did.
logging.info("The add function passed a simple test.")

# Print a final message so notebook users can see a normal output line too.
print("Basic testing and logging example completed successfully.")
````

# 1. Why Testing Matters

Testing checks behavior before bugs reach users. It builds confidence, especially when code changes later.


````python
# This function converts Celsius temperature to Fahrenheit.
def celsius_to_fahrenheit(celsius):
    # Multiply by 9/5 and add 32 using the standard formula.
    return (celsius * 9 / 5) + 32


# Test a known value so we can trust the function.
assert celsius_to_fahrenheit(0) == 32

# Another check helps confirm the formula is consistent.
assert celsius_to_fahrenheit(100) == 212

print("Temperature conversion tests passed.")
````

# 2. Manual Testing vs Automated Testing

Manual testing means running code and checking results by hand. Automated testing encodes those checks so they can run repeatedly.


````python
# Define a function that doubles a number.
def double_number(value):
    # Return a value multiplied by 2.
    return value * 2


# Manual inspection often starts with printing.
print("Manual check:", double_number(4))

# Automated checks remove guesswork and repetition.
assert double_number(4) == 8
assert double_number(-3) == -6

print("Automated checks passed.")
````

# 3. Writing Testable Functions

Small pure functions are easier to test because they depend mainly on inputs and outputs.


````python
# This function takes input and returns output without modifying global state.
def calculate_discount(price, percent):
    # Convert the percentage into a decimal discount amount.
    discount_amount = price * (percent / 100)

    # Subtract the discount from the original price.
    final_price = price - discount_amount

    # Return the computed result.
    return final_price


# Simple functions are straightforward to test.
assert calculate_discount(1000, 10) == 900
assert calculate_discount(250, 20) == 200

print("Testable function examples passed.")
````

# 4. Using assert

`assert` is the simplest way to express that something must be true during a test.


````python
# Create a function that returns whether a value is even.
def is_even(number):
    # Modulo 2 equal to 0 means the number is even.
    return number % 2 == 0


# If this condition is false, Python raises AssertionError.
assert is_even(10) is True

# This second check verifies the opposite case.
assert is_even(7) is False

print("All assert statements succeeded.")
````

# 5. Helpful Assertion Messages

Custom messages make failures easier to understand, especially in beginner practice.


````python
# Define a function that builds a username.
def make_username(first_name, last_name):
    # Lowercase both names and join them with a dot.
    return f"{first_name.lower()}.{last_name.lower()}"


# Add a readable failure message so debugging is easier.
assert make_username("Asha", "Patel") == "asha.patel", "Username format is incorrect."

print("Username assertion passed.")
````

# 6. Basic Edge Cases in Tests

Good tests cover empty inputs, zero values, negative numbers, duplicates, and unusual but valid cases.


````python
# Define a function that returns the average of a list.
def average(numbers):
    # Return 0 for an empty list to avoid division by zero in this example.
    if not numbers:
        return 0

    # Sum all numbers and divide by the count.
    return sum(numbers) / len(numbers)


# Normal case.
assert average([10, 20, 30]) == 20

# Edge case: empty list.
assert average([]) == 0

# Edge case: one value.
assert average([5]) == 5

print("Average edge-case tests passed.")
````

# 7. Common Mistakes in Beginner Tests

A common mistake is testing only the happy path and forgetting invalid or boundary inputs.


````python
# This function returns the first character from text.
def first_character(text):
    # Return an empty string when text is empty so callers get a safe result.
    if text == "":
        return ""

    # Return the first character for non-empty text.
    return text[0]


# Happy path test.
assert first_character("python") == "p"

# Boundary test for empty input.
assert first_character("") == ""

print("Boundary-aware tests passed.")
````

# 8. Organizing Tests with unittest

`unittest` is Python's built-in framework for grouped, repeatable test cases.


````python
import unittest


# Function under test.
def square(number):
    # Multiply the number by itself.
    return number * number


class TestSquare(unittest.TestCase):
    # Test methods must start with the word test.
    def test_positive_value(self):
        # assertEqual checks expected and actual values.
        self.assertEqual(square(4), 16)

    def test_zero(self):
        # Zero squared should stay zero.
        self.assertEqual(square(0), 0)


# Run tests in the notebook without stopping the kernel.
unittest.TextTestRunner().run(unittest.defaultTestLoader.loadTestsFromTestCase(TestSquare))
````

# 9. Test Method Naming

Clear test names explain behavior and make failures easier to read.


````python
import unittest


def normalize_city(city):
    # Strip outer spaces and title-case the text.
    return city.strip().title()


class TestNormalizeCity(unittest.TestCase):
    def test_removes_outer_spaces(self):
        # The name says exactly what behavior is verified.
        self.assertEqual(normalize_city("  delhi  "), "Delhi")

    def test_applies_title_case(self):
        # The output should use title case formatting.
        self.assertEqual(normalize_city("mumbai"), "Mumbai")


unittest.TextTestRunner().run(unittest.defaultTestLoader.loadTestsFromTestCase(TestNormalizeCity))
````

# 10. setUp and tearDown

`setUp` prepares fresh data before each test, which reduces repetition and shared-state bugs.


````python
import unittest


class TestCart(unittest.TestCase):
    def setUp(self):
        # Create a new cart before every test method.
        self.cart = [{"item": "book", "price": 200}, {"item": "pen", "price": 20}]

    def test_cart_has_two_items(self):
        # len returns the number of rows in the cart.
        self.assertEqual(len(self.cart), 2)

    def test_total_price(self):
        # sum adds the price field from each dictionary.
        total = sum(row["price"] for row in self.cart)
        self.assertEqual(total, 220)


unittest.TextTestRunner().run(unittest.defaultTestLoader.loadTestsFromTestCase(TestCart))
````

# 11. Testing Exceptions

Some functions are correct only when they raise an error for invalid input.


````python
import unittest


def divide(a, b):
    # Raise ZeroDivisionError automatically if b is zero.
    return a / b


class TestDivide(unittest.TestCase):
    def test_divide_normal_case(self):
        # Compare normal arithmetic output.
        self.assertEqual(divide(10, 2), 5)

    def test_divide_by_zero_raises_error(self):
        # Verify that invalid input raises the expected exception.
        with self.assertRaises(ZeroDivisionError):
            divide(10, 0)


unittest.TextTestRunner().run(unittest.defaultTestLoader.loadTestsFromTestCase(TestDivide))
````

# 12. Comparing Collections

# 13. Testing Floating Point Values

Floating point math may produce tiny precision differences, so exact equality can be risky.


````python
import unittest


def tax_amount(price, rate):
    # Multiply price by a decimal rate.
    return price * rate


class TestTaxAmount(unittest.TestCase):
    def test_approximate_comparison(self):
        # Use assertAlmostEqual for floating point results.
        self.assertAlmostEqual(tax_amount(199.99, 0.18), 35.9982, places=4)


unittest.TextTestRunner().run(unittest.defaultTestLoader.loadTestsFromTestCase(TestTaxAmount))
````

# 14. Testing Boolean Results

Use clear assertions for truthy and falsy outcomes so intent stays readable.


````python
import unittest


def has_permission(role):
    # Allow only admin and editor roles in this example.
    return role in {"admin", "editor"}


class TestPermission(unittest.TestCase):
    def test_allowed_role(self):
        # Check a True result explicitly.
        self.assertTrue(has_permission("admin"))

    def test_denied_role(self):
        # Check a False result explicitly.
        self.assertFalse(has_permission("guest"))


unittest.TextTestRunner().run(unittest.defaultTestLoader.loadTestsFromTestCase(TestPermission))
````

# 15. Mocking Basics

Mocking replaces a dependency so you can test behavior without calling the real dependency.


````python
from unittest.mock import Mock


# Create a fake function-like object.
email_sender = Mock(return_value=True)

# Call the fake dependency with test data.
result = email_sender("user@example.com", "Welcome!")

# Verify what happened.
assert result is True
email_sender.assert_called_once_with("user@example.com", "Welcome!")

print("Mock basic example passed.")
````

# 16. Mocking External Services

External services such as APIs, email systems, and payment gateways should usually be mocked in unit tests.


````python
from unittest.mock import Mock


def register_user(username, send_email_function):
    # Build a friendly welcome message for the new user.
    message = f"Welcome, {username}!"

    # Call the dependency that sends the email.
    send_email_function(username, message)

    # Return a status string for the caller.
    return "registered"


# Build a fake sender so no real email is sent.
fake_sender = Mock()

# Run the function using the fake dependency.
status = register_user("asha@example.com", fake_sender)

# Check behavior and interaction.
assert status == "registered"
fake_sender.assert_called_once_with("asha@example.com", "Welcome, asha@example.com!")

print("Dependency injection test passed.")
````

# 17. patch with unittest.mock

`patch` temporarily replaces a real object during a test.


````python
import unittest
from unittest.mock import patch


def get_exchange_rate():
    # In real code, this might call an API.
    return 83.5


def convert_usd_to_inr(usd):
    # Multiply the amount by the current exchange rate.
    return usd * get_exchange_rate()


class TestCurrency(unittest.TestCase):
    @patch("__main__.get_exchange_rate", return_value=80)
    def test_conversion_with_mocked_rate(self, mocked_rate):
        # The patched function returns 80 instead of the real value.
        self.assertEqual(convert_usd_to_inr(2), 160)
        mocked_rate.assert_called_once()


unittest.TextTestRunner().run(unittest.defaultTestLoader.loadTestsFromTestCase(TestCurrency))
````

# 18. side_effect in Mocks

`side_effect` can raise exceptions or return changing values across calls.


````python
from unittest.mock import Mock


# Create a mock that returns three different results in order.
reader = Mock(side_effect=["row1", "row2", StopIteration("No more rows")])

# First call returns the first value.
print(reader())

# Second call returns the second value.
print(reader())

# Third call raises the stored exception object.
try:
    reader()
except StopIteration as error:
    print("Handled:", error)
````

# 19. Logging Basics

Logging records runtime events with levels and messages that are easier to manage than plain print statements.


````python
import logging


# Configure one simple logger format for the notebook cell.
logging.basicConfig(level=logging.INFO, format="%(levelname)s:%(message)s", force=True)

# Send a normal informational message.
logging.info("Data import started.")

# Send a warning when something needs attention.
logging.warning("One row had a missing value.")

print("Logging basics completed.")
````

# 20. Logging Levels

Common levels are DEBUG, INFO, WARNING, ERROR, and CRITICAL. They help filter importance.


````python
import logging


logging.basicConfig(level=logging.DEBUG, format="%(levelname)s:%(message)s", force=True)

# DEBUG is useful for detailed internal state.
logging.debug("Loaded configuration dictionary.")

# INFO describes normal progress.
logging.info("Application started successfully.")

# ERROR signals a failure that prevented a step from completing.
logging.error("Could not connect to the reporting database.")
````

# 21. Why logging is better than print for applications

Logging can include timestamps, levels, file names, and can be routed to files or monitoring systems.


````python
import logging


logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s | %(levelname)s | %(message)s",
    force=True,
)

# This message now includes a timestamp and level automatically.
logging.info("A structured log message is more useful than a plain print statement.")
````

# 22. Creating Named Loggers

Named loggers help large applications separate logs by module or feature.


````python
import logging


# Create a logger with a descriptive name.
logger = logging.getLogger("billing")

# Set the minimum level for this logger.
logger.setLevel(logging.INFO)

# Add a handler only if the logger does not already have one.
if not logger.handlers:
    # StreamHandler writes logs to standard output.
    handler = logging.StreamHandler()

    # Formatter controls how each log line looks.
    handler.setFormatter(logging.Formatter("%(name)s | %(levelname)s | %(message)s"))

    # Attach the handler to the logger.
    logger.addHandler(handler)

# Log a message through the named logger.
logger.info("Invoice generated successfully.")
````

# 24. Testing Log Messages

You can test whether code produced the expected logs, not just the expected return values.


````python
import logging
import unittest


def validate_age(age):
    # Create or reuse a named logger for validation messages.
    logger = logging.getLogger("validation")

    # Reject negative ages and record an error.
    if age < 0:
        logger.error("Age cannot be negative.")
        return False

    # Accept valid ages and record an info message.
    logger.info("Age is valid.")
    return True


class TestValidateAge(unittest.TestCase):
    def test_invalid_age_logs_error(self):
        # Capture logs from the validation logger.
        with self.assertLogs("validation", level="ERROR") as captured:
            self.assertFalse(validate_age(-5))

        # Confirm the expected error text appears in the logs.
        self.assertIn("Age cannot be negative.", captured.output[0])


unittest.TextTestRunner().run(unittest.defaultTestLoader.loadTestsFromTestCase(TestValidateAge))
````

# 26. Short-hand and Optimized Testing Patterns

Table-driven tests reduce repetition when the same logic must be tested with many inputs.


````python
# Define one function to test.
def classify_score(score):
    # Return grade labels based on score ranges.
    if score >= 90:
        return "A"
    if score >= 75:
        return "B"
    if score >= 60:
        return "C"
    return "D"


# Store multiple cases in one list of tuples.
test_cases = [
    (95, "A"),
    (80, "B"),
    (62, "C"),
    (40, "D"),
]

# Loop through all cases instead of writing repeated assertions manually.
for score, expected in test_cases:
    assert classify_score(score) == expected

print("Table-driven tests passed.")
````

# 27. Advanced Usage with Parameter-Like Loops

Python's built-in tools do not have pytest-style parameterization, but loops can still express many cases cleanly.


````python
import unittest


def is_palindrome(text):
    # Lowercase the text and remove spaces for a simple comparison.
    compact = text.replace(" ", "").lower()
    return compact == compact[::-1]


class TestPalindrome(unittest.TestCase):
    def test_multiple_cases(self):
        # Group several cases into one test loop.
        cases = [
            ("level", True),
            ("Never odd or even", True),
            ("python", False),
        ]

        # Iterate through each case and compare the result.
        for text, expected in cases:
            self.assertEqual(is_palindrome(text), expected)


unittest.TextTestRunner().run(unittest.defaultTestLoader.loadTestsFromTestCase(TestPalindrome))
````

# 29. Real-world Example: Retry with Logging

Applications often retry unstable operations and log each attempt for observability.


````python
import logging


logging.basicConfig(level=logging.INFO, format="%(levelname)s:%(message)s", force=True)

attempt_results = [False, False, True]

for index, result in enumerate(attempt_results, start=1):
    # Log which attempt is currently running.
    logging.info("Attempt %s started.", index)

    # Stop once the simulated action succeeds.
    if result:
        logging.info("Attempt %s succeeded.", index)
        break
else:
    logging.error("All retry attempts failed.")
````

# 30. Mini Practical Example: Test + Log Together

In real projects, functions return values, log important events, and have tests that verify both behavior and observability.


````python
import logging
import unittest


def withdraw(balance, amount):
    # Create a logger for banking operations.
    logger = logging.getLogger("bank")

    # Reject withdrawals that exceed the balance.
    if amount > balance:
        logger.warning("Withdrawal denied due to insufficient funds.")
        return balance

    # Calculate the new balance after a valid withdrawal.
    new_balance = balance - amount

    # Log the successful operation.
    logger.info("Withdrawal completed successfully.")
    return new_balance


class TestWithdraw(unittest.TestCase):
    def test_successful_withdrawal(self):
        # Confirm the returned balance is correct.
        self.assertEqual(withdraw(1000, 250), 750)

    def test_denied_withdrawal_logs_warning(self):
        # Capture warning logs from the bank logger.
        with self.assertLogs("bank", level="WARNING") as captured:
            self.assertEqual(withdraw(100, 150), 100)

        # Verify the warning text appears in the captured output.
        self.assertIn("insufficient funds", captured.output[0].lower())


unittest.TextTestRunner().run(unittest.defaultTestLoader.loadTestsFromTestCase(TestWithdraw))
````
