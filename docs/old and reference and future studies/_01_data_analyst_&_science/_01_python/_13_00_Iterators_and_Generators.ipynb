# Introduction

- **What is it?** Iterators and generators are tools for producing values one at a time instead of building everything at once.
- **Why it is used?** They help us process data efficiently, especially large or continuous data, without storing unnecessary results in memory.
- **Real-world relevance:** Reading large files line by line, processing API results, building data pipelines, handling logs, streaming records, batching database rows, and interview problems all use iterator-style thinking.


# Basic Syntax Explanation

Important building blocks:

```python
items = [10, 20, 30]
iterator = iter(items)
first = next(iterator)

def count_up_to(n):
    for number in range(1, n + 1):
        yield number
```

Core terms:
- **iterable:** an object you can loop over, like a list, tuple, string, dict, set, or file
- **iterator:** an object that remembers iteration state and provides values one by one
- **generator:** a special kind of iterator created with `yield`

Important functions and keywords:
- `iter()`
- `next()`
- `yield`
- `yield from`
- `for`
- `enumerate()`, `zip()`, `map()`, `filter()`


````python
# Basic Example with detailed comments

# Create a normal list, which is an iterable object.
numbers = [10, 20, 30]

# Convert the iterable into an iterator object.
number_iterator = iter(numbers)

# Pull the first value from the iterator.
print(next(number_iterator))

# Pull the second value from the same iterator.
print(next(number_iterator))

# Pull the third value from the same iterator.
print(next(number_iterator))
````

# 1. Iterable vs Iterator

An iterable is something we can loop over. An iterator is the object that actually remembers where iteration is currently positioned.

Examples of iterables:
- list
- tuple
- string
- dict
- set
- file object

Important idea:
- every iterator is iterable
- not every iterable is already an iterator


````python
# Create a list, which is an iterable.
items = [1, 2, 3]

# Convert the list into an iterator.
iterator = iter(items)

# Print the type checks.
print(hasattr(items, "__iter__"))
print(hasattr(iterator, "__next__"))
````

# 2. The `iter()` Function

`iter()` asks an iterable to produce an iterator.

Why useful:
- manual control over iteration
- helps us understand what `for` loops do behind the scenes
- needed for custom iterator workflows


````python
# Create a tuple iterable.
values = ("a", "b", "c")

# Turn the tuple into an iterator object.
value_iterator = iter(values)

# Print one value at a time manually.
print(next(value_iterator))
print(next(value_iterator))
````

# 3. The `next()` Function

`next()` requests the next item from an iterator.

Edge case:
- when no values remain, Python raises `StopIteration`

Common mistake:
- calling `next()` too many times without handling exhaustion


````python
# Create an iterator from a short list.
iterator = iter([100, 200])

# Read the first value.
print(next(iterator))

# Read the second value.
print(next(iterator))

# Use a default value to avoid StopIteration.
print(next(iterator, "done"))
````

# 4. How `for` Loops Work

A `for` loop automatically:

1. calls `iter()` on the iterable
2. repeatedly calls `next()`
3. stops when `StopIteration` happens

Why important:
- explains why iterators get exhausted
- helps us debug custom iterator code


````python
# Loop over a string iterable.
for character in "abc":
    # Print each character one by one.
    print(character)
````

# 5. Iterator Exhaustion

Once an iterator is fully consumed, it does not restart automatically.

Common mistake:
- expecting a used-up iterator to produce values again

Fix:
- create a new iterator from the original iterable if needed


````python
# Create an iterator from a list.
iterator = iter([1, 2, 3])

# Convert the iterator to a list once.
print(list(iterator))

# Try converting the same iterator again.
print(list(iterator))
````

# 6. Custom Iterator Class

A custom iterator class usually implements:

- `__iter__()`
- `__next__()`

Why useful:
- custom traversal logic
- stateful sequence generation
- interview practice for protocol understanding


````python
# Define a custom iterator class.
class CountUp:
    # Save the starting and ending values.
    def __init__(self, start, end):
        # Store the current state.
        self.current = start

        # Store the stopping point.
        self.end = end

    # Return the iterator object itself.
    def __iter__(self):
        # Return self because this object manages its own state.
        return self

    # Produce the next value.
    def __next__(self):
        # Stop when the current value goes past the end.
        if self.current > self.end:
            raise StopIteration

        # Save the value to return.
        value = self.current

        # Move the internal state forward.
        self.current += 1

        # Return the current value.
        return value


# Loop over the custom iterator.
for value in CountUp(1, 4):
    # Print each produced value.
    print(value)
````

# 7. Generator Functions

A generator function uses `yield` instead of `return` to produce one value at a time.

Why useful:
- simpler than writing a full iterator class
- automatic state preservation between yields
- very common in Python code


````python
# Define a generator function.
def count_up_to(limit):
    # Start from 1 and continue to the limit.
    for number in range(1, limit + 1):
        # Yield one number at a time.
        yield number


# Convert the generator output to a list for display.
print(list(count_up_to(5)))
````

# 8. `yield` vs `return`

`return` ends a function permanently. `yield` pauses the function and remembers its state for the next request.

Why important:
- explains how generators resume later
- helps us choose the right tool for the task


````python
# Define a generator that yields two values.
def demo_generator():
    # Produce the first value.
    yield "first"

    # Produce the second value later.
    yield "second"


# Create the generator object.
generator = demo_generator()

# Read values step by step.
print(next(generator))
print(next(generator))
````

# 9. Generator Expressions

A generator expression is like a list comprehension, but lazy.

List comprehension:
```python
[n * n for n in range(5)]
```

Generator expression:
```python
(n * n for n in range(5))
```

Why useful:
- better memory usage
- great for one-pass pipelines


````python
# Create a generator expression.
squares = (number * number for number in range(1, 6))

# Print all generated values.
print(list(squares))
````

# 10. Memory Efficiency

Generators are memory-friendly because they produce values only when needed.

Real-world use:
- huge files
- continuous event streams
- large database results

Interview insight:
- prefer generators when you do not need random access or repeated traversal


````python
# Build a generator instead of a full list.
numbers = (number for number in range(1, 1000001))

# Read only the first three values without building the whole sequence.
print(next(numbers))
print(next(numbers))
print(next(numbers))
````

# 11. `StopIteration`

`StopIteration` signals that an iterator has no more values.

Common places it appears:
- manual `next()` calls
- incorrectly written custom iterators

Best practice:
- let `for` loops handle it when possible


````python
# Create a short iterator.
iterator = iter(["x"])

# Print the first available value.
print(next(iterator))

# Ask for another value safely with a default.
print(next(iterator, "no more values"))
````

# 12. Infinite Generators

A generator can produce values forever if it never reaches a stopping condition.

Why useful:
- IDs
- timestamps
- repeated retries
- simulation streams

Warning:
- never convert an infinite generator directly to a list


````python
# Define an infinite generator.
def infinite_counter():
    # Start counting at 1.
    number = 1

    # Keep running forever.
    while True:
        # Yield the current value.
        yield number

        # Move to the next number.
        number += 1


# Create the infinite generator.
counter = infinite_counter()

# Read only the first three values.
print(next(counter))
print(next(counter))
print(next(counter))
````

# 13. `yield from`

`yield from` delegates iteration to another iterable or generator.

Why useful:
- cleaner code
- flattening
- combining generators
- nested generator pipelines


````python
# Define a generator that delegates to other iterables.
def combined():
    # Yield all values from the first list.
    yield from [1, 2]

    # Yield all values from the second tuple.
    yield from (3, 4)


# Show the combined generator output.
print(list(combined()))
````

# 15. Enumerate and Zip as Iterator Tools

`enumerate()` and `zip()` return iterator-like objects that are often used in loops and pipelines.

Why useful:
- indexed iteration
- synchronized iteration over multiple sequences


````python
# Use enumerate to get index-value pairs.
for index, value in enumerate(["a", "b", "c"], start=1):
    # Print the numbered values.
    print(index, value)


# Use zip to pair two sequences.
for name, score in zip(["Asha", "Ravi"], [91, 84]):
    # Print matched pairs.
    print(name, score)
````

# 17. One-Pass Nature of Generators

Generators are usually consumed once.

Common mistake:
- using the same generator for two separate loops and expecting both to work

Fix:
- recreate the generator
- or store results if you truly need repeated access


````python
# Create a generator expression.
generator = (number * 2 for number in range(4))

# Consume it once.
print(list(generator))

# Try consuming it again.
print(list(generator))
````

# 18. Returning Values from Generators

A generator can use `return`, but that ends the generator and raises `StopIteration`.

Beginner tip:
- use `yield` for produced values
- use `return` only to stop the generator early


````python
# Define a generator that stops early.
def limited():
    # Yield one value first.
    yield 1

    # Stop the generator.
    return

    # This value is never reached.
    yield 2


# Print the generator output.
print(list(limited()))
````

# 19. Sending Values into Generators

Advanced generators can receive values with `.send()`.

Why useful:
- coroutines
- interactive generator workflows

This is more advanced, but good to recognize when reading mature Python code.


````python
# Define a generator that can receive a value.
def echo():
    # Yield an initial prompt.
    received = yield "ready"

    # Yield the received value back out.
    yield received


# Create the generator.
generator = echo()

# Start the generator and get the first yielded value.
print(next(generator))

# Send a value into the paused generator.
print(generator.send("hello"))
````

# 20. Custom Reverse Iterator

Custom iterators can move in any direction or apply special logic.

Why useful:
- reverse traversal
- skipping patterns
- stepping rules


````python
# Define a reverse iterator class.
class ReverseText:
    # Save the text and starting index.
    def __init__(self, text):
        # Store the original text.
        self.text = text

        # Start from the last valid index.
        self.index = len(text) - 1

    # Return the iterator object.
    def __iter__(self):
        # Return self to support iteration.
        return self

    # Produce the next value from the end.
    def __next__(self):
        # Stop when all characters are consumed.
        if self.index < 0:
            raise StopIteration

        # Read the current character.
        character = self.text[self.index]

        # Move one step left.
        self.index -= 1

        # Return the character.
        return character


# Join the reversed characters into a string.
print("".join(ReverseText("python")))
````

# 21. Batching with Generators

A batching generator groups input values into smaller chunks.

Real-world use:
- database inserts
- API requests
- report pages
- background jobs


````python
# Define a batching generator.
def batches(items, size):
    # Step through the input in batch-sized jumps.
    for start in range(0, len(items), size):
        # Yield one slice at a time.
        yield items[start:start + size]


# Print each batch.
for batch in batches([1, 2, 3, 4, 5], 2):
    # Show the current batch.
    print(batch)
````

# 22. Flattening with Generators

Generators are useful for flattening nested data one piece at a time.

Why useful:
- avoids building temporary large lists
- works well in pipelines


````python
# Define a generator that flattens one level.
def flatten(groups):
    # Loop through each inner group.
    for group in groups:
        # Yield each item from the current group.
        yield from group


# Print the flattened values.
print(list(flatten([[1, 2], [3], [4, 5]])))
````

# 23. Running Totals Generator

A generator can maintain state between yields.

Why useful:
- cumulative reports
- streaming metrics
- dashboards


````python
# Define a generator for running totals.
def running_totals(numbers):
    # Start with total zero.
    total = 0

    # Process numbers one by one.
    for number in numbers:
        # Update the total.
        total += number

        # Yield the new running total.
        yield total


# Show the running totals.
print(list(running_totals([2, 3, 5])))
````

# 25. Edge Cases

Important edge cases:

- empty iterables
- single-item iterables
- exhausted iterators
- infinite generators
- batch size larger than input
- nested empty lists in flattening


````python
# Define a simple generator for an edge-case demo.
def values(items):
    # Yield each item in order.
    for item in items:
        yield item


# Show results for an empty input.
print(list(values([])))

# Show results for a single-item input.
print(list(values([42])))
````

# 26. Short-Hand and Optimized Patterns

Useful compact patterns:

- generator expressions for lazy filtering
- `yield from` for delegation
- `enumerate()` for indexed loops
- `zip()` for paired streams
- `next(iterator, default)` for safe fallback


````python
# Build a lazy generator for even squares.
even_squares = (number * number for number in range(10) if number % 2 == 0)

# Print the generated values.
print(list(even_squares))
````

# 27. Real-World Example: Streaming Log Filter

Suppose we want to process logs lazily and keep only error lines.

Why generator style helps:
- line-by-line processing
- low memory usage
- clean pipeline design


````python
# Create sample log lines.
logs = [
    "INFO Application started",
    "ERROR Database timeout",
    "WARNING Disk usage high",
    "ERROR API key missing",
]

# Build a lazy filter for error lines only.
error_logs = (line for line in logs if line.startswith("ERROR"))

# Print the filtered lines.
for line in error_logs:
    # Show one error log at a time.
    print(line)
````

# 28. Mini Practical Example

This mini project combines multiple concepts:

- a generator pipeline
- lazy filtering
- running totals
- final list conversion for reporting


````python
# Create raw transaction values.
transactions = [100, -20, 50, 0, 30]

# Keep only positive transactions lazily.
positive_transactions = (amount for amount in transactions if amount > 0)

# Convert positive amounts into a running total generator.
def totals(numbers):
    # Start the total at zero.
    total = 0

    # Process each number in order.
    for number in numbers:
        # Add the current number to the running total.
        total += number

        # Yield the updated total.
        yield total


# Print the final running totals report.
print(list(totals(positive_transactions)))
````
