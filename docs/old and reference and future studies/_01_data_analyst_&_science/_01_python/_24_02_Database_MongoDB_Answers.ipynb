## Q30. Show a Try-Except-Finally Client Pattern


```python
# Problem Title: Show a Try-Except-Finally Client Pattern
# Step-by-step logic:
# 1. Read the MongoDB requirement carefully.
# 2. Build the required document, filter, update, stage, or helper.
# 3. Print the result clearly for verification.
# Dry run:
# The script creates the requested MongoDB pattern and prints the expected output shape.
# Time complexity:
# These notebook answers mostly run in O(1) time because they build and print fixed structures.

from pymongo import MongoClient

try:
    client = MongoClient('mongodb://localhost:27017/')
    print('Client created successfully')
except Exception as error:
    print('Handled error:', error)
finally:
    if 'client' in locals():
        client.close()
        print('Client closed')

# Test case:
expected_output = '`Client closed`'
print('Expected output:')
print(expected_output)
```
