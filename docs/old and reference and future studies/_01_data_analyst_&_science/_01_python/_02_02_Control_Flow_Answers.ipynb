## Q9. Count Digits

**Problem:** Return the number of digits in a non-negative integer.

**Approach explanation:**
- Read the input in the required format: `n: int`.
- Use control-flow rules to choose the correct branch or loop behavior.
- Return output in the required format: `int`.
- Handle edge cases from the constraints: n >= 0.
- Prefer the clearest correct solution first, then optimize when repeated scanning or nested loops appear.


````python
def count_digits(n: int) -> int:
    # Zero has one digit.
    if n == 0:
        return 1
    count = 0
    while n > 0:
        count += 1
        n //= 10
    return count

# Time Complexity: O(n) when a loop scans input; O(1) for fixed condition checks.
# Space Complexity: O(1) unless a result list/dictionary is explicitly built.
````

**Approach explanation:**
- Read the input in the required format: `password: str`.
- Use control-flow rules to choose the correct branch or loop behavior.
- Return output in the required format: `str`.
- Handle edge cases from the constraints: Password is string.
- Prefer the clearest correct solution first, then optimize when repeated scanning or nested loops appear.


````python
def password_strength(password: str) -> str:
    has_upper = any(ch.isupper() for ch in password)
    has_lower = any(ch.islower() for ch in password)
    has_digit = any(ch.isdigit() for ch in password)
    if len(password) >= 8 and has_upper and has_lower and has_digit:
        return "Strong"
    if len(password) >= 6:
        return "Medium"
    return "Weak"

# Time Complexity: O(n) when a loop scans input; O(1) for fixed condition checks.
# Space Complexity: O(1) unless a result list/dictionary is explicitly built.
````

## Q20. Prime Check

**Problem:** Return True if n is prime.

**Approach explanation:**
- Read the input in the required format: `n: int`.
- Use control-flow rules to choose the correct branch or loop behavior.
- Return output in the required format: `bool`.
- Handle edge cases from the constraints: n >= 0.
- Prefer the clearest correct solution first, then optimize when repeated scanning or nested loops appear.


````python
def prime_check(n: int) -> bool:
    if n < 2:
        return False
    divisor = 2
    while divisor * divisor <= n:
        if n % divisor == 0:
            return False
        divisor += 1
    return True

# Time Complexity: O(n) when a loop scans input; O(1) for fixed condition checks.
# Space Complexity: O(1) unless a result list/dictionary is explicitly built.
````

### Dry Run and Explanation

**Sample Input**
```python
n = 17
```

**Expected Output**
```python
True
```

Step-by-step:
- Start with the given input.

## Q22. Reverse Integer

**Problem:** Reverse digits of an integer while preserving sign.

**Approach explanation:**
- Read the input in the required format: `n: int`.

- Return output in the required format: `int`.
- Handle edge cases from the constraints: n is integer.
- Prefer the clearest correct solution first, then optimize when repeated scanning or nested loops appear.


````python
def reverse_integer(n: int) -> int:
    sign = -1 if n < 0 else 1
    n = abs(n)
    reversed_value = 0
    while n > 0:
        digit = n % 10
        reversed_value = reversed_value * 10 + digit
        n //= 10
    return sign * reversed_value

# Time Complexity: O(n) when a loop scans input; O(1) for fixed condition checks.
# Space Complexity: O(1) unless a result list/dictionary is explicitly built.
````

## Q30. Sieve Prime Count

**Problem:** Count primes less than n.

**Approach explanation:**
- Read the input in the required format: `n: int`.
- Use control-flow rules to choose the correct branch or loop behavior.
- Return output in the required format: `int`.
- Handle edge cases from the constraints: n >= 0.
- Prefer the clearest correct solution first, then optimize when repeated scanning or nested loops appear.


````python
def sieve_prime_count(n: int) -> int:
    if n < 2:
        return 0
    is_prime = [True] * n
    is_prime[0] = False
    is_prime[1] = False
    p = 2
    while p * p < n:
        if is_prime[p]:
            for multiple in range(p * p, n, p):
                is_prime[multiple] = False
        p += 1
    return sum(is_prime)

# Time Complexity: O(n) when a loop scans input; O(1) for fixed condition checks.
# Space Complexity: O(1) unless a result list/dictionary is explicitly built.
````
