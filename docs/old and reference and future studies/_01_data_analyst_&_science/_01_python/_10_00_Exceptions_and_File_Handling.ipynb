# Basic Syntax Explanation

## Exception syntax
- `try:` contains code that may fail.
- `except:` handles a specific error.
- `else:` runs only if no exception happened.
- `finally:` runs whether an error happened or not.
- `raise` creates an exception manually.

## File handling syntax
- `open(file_name, mode)` opens a file.
- Common modes: `'r'`, `'w'`, `'a'`, `'x'`.
- `with open(...) as file:` is the safest pattern.

## XLSX note
Python usually works with `.xlsx` files through libraries such as `openpyxl` or `pandas`, but the core file-handling ideas still matter first.


```python
# Use a try block around code that may fail
try:
    # Try converting text into an integer
    value = int("25")

    # Print the converted value when conversion succeeds
    print(value)
except ValueError:
    # Print a message if conversion fails
    print("Conversion failed")
```

# Concept 1: What an Exception Is

An exception is an error that happens while the program is running.

## Common built-in exceptions
- `ValueError`
- `TypeError`
- `ZeroDivisionError`
- `FileNotFoundError`
- `KeyError`
- `IndexError`

## Why this matters
Understanding the exception type helps you fix and handle the right problem.


```python
# Create a list with two items
items = [10, 20]

# Use try to safely access an invalid index
try:
    # This index does not exist
    print(items[5])
except IndexError:
    # Handle the exact error type
    print("Index is out of range")
```

# Concept 2: Using `try` and `except`

`try` and `except` let the program continue even when an operation fails.

## Why useful
- Avoids full program crashes
- Lets you show friendly messages
- Helps recover from expected problems

## Common mistake
Using a bare `except:` can hide real bugs if used carelessly.


```python
# Create sample text input
user_input = "hello"

# Try to convert the text into an integer
try:
    # This line raises ValueError because hello is not a number
    number = int(user_input)

    # Print the converted number if successful
    print(number)
except ValueError:
    # Show a helpful user-friendly message
    print("Please enter a valid number")
```

# Concept 3: Handling Multiple Exceptions

One block of code may fail in different ways.

## Why useful
You can write different responses for different problems.

## Example
Conversion may fail with `ValueError`, while division may fail with `ZeroDivisionError`.


```python
# Create sample values
raw_number = "10"
divisor = 0

# Try conversion and division
try:
    # Convert the string into an integer
    number = int(raw_number)

    # Divide by divisor
    result = number / divisor

    # Print the result if all steps succeed
    print(result)
except ValueError:
    # Handle invalid integer conversion
    print("Number conversion failed")
except ZeroDivisionError:
    # Handle division by zero
    print("Cannot divide by zero")
```

# Concept 4: `else` and `finally`

`else` runs only if no exception occurs. `finally` runs no matter what.

## Why useful
- `else` keeps success-only code clean.
- `finally` is useful for cleanup tasks.


```python
# Try a safe conversion
try:
    # Convert the text into an integer
    number = int("42")
except ValueError:
    # Print error text only if conversion fails
    print("Conversion failed")
else:
    # Print success message only when no exception occurs
    print("Converted successfully:", number)
finally:
    # This line always runs
    print("Execution finished")
```

# Concept 5: Raising Exceptions with `raise`

You can create your own error when invalid data is detected.

## Why useful
- Enforces business rules
- Stops bad data early
- Makes bugs easier to trace


```python
# Define a function that validates age
def validate_age(age):
    # Raise an error if age is negative
    if age < 0:
        raise ValueError("Age cannot be negative")

    # Return the valid age
    return age


# Use try to handle the raised exception
try:
    # Call the validation function with invalid data
    print(validate_age(-5))
except ValueError as error:
    # Print the actual error message
    print(error)
```

# Concept 6: Custom Exceptions

Custom exceptions help describe domain-specific problems more clearly.

## Why useful
- Gives meaningful error names
- Makes larger programs easier to understand
- Helps separate different failure cases


```python
# Create a custom exception class
class InvalidSalaryError(Exception):
    # No extra code is needed for this basic custom exception
    pass


# Define a function that validates salary
def set_salary(amount):
    # Raise the custom exception if salary is invalid
    if amount < 0:
        raise InvalidSalaryError("Salary cannot be negative")

    # Return the valid amount
    return amount


# Handle the custom exception
try:
    # Trigger the custom error
    print(set_salary(-1000))
except InvalidSalaryError as error:
    # Print the custom error text
    print(error)
```

# Concept 8: Writing to a File

Python can create and write content to files.

## Why useful
- Save reports
- Save logs
- Save user output for later use

## Edge case
If the file path is invalid or permission is denied, writing may fail.


```python
# Create a small demo file path
file_path = "sample_output.txt"

# Open the file in write mode
with open(file_path, "w", encoding="utf-8") as file:
    # Write the first line into the file
    file.write("Python file handling demo\n")

    # Write another line into the file
    file.write("Second line\n")


# Reopen the file to verify the content
with open(file_path, "r", encoding="utf-8") as file:
    # Read all content from the file
    content = file.read()

    # Print the file content
    print(content)
```

# Concept 9: Reading from a File

Python supports reading:
- the whole file,
- one line,
- all lines as a list.

## Methods
- `.read()`
- `.readline()`
- `.readlines()`


```python
# Create a demo file for reading
with open("sample_read.txt", "w", encoding="utf-8") as file:
    # Write multiple lines into the file
    file.write("Line 1\nLine 2\nLine 3")


# Read the file back
with open("sample_read.txt", "r", encoding="utf-8") as file:
    # Read all lines into a list
    lines = file.readlines()

    # Print the list of lines
    print(lines)
```

# Concept 10: Context Managers with `with`

`with open(...)` automatically closes the file even if an error occurs.

## Why this is best practice
- Cleaner code
- Safer resource handling
- Avoids forgetting `close()`


```python
# Use a context manager to read a file safely
with open("sample_read.txt", "r", encoding="utf-8") as file:
    # Read one line from the file
    first_line = file.readline()

    # Print the line
    print(first_line.strip())

# The file is automatically closed after the block ends
```

# Concept 11: Handling File Errors

Files can fail to open for many reasons.

## Common file-related exceptions
- `FileNotFoundError`
- `PermissionError`
- `IsADirectoryError`

## Real-world example
A scheduled script should not crash just because one expected input file is missing.


```python
# Try opening a file that does not exist
try:
    # This file is intentionally missing
    with open("missing_file.txt", "r", encoding="utf-8") as file:
        # Read the file if it exists
        print(file.read())
except FileNotFoundError:
    # Handle the missing file gracefully
    print("The requested file was not found")
```

# Concept 12: Parsing File Content Safely

File content is often messy, so conversion errors are common.

## Why useful
You may need to skip bad rows while still processing good ones.

## Real-world example
Reading numeric values from a report where some cells contain invalid text.


```python
# Create a file with mixed good and bad values
with open("numbers.txt", "w", encoding="utf-8") as file:
    # Write valid and invalid lines
    file.write("10\n20\nhello\n30")


# Start a running total
total = 0

# Read and parse the file safely
with open("numbers.txt", "r", encoding="utf-8") as file:
    # Loop through every line in the file
    for line in file:
        # Remove surrounding spaces and newline characters
        clean_line = line.strip()

        try:
            # Try converting the line to an integer
            total += int(clean_line)
        except ValueError:
            # Skip lines that are not valid integers
            print("Skipped invalid line:", clean_line)


# Print the sum of valid values
print(total)
```

# Concept 15: Real-World Mini Example

Let us combine exception handling and file processing.

## Goal
- Read values from a file
- Skip bad entries
- Compute a clean total
- Write a summary file


```python
# Create an input file with mixed values
with open("report_values.txt", "w", encoding="utf-8") as file:
    # Write both valid and invalid data
    file.write("100\n250\ninvalid\n300")


# Create variables for the summary
total = 0
skipped = 0

# Read the input file safely
with open("report_values.txt", "r", encoding="utf-8") as file:
    # Loop through each line
    for line in file:
        # Clean the line text
        value = line.strip()

        try:
            # Add valid integer values to the total
            total += int(value)
        except ValueError:
            # Count invalid values
            skipped += 1


# Write the summary report into an output file
with open("report_summary.txt", "w", encoding="utf-8") as file:
    # Save total and skipped count into the file
    file.write(f"Total: {total}\n")
    file.write(f"Skipped: {skipped}\n")


# Read the summary back and print it
with open("report_summary.txt", "r", encoding="utf-8") as file:
    # Print the generated summary text
    print(file.read())
```

# Concept 16: Common Mistakes and Best Practices

## Common mistakes
- Using bare `except:` too broadly
- Forgetting `with open(...)`
- Using `'w'` when you meant `'a'`
- Hiding exceptions without logging or printing useful details
- Ignoring encoding when working with text files

## Best practices
- Catch specific exceptions
- Use context managers
- Validate data before writing or processing
- Keep file paths clear and predictable
- Separate reading, cleaning, and writing into small steps


```python
# Define a clean helper function for safe integer conversion
def safe_to_int(text):
    # Try converting the text into an integer
    try:
        return int(text)
    except ValueError:
        # Return None when conversion fails
        return None


# Test the helper with valid and invalid values
print(safe_to_int("45"))
print(safe_to_int("oops"))
```
