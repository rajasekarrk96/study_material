## Q12. Queue Process All

**Problem:** Use `queue.Queue` to process all tasks in FIFO order and return the processed list.

**Step-by-step logic:**
- Read the required input shape: `tasks: list[str]`.
- Choose the correct concurrency or asyncio pattern for the task.
- Keep ordering, waiting, and coordination rules clear.
- Handle the stated constraint carefully: Use `put` and `get`.
- Return the final answer in the required format `list[str]`.


````python
def queue_process_all(tasks: list[str]) -> list[str]:
    # Import Queue for FIFO processing.
    from queue import Queue

    # Create a queue and load all tasks.
    queue = Queue()
    for task in tasks:
        queue.put(task)

    # Process every queued task in FIFO order.
    processed = []
    while not queue.empty():
        processed.append(queue.get())
    return processed

# Time Complexity: depends on the number of tasks, responses, or items processed above.
# Space Complexity: depends on extra queues, gathered results, counters, or formatted outputs created above.
````

````python
def async_queue_drain(items: list[str]) -> list[str]:
    # Import asyncio for async queue usage.
    import asyncio

    # Define an async queue workflow.
    async def main():
        queue = asyncio.Queue()
        for item in items:
            await queue.put(item)
        drained = []
        while not queue.empty():
            drained.append(await queue.get())
        return drained

    # Run and return the drained items.
    return asyncio.run(main())

# Time Complexity: depends on the number of tasks, responses, or items processed above.
# Space Complexity: depends on extra queues, gathered results, counters, or formatted outputs created above.
````

## Q35. Worker Result Formatter

**Problem:** Given worker result dictionaries with `task` and `ok`, return strings like `task1: ok` or `task2: failed`.

**Step-by-step logic:**
- Read the required input shape: `results: list[dict]`.
- Choose the correct concurrency or asyncio pattern for the task.
- Keep ordering, waiting, and coordination rules clear.
- Handle the stated constraint carefully: Preserve original order.
- Return the final answer in the required format `list[str]`.


````python
def worker_result_formatter(results: list[dict]) -> list[str]:
    # Convert worker results into readable status strings.
    return [f"{result['task']}: {'ok' if result['ok'] else 'failed'}" for result in results]

# Time Complexity: depends on the number of tasks, responses, or items processed above.
# Space Complexity: depends on extra queues, gathered results, counters, or formatted outputs created above.
````
