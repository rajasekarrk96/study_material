````python
def remove_duplicates_keep_order(items: list[int]) -> list[int]:
    seen = set()
    result = []
    for item in items:
        if item not in seen:
            seen.add(item)
            result.append(item)
    return result

# Time Complexity: based on loops/sorting/data-structure operations used above.
# Space Complexity: based on extra lists/sets/dicts/stacks created above.
````

## Q16. Stack Valid Parentheses

**Problem:** Check valid brackets containing (), {}, [].

**Approach explanation:**
- Identify the best data structure for the operation.
- Use input format `text: str` and return output format `bool`.
- Handle edge cases from constraints: String may be empty.
- Use brute force only when simple enough; use hash maps, sets, stacks, queues, heaps, or sorting when they improve clarity/performance.
- Keep the solution readable first, then reason about time and space complexity.


````python
def stack_valid_parentheses(text: str) -> bool:
    pairs = {")": "(", "}": "{", "]": "["}
    stack = []
    for ch in text:
        if ch in "({[":
            stack.append(ch)
        elif ch in pairs:
            if not stack or stack.pop() != pairs[ch]:
                return False
    return not stack

# Time Complexity: based on loops/sorting/data-structure operations used above.
# Space Complexity: based on extra lists/sets/dicts/stacks created above.
````

## Q17. Queue Simulation

**Problem:** Process queue by removing first k items.

**Approach explanation:**
- Identify the best data structure for the operation.
- Use input format `items: list[str], k: int` and return output format `list[str]`.
- Handle edge cases from constraints: 0 <= k.
- Use brute force only when simple enough; use hash maps, sets, stacks, queues, heaps, or sorting when they improve clarity/performance.
- Keep the solution readable first, then reason about time and space complexity.


````python
def queue_simulation(items: list[str], k: int) -> list[str]:
    from collections import deque
    queue = deque(items)
    for _ in range(min(k, len(queue))):
        queue.popleft()
    return list(queue)

# Time Complexity: based on loops/sorting/data-structure operations used above.
# Space Complexity: based on extra lists/sets/dicts/stacks created above.
````

- Keep the solution readable first, then reason about time and space complexity.


````python
def longest_consecutive(nums: list[int]) -> int:
    values = set(nums)
    best = 0
    for num in values:
        if num - 1 not in values:
            current = num
            streak = 1
            while current + 1 in values:
                current += 1
                streak += 1
            best = max(best, streak)
    return best

# Time Complexity: based on loops/sorting/data-structure operations used above.
# Space Complexity: based on extra lists/sets/dicts/stacks created above.
````

**Approach explanation:**
- Identify the best data structure for the operation.
- Use input format `nums: list[int], k: int` and return output format `list[int]`.
- Handle edge cases from constraints: 1 <= k <= len(nums).
- Use brute force only when simple enough; use hash maps, sets, stacks, queues, heaps, or sorting when they improve clarity/performance.
- Keep the solution readable first, then reason about time and space complexity.


````python
def sliding_window_maximum(nums: list[int], k: int) -> list[int]:
    from collections import deque
    dq = deque()
    result = []
    for i, num in enumerate(nums):
        while dq and dq[0] <= i - k:
            dq.popleft()
        while dq and nums[dq[-1]] <= num:
            dq.pop()
        dq.append(i)
        if i >= k - 1:
            result.append(nums[dq[0]])
    return result

# Time Complexity: based on loops/sorting/data-structure operations used above.
# Space Complexity: based on extra lists/sets/dicts/stacks created above.
````
