# Introduction

- **What is it?** A web API lets one program communicate with another over HTTP, usually by sending requests and receiving structured responses such as JSON.
- **Why it is used?** APIs let applications fetch data, submit forms, update records, trigger workflows, and integrate with online services.
- **Real-world relevance:** Weather apps, payment systems, dashboards, log collectors, machine learning services, and automation scripts all use APIs and HTTP requests regularly.


# Basic Syntax Explanation

The most common Python library for beginner-friendly HTTP work is `requests`.

```python
import requests

response = requests.get("https://example.com/api")
print(response.status_code)
data = response.json()
```

Important ideas:
- a **request** goes to a URL
- a **response** comes back with status code, headers, and body
- JSON is a very common response format
- APIs may require parameters, authentication, or request bodies


````python
# Basic Example with detailed comments

# Import the requests library used for HTTP requests.
import requests

# Store an example API URL as text.
url = "https://example.com/api/users"

# Build a small dictionary of query parameters.
params = {"page": 1}

# Print the example request parts instead of sending a real network call.
print("URL:", url)
print("Params:", params)

# Show what a typical response-handling flow would inspect.
print("Typical next steps: check status_code, read JSON, handle errors")
````

# 11. Status Codes

Important status code groups:

- `2xx` success
- `3xx` redirection
- `4xx` client error
- `5xx` server error

Common examples:
- `200 OK`
- `201 Created`
- `400 Bad Request`
- `401 Unauthorized`
- `404 Not Found`
- `500 Internal Server Error`


````python
# Store common status codes in a dictionary.
status_meanings = {200: "OK", 201: "Created", 404: "Not Found", 500: "Server Error"}

# Print all status code meanings.
for code, meaning in status_meanings.items():
    print(code, meaning)
````

# 23. Shorthand and Optimized Patterns

Useful patterns:

- use `response.json()` only after checking response type
- use `dict.get()` for missing fields
- use list comprehensions for extraction
- keep URLs, params, headers, and payloads separate
- centralize error handling for repeated request logic


````python
# Extract a field safely from possibly incomplete data.
record = {"name": "Asha"}

# Use get to avoid KeyError.
print(record.get("city", "Unknown"))
````

# 24. Real-World Mini Practical Example

This example simulates a typical API workflow without making a real network call.

Goal:
- prepare URL, headers, and params
- simulate a JSON response
- validate status
- extract a clean report


````python
# Build a request configuration.
url = "https://api.example.com/users"
headers = {"Accept": "application/json"}
params = {"page": 1}

# Simulate an API response object with status and JSON-like content.
status_code = 200
response_json = {
    "users": [
        {"name": "Asha", "city": "Pune"},
        {"name": "Ravi", "city": "Delhi"},
    ]
}

# Validate the simulated response.
if status_code == 200:
    # Extract user names into a simple report.
    names = [user["name"] for user in response_json["users"]]
    print(url)
    print(headers)
    print(params)
    print(names)
````
