## Introduction

### What is it?
CSV is a plain-text tabular format; Excel is a workbook format with sheets and rich metadata.

### Why it is used?
Most business tools export reports as CSV/XLSX. Analysts clean, validate, and transform these files.

### Real-world relevance
Data pipelines fail often because of wrong delimiter, wrong encoding, malformed rows, duplicate headers, and type inconsistencies.

## Basic Syntax Explanation

- `csv.reader()` -> rows as lists
- `csv.DictReader()` -> rows as dictionaries
- `csv.writer()` / `csv.DictWriter()` -> write output
- `open(..., encoding=...)` controls decoding/encoding
- `newline=""` avoids extra blank lines on Windows
- Excel read/write commonly uses `pandas.read_excel` and `DataFrame.to_excel`

```python
# Basic example: read a CSV and print rows
import csv
from pathlib import Path

sample_path = Path('basic_people.csv')
sample_path.write_text('name,age,city\nAsha,22,Delhi\nRavi,24,Pune\n', encoding='utf-8')

with open(sample_path, mode='r', encoding='utf-8', newline='') as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)
```

## Types / Variations of CSV

Detailed explanation of `Types / Variations of CSV` with practical usage, risks, and best practices for analyst workflows.

```python
import csv, io
csv_text = 'a,b,c\n1,2,3\n'
print(list(csv.reader(io.StringIO(csv_text), delimiter=',')))
sc_text = 'a;b;c\n1;2;3\n'
print(list(csv.reader(io.StringIO(sc_text), delimiter=';')))
tsv_text = 'a\tb\tc\n1\t2\t3\n'
print(list(csv.reader(io.StringIO(tsv_text), delimiter='\t')))
```

## Edge Cases

Detailed explanation of `Edge Cases` with practical usage, risks, and best practices for analyst workflows.

```python
import csv, io
text = 'name,comment\nAsha,"good, consistent progress"\n'
for row in csv.DictReader(io.StringIO(text)):
    print(row['name'], '|', row['comment'])
text2 = 'id,name,score\n1,Asha,\n'
for row in csv.DictReader(io.StringIO(text2)):
    print('missing score ->', row['score'])
```

## Advanced Usage: Dialects and Sniffer

Detailed explanation of `Advanced Usage: Dialects and Sniffer` with practical usage, risks, and best practices for analyst workflows.

```python
import csv, io
sample = 'col1;col2;col3\n10;20;30\n'
dialect = csv.Sniffer().sniff(sample)
reader = csv.reader(io.StringIO(sample), dialect)
print('Guessed delimiter:', dialect.delimiter)
print(list(reader))
```

## Real-world Example: Sales Import Cleanup

Detailed explanation of `Real-world Example: Sales Import Cleanup` with practical usage, risks, and best practices for analyst workflows.

```python
import csv
cleaned = []
with open('sales_dirty.csv', 'w', encoding='utf-8', newline='') as f:
    f.write('order_id,amount,status\n1001,120.5,paid\n1002,,pending\n1003,90,paid\n')
with open('sales_dirty.csv', 'r', encoding='utf-8', newline='') as f:
    for row in csv.DictReader(f):
        amount = float(row['amount']) if row['amount'] else 0.0
        cleaned.append({'order_id': row['order_id'], 'amount': amount, 'status': row['status']})
print(cleaned)
```

## Mini Practical: Encoding Recovery

Detailed explanation of `Mini Practical: Encoding Recovery` with practical usage, risks, and best practices for analyst workflows.

```python
from pathlib import Path
p = Path('encoding_demo.txt')
p.write_bytes('caf\xe9'.encode('latin-1'))
try:
    text = p.read_text(encoding='utf-8')
except UnicodeDecodeError:
    text = p.read_text(encoding='latin-1')
print(text)
```

## Excel Interop Basics with pandas

Detailed explanation of `Excel Interop Basics with pandas` with practical usage, risks, and best practices for analyst workflows.

```python
try:
    import pandas as pd
    df = pd.DataFrame({'name':['Asha','Ravi'], 'score':[88,92]})
    df.to_excel('scores.xlsx', index=False, sheet_name='Scores')
    read_back = pd.read_excel('scores.xlsx', sheet_name='Scores')
    print(read_back.to_dict(orient='records'))
except Exception as e:
    print('Install pandas/openpyxl for Excel demo:', e)
```
