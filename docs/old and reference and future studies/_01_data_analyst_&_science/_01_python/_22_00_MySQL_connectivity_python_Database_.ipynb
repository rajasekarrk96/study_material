# Basic Syntax Explanation

## Standard workflow
1. Install `mysql-connector-python`
2. Import `mysql.connector`
3. Create a connection with host, user, password, database, and port
4. Create a cursor with `connection.cursor()`
5. Run SQL using `execute()` or `executemany()`
6. Use `commit()` for `INSERT`, `UPDATE`, and `DELETE`
7. Use fetch methods for `SELECT`
8. Close the cursor and connection

## Operators and patterns involved
- Python assignment: `=`
- SQL comparison: `=`, `>`, `<`, `>=`, `<=`, `!=`
- SQL logic: `AND`, `OR`, `NOT`
- SQL pattern matching: `LIKE`
- SQL range and grouping: `BETWEEN`, `IN`, `GROUP BY`, `ORDER BY`
- MySQL placeholder for parameterized queries: `%s`

## Safety rule
Always use parameterized queries instead of directly joining user input into SQL strings.


```python
# Import the MySQL connector package so Python can talk to a MySQL server.
import mysql.connector

# Import the Error class so connection and query problems can be handled clearly.
from mysql.connector import Error

try:
    # Open a connection to the MySQL server.
    # Replace the values below with your real server details before running.
    connection = mysql.connector.connect(
        host='localhost',
        user='root',
        password='your_password',
        database='school_db',
        port=3306
    )

    # Check whether the connection was established successfully.
    if connection.is_connected():
        print('Connected to MySQL successfully')

        # Create a cursor object.
        # The cursor sends SQL statements from Python to MySQL.
        cursor = connection.cursor()

        # Run a very small query to confirm that the server responds.
        cursor.execute('SELECT DATABASE()')

        # Read one result row from the query.
        active_database = cursor.fetchone()

        # Print the active database name.
        print('Active database:', active_database)

        # Close the cursor after the query work is complete.
        cursor.close()
except Error as error:
    # Print the exact MySQL-related error if something fails.
    print('MySQL error:', error)
finally:
    # Close the connection only if it was created and is still open.
    if 'connection' in locals() and connection.is_connected():
        connection.close()
        print('Connection closed')
```

## Concept 4: Establishing a Connection


A connection is the live communication link between Python and MySQL.

### Why it matters
- No SQL query can run without a connection
- Connection status should be checked before using it

### Common mistakes
- Forgetting to verify `is_connected()`
- Leaving idle connections open
- Opening too many connections unnecessarily


```python
import mysql.connector
from mysql.connector import Error

try:
    # Open the MySQL connection.
    connection = mysql.connector.connect(
        host='localhost',
        user='root',
        password='your_password',
        database='school_db'
    )

    # Confirm the connection state.
    if connection.is_connected():
        print('Connection established')
except Error as error:
    print('Connection failed:', error)
finally:
    # Close the connection when finished.
    if 'connection' in locals() and connection.is_connected():
        connection.close()
        print('Connection closed')
```

## Concept 5: Connection Object vs Cursor Object


The connection and cursor play different roles.

### Connection object
- Opens the database session
- Commits or rolls back transactions
- Creates cursor objects
- Closes the session

### Cursor object
- Sends SQL commands
- Fetches rows
- Keeps metadata such as `rowcount` or `description`

### Common mistake
- Trying to call `fetchall()` on the connection object instead of the cursor


```python
import mysql.connector

# This snippet shows the normal object creation pattern.
connection = mysql.connector.connect(
    host='localhost',
    user='root',
    password='your_password',
    database='school_db'
)

# Create a cursor from the connection.
cursor = connection.cursor()

# Print the object types to show that they are different objects.
print('Connection type:', type(connection).__name__)
print('Cursor type:', type(cursor).__name__)

# Close objects in reverse order of use.
cursor.close()
connection.close()
```

## Concept 6: Creating a Database from Python


Python can create a database by connecting to the server first and then executing a `CREATE DATABASE` statement.

### Variation
- Create database only if it does not exist using `IF NOT EXISTS`

### Common mistakes
- Connecting directly to a missing database before creating it
- Forgetting privileges


```python
import mysql.connector
from mysql.connector import Error

try:
    # Connect to the server without selecting a specific database first.
    connection = mysql.connector.connect(
        host='localhost',
        user='root',
        password='your_password'
    )
    cursor = connection.cursor()

    # Create the database only if it does not already exist.
    cursor.execute('CREATE DATABASE IF NOT EXISTS school_db')
    print('Database checked or created successfully')

    cursor.close()
except Error as error:
    print('Database creation error:', error)
finally:
    if 'connection' in locals() and connection.is_connected():
        connection.close()
```

## Concept 7: Creating Tables and Choosing Data Types


Tables should be designed carefully before inserting data.

### Important table design ideas
- Use a primary key
- Use suitable data types like `INT`, `VARCHAR`, `DATE`, `DECIMAL`
- Use `NOT NULL` for required columns
- Use `UNIQUE` where duplicates are not allowed

### Common mistakes
- Using text for everything

- Creating unclear column names


```python
create_table_query = '''
CREATE TABLE IF NOT EXISTS students (
    id INT AUTO_INCREMENT PRIMARY KEY,
    name VARCHAR(100) NOT NULL,
    email VARCHAR(150) UNIQUE NOT NULL,
    age INT,
    course VARCHAR(100) NOT NULL
)
'''

# Print the query so the structure can be studied clearly.
print(create_table_query)
```

## Concept 8: Inserting Data Safely


`INSERT` adds new rows to a table.

### Safe approach
Use parameterized queries with `%s` placeholders.

### Why it matters
- Prevents SQL injection
- Keeps data values separate from SQL structure
- Reduces quoting mistakes

### Common mistakes
- Using f-strings for SQL values
- Forgetting `commit()`
- Mismatching placeholder count and value count


```python
insert_query = 'INSERT INTO students (name, email, age, course) VALUES (%s, %s, %s, %s)'

# Prepare one row of values that matches the placeholder order exactly.
student_values = ('Aman', 'aman@example.com', 20, 'Python')

# Print the query and value tuple.
print(insert_query)
print(student_values)
```

## Concept 9: Reading Data with `SELECT`


`SELECT` reads stored rows.

### Variations
- `SELECT *`
- Specific columns only
- `WHERE` filters
- `ORDER BY`
- `LIMIT`

### Edge cases
- Zero matching rows
- Very large result sets
- Result order not guaranteed without `ORDER BY`


```python
select_query = '''
SELECT name, course
FROM students
WHERE age >= %s
ORDER BY name ASC
LIMIT 5
'''

# Store the filter value separately for safe execution.
filter_values = (18,)

print(select_query)
print(filter_values)
```

## Concept 10: Fetch Methods


After a `SELECT`, the cursor can fetch results in different ways.

### Variations
- `fetchone()`: one row
- `fetchmany(size)`: next group of rows
- `fetchall()`: all remaining rows

### Common mistakes
- Calling fetch methods before executing a query
- Expecting `fetchall()` to repeat the same result without re-running the query


```python
# Show the method names and their meanings for easy review.
fetch_methods = {
    'fetchone()': 'Returns one row or None',
    'fetchmany(3)': 'Returns the next 3 rows as a list',
    'fetchall()': 'Returns all remaining rows'
}

for method, meaning in fetch_methods.items():
    print(method, '->', meaning)
```

## Concept 11: Updating Rows


`UPDATE` changes rows that already exist.

### Important rule
Use a `WHERE` clause unless you intentionally want to update every row.

### Common mistakes
- Missing `WHERE`
- Updating the wrong column
- Forgetting `commit()`


```python
update_query = 'UPDATE students SET course = %s WHERE email = %s'

# First value becomes the new course.
# Second value identifies which row should be updated.
update_values = ('Advanced Python', 'aman@example.com')

print(update_query)
print(update_values)
```

## Concept 12: Deleting Rows


`DELETE` removes rows.

### Variations
- Delete one row
- Delete many rows with a broader condition
- Delete old log rows by date

### Edge case
Deleting a non-existing row usually affects zero rows instead of raising an error.

### Common mistake
- Running `DELETE` without `WHERE`


```python
delete_query = 'DELETE FROM students WHERE email = %s'

# Identify exactly which row to remove.
delete_values = ('aman@example.com',)

print(delete_query)
print(delete_values)
```

## Concept 13: Transactions, Commit, and Rollback


A transaction groups related SQL changes into one logical unit.

### Why it is used
- Keeps data consistent
- Prevents partial updates
- Lets the program undo changes with `rollback()` if an error happens

### Real-world example
Money transfer between two bank accounts should either fully succeed or fully fail.

### Common mistakes
- Forgetting to commit successful work
- Forgetting to roll back failed multi-step updates


```python
# Write transaction-related queries for a transfer operation.
withdraw_query = 'UPDATE accounts SET balance = balance - %s WHERE account_id = %s'
deposit_query = 'UPDATE accounts SET balance = balance + %s WHERE account_id = %s'

# Store values separately for safety and clarity.
withdraw_values = (500, 101)
deposit_values = (500, 202)

print(withdraw_query)
print(withdraw_values)
print(deposit_query)
print(deposit_values)
```

## Concept 14: Error Handling


Database code can fail because of invalid credentials, duplicate keys, bad SQL, missing tables, or server outages.

### Good practice
- Use `try` / `except` / `finally`
- Catch `mysql.connector.Error`
- Roll back failed writes
- Close the connection in `finally`

### Common mistakes
- Swallowing errors silently
- Not closing resources after failure


```python
import mysql.connector
from mysql.connector import Error

try:
    # Try a connection that may fail if server details are wrong.
    connection = mysql.connector.connect(
        host='localhost',
        user='root',
        password='wrong_password',
        database='school_db'
    )
except Error as error:
    # Display the error so the learner can diagnose the issue.
    print('Handled MySQL error:', error)
finally:
    # Close the connection only if it exists and is open.
    if 'connection' in locals() and connection.is_connected():
        connection.close()
```

## Concept 15: Bulk Insert with `executemany()`


When many rows must be inserted, `executemany()` is cleaner than repeating `execute()` manually.

### Benefits
- Less repeated code
- Easier imports from lists or CSV-derived data
- Better readability

### Common mistake
- Passing one tuple instead of a list of tuples


```python
bulk_insert_query = 'INSERT INTO students (name, email, age, course) VALUES (%s, %s, %s, %s)'

# Store multiple student rows inside a list of tuples.
bulk_values = [
    ('Aman', 'aman@example.com', 20, 'Python'),
    ('Bina', 'bina@example.com', 21, 'SQL'),
    ('Charu', 'charu@example.com', 22, 'MySQL')
]

print(bulk_insert_query)
for row in bulk_values:
    print(row)
```

## Concept 16: Filtering, Sorting, Aggregation, and Grouping


Real queries often need reporting features, not just raw row retrieval.

### Important tools
- `WHERE` for filtering
- `ORDER BY` for sorting
- `COUNT`, `SUM`, `AVG`, `MIN`, `MAX`
- `GROUP BY` for summary reports

### Real-world examples
- Count students per course
- Total sales per seller
- Average score by subject


```python
report_query = '''
SELECT course, COUNT(*) AS total_students, AVG(age) AS average_age
FROM students
GROUP BY course
ORDER BY total_students DESC
'''

print(report_query)
```

## Concept 17: Joins


Joins combine related tables.

### Basic variation
- `INNER JOIN` returns matching rows from both tables

### Example
- `students.id`
- `enrollments.student_id`

### Common mistakes
- Joining on the wrong columns
- Forgetting table prefixes when the same column names appear in multiple tables


```python
join_query = '''
SELECT students.name, enrollments.batch_name
FROM students
INNER JOIN enrollments
    ON students.id = enrollments.student_id
ORDER BY students.name
'''

print(join_query)
```

## Concept 21: Mini Practical Example - Student Management with MySQL


This mini practical combines many ideas:
- Connect to MySQL
- Create a table
- Insert rows
- Read filtered rows
- Update one row
- Display final results

This is the type of workflow beginners often build in real projects.


```python
import mysql.connector
from mysql.connector import Error

try:
    # Connect to the target database.
    connection = mysql.connector.connect(
        host='localhost',
        user='root',
        password='your_password',
        database='school_db'
    )
    cursor = connection.cursor()

    # Create a table if it does not already exist.
    cursor.execute(
        '''
        CREATE TABLE IF NOT EXISTS student_demo (
            id INT AUTO_INCREMENT PRIMARY KEY,
            name VARCHAR(100) NOT NULL,
            course VARCHAR(100) NOT NULL,
            marks INT NOT NULL
        )
        '''
    )

    # Insert multiple sample rows.
    cursor.executemany(
        'INSERT INTO student_demo (name, course, marks) VALUES (%s, %s, %s)',
        [
            ('Aman', 'Python', 88),
            ('Bina', 'Python', 94),
            ('Charu', 'MySQL', 81)
        ]
    )
    connection.commit()

    # Read only Python learners.
    cursor.execute(
        'SELECT name, marks FROM student_demo WHERE course = %s ORDER BY marks DESC',
        ('Python',)
    )
    print('Python learners:', cursor.fetchall())

    # Update one row and save the change.
    cursor.execute(
        'UPDATE student_demo SET marks = marks + %s WHERE name = %s',
        (5, 'Aman')
    )
    connection.commit()

    # Display final results.
    cursor.execute('SELECT name, course, marks FROM student_demo ORDER BY name')
    print('Final rows:', cursor.fetchall())

    cursor.close()
except Error as error:
    print('MySQL practical error:', error)
finally:
    if 'connection' in locals() and connection.is_connected():
        connection.close()
```
