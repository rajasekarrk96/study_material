# Introduction

## What is it?
MongoDB is a NoSQL document database. Instead of storing data in fixed rows and columns, MongoDB stores data as flexible JSON-like documents.

## Why is it used?
- Real applications often store data that changes shape over time
- Nested objects and arrays fit naturally into document databases
- MongoDB is useful for fast development and flexible schema design
- Python can automate inserts, queries, updates, reporting, and data pipelines with MongoDB

## Real-world relevance
- Product catalogs with different attributes per product
- User profiles and activity logs
- Content management systems
- Event tracking and analytics
- IoT and sensor data
- Modern web backends with nested JSON data


# Basic Syntax Explanation

## Standard workflow
1. Install `pymongo`
2. Import `MongoClient` from `pymongo`
3. Create a client connection with the MongoDB URI
4. Select a database
5. Select a collection
6. Insert, find, update, or delete documents
7. Close the client if needed

## Important concepts and operators
- Document access in Python: `document['name']`
- Common query operators: `$gt`, `$gte`, `$lt`, `$lte`, `$in`, `$ne`
- Common update operators: `$set`, `$inc`, `$push`, `$unset`
- Projection selects which fields to return
- Sorting uses `.sort()`
- Limiting uses `.limit()`

## Safety rule
Validate document structure carefully and avoid assuming every field exists in every document.


```python
# Import MongoClient so Python can connect to a MongoDB server.
from pymongo import MongoClient

try:
    # Create a MongoDB client connection.
    # Replace the URI if your MongoDB server uses different settings.
    client = MongoClient('mongodb://localhost:27017/')

    # Select a database named school_db.
    database = client['school_db']

    # Select a collection named students.
    collection = database['students']

    # Print a simple success message and names for confirmation.
    print('Connected to MongoDB successfully')
    print('Database name:', database.name)
    print('Collection name:', collection.name)
finally:
    # Close the client connection after the demonstration is complete.
    client.close()
```

## Concept 1: What MongoDB Is and Why It Is Used


MongoDB is a document database that stores data in flexible documents rather than fixed relational rows.

### Basic concepts
- Database: container of collections
- Collection: group of related documents
- Document: one record, usually like a Python dictionary
- Field: one key-value pair inside a document

### Variations
- Local MongoDB server
- MongoDB Atlas cloud database
- Replica sets and sharded clusters in larger systems

### Common mistakes
- Thinking MongoDB documents must all have identical fields
- Ignoring nested structures
- Treating collections exactly like SQL tables without adjusting design thinking


```python
# Store the core MongoDB structure idea in Python variables.
database_name = 'school_db'               # Name of the database
collection_name = 'students'             # Name of the collection
document_fields = ['name', 'course']     # Example document fields

# Print the structure so it is easy to visualize.
print('Database:', database_name)
print('Collection:', collection_name)
print('Fields:', document_fields)
```

## Concept 4: Database and Collection Access


After connecting, Python code selects a database and then a collection.

### Important idea
MongoDB creates databases and collections lazily when data is first inserted.

### Common mistakes
- Assuming the database must already contain collections before code can reference them
- Confusing collection names with document fields


```python
from pymongo import MongoClient

# Create the client connection object.
client = MongoClient('mongodb://localhost:27017/')

# Select a database using dictionary-style access.
database = client['school_db']

# Select a collection inside the database.
collection = database['students']

print('Database:', database.name)
print('Collection:', collection.name)

client.close()
```

## Concept 5: Understanding Documents and Flexible Schema


MongoDB documents are usually similar to Python dictionaries and can contain nested documents and arrays.

### Why flexibility matters
- Different products may have different fields
- User profiles may grow over time
- Nested information can be stored naturally

### Edge cases
- Some documents may miss fields that others have
- Nested data may require careful access


```python
# Create two documents with slightly different shapes.
student_one = {'name': 'Aman', 'course': 'Python', 'marks': 88}
student_two = {'name': 'Bina', 'course': 'MongoDB', 'skills': ['Python', 'MongoDB']}

print(student_one)
print(student_two)
```

## Concept 6: The `_id` Field


Every MongoDB document has a unique `_id` field.

### Why it matters
- Uniquely identifies each document
- Helps in updates, deletes, and direct lookups

### Common mistakes
- Forgetting that MongoDB auto-generates `_id` if it is not supplied
- Creating duplicate custom `_id` values


```python
# Show what a custom _id field might look like.
student_document = {
    '_id': 1,
    'name': 'Aman',
    'course': 'Python'
}

print(student_document)
```

## Concept 7: Inserting One Document


`insert_one()` adds one document to a collection.

### Why it is useful
- Good for single records
- Simple way to begin learning document insertion

### Common mistakes
- Inserting malformed documents
- Assuming insert automatically validates every field unless validation rules exist


```python
document = {
    'name': 'Aman',
    'email': 'aman@example.com',
    'course': 'Python',
    'marks': 88
}

print('Document to insert:')
print(document)
```

## Concept 8: Inserting Many Documents


`insert_many()` is useful when adding multiple documents in one operation.

### Benefits
- Cleaner batch inserts
- Better for imported lists of records

### Common mistakes
- Passing one dictionary instead of a list of dictionaries


```python
documents = [
    {'name': 'Aman', 'course': 'Python'},
    {'name': 'Bina', 'course': 'MongoDB'},
    {'name': 'Charu', 'course': 'Data Science'}
]

for document in documents:
    print(document)
```

## Concept 9: Reading Documents with `find_one()` and `find()`


MongoDB reads documents with methods like `find_one()` and `find()`.

### Variations
- `find_one()` returns one document
- `find()` returns a cursor over many documents

### Edge cases
- `find_one()` may return `None`
- A query may match zero documents


```python
single_filter = {'name': 'Aman'}
many_filter = {'course': 'Python'}

print('find_one filter:', single_filter)
print('find filter:', many_filter)
```

## Concept 10: Query Filters and Comparison Operators


Filters decide which documents should match a query.

### Common operators
- Equality: `{'course': 'Python'}`
- Greater than: `{'marks': {'$gt': 80}}`
- Less than: `{'marks': {'$lt': 50}}`
- In list: `{'course': {'$in': ['Python', 'MongoDB']}}`

### Common mistakes
- Using SQL syntax in a MongoDB filter
- Forgetting braces around operator dictionaries


```python
filter_examples = [
    {'course': 'Python'},
    {'marks': {'$gt': 80}},
    {'course': {'$in': ['Python', 'MongoDB']}}
]

for example in filter_examples:
    print(example)
```

## Concept 11: Projection


Projection controls which fields are returned from matching documents.

### Why it matters
- Reduces unnecessary data transfer
- Hides fields that are not needed

### Common mistakes
- Mixing include and exclude fields incorrectly in one projection
- Forgetting `_id` is included by default unless excluded


```python
projection = {
    'name': 1,
    'course': 1,
    '_id': 0
}

print(projection)
```

## Concept 12: Sorting and Limiting


MongoDB can sort result documents and limit how many are returned.

### Why it matters
- Useful for top students, latest events, highest scores, or pagination-style displays

### Common mistakes
- Sorting after converting the cursor to a list in a way that loses database efficiency


```python
sort_field = 'marks'
sort_order = -1   # -1 means descending, 1 means ascending
limit_count = 5

print('Sort field:', sort_field)
print('Sort order:', sort_order)
print('Limit:', limit_count)
```

## Concept 13: Updating Documents with `$set`


`$set` updates one or more fields in a document without replacing the whole document.

### Common mistakes
- Forgetting `$set` and accidentally replacing the full document
- Using the wrong filter and updating the wrong document


```python
update_filter = {'email': 'aman@example.com'}
update_operation = {'$set': {'course': 'Advanced Python'}}

print(update_filter)
print(update_operation)
```

## Concept 14: Numeric Updates with `$inc`


`$inc` increases or decreases numeric values.

### Real-world uses
- Increment page views
- Add marks
- Track counters


```python
update_filter = {'name': 'Aman'}
update_operation = {'$inc': {'marks': 5}}

print(update_filter)
print(update_operation)
```

## Concept 15: Array Updates with `$push` and `$addToSet`


Arrays are common in MongoDB for tags, skills, and related items.

### Variations
- `$push` always adds an item
- `$addToSet` adds only if it is not already present

### Common mistake
- Using `$push` when duplicate values should be prevented


```python
push_operation = {'$push': {'skills': 'MongoDB'}}
add_to_set_operation = {'$addToSet': {'skills': 'MongoDB'}}

print(push_operation)
print(add_to_set_operation)
```

## Concept 16: Deleting Documents


MongoDB deletes documents with `delete_one()` or `delete_many()`.

### Edge case
- A delete filter may match zero documents

### Common mistake
- Using a broad filter accidentally and deleting too many documents


```python
delete_filter = {'email': 'aman@example.com'}

print(delete_filter)
```

## Concept 17: Replacing Documents


`replace_one()` replaces the full matching document.

### Why it is different
- Unlike `$set`, replacement overwrites the old document structure

### Common mistake
- Forgetting required fields during replacement


```python
replace_filter = {'_id': 1}
new_document = {'_id': 1, 'name': 'Aman', 'course': 'MongoDB', 'marks': 93}

print(replace_filter)
print(new_document)
```

## Concept 18: Aggregation Basics


Aggregation pipelines process documents in stages.

### Common stages
- `$match`
- `$group`
- `$sort`
- `$project`

### Real-world use
- Total sales per product
- Average marks by course
- Count documents by category


```python
pipeline = [
    {'$match': {'marks': {'$gte': 80}}},
    {'$group': {'_id': '$course', 'average_marks': {'$avg': '$marks'}}},
    {'$sort': {'average_marks': -1}}
]

for stage in pipeline:
    print(stage)
```

## Concept 19: Error Handling and Connection Safety


MongoDB operations can fail because of server availability, authentication issues, network problems, or malformed operations.

### Good practice
- Use `try` / `except` / `finally`
- Close the client when the work is finished
- Validate document structure before update logic depends on it


```python
from pymongo import MongoClient

try:
    client = MongoClient('mongodb://localhost:27017/')
    print('MongoDB client created')
except Exception as error:
    print('Handled MongoDB error:', error)
finally:
    if 'client' in locals():
        client.close()
```

## Concept 22: Mini Practical Example - Student Profiles in MongoDB


This mini practical combines:
- Connection setup
- Database and collection access
- Insert documents
- Query by filter
- Update a field
- Add an array skill

This is close to how MongoDB is used in small application modules.


```python
from pymongo import MongoClient

client = MongoClient('mongodb://localhost:27017/')
database = client['school_db']
collection = database['student_demo']

# Build example documents that could be inserted into MongoDB.
documents = [
    {'name': 'Aman', 'course': 'Python', 'marks': 88, 'skills': ['Python']},
    {'name': 'Bina', 'course': 'MongoDB', 'marks': 94, 'skills': ['MongoDB']},
    {'name': 'Charu', 'course': 'Python', 'marks': 81, 'skills': []}
]

# Print the steps that a practical MongoDB workflow would use.
print('Collection name:', collection.name)
print('Documents to insert:')
for document in documents:
    print(document)

print('Filter for Python learners:', {'course': 'Python'})
print('Update Aman marks:', {'$inc': {'marks': 5}})
print('Add skill to Charu:', {'$addToSet': {'skills': 'MongoDB'}})

client.close()
```
