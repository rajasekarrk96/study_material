## Q11. Custom Countdown Iterator

**Problem:** Create a custom iterator class that counts down from n to 1 and return its output as a list.

**Step-by-step logic:**
- Read the required input shape: `n: int`.
- Choose the correct iterator or generator pattern for the task.
- Handle exhaustion, lazy evaluation, and edge cases carefully.
- Return the final value in the required format `list[int]`.
- Respect the constraint: 0 <= n <= 50.


````python
def custom_countdown_iterator(n: int) -> list[int]:
    # Define a custom iterator class.
    class Countdown:
        # Save the starting value.
        def __init__(self, start):
            # Store the current state.
            self.current = start

        # Return the iterator object itself.
        def __iter__(self):
            # Return self because this object manages the iteration state.
            return self

        # Produce the next value.
        def __next__(self):
            # Stop when the counter reaches zero.
            if self.current <= 0:
                raise StopIteration

            # Save the current value.
            value = self.current

            # Decrease the state for the next call.
            self.current -= 1

            # Return the produced value.
            return value

    # Return all values from the iterator as a list.
    return list(Countdown(n))

# Time Complexity: depends on the number of elements processed or generated above.
# Space Complexity: depends on whether results are streamed lazily or collected into lists.
````

## Q13. Batch Items

- Choose the correct iterator or generator pattern for the task.
- Handle exhaustion, lazy evaluation, and edge cases carefully.
- Return the final value in the required format `list[list[int]]`.
- Respect the constraint: size >= 1.


````python
def batch_items(items: list[int], size: int) -> list[list[int]]:
    # Define a generator for batches.
    def batches():
        # Move forward in batch-sized jumps.
        for start in range(0, len(items), size):
            # Yield one slice at a time.
            yield items[start:start + size]

    # Convert the batches to a list.
    return list(batches())

# Time Complexity: depends on the number of elements processed or generated above.
# Space Complexity: depends on whether results are streamed lazily or collected into lists.
````

## Q35. Task Queue Iterator

**Problem:** Create a custom iterator class over a list of tasks and return the tasks in processing order.

**Step-by-step logic:**
- Read the required input shape: `tasks: list[str]`.
- Choose the correct iterator or generator pattern for the task.
- Handle exhaustion, lazy evaluation, and edge cases carefully.
- Return the final value in the required format `list[str]`.
- Respect the constraint: The list may be empty.


````python
def task_queue_iterator(tasks: list[str]) -> list[str]:
    # Define an iterator class over the task list.
    class TaskQueue:
        # Save the task values and starting position.
        def __init__(self, values):
            self.values = values
            self.index = 0

        # Return the iterator object.
        def __iter__(self):
            return self

        # Return one task at a time.
        def __next__(self):
            if self.index >= len(self.values):
                raise StopIteration
            current = self.values[self.index]
            self.index += 1
            return current

    # Return all tasks in processing order.
    return list(TaskQueue(tasks))

# Time Complexity: depends on the number of elements processed or generated above.
# Space Complexity: depends on whether results are streamed lazily or collected into lists.
````
