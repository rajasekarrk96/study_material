### Q1: Palindrome Check
Write a function `is_palindrome(s: str) -> bool` that takes a string `s` and returns `True` if it is a palindrome, and `False` otherwise. Use string slicing.

**Example 1:**
```
Input: s = "racecar"
Output: true
```
**Example 2:**
```
Input: s = "hello"
Output: false
```

**Constraints:**
* `1 <= s.length <= 10^5`
* `s` consists only of printable ASCII characters.

````python
def is_palindrome(s: str) -> bool:
    # Using string slicing to reverse the string
    return s == s[::-1]

# Test
print(is_palindrome("racecar")) # True
print(is_palindrome("hello"))   # False
````

### Q2: Length of Last Word
Write a function `length_of_last_word(s: str) -> int` that returns the length of the last word in the string. The string may contain trailing spaces. Hint: Use `split()` and `strip()`.

**Example 1:**
```
Input: s = "Hello World"
Output: 5
Explanation: The last word is "World" with length 5.
```
**Example 2:**
```
Input: s = "   fly me   to   the moon  "
Output: 4
Explanation: The last word is "moon" with length 4.
```

**Constraints:**
* `1 <= s.length <= 10^4`
* `s` consists of only English letters and spaces ' '.
* There will be at least one word in `s`.

````python
def length_of_last_word(s: str) -> int:
    # strip() removes trailing spaces, split() splits by space
    words = s.strip().split()
    if not words:
        return 0
    return len(words[-1])

# Test
print(length_of_last_word("Hello World")) # 5
print(length_of_last_word("   fly me   to   the moon  ")) # 4
````

### Q3: Float to Int Casting & Math
Write a function `process_number(s: str) -> int`. You are given a string `s` representing a decimal number (e.g., "15.75"). Cast it to a float, then cast it to an integer (truncating the decimal), and finally return the integer raised to the power of 2 using the `**` operator.

**Example 1:**
```
Input: s = "15.75"
Output: 225
Explanation: float("15.75") is 15.75. int(15.75) is 15. 15^2 is 225.
```
**Example 2:**
```
Input: s = "9.99"
Output: 81
```

**Constraints:**
* `1 <= s.length <= 10`

### Q4: Check if Number is Even or Odd using Bitwise
Write a function `is_even(n: int) -> bool` that returns `True` if `n` is even and `False` if `n` is odd. You must use the bitwise `&` operator, not the modulo `%` operator.

**Example 1:**
```
Input: n = 10
Output: true
```
**Example 2:**
```
Input: n = 7
Output: false
```

**Constraints:**
* `0 <= n <= 10^9`

````python
def is_even(n: int) -> bool:
    # Bitwise AND with 1 checks the least significant bit.
    # If it's 0, the number is even.
    return (n & 1) == 0

# Test
print(is_even(10)) # True
print(is_even(7))  # False
````

### Q5: Single Number
Given a list of integers where every element appears twice except for one, find that single one. Write a function `single_number(nums: list[int]) -> int`. Use the XOR `^` operator to solve this optimally.

**Example 1:**
```
Input: nums = [4,1,2,1,2]
Output: 4
```
**Example 2:**
```
Input: nums = [2,2,1]
Output: 1
```

**Constraints:**
* `1 <= nums.length <= 3 * 10^4`
* `-3 * 10^4 <= nums[i] <= 3 * 10^4`
* Each element in the array appears twice except for one element which appears only once.

````python
def single_number(nums: list[int]) -> int:
    # XORing a number with itself gives 0.
    # XORing all numbers will leave only the single number.
    result = 0
    for num in nums:
        result ^= num
    return result

# Test
print(single_number([4, 1, 2, 1, 2])) # 4
````

### Q7: Find the Difference
You are given two strings `s` and `t`. String `t` is generated by randomly shuffling `s` and adding one more letter at a random position. Write a function `find_the_difference(s: str, t: str) -> str` that returns the letter that was added. Hint: Use `ord()` and `chr()`, or XOR.

**Example 1:**
```
Input: s = "abcd", t = "abcde"
Output: "e"
Explanation: 'e' is the letter that was added.
```
**Example 2:**
```
Input: s = "", t = "y"
Output: "y"
```

**Constraints:**
* `0 <= s.length <= 1000`
* `t.length == s.length + 1`
* `s` and `t` consist of lowercase English letters.

````python
def find_the_difference(s: str, t: str) -> str:
    # Using ASCII values sum difference
    sum_s = sum(ord(c) for c in s)
    sum_t = sum(ord(c) for c in t)
    return chr(sum_t - sum_s)

# Test
print(find_the_difference("abcd", "abcde")) # e
````

### Q8: Reverse Words in a String
Write a function `reverse_words(s: str) -> str`. Given an input string `s`, reverse the order of the words. The words should be separated by a single space, and there should be no leading or trailing spaces.

**Example 1:**
```
Input: s = "the sky is blue"
Output: "blue is sky the"
```
**Example 2:**
```
Input: s = "  hello world  "
Output: "world hello"
Explanation: Your reversed string should not contain leading or trailing spaces.
```
**Example 3:**
```
Input: s = "a good   example"
Output: "example good a"
Explanation: You need to reduce multiple spaces between two words to a single space in the reversed string.
```

**Constraints:**
* `1 <= s.length <= 10^4`
* `s` contains English letters (upper-case and lower-case), digits, and spaces ' '.
* There is at least one word in `s`.

````python
def reverse_words(s: str) -> str:
    # split() without arguments handles multiple spaces automatically
    words = s.split()
    # Reverse the list and join with a single space
    return " ".join(words[::-1])

# Test
print(reverse_words("  hello world  ")) # "world hello"
print(reverse_words("a good   example")) # "example good a"
````

### Q9: Valid IP Address format
Write a function `is_valid_ipv4(ip: str) -> bool`. An IP address is valid if it has exactly 4 parts separated by dots, each part contains only digits, and each part is an integer between 0 and 255 inclusive. Leading zeros are not allowed unless the number is exactly "0".

**Example 1:**
```
Input: ip = "192.168.1.1"
Output: true
```
**Example 2:**
```
Input: ip = "192.168.01.1"
Output: false
Explanation: Leading zeros are not permitted.
```
**Example 3:**
```
Input: ip = "256.1.1.1"
Output: false
```

**Constraints:**
* `1 <= ip.length <= 15`
* `ip` consists of digits and dots '.'.

````python
def is_valid_ipv4(ip: str) -> bool:
    parts = ip.split(".")
    if len(parts) != 4:
        return False
    
    for part in parts:
        if not part.isdigit():
            return False
        if len(part) > 1 and part.startswith("0"):
            return False
        if not (0 <= int(part) <= 255):
            return False
            
    return True

# Test
print(is_valid_ipv4("192.168.1.1")) # True
print(is_valid_ipv4("192.168.01.1")) # False
````

### Q10: Calculate Time Difference in Seconds
Write a function `time_diff(time1: str, time2: str) -> int`. The inputs are two strings in the format "HH:MM:SS". Find the absolute difference in seconds between the two times. You can use string splitting and math, or the `datetime` module.

**Example 1:**
```
Input: time1 = "12:00:00", time2 = "13:30:15"
Output: 5415
```
**Example 2:**
```
Input: time1 = "23:59:59", time2 = "00:00:00"
Output: 86399
```

**Constraints:**
* `time1` and `time2` are valid times in the format "HH:MM:SS".
* `00 <= HH <= 23`, `00 <= MM <= 59`, `00 <= SS <= 59`.

````python
def time_diff(time1: str, time2: str) -> int:
    # Using split and arithmetic
    def to_seconds(t: str) -> int:
        h, m, s = [int(x) for x in t.split(":")]
        return h * 3600 + m * 60 + s
        
    return abs(to_seconds(time1) - to_seconds(time2))

# Test
print(time_diff("12:00:00", "13:30:15")) # 5415
````

### Q11: Caesar Cipher
Write a function `caesar_cipher(s: str, k: int) -> str`. The function should shift every lowercase English letter in `s` by `k` positions forward in the alphabet. If it goes past 'z', it wraps around to 'a'. Non-lowercase characters should remain unchanged. Use `ord()`, `chr()`, and modulo `%`.

**Example 1:**
```
Input: s = "xyz", k = 2
Output: "zab"
```
**Example 2:**
```
Input: s = "hello world!", k = 3
Output: "khoor zruog!"
Explanation: Non-alphabet characters remain unchanged.
```

**Constraints:**
* `1 <= s.length <= 10^4`
* `s` consists of printable ASCII characters.
* `0 <= k <= 100`

````python
def caesar_cipher(s: str, k: int) -> str:
    result = []
    for char in s:
        if 'a' <= char <= 'z':
            # Normalize ASCII to 0-25 range, add shift, modulo 26, then convert back to ASCII
            shifted_ascii = (ord(char) - ord('a') + k) % 26 + ord('a')
            result.append(chr(shifted_ascii))
        else:
            result.append(char)
    return "".join(result)

# Test
print(caesar_cipher("xyz", 2)) # "zab"
print(caesar_cipher("hello world!", 3)) # "khoor zruog!"
````

### Q12: Implement Custom `split()`
Write a function `custom_split(s: str, delimiter: str) -> list[str]` that mimics the behavior of the built-in `split()` function (for a specific string delimiter). Do not use the built-in `split()` or `partition()`. Use `find()` and string slicing.

**Example 1:**
```
Input: s = "apple-banana-orange", delimiter = "-"
Output: ["apple", "banana", "orange"]
```
**Example 2:**
```
Input: s = "hello::world::python", delimiter = "::"
Output: ["hello", "world", "python"]
```

**Constraints:**
* `1 <= s.length <= 10^4`
* `1 <= delimiter.length <= 10`
* `s` and `delimiter` consist of printable ASCII characters.

````python
def custom_split(s: str, delimiter: str) -> list[str]:
    if not delimiter:
        raise ValueError("empty separator")
        
    result = []
    start = 0
    
    while True:
        index = s.find(delimiter, start)
        if index == -1:
            # Append the remaining part of the string
            result.append(s[start:])
            break
        # Append the part before the delimiter
        result.append(s[start:index])
        # Move the start pointer past the delimiter
        start = index + len(delimiter)
        
    return result

# Test
print(custom_split("apple-banana-orange", "-")) # ['apple', 'banana', 'orange']
print(custom_split("hello::world::python", "::")) # ['hello', 'world', 'python']
````

### Q13: Add Two Numbers without `+`
Write a function `get_sum(a: int, b: int) -> int` that calculates the sum of two positive integers without using the `+` or `-` operators. Use bitwise `^`, `&`, and `<<`.

**Example 1:**
```
Input: a = 5, b = 7
Output: 12
```
**Example 2:**
```
Input: a = 20, b = 30
Output: 50
```

**Constraints:**
* `0 <= a, b <= 1000`

````python
def get_sum(a: int, b: int) -> int:
    # In Python, integers have arbitrary precision, so we don't need 32-bit masking 
    # for positive numbers. If dealing with negative numbers, 32-bit mask is required.
    while b != 0:
        carry = a & b
        a = a ^ b
        b = carry << 1
    return a

# Test
print(get_sum(5, 7)) # 12
````

### Q14: String to Integer (atoi)
Write a function `my_atoi(s: str) -> int` that converts a string to a 32-bit signed integer. 
1. Ignore leading whitespace. 
2. Check for '+' or '-'. 
3. Read in characters until the next non-digit character. 
4. Convert to integer. 
5. Clamp to 32-bit signed integer range `[-2^31, 2^31 - 1]`.

**Example 1:**
```
Input: s = "   -42"
Output: -42
Explanation: Read the '-' then "42".
```
**Example 2:**
```
Input: s = "4193 with words"
Output: 4193
Explanation: Stop reading at " with words".
```

**Constraints:**
* `0 <= s.length <= 200`
* `s` consists of English letters (lower-case and upper-case), digits (0-9), ' ', '+', '-', and '.'.

````python
def my_atoi(s: str) -> int:
    s = s.lstrip()
    if not s:
        return 0
        
    sign = 1
    start = 0
    if s[0] == '-':
        sign = -1
        start = 1
    elif s[0] == '+':
        start = 1
        
    result = 0
    for i in range(start, len(s)):
        if not s[i].isdigit():
            break
        # Convert char to int using ASCII logic (or int(s[i]))
        result = result * 10 + (ord(s[i]) - ord('0'))
        
    result *= sign
    
    # Clamp to 32-bit signed integer range
    INT_MIN = -2**31
    INT_MAX = 2**31 - 1
    
    if result < INT_MIN:
        return INT_MIN
    if result > INT_MAX:
        return INT_MAX
        
    return result

# Test
print(my_atoi("   -42")) # -42
print(my_atoi("4193 with words")) # 4193
````

### Q15: Generate a Strong Password
Write a function `generate_password() -> str` that generates a random 12-character strong password. It MUST contain at least one uppercase letter, one lowercase letter, one digit, and one special character (from "!@#$%^&*"). Use the `random` module (`choice`, `shuffle`) and ASCII ranges with `chr()` to build the pools dynamically.

**Example 1:**
```
Input: None
Output: "aB3$xyz!pqR9" (Output will vary)
Explanation: Contains uppercase, lowercase, digit, and special character. Total length is 12.
```

**Constraints:**
* Output length must be exactly 12.
* Contains at least one char from `[A-Z]`, `[a-z]`, `[0-9]`, and `[!@#$%^&*]`.

````python
import random

def generate_password() -> str:
    # Generate pools using list comprehension and ASCII
    uppercase = [chr(i) for i in range(65, 91)]
    lowercase = [chr(i) for i in range(97, 123)]
    digits = [chr(i) for i in range(48, 58)]
    specials = list("!@#$%^&*")
    
    all_chars = uppercase + lowercase + digits + specials
    
    # Ensure at least one from each category
    password_chars = [
        random.choice(uppercase),
        random.choice(lowercase),
        random.choice(digits),
        random.choice(specials)
    ]
    
    # Fill the rest of the 12 characters
    for _ in range(8):
        password_chars.append(random.choice(all_chars))
        
    # Shuffle to ensure randomness
    random.shuffle(password_chars)
    
    return "".join(password_chars)

# Test
print(generate_password())
````
