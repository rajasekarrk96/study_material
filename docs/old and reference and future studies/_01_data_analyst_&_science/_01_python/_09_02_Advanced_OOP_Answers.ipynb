### Q11. Rectangle and Circle Area

Dry run: rectangle area is `4 * 5 = 20`; circle area is `3.14159 * 3 * 3 = 28.27431`. Time complexity is `O(1)` per area call.


```python
# Define the base Shape class
class Shape:
    # Create a shared interface method
    def area(self):
        # Raise an error when a child does not implement the method
        raise NotImplementedError('Child class must implement area')

# Define the Rectangle child class
class Rectangle(Shape):
    # Initialize width and height
    def __init__(self, width, height):
        # Store width
        self.width = width
        # Store height
        self.height = height

    # Return the rectangle area
    def area(self):
        # Multiply width by height
        return self.width * self.height

# Define the Circle child class
class Circle(Shape):
    # Initialize radius
    def __init__(self, radius):
        # Store the radius
        self.radius = radius

    # Return the circle area
    def area(self):
        # Apply the circle area formula
        return 3.14159 * self.radius * self.radius

# Create shape objects
rectangle = Rectangle(4, 5)
circle = Circle(3)

# Test case 1
expected_output = (20, 28.27431)
print((rectangle.area(), circle.area()))  # should match expected
print(expected_output)
```

### Q16. Validated Percentage Property

Dry run: setting `85` passes validation and is stored; values outside `0` to `100` would raise a `ValueError`. Time complexity is `O(1)`.


```python
# Define the Percentage class
class Percentage:
    # Initialize the object through the property setter
    def __init__(self, value):
        # Set the validated value
        self.value = value

    # Read the stored value
    @property
    def value(self):
        # Return the internal percentage value
        return self._value

    # Validate and store new values
    @value.setter
    def value(self, new_value):
        # Reject values outside the valid range
        if not 0 <= new_value <= 100:
            raise ValueError('Percentage must be between 0 and 100')
        # Store the validated value
        self._value = new_value

# Create one valid object
percentage = Percentage(85)

# Test case 1
expected_output = 85
print(percentage.value)  # should match expected
print(expected_output)
```

### Q19. Statement Account

Dry run: start from `1000`, deposit `100`, withdraw `50`, and track both balance and transaction count. Time complexity is `O(1)` per operation.


```python
# Define the StatementAccount class
class StatementAccount:
    # Initialize balance and transaction count
    def __init__(self, balance):
        # Store the starting balance
        self.balance = balance
        # Start with zero transactions
        self.count = 0

    # Deposit money and count the action
    def deposit(self, amount):
        # Increase balance and transaction count
        self.balance += amount
        self.count += 1

    # Withdraw money and count the action
    def withdraw(self, amount):
        # Decrease balance and transaction count
        self.balance -= amount
        self.count += 1

    # Return statement data
    def statement(self):
        # Return the final summary dictionary
        return {'balance': self.balance, 'count': self.count}

# Create one statement account object
account = StatementAccount(1000)
account.deposit(100)
account.withdraw(50)

# Test case 1
expected_output = {'balance': 1050, 'count': 2}
print(account.statement())  # should match expected
print(expected_output)
```

### Q21. Positive Descriptor

Dry run: assigning `500` calls the descriptor `__set__()` which accepts non-negative values and stores them. Time complexity is `O(1)` per set/get.


```python
# Define the descriptor class
class PositiveNumber:
    # Learn the managed attribute name
    def __set_name__(self, owner, name):
        # Store the internal private attribute name
        self.private_name = '_' + name

    # Return the stored value
    def __get__(self, instance, owner):
        # Read the internal stored attribute value
        return getattr(instance, self.private_name)

    # Validate and store new values
    def __set__(self, instance, value):
        # Reject negative numbers
        if value < 0:
            raise ValueError('Value must be non-negative')
        # Store the validated value
        setattr(instance, self.private_name, value)

# Use the descriptor in a Product class
class Product:
    # Attach the descriptor to price
    price = PositiveNumber()

    # Initialize the product with a price
    def __init__(self, price):
        # Set the price through the descriptor
        self.price = price

# Create one valid product object
product = Product(500)

# Test case 1
expected_output = 500
print(product.price)  # should match expected
print(expected_output)
```

### Q28. Polymorphic Area Sum

Dry run: sum `Rectangle(4, 5).area()` which is `20` and `Circle(3).area()` which is `28.27431`, giving `48.27431`. Time complexity is `O(n)` for `n` shapes.


```python
# Define the base Shape class
class Shape:
    # Provide an area interface
    def area(self):
        raise NotImplementedError('Child class must implement area')

# Define Rectangle
class Rectangle(Shape):
    # Initialize rectangle dimensions
    def __init__(self, width, height):
        self.width = width
        self.height = height

    # Return rectangle area
    def area(self):
        return self.width * self.height

# Define Circle
class Circle(Shape):
    # Initialize circle radius
    def __init__(self, radius):
        self.radius = radius

    # Return circle area
    def area(self):
        return 3.14159 * self.radius * self.radius

# Define a function that sums shape areas
def total_area(shapes):
    # Sum the area of every shape object
    return sum(shape.area() for shape in shapes)

# Test case 1
shape_list = [Rectangle(4, 5), Circle(3)]
expected_output = 48.27431
print(total_area(shape_list))  # should match expected
print(expected_output)
```

### Q32. Student Report Object

Dry run: average of `[88, 76, 91, 84]` is `84.75`, which falls in the `B` range. Time complexity is `O(n)`.


```python
# Define the StudentReport class
class StudentReport:
    # Initialize name and marks list
    def __init__(self, name, marks):
        # Store the student name
        self.name = name
        # Store the marks list
        self.marks = marks

    # Return the average marks
    def average(self):
        # Compute the arithmetic mean
        return sum(self.marks) / len(self.marks)

    # Return the final grade
    def grade(self):
        # Use the average to assign the grade
        avg = self.average()
        if avg >= 90:
            return 'A'
        if avg >= 75:
            return 'B'
        if avg >= 60:
            return 'C'
        return 'D'

    # Return a report dictionary
    def report(self):
        # Build the final report output
        return {'name': self.name, 'average': self.average(), 'grade': self.grade()}

# Create one report object
report = StudentReport('Asha', [88, 76, 91, 84])

# Test case 1
expected_output = {'name': 'Asha', 'average': 84.75, 'grade': 'B'}
print(report.report())  # should match expected
print(expected_output)
```

### Q33. Shopping Cart OOP

Dry run: `Book` contributes `300 * 2 = 600` and `Pen` contributes `20 * 5 = 100`, so total is `700`. Time complexity is `O(n)` for `n` items.


```python
# Define the ShoppingCart class
class ShoppingCart:
    # Initialize an empty item list
    def __init__(self):
        # Store cart items in a list
        self.items = []

    # Add one item to the cart
    def add_item(self, name, price, quantity):
        # Append the item data as a dictionary
        self.items.append({'name': name, 'price': price, 'quantity': quantity})

    # Return the total bill
    def total_bill(self):
        # Sum price times quantity for every cart item
        return sum(item['price'] * item['quantity'] for item in self.items)

# Create one shopping cart object
cart = ShoppingCart()
cart.add_item('Book', 300, 2)
cart.add_item('Pen', 20, 5)

# Test case 1
expected_output = 700
print(cart.total_bill())  # should match expected
print(expected_output)
```

### Q34. ATM Simulation

Dry run: start at `1000`, deposit `500` to reach `1500`, withdraw `200` to reach `1300`, and count two transactions. Time complexity is `O(1)` per operation.


```python
# Define the ATMAccount class
class ATMAccount:
    # Initialize balance and transaction count
    def __init__(self, balance):
        # Store the starting balance
        self.balance = balance
        # Start with zero transactions
        self.transactions = 0

    # Deposit money into the account
    def deposit(self, amount):
        # Increase balance and transaction count
        self.balance += amount
        self.transactions += 1

    # Withdraw money from the account
    def withdraw(self, amount):
        # Decrease balance and transaction count
        self.balance -= amount
        self.transactions += 1

    # Return a summary dictionary
    def mini_statement(self):
        # Build the final statement output
        return {'balance': self.balance, 'transactions': self.transactions}

# Create one ATM account object
account = ATMAccount(1000)
account.deposit(500)
account.withdraw(200)

# Test case 1
expected_output = {'balance': 1300, 'transactions': 2}
print(account.mini_statement())  # should match expected
print(expected_output)
```

### Q35. Course Enrollment Tracker

Dry run: after enrolling three names, the count is `3` and the sorted name list becomes `['Asha', 'Meena', 'Ravi']`. Time complexity is `O(n log n)` because of sorting.


```python
# Define the CourseEnrollment class
class CourseEnrollment:
    # Initialize an empty student list
    def __init__(self):
        # Store enrolled student names
        self.students = []

    # Enroll one student
    def enroll(self, name):
        # Add the student name to the list
        self.students.append(name)

    # Return enrollment summary
    def summary(self):
        # Return count and sorted names in one dictionary
        return {'count': len(self.students), 'students': sorted(self.students)}

# Create one enrollment tracker object
enrollment = CourseEnrollment()
enrollment.enroll('Ravi')
enrollment.enroll('Asha')
enrollment.enroll('Meena')

# Test case 1
expected_output = {'count': 3, 'students': ['Asha', 'Meena', 'Ravi']}
print(enrollment.summary())  # should match expected
print(expected_output)
```
