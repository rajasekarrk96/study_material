## Starter Code Examples
Run these examples to see the core ideas in action.


```python
def derivative(f, x, h=1e-6):
    return (f(x+h) - f(x)) / h
print('approx derivative of x^2 at x=3', derivative(lambda t: t**2, 3))
```
