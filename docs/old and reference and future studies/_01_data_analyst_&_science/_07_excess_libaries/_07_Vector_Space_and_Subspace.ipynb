# Vector Space and Subspace
A vector space is a set of vectors that supports addition and scalar multiplication. A subspace is a subset that is itself a vector space.


## Requirements for a vector space
- Contains the zero vector.
- Closed under addition.
- Closed under scalar multiplication.


```python
import numpy as np
u = np.array([1, 2])
v = np.array([2, 4])
print("u + v =", u + v)
print("2 * u =", 2 * u)
```

## Subspace example
The set of all vectors of the form `[a, 2a]` is a subspace of `R^2`. It contains the zero vector and is closed under addition and scalar multiplication.


## Practical interpretation
Subspaces are important for understanding spaces of solutions, such as the null space of a matrix or the span of basis vectors.


## Key Takeaways
- A vector space is closed under addition and scalar multiplication.
- A subspace must contain the zero vector.
- Subspaces appear as solution sets (nullspace, span).


### Practice
Test whether [1,2] and [2,4] span a subspace and are linearly dependent.


```python
import numpy as np
u=np.array([1,2])
v=np.array([2,4])
print('dependent?', np.linalg.matrix_rank(np.column_stack((u,v)))<2)
```
