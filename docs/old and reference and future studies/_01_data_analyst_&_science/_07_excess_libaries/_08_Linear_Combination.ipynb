# Linear Combination
A linear combination multiplies vectors by scalars and then adds them. This concept is central to span, basis, and linear independence.


## Definition
Given vectors `v1, v2, ..., vn` and scalars `a1, a2, ..., an`, the expression `a1 v1 + a2 v2 + ... + an vn` is a linear combination.


```python
import numpy as np
v1 = np.array([1, 0])
v2 = np.array([0, 1])
comb = 3 * v1 + (-2) * v2
print("Linear combination:", comb)
```

## Span
The span of a set of vectors is the set of all possible linear combinations of those vectors.


## Linear independence
Vectors are linearly independent if the only solution to `a1 v1 + ... + an vn = 0` is all coefficients equal to zero.


## Example
In `R^2`, `[1, 0]` and `[0, 1]` are independent and span the plane, while `[1, 2]` and `[2, 4]` are dependent.


## Key Takeaways
- Linear combination = sum of scalars times vectors.
- Span is set of all linear combinations.
- Independence determines if vectors form a basis.
