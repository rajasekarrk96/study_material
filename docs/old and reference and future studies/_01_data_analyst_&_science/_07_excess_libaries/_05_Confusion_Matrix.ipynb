# Confusion Matrix
A confusion matrix evaluates classification results by comparing predicted labels and actual labels.


## Binary classification layout
- True Positive (TP)
- False Positive (FP)
- False Negative (FN)
- True Negative (TN)


## Performance metrics
- Accuracy = `(TP + TN) / (TP + TN + FP + FN)`
- Precision = `TP / (TP + FP)`
- Recall = `TP / (TP + FN)`
- F1 score = `2 * precision * recall / (precision + recall)`


```python
TP = 50
FP = 10
FN = 5
TN = 35
accuracy = (TP + TN) / (TP + TN + FP + FN)
precision = TP / (TP + FP)
recall = TP / (TP + FN)
f1 = 2 * precision * recall / (precision + recall)
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 score:", f1)
```

## When to use a confusion matrix
Use it when classes are imbalanced or when false positives and false negatives have different consequences.


## Example scenario
A model that predicts whether an email is spam can be evaluated with this matrix to understand its performance beyond simple accuracy.


## Key Takeaways
- Confusion matrix summarizes prediction vs truth.
- Important metrics: accuracy, precision, recall, F1.
- Use when classes are imbalanced or costs differ.
