# Calculus I: Limits and Derivatives
This notebook covers the fundamental ideas of limits and derivatives, which describe change in mathematical functions.


## Limits
A limit describes the value a function approaches as the input approaches a specific point.


```python
def f(x):
    return x**2 + 2*x + 1
for x in [1.9, 1.99, 2.01, 2.1]:
    print(x, f(x))
```

## Derivative definition
The derivative is the limit of the difference quotient: `(f(x+h) - f(x)) / h` as `h` approaches zero. It measures instantaneous rate of change.


```python
def derivative(f, x, h=1e-6):
    return (f(x + h) - f(x)) / h
print("Derivative approx at x=3:", derivative(lambda t: t**2, 3))
```

## Power rule
For `f(x)=x^n`, the derivative is `f'(x)=n*x^{n-1}`.


```python
import sympy as sp
x = sp.symbols("x")
f = x**3 - 2*x**2 + x
sp.diff(f, x)
```

## Key Takeaways
- Limits describe approach behavior of functions.
- Derivative is the instantaneous rate of change.
- Power rule simplifies many derivatives.
