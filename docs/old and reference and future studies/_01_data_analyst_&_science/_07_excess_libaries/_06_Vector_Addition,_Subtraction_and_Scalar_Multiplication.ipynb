# Vector Addition, Subtraction, and Scalar Multiplication
Vectors represent direction and magnitude. This notebook covers basic vector arithmetic and geometric intuition.


## Vector addition and subtraction
Add or subtract corresponding components.
Example: `[2, 3] + [1, -1] = [3, 2]`.


```python
import numpy as np
v = np.array([2, 3])
w = np.array([1, -1])
print("v + w =", v + w)
print("v - w =", v - w)
```

## Scalar multiplication
Multiply each component by a scalar.
Example: `3 * [2, 3] = [6, 9]`.


```python
scalar = 3
print(scalar * v)
```

## Geometric meaning
Vector addition corresponds to moving one vector along another. Scalar multiplication changes length and direction depending on the sign of the scalar.


## Key Takeaways
- Vector addition and subtraction are component-wise.
- Scalar multiplication scales magnitude and can reverse direction.
- Visualize with arrows for intuition.
