## Starter Code Examples
Run these examples to see the core ideas in action.


```python
import matplotlib.pyplot as plt
x = list(range(-10, 11))
y = [1.5 * xi - 2 for xi in x]
plt.plot(x, y, marker='o')
plt.title('Line: y = 1.5x - 2')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)
plt.show()
```
