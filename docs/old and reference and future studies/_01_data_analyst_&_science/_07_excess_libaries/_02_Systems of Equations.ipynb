# Systems of Equations
A system is a set of equations that share variables. The solution must satisfy every equation in the set.


## Example system
`2x + y = 10`
`x - y = 2`
The solution pair `(x, y)` satisfies both equations.


## Solve by elimination
Add the equations to remove one variable.


```python
x = 4
y = 2
print("Solution: x=", x, "y=", y)
print("Check 1:", 2 * x + y == 10)
print("Check 2:", x - y == 2)
```

## Matrix method
Represent the system as `A x = b`. Use NumPy to solve for `x`.


```python
import numpy as np
A = np.array([[2, 1], [1, -1]])
b = np.array([10, 2])
solution = np.linalg.solve(A, b)
print("Solution:", solution)
```

## Interpretation
Graphically, each equation is a line. The intersection point is the solution.
If the lines are parallel, there is no solution. If the lines coincide, there are infinitely many solutions.


## Key Takeaways
- Systems are sets of equations with common variables.
- Solve by elimination, substitution, or matrix methods.
- Check for no solution (parallel) or infinite solutions (coincident).
