# Linear Equations
A linear equation has variables of degree 1 and forms a straight line when graphed in 2D.


## Slope-intercept form
`y = mx + b`, where `m` is the slope and `b` is the y-intercept.


```python
import pandas as pd
import matplotlib.pyplot as plt
x = list(range(-5, 6))
y = [2 * xi + 1 for xi in x]
df = pd.DataFrame({"x": x, "y": y})
print(df.head())
```

```python
plt.plot(x, y, marker="o")
plt.title("Line: y = 2x + 1")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

## Intercepts
- x-intercept: set `y=0` and solve for `x`.
- y-intercept: set `x=0` and solve for `y`.


```python
m = 1.5
b = -2
x_intercept = -b / m
y_intercept = b
print("x-intercept:", x_intercept)
print("y-intercept:", y_intercept)
```

## Key Takeaways
- Lines follow y = mx + b.
- Slope m controls steepness; intercept b shifts vertically.
- Intercepts found by setting x or y to zero.
