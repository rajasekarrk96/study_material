## Starter Code Examples
Run these examples to see the core ideas in action.


```python
import pandas as pd
df = pd.DataFrame({'actual':[1,0,1,1,0,1,0,0],'pred':[1,0,0,1,0,1,1,0]})
TP = ((df.actual==1)&(df.pred==1)).sum()
FP = ((df.actual==0)&(df.pred==1)).sum()
FN = ((df.actual==1)&(df.pred==0)).sum()
TN = ((df.actual==0)&(df.pred==0)).sum()
print(dict(TP=TP,FP=FP,FN=FN,TN=TN))
```
