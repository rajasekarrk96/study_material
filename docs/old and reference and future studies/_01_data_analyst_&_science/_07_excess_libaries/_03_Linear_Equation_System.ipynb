# Linear Equation Systems
This notebook explains how to represent and solve systems of linear equations using matrices and Python.


## Matrix form
For the system:
`2x + 3y - z = 5`
`x - y + 2z = 4`
`3x + y + z = 6`
we can write `A x = b`.


```python
import numpy as np
A = np.array([[2, 3, -1], [1, -1, 2], [3, 1, 1]])
b = np.array([5, 4, 6])
x = np.linalg.solve(A, b)
print("Solution vector x =", x)
```

## Determinant and uniqueness
If the determinant of `A` is non-zero, the system has exactly one solution.


```python
det_A = np.linalg.det(A)
print("Determinant of A =", det_A)
```

## Gaussian elimination concept
Gaussian elimination converts the system to upper triangular form, then solves by back substitution.


## Practical note
This matrix view is useful for solving larger systems and for understanding linear algebra in data science.


## Key Takeaways
- Represent systems as matrix equation Ax = b.
- If det(A) != 0, unique solution exists.
- Use numerical libraries for larger systems.
