# Calculus II: Integrals and Partial Derivatives
This notebook introduces definite integrals and partial derivatives for functions of more than one variable.


## Definite integrals
A definite integral represents the signed area under a curve between two limits.


```python
import sympy as sp
x = sp.symbols("x")
f = x**2
area = sp.integrate(f, (x, 0, 3))
print("Integral of x^2 from 0 to 3 =", area)
```

## Partial derivatives
For functions of several variables, a partial derivative measures change with respect to one variable while holding the others constant.


```python
x, y = sp.symbols("x y")
g = x**2 * y + sp.sin(y)
print("dg/dx =", sp.diff(g, x))
print("dg/dy =", sp.diff(g, y))
```

## Gradient
The gradient is a vector of all first-order partial derivatives. It points in the direction of greatest increase of the function.


## Key Takeaways
- Definite integrals compute signed area under a curve.
- Partial derivatives measure change w.r.t one variable keeping others fixed.
- Gradient collects first-order partials and points to max increase.
