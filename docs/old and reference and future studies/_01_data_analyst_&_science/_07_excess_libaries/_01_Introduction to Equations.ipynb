# Introduction to Equations
Equations express that two expressions are equal. This notebook explains variables, constants, basic equation types, and solving techniques.


## Key concepts
- Variable: an unknown represented by a symbol such as `x`.
- Constant: a fixed number.
- Expression: a combination of variables, numbers, and operators.
- Equation: asserts equality between two expressions.


## One-step equations
Solve by reversing one operation.
Example: `x + 4 = 9` becomes `x = 9 - 4`.


```python
x = 9 - 4
print("x =", x)
print("Verify:", x + 4 == 9)
```

## Two-step equations
Solve by reversing operations in reverse order.
Example: `3x - 2 = 10` becomes `3x = 12`, then `x = 4`.


```python
x = (10 + 2) / 3
print("x =", x)
print("Verify:", 3 * x - 2 == 10)
```

## Practice problem
Solve `2(x - 3) = 8`.


```python
x = 8 / 2 + 3
print("x =", x)
print("Verify:", 2 * (x - 3) == 8)
```
