# Edge Cases
- Real data has missing values, inconsistent types, empty inputs, duplicate records, outliers, rare classes, and invalid categories.
- Advanced systems also face memory limits, version mismatch, unstable training, silent leakage, and non-deterministic results.
- Professional code checks assumptions before trusting output.


````python
# Edge case handling example for Bias and Fairness
raw_values = [10, None, 25, '30', '', 40]
clean_values = []

for value in raw_values:
    # Skip missing values.
    if value is None or value == '':
        continue

    # Convert numeric strings into numbers.
    numeric_value = float(value)

    # Keep only valid non-negative values.
    if numeric_value >= 0:
        clean_values.append(numeric_value)

print('Original:', raw_values)
print('Cleaned:', clean_values)
print('Clean average:', sum(clean_values) / len(clean_values))
````

# Common Mistakes
- Skipping data validation and trusting raw input.
- Mixing training and testing data during preprocessing.
- Choosing a complex tool before building a baseline.
- Ignoring units, data types, sample size, and business meaning.
- Reporting only successful metrics without explaining failure cases.


````python
# Common mistake example for Bias and Fairness
train_scores = [70, 80, 90]
test_scores = [40, 50]

# Mistake: calculating preprocessing information on train + test together.
wrong_average = sum(train_scores + test_scores) / len(train_scores + test_scores)

# Correct: learn preprocessing information from training data only.
correct_train_average = sum(train_scores) / len(train_scores)

print('Wrong average using all data:', wrong_average)
print('Correct training-only average:', correct_train_average)
````

# Short-hand and Optimized Methods
- Professional code uses concise patterns only when they remain readable.
- Optimization can mean fewer loops, vectorized operations, cached computation, correct indexing, smaller data types, or reusable pipelines.
- Short-hand is useful for exploration, but production code should still be clear and testable.


````python
# Short-hand and optimized method example for Bias and Fairness
records = [
    {'department': 'sales', 'amount': 100},
    {'department': 'sales', 'amount': 150},
    {'department': 'support', 'amount': 90},
]

# Clear short-hand: list comprehension for one transformation.
amounts = [record['amount'] for record in records]

# Clear aggregation without unnecessary repeated code.
totals = {}
for record in records:
    department = record['department']
    totals[department] = totals.get(department, 0) + record['amount']

print('Amounts:', amounts)
print('Totals by department:', totals)
````

# Advanced Usage
- Advanced work includes custom configuration, pipeline composition, performance tuning, reproducibility, and integration with other tools.
- For modeling topics, advanced work also includes validation design, hyperparameter tuning, interpretation, monitoring, and fairness checks.
- For visualization and reporting topics, advanced work includes interactivity, layout decisions, accessibility, and stakeholder-ready communication.


````python
# Advanced usage example for Bias and Fairness
# Build a reusable pipeline-style function.
def validate_transform_summarize(values, minimum_allowed=0):
    # Validate input type.
    if not isinstance(values, list):
        raise TypeError('values must be a list')

    cleaned = []
    for value in values:
        # Ignore missing values.
        if value is None:
            continue

        # Convert values to float for consistent numeric processing.
        numeric_value = float(value)

        # Enforce a domain rule.
        if numeric_value >= minimum_allowed:
            cleaned.append(numeric_value)

    # Return a production-friendly result dictionary.
    return {
        'cleaned_values': cleaned,
        'count': len(cleaned),
        'average': sum(cleaned) / len(cleaned) if cleaned else 0,
    }

result = validate_transform_summarize([10, None, '20', 30], minimum_allowed=0)
print(result)
````

# Mini Practical Example
- Small practical examples build confidence before large projects.
- A good mini project includes input data, transformation logic, validation, output, and a short interpretation.
- The learner should change the input and rerun the code to understand cause and effect.


````python
# Mini practical example for Bias and Fairness
# Goal: create a small quality report.
dataset = [
    {'id': 1, 'value': 100, 'status': 'valid'},
    {'id': 2, 'value': None, 'status': 'missing'},
    {'id': 3, 'value': 130, 'status': 'valid'},
    {'id': 3, 'value': 130, 'status': 'duplicate'},
]

# Count total rows.
total_rows = len(dataset)

# Count missing values.
missing_rows = sum(1 for row in dataset if row['value'] is None)

# Detect duplicate IDs.
seen_ids = set()
duplicate_ids = set()
for row in dataset:
    if row['id'] in seen_ids:
        duplicate_ids.add(row['id'])
    seen_ids.add(row['id'])

quality_report = {
    'total_rows': total_rows,
    'missing_rows': missing_rows,
    'duplicate_ids': sorted(duplicate_ids),
}

print(quality_report)
````
