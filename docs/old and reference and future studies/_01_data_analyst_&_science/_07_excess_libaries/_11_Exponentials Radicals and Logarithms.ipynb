# Exponentials, Radicals, and Logarithms
This notebook reviews exponent rules, radical expressions, and logarithms with examples.


## Exponent rules
- `a^m * a^n = a^{m+n}`
- `(a^m)^n = a^{m*n}`
- `a^0 = 1` (for `a != 0`).


```python
print("2**3 * 2**2 =", 2**3 * 2**2)
print("2**5 =", 2**5)
```

## Radicals and fractional exponents
The square root is `a^{1/2}` and the nth root is `a^{1/n}`.


```python
import math
print("sqrt(16) =", math.sqrt(16))
print("4**0.5 =", 4**0.5)
```

## Logarithms
If `a^x = b`, then `x = log_a(b)`. Common logarithms use base `10` or `e`.


```python
print("log10(100) =", math.log10(100))
print("ln(e) =", math.log(math.e))
```

## Key Takeaways
- Exponent rules and radical (fractional exponent) identities simplify expressions.
- Logarithms are inverses of exponentials and useful for solving for exponents.
