# PART 1: What is Machine Learning?

## The old way: Programming with rules

Imagine you want a program that decides if an email is **spam** or **not spam**.

The old-fashioned way:
```
IF email contains "FREE MONEY"     -> spam
IF email contains "Nigerian prince" -> spam
```

Problem? You need **thousands of rules**. Spammers just change their wording.

---

## The ML way: Learn from examples

1. Show the computer **thousands of emails** already labelled (spam / not spam)
2. The computer **finds patterns on its own**
3. Now it classifies new emails it has never seen

> **Machine Learning = teaching computers to learn from data, not rules.**

---

## Analogy: Teaching a child

You don't teach a child what a cat is by writing rules.
You show them pictures: *'That is a cat. That is a dog.'*
After enough examples the child recognises cats it has never seen.
**ML works the same way.**

---

## AI vs Machine Learning vs Deep Learning

```
+------------------------------------------+
```

- **AI** = any technique that makes computers seem intelligent
- **ML** = a type of AI where the computer learns from data
- **Deep Learning** = ML using large neural networks (ChatGPT, image recognition)

# PART 2: Types of ML Problems

## Supervised Learning -- Learn with a teacher

You have data **with labels** (correct answers already given).

| House Size | Bedrooms | Price (label) |
|---|---|---|
| 1500 sqft | 3 | $250,000 |
| 2000 sqft | 4 | $320,000 |

- **Regression**: predict a number (house price, temperature)
- **Classification**: predict a category (spam/not-spam, cat/dog)

---

## Unsupervised Learning -- Find patterns alone

No labels. The algorithm discovers hidden structure.
- **Clustering**: group 1 million customers into segments
- **Dimensionality reduction**: compress 1000-feature data to 2 features

---

## Reinforcement Learning -- Learn by trial and error

An agent takes actions, gets rewards or penalties, learns to maximise reward.
Examples: AlphaGo, self-driving cars, game-playing AI.

**This course focuses on supervised learning** -- the most common type.

# PART 4: Why Math? Why Linear Algebra?

## Computers only understand numbers

| Raw data | What it becomes |
|---|---|
| Email text | Word count list |
| Image | Grid of pixel brightness values (0-255) |
| Audio | Sequence of sound amplitude values |
| House info | [1500, 3, 10, 5.2] (size, bedrooms, age, distance) |

Your dataset is a **table of numbers** -- that table is a **matrix**.

---

## Why specifically Linear Algebra?

Linear Algebra is the math of **lists and tables of numbers**.

| ML question | LA answer |
|---|---|
| How to represent 10,000 data points? | Matrices |
| How to transform data? | Matrix multiplication |
| Most important directions in data? | Eigenvalues / SVD |
| How to measure similarity? | Dot products / Norms |
| How to find the best model? | Solving Ax = b |

# PART 5: NumPy -- The Math Library

## Why not plain Python lists?

Python lists are **slow** for math.
For 1 million houses, Python loops take minutes. NumPy does it in milliseconds.

NumPy also gives you:
- Add two lists element by element (not possible with + on plain lists)
- Multiply every element by a number
- Mean, std, max, min in one call
- Matrix multiplication

**NumPy = fast arrays + math operations**

````python
# Always import numpy as 'np' -- everyone uses this abbreviation
import numpy as np
import matplotlib.pyplot as plt

print('NumPy version:', np.__version__)
print('Ready!')
````

## 5.1 Creating Arrays

````python
# ---------------------------------------------------------
# 1D array: one house described by 4 features
# [size_sqft, bedrooms, age_years, distance_km]
# ---------------------------------------------------------
house = np.array([1500, 3, 10, 5.2])

print('The house:', house)
print('Type:     ', type(house))
print('Shape:    ', house.shape)   # (4,) means 4 elements, 1 dimension
print('ndim:     ', house.ndim)    # 1 dimension
print('dtype:    ', house.dtype)
````

````python
# ---------------------------------------------------------
# 2D array: our dataset -- 5 houses, 4 features each
# Each ROW = one house. Each COLUMN = one feature.
# ---------------------------------------------------------
dataset = np.array([
    [1500, 3, 10, 5.2],   # house 0
    [2000, 4,  5, 3.1],   # house 1
    [1200, 2, 20, 8.0],   # house 2
    [1800, 3,  8, 4.5],   # house 3
    [2500, 5,  2, 1.0],   # house 4
])

print('Dataset:')
print(dataset)
print()
print('Shape:', dataset.shape)          # (5, 4) = 5 rows, 4 columns
print('Rows (houses):', dataset.shape[0])
print('Cols (features):', dataset.shape[1])
print('Total elements:', dataset.size)  # 5 x 4 = 20
print('ndim:', dataset.ndim)
````

````python
# ---------------------------------------------------------
# Useful ways to CREATE arrays
# ---------------------------------------------------------

zeros  = np.zeros((3, 4))        # 3 rows, 4 columns of zeros
ones   = np.ones((2, 3))         # 2 rows, 3 columns of ones
nums   = np.arange(0, 10, 2)     # [0, 2, 4, 6, 8]
evenly = np.linspace(0, 1, 5)    # 5 numbers from 0 to 1 inclusive

print('zeros (3x4):\n', zeros)
print('ones (2x3):\n', ones)
print('arange(0,10,2):', nums)
print('linspace(0,1,5):', evenly)

# Random numbers -- used everywhere in ML
np.random.seed(42)              # set seed for reproducibility
rand = np.random.randn(3, 3)    # 3x3 standard normal distribution
print('Random 3x3:\n', rand.round(3))
````

## 5.2 Indexing and Slicing

````python
# ---------------------------------------------------------
# Indexing: dataset[row, col]  -- row first, then column (0-based)
# ---------------------------------------------------------
print('First house (row 0):    ', dataset[0])
print('Third house (row 2):    ', dataset[2])
print('Size of house 0:        ', dataset[0, 0])
print('Bedrooms of house 1:    ', dataset[1, 1])
print('Distance of last house: ', dataset[-1, 3])  # -1 = last row

# Slicing: get multiple rows or columns
print('First 3 houses:')
print(dataset[:3])

print('All sizes (column 0):   ', dataset[:, 0])
print('All bedrooms (col 1):   ', dataset[:, 1])

# Boolean indexing
sizes    = dataset[:, 0]
big_mask = sizes > 1700
print('Houses larger than 1700 sqft:')
print(dataset[big_mask])
````

## 5.3 Math Operations

````python
# ---------------------------------------------------------
# Math on whole arrays -- no loops needed!
# ---------------------------------------------------------
prices = np.array([250000., 320000., 190000., 280000., 410000.])

print('Prices:             ', prices)
print('After 5% increase:  ', prices * 1.05)
print('After 10k tax:      ', prices - 10000)
print('In thousands ($k):  ', prices / 1000)

# Two arrays -- operations work element-by-element
old_prices = np.array([240000., 300000., 180000., 270000., 390000.])
change = prices - old_prices
pct    = (change / old_prices) * 100
print('Price changes:', change)
print('Pct changes:  ', pct.round(1))
````

````python
# ---------------------------------------------------------
# Statistical operations -- used constantly in ML
# ---------------------------------------------------------
print('=== Stats on prices ===')
print(f'Mean:   ${prices.mean():,.0f}')
print(f'Median: ${np.median(prices):,.0f}')
print(f'Std:    ${prices.std():,.0f}')
print(f'Min:    ${prices.min():,.0f}')
print(f'Max:    ${prices.max():,.0f}')

print()
print('=== Stats on the dataset matrix ===')
# axis=0 means per column (across all rows)
print('Column means:', dataset.mean(axis=0))
print('  -> mean size, mean bedrooms, mean age, mean distance')

# axis=1 means per row (across all columns)
print('Row means:   ', dataset.mean(axis=1).round(2))
````

````python
# ---------------------------------------------------------
# Reshaping -- change shape without changing data
# Used constantly to fix shape mismatches
# ---------------------------------------------------------
flat = np.arange(1, 13)            # [1, 2, 3, ..., 12]
print('Original:', flat, '| shape:', flat.shape)

reshaped = flat.reshape(3, 4)      # 3 rows, 4 columns
print('Reshaped (3, 4):')
print(reshaped)

auto = flat.reshape(4, -1)         # 4 rows, NumPy figures out columns
print('reshape(4, -1) -> shape:', auto.shape)

back = reshaped.flatten()          # back to 1D
print('Flattened:', back)
````

# PART 6: Matplotlib -- Visualising Your Data

> **Why visualise?** You cannot understand 1000 numbers by staring at them.
> A picture reveals patterns in seconds.

````python
# ---------------------------------------------------------
# Line plot -- trend over time (e.g. training loss)
# ---------------------------------------------------------
epochs     = np.arange(1, 11)
train_loss = [2.5, 1.8, 1.3, 1.0, 0.8, 0.7, 0.6, 0.55, 0.52, 0.50]
val_loss   = [2.6, 2.0, 1.5, 1.2, 1.0, 0.9, 0.85, 0.84, 0.83, 0.82]

plt.figure(figsize=(8, 4))
plt.plot(epochs, train_loss, 'b-o', label='Training loss',   linewidth=2)
plt.plot(epochs, val_loss,   'r-s', label='Validation loss', linewidth=2)
plt.xlabel('Epoch (training round)')
plt.ylabel('Loss (error)')
plt.title('How Loss Decreases During Training')
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
````

# PART 7: Vocabulary You Will Hear Constantly

| Term | Plain English | Example |
|---|---|---|
| **Dataset** | Your collection of examples | 10,000 house records |
| **Sample / Data point** | One example in the dataset | One house |
| **Feature / Input** | A property describing a sample | House size, bedrooms |
| **Label / Target** | The answer you want to predict | House price |
| **Training set** | Data used to teach the model | 80% of your houses |
| **Test set** | Data held back to measure accuracy | 20% of your houses |
| **Model** | The thing that learns and makes predictions | Linear regression |
| **Parameters / Weights** | Numbers the model learns | The coefficients |
| **Training** | Adjusting weights to fit the data | Running gradient descent |
| **Prediction** | The model's output for a new input | Predicted house price |
| **Loss / Error** | How wrong the predictions are | Mean squared error |
| **Gradient descent** | Algorithm that adjusts weights to reduce loss | |
| **Epoch** | One full pass through the training set | |
| **Batch** | A small chunk of data processed at once | 32 houses at a time |
| **Hyperparameter** | A setting you choose (not learned) | Learning rate = 0.01 |

````python
# ---------------------------------------------------------
# Mini ML pipeline preview
# Every step uses linear algebra -- which is what this course teaches.
# ---------------------------------------------------------
np.random.seed(42)

# STEP 1: create dataset
n        = 100
sizes    = np.random.uniform(800, 3000, n)
bedrooms = np.random.randint(1, 6, n).astype(float)
prices   = 150*sizes + 20000*bedrooms + np.random.randn(n)*20000

# STEP 2: feature matrix and labels
X = np.column_stack([sizes, bedrooms])   # shape (100, 2)
y = prices
print(f'X shape: {X.shape}  (100 houses x 2 features)')
print(f'y shape: {y.shape}  (100 price labels)')

# STEP 3: train/test split
X_train, X_test = X[:80], X[80:]
y_train, y_test = y[:80], y[80:]

# STEP 4: fit linear regression (OLS)
Xb = np.column_stack([np.ones(80), X_train])   # add bias column
w  = np.linalg.solve(Xb.T @ Xb, Xb.T @ y_train)
print(f'Learned: bias=${w[0]:,.0f}, per_sqft=${w[1]:.2f}, per_bed=${w[2]:,.0f}')

# STEP 5: evaluate
Xb_test = np.column_stack([np.ones(20), X_test])
y_pred  = Xb_test @ w
rmse    = np.sqrt(np.mean((y_test - y_pred)**2))
print(f'Test RMSE: ${rmse:,.0f}  ({rmse/y_test.mean()*100:.1f}% relative error)')
````

````python
# Visualise results
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

axes[0].scatter(y_test/1000, y_pred/1000, alpha=0.7, s=50, color='steelblue')
line = np.linspace(y_test.min(), y_test.max(), 100) / 1000
axes[0].plot(line, line, 'r--', lw=2, label='Perfect prediction')
axes[0].set_xlabel('Actual Price ($000s)')
axes[0].set_ylabel('Predicted Price ($000s)')
axes[0].set_title('Actual vs Predicted House Prices')
axes[0].legend(); axes[0].grid(True, alpha=0.3)

residuals = (y_test - y_pred) / 1000
axes[1].hist(residuals, bins=15, color='orange', edgecolor='white')
axes[1].axvline(0, color='red', lw=2, linestyle='--')
axes[1].set_xlabel('Prediction Error ($000s)')
axes[1].set_ylabel('Count')
axes[1].set_title(f'Errors  (RMSE = ${rmse/1000:.1f}k)')
axes[1].grid(True, alpha=0.3)

plt.suptitle('Our First Machine Learning Model!', fontsize=13, fontweight='bold')
plt.tight_layout()
plt.show()
print('Congratulations! You just ran a complete ML pipeline.')
print('Every step uses linear algebra -- which is what this course teaches.')
````
