## What problem are we solving?

Before a machine learning model can learn anything, your data has to be turned into **numbers**. But not just a random pile of numbers — they need to be organized in a specific structure.

This notebook teaches the **four structures** used to hold numbers in ML:

| Structure | Short description | Example |
|---|---|---|
| **Scalar** | A single number | The price of one house |
| **Vector** | A list of numbers | All features of one house |
| **Matrix** | A table of numbers | Your entire dataset |
| **Tensor** | A multi-dimensional array | A batch of images |

Think of them as a family: each one is a bigger, more capable version of the last.

````python
# Always run this cell first — import our libraries
import numpy as np
import matplotlib.pyplot as plt

print("Libraries loaded. Let's go!")
````

## 1. Scalar — A Single Number

### What is it?
A **scalar** is just **one number**. Nothing more, nothing less.

### Everyday examples
- The temperature outside: **72°F** — one number
- Your age: **25** — one number
- The price of a coffee: **$4.50** — one number
- A model's accuracy: **94.3%** — one number

### In ML, scalars represent:
- The **loss** (how wrong the model is) — e.g., loss = 0.045
- The **learning rate** — e.g., lr = 0.001
- A **single prediction** — e.g., predicted price = $312,000
- **Hyperparameters** like dropout rate, regularization strength

### Notation
Mathematicians write scalars in lowercase italic: *x*, *y*, *α*, *λ*

### Why does a "single number" need its own special name?
Because in linear algebra, we work with numbers that have **dimensions**. A scalar has **zero dimensions** — it's just a point on the number line. Everything above (vectors, matrices, tensors) has more structure. Knowing something is a scalar tells you it has **no direction, no shape — just a value**.

````python
# ─────────────────────────────────────────────────────────
# SCALARS IN PYTHON AND NUMPY
# ─────────────────────────────────────────────────────────

# Regular Python numbers ARE scalars
house_price    = 312000      # price of one house
learning_rate  = 0.001       # ML hyperparameter
model_accuracy = 0.943       # model performance

print("=== Regular Python scalars ===")
print(f"House price:    ${house_price:,}")
print(f"Learning rate:  {learning_rate}")
print(f"Accuracy:       {model_accuracy:.1%}")

print()
print("=== NumPy 0-dimensional scalar ===")
# NumPy also has a scalar — a 0-dimensional array
np_scalar = np.float64(312000)
print(f"NumPy scalar:   {np_scalar}")
print(f"Shape:          {np_scalar.shape}   ← empty parentheses = 0 dimensions")
print(f"ndim:           {np_scalar.ndim}    ← 0 dimensions")
print(f"Type:           {type(np_scalar)}")

print()
print("=== Scalars come from array operations ===")
prices = np.array([250000, 320000, 190000, 280000, 410000])
avg_price = prices.mean()   # .mean() returns a scalar
print(f"Array of prices:  {prices}")
print(f"Average price:    {avg_price}   ← this is a scalar")
print(f"Shape of avg:     {np.array(avg_price).shape}")
````

## 2. Vector — An Ordered List of Numbers

### What is it?
A **vector** is an **ordered list of numbers**. Order matters — [1, 2, 3] is different from [3, 2, 1].

### Everyday analogy: A recipe
A chocolate cake recipe might be:
```
[flour=200g, sugar=150g, butter=100g, eggs=3, cocoa=50g]
```
That's a vector! It's a list of values that together describe **one thing** (the recipe).

### In ML, one row of your dataset is a vector:
```
House #1 = [1500, 3, 10, 5.2]
             ↑     ↑   ↑    ↑
            size  beds age  distance
```
This is a **feature vector** — a list of numbers that fully describes one data point.

### Geometric interpretation (in 2D)
A 2D vector [3, 4] is like an **arrow** from the origin (0,0) to the point (3,4).
- The direction tells us "which way"
- The length tells us "how far"
This becomes crucial when we talk about similarity and projections later.

### Notation
Vectors are written in bold lowercase: **x**, **w**, **b**  
(or with an arrow: x⃗)

````python
# ─────────────────────────────────────────────────────────
# VECTORS IN NUMPY
# ─────────────────────────────────────────────────────────

# A house described by 4 features: [size_sqft, bedrooms, age_years, distance_km]
house_1 = np.array([1500.0, 3.0, 10.0, 5.2])
house_2 = np.array([2100.0, 4.0, 5.0,  2.8])

print("=== Vectors: feature descriptions of houses ===")
print(f"House 1: {house_1}")
print(f"  Shape: {house_1.shape}   ← (4,) means: 4 elements in 1 dimension")
print(f"  ndim:  {house_1.ndim}")
print()
print(f"House 2: {house_2}")

print()
print("=== Accessing individual elements ===")
print(f"house_1[0] = {house_1[0]}  ← size in sqft")
print(f"house_1[1] = {house_1[1]}  ← bedrooms")
print(f"house_1[2] = {house_1[2]} ← age in years")
print(f"house_1[3] = {house_1[3]} ← distance from city center")

print()
print("=== More vector examples in ML ===")
# Weights of a linear model (one weight per feature)
model_weights = np.array([150.0, 20000.0, -500.0, -8000.0])
print(f"Model weights: {model_weights}")
print(f"  Each weight says how much each feature affects the price")
print(f"  Size: +150 per sqft, Bedrooms: +20000, Age: -500, Distance: -8000")

# A prediction: predicted_price = house_features · weights + bias
bias = 50000
predicted_price = np.dot(house_1, model_weights) + bias  # dot product (covered in notebook 08)
print(f"\nPredicted price for house 1: ${predicted_price:,.0f}")
````

````python
# ─────────────────────────────────────────────────────────
# ROW VECTOR vs COLUMN VECTOR
# ─────────────────────────────────────────────────────────
# This distinction matters for matrix multiplication!

# Row vector: shape (4,) — the default
row = np.array([1500.0, 3.0, 10.0, 5.2])
print("Row vector (default 1D array):")
print(row)
print(f"Shape: {row.shape}   ← just (4,), no second dimension")

print()
# Column vector: shape (4, 1) — one column, multiple rows
col = row.reshape(-1, 1)  # -1 means 'figure it out' → gives shape (4, 1)
print("Column vector (reshaped to column):")
print(col)
print(f"Shape: {col.shape}  ← (4 rows, 1 column)")

print()
print("Transpose: turn row into column and vice versa")
print(f"row.T  shape: {row.T.shape}   ← TRAP! .T on 1D doesn't change shape")
print(f"col.T  shape: {col.T.shape}  ← 2D transpose works correctly")
print()
print("TIP: If you need a true column vector, use reshape(-1, 1)")
````

````python
# ─────────────────────────────────────────────────────────
# VISUALIZING VECTORS AS ARROWS (2D)
# ─────────────────────────────────────────────────────────
# This helps build geometric intuition

fig, axes = plt.subplots(1, 2, figsize=(13, 5))

# Left: Two 2D feature vectors as arrows
ax = axes[0]
vectors_2d = [
    (np.array([3.0, 2.0]), 'steelblue', 'House 1: [size=3, beds=2]'),
    (np.array([4.0, 4.0]), 'orange',    'House 2: [size=4, beds=4]'),
    (np.array([1.0, 3.0]), 'green',     'House 3: [size=1, beds=3]'),
]
for vec, color, label in vectors_2d:
    ax.annotate("", xy=vec, xytext=(0, 0),
                arrowprops=dict(arrowstyle="->", color=color, lw=2.5))
    ax.plot(*vec, 'o', color=color, markersize=8)
    ax.text(vec[0]+0.1, vec[1]+0.1, label, color=color, fontsize=9)
ax.set_xlim(-0.5, 5.5); ax.set_ylim(-0.5, 5.5)
ax.set_xlabel('Feature 1: House Size (scaled)'); ax.set_ylabel('Feature 2: Bedrooms')
ax.axhline(0, color='gray', lw=0.5); ax.axvline(0, color='gray', lw=0.5)
ax.grid(True, alpha=0.3); ax.set_aspect('equal')
ax.set_title('Each house = a vector = an arrow from origin', fontsize=10)

# Right: vector components
ax2 = axes[1]
vec = np.array([3.0, 2.0])
ax2.annotate("", xy=vec, xytext=(0,0),
             arrowprops=dict(arrowstyle="->", color='steelblue', lw=3))
ax2.annotate("", xy=(vec[0], 0), xytext=(0,0),
             arrowprops=dict(arrowstyle="->", color='red', lw=2, linestyle='dashed'))
ax2.annotate("", xy=(vec[0], vec[1]), xytext=(vec[0], 0),
             arrowprops=dict(arrowstyle="->", color='green', lw=2, linestyle='dashed'))
ax2.text(1.5, -0.25, f'x-component = {vec[0]}', color='red', fontsize=10)
ax2.text(vec[0]+0.1, vec[1]/2, f'y-component = {vec[1]}', color='green', fontsize=10)
ax2.text(1.2, 1.3, f'v = [{vec[0]}, {vec[1]}]', color='steelblue', fontsize=12, fontweight='bold')
ax2.set_xlim(-0.5, 4.5); ax2.set_ylim(-0.5, 3.5)
ax2.axhline(0, color='gray', lw=0.5); ax2.axvline(0, color='gray', lw=0.5)
ax2.grid(True, alpha=0.3); ax2.set_aspect('equal')
ax2.set_title('Vector components: the numbers are coordinates', fontsize=10)

plt.suptitle('Vectors = arrows in space, OR lists of feature values', fontsize=12)
plt.tight_layout()
plt.show()
````

## 3. Matrix — A Table of Numbers

### What is it?
A **matrix** is a **rectangular table of numbers** arranged in rows and columns — exactly like a spreadsheet.

### The key insight for ML
> **Your entire dataset is a matrix.**

If you have 1000 houses and 4 features per house, your dataset is a **1000 × 4 matrix**:
- Each **row** = one house (one data sample)
- Each **column** = one feature (one property being measured)

```
         [size,  beds, age,  dist]
House 1: [1500,  3,    10,   5.2 ]
House 2: [2100,  4,    5,    2.8 ]
House 3: [1200,  2,    20,   8.0 ]
   ⋮
```

### Dimensions
We always say **m × n** ("m by n"):
- m = number of **rows** (usually = number of samples)
- n = number of **columns** (usually = number of features)

### Notation
Matrices are written in bold uppercase: **X**, **W**, **A**

Element at row *i*, column *j* is written A[i, j] or Aᵢⱼ

````python
# ─────────────────────────────────────────────────────────
# MATRICES IN NUMPY
# ─────────────────────────────────────────────────────────

# Our dataset: 6 houses × 4 features
# Columns: [size_sqft, bedrooms, age_years, distance_km]
X = np.array([
    [1500, 3, 10, 5.2],   # House 0
    [2100, 4,  5, 2.8],   # House 1
    [1200, 2, 20, 8.0],   # House 2
    [1800, 3,  8, 4.5],   # House 3
    [2500, 5,  2, 1.0],   # House 4
    [950,  1, 30, 12.0],  # House 5
], dtype=float)

print("=== Dataset matrix X ===")
print(X)
print(f"\nShape:   {X.shape}   ← (6 houses, 4 features)")
print(f"Rows:    {X.shape[0]}  (number of houses)")
print(f"Columns: {X.shape[1]}  (number of features)")
print(f"ndim:    {X.ndim}")
````

````python
# ─────────────────────────────────────────────────────────
# INDEXING INTO A MATRIX
# X[row, col] — always: row first, then column
# ─────────────────────────────────────────────────────────

print("=== Indexing ===")
print(f"X[0]       = {X[0]}    ← entire first house (row 0)")
print(f"X[2]       = {X[2]}  ← entire third house (row 2)")
print(f"X[0, 0]    = {X[0,0]}  ← size of house 0")
print(f"X[1, 1]    = {X[1,1]}  ← bedrooms of house 1")
print(f"X[-1, 0]   = {X[-1,0]}  ← size of LAST house (negative indexing)")

print()
print("=== Slicing: getting rows/columns ===")
print(f"All sizes   (col 0): {X[:, 0]}")
print(f"All bedrooms(col 1): {X[:, 1]}")
print(f"First 3 houses:")
print(X[:3])               # rows 0, 1, 2

print(f"\nOnly size and bedrooms (first 2 columns):")
print(X[:, :2])
````

````python
# ─────────────────────────────────────────────────────────
# MATRIX vs VECTOR: the connection
# A matrix is just a collection of vectors stacked together
# ─────────────────────────────────────────────────────────

print("=== Matrix = stacked row vectors ===")
print(f"X[0] = {X[0]}  ← House 0 vector")
print(f"X[1] = {X[1]}  ← House 1 vector")
print("Stack them:")
print(np.array([X[0], X[1]]))

print()
print("=== Each column is also a vector ===")
size_col = X[:, 0]   # all sizes
print(f"Size column: {size_col}")
print(f"→ This is just a 1D array (vector) of all house sizes")

print()
print("=== Other matrix examples in ML ===")
print()
# Weight matrix: connecting 4 input features to 3 output neurons
np.random.seed(42)
W = np.random.randn(4, 3).round(3)   # 4 inputs → 3 neurons
print(f"Neural network weight matrix W (4 inputs → 3 neurons):")
print(W)
print(f"Shape: {W.shape}")
print("Each column = weights going into one neuron")
print("Each row = weights coming from one input feature")
````

````python
# ─────────────────────────────────────────────────────────
# VISUALIZING A MATRIX AS A HEATMAP
# ─────────────────────────────────────────────────────────

fig, axes = plt.subplots(1, 2, figsize=(14, 4))

# Left: dataset matrix as heatmap
# First, normalize columns so they're on comparable scales
X_norm = (X - X.mean(axis=0)) / X.std(axis=0)

im = axes[0].imshow(X_norm, cmap='RdBu', aspect='auto')
plt.colorbar(im, ax=axes[0])
axes[0].set_xticks(range(4))
axes[0].set_xticklabels(['Size', 'Bedrooms', 'Age', 'Distance'])
axes[0].set_yticks(range(6))
axes[0].set_yticklabels([f'House {i}' for i in range(6)])
for i in range(6):
    for j in range(4):
        axes[0].text(j, i, f'{X_norm[i,j]:.1f}', ha='center', va='center', fontsize=9)
axes[0].set_title('Dataset Matrix as Heatmap\n(normalized, red=high, blue=low)', fontsize=10)

# Right: same matrix as a table with actual numbers
axes[1].axis('off')
table_data = [[f'{X[i,j]:.0f}' for j in range(4)] for i in range(6)]
table = axes[1].table(
    cellText=table_data,
    colLabels=['Size (sqft)', 'Bedrooms', 'Age (yrs)', 'Distance (km)'],
    rowLabels=[f'House {i}' for i in range(6)],
    cellLoc='center',
    loc='center'
)
table.auto_set_font_size(False)
table.set_fontsize(10)
table.scale(1.3, 1.8)
axes[1].set_title('Same Matrix as a Table', fontsize=11, pad=20)

plt.suptitle('A matrix is just a table of numbers — your entire dataset!', fontsize=12)
plt.tight_layout()
plt.show()
````

## 4. Tensor — Going Beyond 2D

### What is it?
A **tensor** is the generalization of scalars, vectors, and matrices to **any number of dimensions**.

| Name | Dimensions | Example | Shape |
|---|---|---|---|
| Scalar | 0D | A single number: 42 | () |
| Vector | 1D | A list: [1, 2, 3, 4] | (4,) |
| Matrix | 2D | A table: rows × cols | (100, 4) |
| Tensor | 3D+ | A cube of numbers | (32, 28, 28) |

### Why do we need more than 2D?
Some data is naturally multi-dimensional:

**Color image:**
- Height × Width × Channels (RGB)
- A 256×256 color image → shape (256, 256, 3)
- Those 3 channels are Red, Green, Blue — each a 256×256 matrix!

**Batch of images (in deep learning):**
- Batch × Height × Width × Channels
- 32 images at once → shape (32, 256, 256, 3)

**Text sequences:**
- Batch × Sequence Length × Embedding Dimension
- 16 sentences, each 50 words, each word as 768 numbers → shape (16, 50, 768)

````python
# ─────────────────────────────────────────────────────────
# TENSORS IN NUMPY
# ─────────────────────────────────────────────────────────

print("=== The 4-level hierarchy ===")
print()

# Scalar (0D)
scalar = np.float64(42.0)
print(f"Scalar:  {scalar}")
print(f"  shape={scalar.shape}, ndim={scalar.ndim}")
print()

# Vector (1D)
vector = np.array([1.0, 2.0, 3.0, 4.0])
print(f"Vector:  {vector}")
print(f"  shape={vector.shape}, ndim={vector.ndim}")
print()

# Matrix (2D)
matrix = np.array([[1., 2., 3.], [4., 5., 6.]])
print(f"Matrix:\n{matrix}")
print(f"  shape={matrix.shape}, ndim={matrix.ndim}")
print()

# 3D Tensor: a color image (height=4, width=4, 3 channels)
np.random.seed(7)
image_tensor = np.random.randint(0, 256, size=(4, 4, 3))  # tiny 4x4 color image
print(f"3D Tensor (tiny image):")
print(f"  shape={image_tensor.shape}, ndim={image_tensor.ndim}")
print(f"  Red channel (first pixel row): {image_tensor[0, :, 0]}")
print()

# 4D Tensor: batch of 8 images
batch = np.random.randint(0, 256, size=(8, 4, 4, 3))  # 8 tiny images
print(f"4D Tensor (batch of 8 images):")
print(f"  shape={batch.shape}, ndim={batch.ndim}")
print(f"  Meaning: {batch.shape[0]} images, {batch.shape[1]}×{batch.shape[2]} pixels, {batch.shape[3]} channels")
````

````python
# ─────────────────────────────────────────────────────────
# VISUALIZING AN IMAGE AS A 3D TENSOR
# ─────────────────────────────────────────────────────────

np.random.seed(42)
# Create a simple 8×8 color image (small but visible)
img = np.random.randint(0, 256, size=(8, 8, 3), dtype=np.uint8)

fig, axes = plt.subplots(1, 4, figsize=(14, 3.5))

# Show full color image
axes[0].imshow(img)
axes[0].set_title(f'Color Image\nshape = {img.shape}\n(8×8 pixels, 3 channels)', fontsize=9)
axes[0].axis('off')

# Show each channel separately
channel_names = ['Red Channel', 'Green Channel', 'Blue Channel']
colormaps = ['Reds', 'Greens', 'Blues']
for i in range(3):
    axes[i+1].imshow(img[:, :, i], cmap=colormaps[i])
    axes[i+1].set_title(f'{channel_names[i]}\nimg[:, :, {i}]\nshape = {img[:,:,i].shape}', fontsize=9)
    axes[i+1].axis('off')

plt.suptitle('A color image is a 3D tensor — it\'s just 3 stacked matrices (one per color channel)',
             fontsize=11)
plt.tight_layout()
plt.show()
````

## 5. Putting It All Together

### The whole picture

```
                        ML DATA STRUCTURES

   SCALAR          VECTOR          MATRIX          TENSOR
   ------          ------          ------          ------
   One number      List of #s      Table of #s     3D+ array

   shape: ()       shape: (n,)     shape: (m, n)   shape: (a,b,c,...)
   ndim: 0         ndim: 1         ndim: 2         ndim: 3+

   Example:        Example:        Example:        Example:
   loss = 0.03     one house       whole dataset   batch of images
   accuracy        one prediction  weight matrix   video frames
   learning rate   word embedding  attention mat.  text embeddings
```

````python
# ─────────────────────────────────────────────────────────
# FINAL SUMMARY: quick reference for shapes
# ─────────────────────────────────────────────────────────

print("="*55)
print("SHAPE CHEAT SHEET")
print("="*55)
print(f"{'Name':<12} {'ndim':<8} {'shape example':<20} {'ML role'}")
print("-"*55)

examples = [
    ("Scalar",  np.float64(3.14),               "loss, lr, accuracy"),
    ("Vector",  np.zeros(5),                    "one data sample, weights"),
    ("Matrix",  np.zeros((100, 5)),             "dataset, weight matrix"),
    ("Tensor3", np.zeros((32, 28, 28)),         "batch of grayscale images"),
    ("Tensor4", np.zeros((32, 28, 28, 3)),      "batch of color images"),
]

for name, obj, role in examples:
    print(f"{name:<12} {obj.ndim:<8} {str(obj.shape):<20} {role}")

print()
print("KEY RULE:")
print("  shape[0] = number of SAMPLES (rows) in your dataset")
print("  shape[1] = number of FEATURES (columns) in your dataset")
````
