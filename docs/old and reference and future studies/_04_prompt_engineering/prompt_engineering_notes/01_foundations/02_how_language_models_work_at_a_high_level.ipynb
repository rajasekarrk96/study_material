## Core Notes

### A language model predicts text one piece at a time

The core mechanism is next-token prediction. The model looks at the current context and predicts likely continuations. This explains why small prompt changes can yield major behavioral shifts.

### Training and prompting are different layers

Training teaches the model broad language patterns and task associations. Prompting does not retrain the model. It activates and constrains the behavior available at inference time.

### Fluency is not the same as correctness

Because the model is optimized to generate plausible language, confident wording can hide factual or reasoning errors. Expert prompt design therefore includes evidence, verification, or tool support when accuracy matters.



## Frameworks And Design Patterns

### Pattern activation view

**Use when:** You want to explain why examples, tone, and format requests matter.

**How it works:** A prompt cues the model toward certain language and task patterns. Clearer patterns lead to more stable outputs.

**Watch out for:** This mental model is useful, but do not assume it gives perfect control.

**Scenario:** Same topic, different audience and format.

**Why this example matters:** It demonstrates that prompt conditions steer which learned patterns the model uses.



## Common Mistakes And Corrections

### Mistake 1

**Problem:** Assuming the model automatically knows the intended use case.

**Why it hurts:** Outputs drift into generic or mismatched styles.

**Better move:** State audience, task, and desired outcome explicitly.

### Mistake 2

**Problem:** Confusing coherent language with reliable reasoning.

**Why it hurts:** You may trust polished but wrong outputs.

**Better move:** Use verification, decomposition, citations, or tools when stakes are high.



## Evaluation Checklist

- Can the prompt nudge the model toward the right pattern family?
- Does the output need fluency only, or verified truth?
- Would adding examples reduce ambiguity?
