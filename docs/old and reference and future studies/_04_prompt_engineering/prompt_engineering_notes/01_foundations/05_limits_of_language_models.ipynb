# Limits Of Language Models

**Module:** Foundations

**Level:** Beginner to Intermediate

**Topic Focus:** Understanding where prompting alone cannot guarantee correctness or reliability.


## Learning Objectives

- Recognize common limits such as hallucination, ambiguity sensitivity, and context overload.
- Learn when to add evidence, structure, or external systems.
- Avoid over-trusting polished outputs.


## Why This Topic Matters

Expert prompt engineers know when prompting is enough and when system support, retrieval, tools, or human review are required.

Prompt engineering improves when you can identify the exact job a prompt must do, the failure modes it creates, and the levers available to improve it.

## Core Notes

### Prompting improves behavior but does not eliminate uncertainty

A well-written prompt can reduce mistakes, but it cannot guarantee truth in the absence of evidence or external verification. This is especially important in legal, financial, medical, and scientific tasks.

### Models can follow surface patterns while missing deeper correctness

Outputs may look structured and persuasive even when they contain fabricated details, missed edge cases, or shallow reasoning. Expert users therefore judge outputs against requirements rather than presentation quality alone.

### Some failures are system failures, not prompt failures

If a task requires current data, exact calculations, private documents, or external actions, the right solution may involve tools, retrieval, code execution, or human review instead of endlessly rewriting the prompt.



## Frameworks And Design Patterns

### Prompt vs system boundary

**Use when:** You are deciding whether to keep iterating on wording or add infrastructure.

**How it works:** Ask what the model can infer from prompt text alone and what requires external evidence or enforcement.

**Watch out for:** Do not use prompt edits to mask a missing system capability.



## Prompt Examples

### Evidence-bounded answer prompt

**Scenario:** You want to prevent unsupported claims.

```text
Answer only using the provided document. If the answer is not supported by the text, say 'Not supported by the provided material.' Then cite the exact section used.
```

**Why this example matters:** It shows how prompts can narrow behavior even though they cannot create true guarantees by themselves.



## Common Mistakes And Corrections

### Mistake 1

**Problem:** Assuming more detailed wording always fixes factual errors.

**Why it hurts:** Time is wasted on prompts when evidence is the real missing ingredient.

**Better move:** Introduce retrieval, references, or tooling when truth depends on external data.

### Mistake 2

**Problem:** Ignoring failure categories.

**Why it hurts:** You cannot tell whether a problem is about task clarity, model capability, or missing data.

**Better move:** Classify failures before revising prompts.



## Evaluation Checklist

- Is the task closed-world or open-world?
- Does the answer require verified facts, computation, or current data?
- Can the model satisfy the task from prompt context alone?


## Practice Prompts

Use these as writing drills, not as code exercises.

1. Take a high-stakes task and list which parts can be handled by prompting and which require a larger system.
2. Write a prompt that forces the model to admit uncertainty instead of guessing.


## Summary

Prompt engineering becomes stronger when you understand its limits. Reliability often depends on combining prompts with data, tools, and evaluation.

A strong prompt engineer can explain the topic simply, apply it deliberately, and diagnose when it is missing from a workflow.
