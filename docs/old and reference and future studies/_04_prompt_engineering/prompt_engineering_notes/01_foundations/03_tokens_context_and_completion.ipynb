Prompt engineers need to manage limited context space, prioritize information, and control output length and structure.

Prompt engineering improves when you can identify the exact job a prompt must do, the failure modes it creates, and the levers available to improve it.

## Core Notes

### Tokens are the model's working units

A token is often smaller than a word and sometimes larger than a character. Since models operate on tokens, all prompt and output budgeting is fundamentally token budgeting.

### Context includes more than the latest user message

System instructions, prior turns, examples, retrieved documents, and tool traces may all consume context. Expert prompting often means deciding what not to include.

### Completion design begins before output

If you need a brief, structured, or evidence-backed output, that must be designed in the prompt. Completion quality is partly determined before generation starts.



## Frameworks And Design Patterns

### Priority ordering

**Use when:** Context is tight and not everything can fit.

**How it works:** Place critical instructions and evidence first, compress supporting details, and remove irrelevant material.

**Watch out for:** Important instructions buried inside long context are easy for the model to underweight.



## Prompt Examples

### Compression-aware summary prompt

**Scenario:** You need a concise answer from a long source text.

```text
Summarize the following report for an operations manager. Focus only on deadlines, blockers, budget risks, and decisions needed this week. Ignore historical background unless it changes an action item.
```

**Why this example matters:** It shows how context selection and output priorities work together.



## Common Mistakes And Corrections

### Mistake 1

**Problem:** Stuffing every available detail into the prompt.

**Why it hurts:** Signal is drowned by noise and key instructions weaken.

**Better move:** Include only task-relevant material and label it clearly.

### Mistake 2

**Problem:** Asking for long outputs after using most of the context budget on input.

**Why it hurts:** Truncation or shallow answers become more likely.

**Better move:** Balance input size and output needs intentionally.



## Evaluation Checklist

- Which parts of the context are essential, optional, or removable?
- Is the desired output length realistic given the input size?
- Will conversation history help or distract?


## Practice Prompts

Use these as writing drills, not as code exercises.

1. Take a long article and write a prompt that clearly defines what should be ignored.
2. Write two versions of a prompt: one for a short answer and one for a deep answer, while keeping the same source context.

Token, context, and completion awareness turns prompting from vague writing into constrained design under limited capacity.
