## Why This Topic Matters

Developer-facing prompt systems often process logs, tickets, documents, screenshots, and user conversations that may contain sensitive information.

Prompt engineering improves when you can identify the exact job a prompt must do, the failure modes it creates, and the levers available to improve it.

## Core Notes

### The best sensitive data prompt is often a smaller prompt

If a workflow does not need a full raw log, ticket, or document, do not send it. Data minimization is both a privacy principle and a context-quality improvement.

### Redaction can be a prompt task of its own

Some workflows benefit from a first-stage prompt that strips or masks sensitive details before a second-stage analysis prompt sees the content.

### Memory and logging policies matter here too

Privacy risk does not end at generation. Stored traces, retained context, and derived summaries can all leak sensitive information if governance is weak.



## Frameworks And Design Patterns

### Minimize, redact, then analyze

**Use when:** Inputs may contain sensitive user or business data.

**How it works:** Remove unnecessary identifiers first, then run the core task on reduced or sanitized content.

**Watch out for:** Over-redaction can remove details needed for correct reasoning, so balance minimization with task fidelity.



## Prompt Examples

### Redact-before-analysis workflow

**Scenario:** Support logs may contain secrets and PII.

```text
First detect and mask emails, phone numbers, account IDs, tokens, and passwords in the text. Then summarize the technical issue and next action items using only the redacted version.
```

**Why this example matters:** It shows how privacy-aware prompting can be built into the workflow itself.



## Common Mistakes And Corrections

### Mistake 1

**Problem:** Sending raw sensitive content when the task needs only a subset.

**Why it hurts:** Risk rises without improving result quality.

**Better move:** Minimize and redact before prompting where possible.

### Mistake 2

**Problem:** Ignoring sensitive information in stored traces and memories.

**Why it hurts:** Operational data becomes a secondary leak channel.

**Better move:** Apply governance to retention, tracing, and memory design too.



## Evaluation Checklist

- What sensitive information is truly necessary for this task?
- Can a redaction or minimization stage reduce risk safely?
- How is sensitive content handled in memory and logs?
