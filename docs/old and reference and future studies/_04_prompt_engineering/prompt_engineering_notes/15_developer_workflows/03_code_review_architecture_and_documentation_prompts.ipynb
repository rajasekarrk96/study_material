## Core Notes

### Review prompts need a review lens

A generic request to review code often produces low-signal commentary. Better review prompts specify what to inspect such as correctness, regressions, performance, API design, edge cases, or missing tests.

### Architecture prompts should expose tradeoffs

Design discussions are strongest when prompts ask for alternatives, failure modes, bottlenecks, coupling concerns, and operational consequences rather than only requesting an opinion.

### Documentation prompts should optimize for the reader

A summary for maintainers differs from onboarding notes for new developers or migration notes for integrators. Audience again matters strongly in technical prompting.

### Architecture critique prompts should ask for alternatives, not only criticism

Useful design review prompts surface tradeoffs, possible alternative approaches, migration paths, failure modes, and operational consequences instead of only asking whether a design is good or bad.

### Technical documentation prompts benefit from structure contracts

Strong prompts for technical docs often specify required sections such as purpose, assumptions, dependencies, failure cases, rollout notes, and unresolved questions. This makes the output more maintainable and more useful to teams.

### Different technical artifacts require different review rubrics

A database migration, an API contract, a caching design, and a new service boundary all deserve different prompt checklists. Reusing one generic review prompt usually leaves important risks invisible.



## Frameworks And Design Patterns

### Findings first, rationale second

**Use when:** You want actionable code review or architecture critique output.

**How it works:** Ask for concrete findings in severity order before broader commentary or suggestions.

**Watch out for:** If the review lens is too broad, output may become generic and low-priority.

### Artifact-specific review checklist

**Use when:** The technical artifact has domain-specific failure modes, such as APIs, migrations, or distributed systems changes.

**How it works:** Embed the key risk dimensions for that artifact type directly into the prompt.

**Watch out for:** A checklist that is too long may dilute focus unless priorities are stated clearly.

### Reader-first technical docs prompt

**Use when:** You need maintainers, new team members, or integrators to understand a system artifact quickly.

**How it works:** Specify who the document is for and which questions it should let that reader answer confidently.

**Watch out for:** Documentation that sounds polished but omits assumptions or failure cases is often less useful than plainer but operationally complete notes.



## Prompt Examples

### Risk-focused code review prompt

**Scenario:** Reviewing a pull request for correctness and regressions.

```text
Review this change for correctness, behavioral regressions, missing tests, and operational risk. List findings first in severity order, with file references and a short explanation of impact. Keep style comments secondary unless they affect maintainability or correctness.
```

**Why this example matters:** It aligns the prompt with how strong engineering reviews are usually structured.

### API contract review prompt

**Scenario:** A backend team is reviewing a new endpoint design.

```text
Review this API proposal for request/response clarity, validation gaps, backward compatibility risk, pagination behavior, error modeling, and observability implications. Prioritize any breaking-change risk or ambiguous contract behavior above style suggestions.
```

**Why this example matters:** It demonstrates an artifact-specific review rubric inside the prompt itself.

### Architecture ADR prompt

**Scenario:** A team wants a concise design record from a technical decision.

```text
Turn the design discussion below into an architecture decision record for future maintainers. Include context, chosen option, rejected alternatives, tradeoffs, rollout implications, operational risks, and open questions. Write for engineers who join the team later and need to understand why this decision was made.
```

**Why this example matters:** It shows documentation prompting as audience-aware technical communication, not just summarization.



## Common Mistakes And Corrections

### Mistake 1

**Problem:** Asking for general feedback on technical artifacts.

**Why it hurts:** The model produces vague praise or broad advice.

**Better move:** Define the review lens and output order clearly.

### Mistake 2

**Problem:** Treating documentation prompts as simple summarization.

**Why it hurts:** Reader needs and operational context are missed.

**Better move:** Specify audience, purpose, and maintenance use case.

### Mistake 3

**Problem:** Using the same review prompt for diffs, APIs, and architecture docs.

**Why it hurts:** Key risk categories specific to each artifact are missed.

**Better move:** Customize the review lens to the artifact type.

### Mistake 4

**Problem:** Requesting architecture critique without asking for alternatives or tradeoffs.

**Why it hurts:** The output becomes opinionated but not decision-useful.

**Better move:** Ask for competing options, failure modes, and operational consequences explicitly.



## Evaluation Checklist

- Does the prompt name the technical review criteria explicitly?
- Are findings prioritized by impact rather than by style?
- Is the documentation output shaped for a real developer audience?
- Does the review rubric match the artifact type?
- Would the output help a real engineering decision or maintenance task?
