# Code Generation Refactoring And Implementation Prompts

**Module:** Developer Workflows Code Prompting And Technical Work

**Level:** Intermediate to Advanced

**Topic Focus:** Prompt patterns for asking models to write, update, or refactor code responsibly.


## Learning Objectives

- Write clearer implementation prompts for coding tasks.
- Constrain code generation by scope, style, and acceptance criteria.
- Reduce vague requests that produce partial or mismatched code.


## Why This Topic Matters

Developers often use prompt engineering first through coding tasks, but weak prompting can create brittle code, missed constraints, and expensive rewrites.

Prompt engineering improves when you can identify the exact job a prompt must do, the failure modes it creates, and the levers available to improve it.

## Core Notes

### Good coding prompts look like lightweight engineering specs

The model should know what to build, where to change it, what constraints to respect, and how success will be judged. A vague 'fix this' prompt often leads to shallow or risky edits.

### Implementation prompts should define boundaries

State the files, modules, or behavior in scope. This reduces unnecessary rewrites and helps keep code changes aligned with the real task.

### Acceptance criteria matter as much as code style

A prompt that defines expected behavior, tests, edge cases, and non-goals often produces much better implementation work than one that asks only for code.

### Implementation prompts should expose hidden assumptions

If framework versions, database behavior, deployment constraints, or API compatibility expectations matter, they should be stated directly. Otherwise the model may make locally reasonable but globally dangerous assumptions.

### Refactors need preservation rules

A refactor prompt should define which behaviors must remain identical, which APIs must stay stable, and which architectural improvements are desired. This helps prevent clean-looking but semantically risky rewrites.

### Code prompts should match the real development unit

Sometimes the right prompt is feature-level, sometimes function-level, sometimes repository-level. If the task granularity is wrong, the model either sees too little context or attempts too much change at once.



## Frameworks And Design Patterns

### Goal + scope + constraints + acceptance

**Use when:** You want reliable implementation or refactor results.

**How it works:** Define the coding goal, the exact scope, any style or architectural constraints, and the success conditions.

**Watch out for:** Do not overload one prompt with several independent code tasks unless they truly belong together.

### Behavior preservation checklist

**Use when:** You are asking for a refactor, migration, or cleanup pass.

**How it works:** State what must stay unchanged, such as public API shape, query semantics, auth behavior, logging expectations, or error contracts.

**Watch out for:** Without preservation rules, the model may treat cleanup as permission to redesign behavior.

### Assumptions-first implementation prompt

**Use when:** The task has missing environment detail but still seems safe to draft.

**How it works:** Ask the model to list working assumptions first, then implement against them, and clearly mark where verification is still needed.

**Watch out for:** If critical assumptions affect correctness, clarification may be better than drafting.



## Prompt Examples

### Scoped implementation prompt

**Scenario:** A developer wants a targeted feature addition.

```text
Add pagination support to the existing users endpoint. Scope: update only the controller, service, and response schema related to user listing. Constraints: preserve current filtering behavior, do not change authentication, and keep compatibility with existing clients. Acceptance: support page and page_size inputs, return total count, and add tests for default and custom pagination.
```

**Why this example matters:** It frames the task like a practical development ticket rather than a vague ask.

### Safe refactor prompt

**Scenario:** A developer wants cleanup without behavioral drift.

```text
Refactor the payment reconciliation service for readability and testability. Scope: only the reconciliation module and its direct tests. Preserve: output schema, retry behavior, logging fields, and public method signatures. Improve: reduce nested conditionals, isolate provider-specific rules into helpers, and add regression tests around duplicate-event handling. Do not change queueing, persistence, or external API calls.
```

**Why this example matters:** It shows how preservation rules and non-goals protect a refactor request.

### Assumption-aware implementation prompt

**Scenario:** A developer wants a draft despite partial context.

```text
Implement a rate-limiter middleware for the API gateway. If framework-specific details are unclear, state assumptions first and proceed with a draft using standard middleware patterns. Include where the assumptions affect correctness, and add tests for burst limits, reset windows, and per-user isolation.
```

**Why this example matters:** It balances forward progress with explicit assumption management.



## Common Mistakes And Corrections

### Mistake 1

**Problem:** Asking for code without naming the target behavior clearly.

**Why it hurts:** The model may change code that was not meant to move.

**Better move:** Specify scope, files, and acceptance criteria.

### Mistake 2

**Problem:** Requesting refactors without non-goal boundaries.

**Why it hurts:** The model may rewrite too much under the label of cleanup.

**Better move:** State what must remain unchanged.

### Mistake 3

**Problem:** Omitting compatibility or migration expectations.

**Why it hurts:** Generated changes may look correct but break callers, data flows, or deployment assumptions.

**Better move:** Call out backward-compatibility, migration, and rollout constraints explicitly.

### Mistake 4

**Problem:** Asking for implementation without test expectations.

**Why it hurts:** The code may compile but remain weakly verified around the real edge cases.

**Better move:** Specify the behaviors and scenarios that tests must cover.



## Evaluation Checklist

- Does the prompt define scope and acceptance criteria clearly?
- Are architectural constraints or non-goals visible?
- Could a teammate review the requested change from the prompt alone?
- Are compatibility, rollout, or migration risks described?
- Would the model know what tests or evidence of correctness are expected?

Developer prompts for code generation work best when they read like precise implementation briefs with clear scope and testable outcomes.
