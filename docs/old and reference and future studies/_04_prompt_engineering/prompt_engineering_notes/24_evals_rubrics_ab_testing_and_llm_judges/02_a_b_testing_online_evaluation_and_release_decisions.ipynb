## Core Notes

### Offline evals and online behavior answer different questions

Offline evals show whether a prompt handles known test cases well. Online tests reveal how the prompt behaves in live usage, including user satisfaction, follow-up frequency, and operational friction.

### A/B testing should reflect real product goals

A prompt that produces more detailed answers may look better offline but frustrate users if it increases latency or hides the direct answer. Online comparison helps reveal these tradeoffs.

### Release decisions should combine multiple signals

Strong prompt ops often considers offline task scores, online user outcomes, safety metrics, and rollback readiness together instead of relying on one metric alone.



## Frameworks And Design Patterns

### Offline screen, online validate, then release

**Use when:** A prompt revision affects a user-facing workflow.

**How it works:** Use offline evaluation to filter weak candidates, then compare stronger variants in controlled live settings before full rollout.

**Watch out for:** Online tests without safety monitoring can expose users to regressions.

## Common Mistakes And Corrections

### Mistake 1

**Problem:** Promoting a prompt based only on offline examples.

**Why it hurts:** Live user experience regressions may go unseen.

**Better move:** Add controlled online validation where appropriate.

### Mistake 2

**Problem:** Running A/B tests without defining success metrics first.

**Why it hurts:** Teams argue after the fact about what counted as improvement.

**Better move:** Define decision criteria before rollout.



## Evaluation Checklist

- What does this prompt change aim to improve in actual user outcomes?
- Which offline metrics must pass before online testing begins?
- What online signals would justify rollout or rollback?
