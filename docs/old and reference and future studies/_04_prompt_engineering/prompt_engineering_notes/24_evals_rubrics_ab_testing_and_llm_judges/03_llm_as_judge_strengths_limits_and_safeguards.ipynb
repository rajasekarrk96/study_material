## Core Notes

### LLM judges are tools, not truth oracles

They can help compare outputs at scale, but their ratings still depend on rubric quality, prompt phrasing, examples, and the judge model's own weaknesses.

### Judge prompts should be as carefully designed as generator prompts

A weak judge prompt can produce shallow or unstable scoring. Good judge prompts specify the rubric, evidence basis, allowed outputs, and how to handle ties or insufficient information.

### Human calibration remains important

Use spot checks, disagreement analysis, and periodic recalibration to make sure automated judging still reflects the real quality standards you care about.



## Frameworks And Design Patterns

### Judge with rubric, audit with humans

**Use when:** You want scalable evaluation but cannot rely entirely on manual review.

**How it works:** Use model-based judging for breadth and human review for calibration, edge cases, and disputed outputs.

**Watch out for:** Do not let a judge prompt quietly redefine quality standards away from product needs.



## Prompt Examples

### Pairwise judge setup

**Scenario:** Two answer variants need comparative scoring.

```text
Compare answer A and answer B using only the rubric dimensions of factual faithfulness, task completeness, and usefulness to the intended user. Choose A, B, or Tie. Explain the decisive criterion briefly.
```

**Why this example matters:** It shows a constrained judge prompt that avoids vague all-around scoring.



## Common Mistakes And Corrections

### Mistake 1

**Problem:** Assuming automated judging is objective by default.

**Why it hurts:** Hidden biases and rubric drift can accumulate.

**Better move:** Audit and calibrate judge behavior regularly.

### Mistake 2

**Problem:** Using unconstrained judge prompts.

**Why it hurts:** Scores become noisy, unstable, or hard to interpret.

**Better move:** Constrain the rubric and output shape explicitly.



## Evaluation Checklist

- What parts of judging can be automated safely?
- How will judge outputs be calibrated against human review?
- Could the judge prompt itself be biasing results or rewarding the wrong features?
