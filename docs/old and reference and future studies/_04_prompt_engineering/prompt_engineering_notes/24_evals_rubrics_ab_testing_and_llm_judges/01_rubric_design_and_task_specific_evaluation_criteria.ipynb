# Rubric Design And Task Specific Evaluation Criteria

**Module:** Evals Rubrics A B Testing And LLM Judges

**Level:** Expert

**Topic Focus:** Building evaluation rubrics that reflect real task success instead of generic preferences.


## Learning Objectives

- Design rubrics tailored to specific prompt tasks.
- Distinguish core criteria from cosmetic preferences.

## Why This Topic Matters

Prompt quality improves only when teams measure the dimensions that actually matter for the task, and that begins with strong rubric design.

Prompt engineering improves when you can identify the exact job a prompt must do, the failure modes it creates, and the levers available to improve it.

## Core Notes

### A rubric should capture what success means operationally

Good rubrics define whether the output is accurate, complete, grounded, concise, helpful, safe, or structurally compliant in the specific ways the feature needs.

### Not every criterion deserves equal weight

For some tasks, accuracy dominates style. For others, groundedness and escalation behavior may matter more than completeness. Weighting helps evaluators make tradeoffs consistently.

### Rubrics should be concrete enough to apply repeatedly

Criteria like 'good' or 'smart' are too vague. Strong rubrics use observable standards that different reviewers can apply with reasonable agreement.



## Frameworks And Design Patterns

### Task goal -> failure risks -> eval rubric

**Use when:** You are defining how to compare prompts or judge outputs.

**How it works:** Start from the task goal, identify the major failure modes, and convert them into explicit evaluation dimensions.

**Watch out for:** A rubric that is too broad becomes hard to score consistently.



## Prompt Examples

### Support assistant rubric sketch

**Scenario:** A grounded support workflow needs reliable evaluation.

```text
Score each answer on: documentation faithfulness, usefulness to the user, proper handling of uncertainty, citation quality, and escalation judgment.
```

**Why this example matters:** It shows how a rubric can reflect the true job of the assistant rather than only surface polish.



## Common Mistakes And Corrections

### Mistake 1

**Problem:** Using the same rubric for very different prompt tasks.

**Why it hurts:** Important task-specific failures may be missed.

**Better move:** Design rubrics around the actual task and risk profile.

### Mistake 2

**Problem:** Mixing essential criteria with taste preferences.

**Why it hurts:** Scoring becomes noisy and less meaningful.

**Better move:** Separate functional criteria from optional style preferences.



## Evaluation Checklist

- Do the rubric dimensions match the real use case?
- Can two reviewers apply the rubric with similar conclusions?
- Which criteria should be weighted most heavily and why?
