## Why This Topic Matters

Without evaluation, prompt work becomes anecdotal. Reliable improvement requires explicit criteria and repeated comparison.

Prompt engineering improves when you can identify the exact job a prompt must do, the failure modes it creates, and the levers available to improve it.

## Core Notes

### You cannot improve what you do not measure

Prompt quality should be evaluated against concrete goals such as accuracy, coverage, formatting compliance, groundedness, or decision usefulness. Informal impressions are not enough for repeatable work.

### Small eval sets are still valuable

Even a modest set of representative examples can reveal whether a revision improved consistency or only helped one lucky case.

### Criteria should match the task

A creative writing prompt and a data extraction prompt do not need the same rubric. Experts choose the evaluation dimensions that matter to the actual use case.



## Frameworks And Design Patterns

### Input set + rubric + comparison

**Use when:** You want to compare prompt versions fairly.

**How it works:** Run the same examples through different prompts and score the outputs against a fixed rubric.

**Watch out for:** Changing both the prompt and the test set at once makes conclusions weak.

## Common Mistakes And Corrections

### Mistake 1

**Problem:** Judging prompt quality from one strong output.

**Why it hurts:** You miss variability and edge-case failures.

**Better move:** Use multiple test cases with different patterns.

### Mistake 2

**Problem:** Using vague evaluation language.

**Why it hurts:** Reviews become subjective and hard to compare.

**Better move:** Turn quality into observable criteria.



## Evaluation Checklist

- What does success mean for this task specifically?
- Do the test inputs cover easy, typical, and edge cases?
- Can two reviewers apply the rubric similarly?
