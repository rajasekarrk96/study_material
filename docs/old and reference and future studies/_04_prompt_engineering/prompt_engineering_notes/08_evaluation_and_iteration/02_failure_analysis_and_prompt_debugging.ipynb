# Failure Analysis And Prompt Debugging

**Module:** Evaluation Debugging And Iteration

**Level:** Advanced

**Topic Focus:** Diagnosing why outputs fail so revisions become targeted instead of random.


## Learning Objectives

- Categorize failures precisely.
- Map failures to likely prompt or system causes.
- Debug by changing one major variable at a time.


## Why This Topic Matters

Most prompting time is not spent writing first drafts; it is spent understanding why outputs go wrong and what change is actually needed.

Prompt engineering improves when you can identify the exact job a prompt must do, the failure modes it creates, and the levers available to improve it.

## Core Notes

### Name the failure before fixing it

An output can fail by being factually wrong, incomplete, badly structured, overconfident, off-tone, ungrounded, or simply aimed at the wrong audience. Different failure types call for different interventions.

### Some prompt bugs are actually system bugs

If the model lacks the required evidence, no wording trick will fully solve the task. Debugging starts by deciding whether the failure came from unclear prompting or missing capability.

### Isolate variables in revision

When you change role, examples, context, and output format all at once, you learn very little. Experts revise prompts deliberately and compare against the same cases.



## Frameworks And Design Patterns

### Failure type -> probable cause -> intervention

**Use when:** An output is bad and the next revision is unclear.

**How it works:** Classify the failure, hypothesize the cause, then choose a specific prompt or system change.

**Watch out for:** Do not assume the first plausible cause is the right one without comparison.



## Prompt Examples

### Debug note example

**Scenario:** A review prompt keeps missing edge cases.

```text
Observed failure: output lists obvious issues but misses contradictions across sections. Likely cause: the prompt asks for 'issues' broadly but never directs cross-section comparison. Proposed change: add a dedicated instruction to compare claims across sections before listing findings.
```

**Why this example matters:** It models targeted debugging rather than random rewriting.



## Common Mistakes And Corrections

### Mistake 1

**Problem:** Calling every bad output a hallucination.

**Why it hurts:** You lose diagnostic precision.

**Better move:** Separate unsupported invention from incompleteness, ambiguity, tone mismatch, and formatting failure.

### Mistake 2

**Problem:** Making large revisions without a theory of failure.

**Why it hurts:** Improvement becomes slow and untraceable.

**Better move:** State the suspected cause before revising.



## Evaluation Checklist

- What exact failure occurred?
- Did the prompt omit a needed instruction, context block, or constraint?
- Would a system change solve the problem more directly?


## Practice Prompts

Use these as writing drills, not as code exercises.

1. Take a weak output and write a debugging note with failure type, likely cause, and planned revision.
2. Create a mini checklist for diagnosing prompt failures in your own workflow.


## Summary

Prompt debugging improves sharply when failures are named precisely and revisions are tied to specific hypotheses.

A strong prompt engineer can explain the topic simply, apply it deliberately, and diagnose when it is missing from a workflow.
