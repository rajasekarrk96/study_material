## Core Notes

### Agentic prompting is workflow prompting

Instead of asking for one answer, you are prompting for a sequence of behaviors: gather information, plan, act, check, and continue or stop. The prompt becomes part of a control loop.

### State matters

An agent-like system usually needs memory of what has been learned, done, or promised. Prompt design must account for state representation and state updates.

### Guardrails and stopping conditions are essential

Without boundaries, agentic systems may loop, over-act, or take unsafe actions. Prompting must define when to stop, when to ask for help, and when to escalate uncertainty.



## Frameworks And Design Patterns

### Goal -> plan -> act -> check -> stop

**Use when:** A task requires several dependent steps and possible tool use.

**How it works:** The model works through a repeated control loop until a stopping rule is met.

**Watch out for:** Unclear stop conditions lead to waste, drift, or unsafe persistence.



## Prompt Examples

### Research assistant workflow prompt

**Scenario:** You want careful staged behavior rather than one big answer.

```text
Your job is to produce a short research brief. First list what information is needed. Then identify which facts must come from sources. Then summarize the gathered evidence, highlight uncertainty, and stop before making recommendations if key evidence is missing.
```

**Why this example matters:** It shows goal-driven staged behavior with an uncertainty stop rule.



## Common Mistakes And Corrections

### Mistake 1

**Problem:** Calling a workflow agentic when it has no explicit state or loop control.

**Why it hurts:** The system looks advanced but is actually brittle.

**Better move:** Define goals, steps, state, and stopping conditions explicitly.

### Mistake 2

**Problem:** Letting the model continue despite missing evidence.

**Why it hurts:** The workflow produces confident but unsafe conclusions.

**Better move:** Add explicit escalation or stop rules.



## Evaluation Checklist

- What state must persist between steps?
- Which actions require external confirmation?
- When should the workflow stop, ask, or escalate?
