# Output Format Control

**Module:** Core Prompting Patterns

**Level:** Intermediate

**Topic Focus:** Shaping answers into reusable forms such as bullets, tables, labels, and schemas.


## Learning Objectives

- Specify output shape clearly.
- Choose formats that match the task and downstream use.
- Reduce variability through stronger output contracts.


## Why This Topic Matters

Even a good answer is less useful if it arrives in an unpredictable form. Format control is essential for workflow integration.

Prompt engineering improves when you can identify the exact job a prompt must do, the failure modes it creates, and the levers available to improve it.

## Core Notes

### Output form affects usability

A dense paragraph may be harder to scan than a table. A table may be harder to narrate than a concise memo. Expert prompting matches the format to the human or system consuming the answer.

### Specify columns, fields, or sections explicitly

The more predictable the output needs to be, the more explicit you should be about headings, labels, and allowed values.

### Format control helps evaluation

A stable structure makes it easier to compare runs, score outputs, and feed results into other steps.



## Frameworks And Design Patterns

### Schema-first prompting

**Use when:** The output will be consumed by another process or must remain stable.

**How it works:** Define the exact fields and expected values before generation begins.

**Watch out for:** If the schema is too rigid for the task, information may be forced into poor slots.



## Prompt Examples

### Table-based review prompt

**Scenario:** You want consistent critique output.

```text
Review the proposal and return a table with columns: section, issue, severity, evidence, suggested revision. Use severity values only from: low, medium, high.
```

**Why this example matters:** It makes the critique structured and easier to compare across documents.



## Common Mistakes And Corrections

### Mistake 1

**Problem:** Asking for structure but leaving field names vague.

**Why it hurts:** Outputs vary in labeling and ordering.

**Better move:** Name the exact columns or sections.

### Mistake 2

**Problem:** Choosing a format that does not match the user need.

**Why it hurts:** The answer may be correct but awkward to use.

**Better move:** Design output for the consumer, not only the model.



## Evaluation Checklist

- Is the output format fully specified?
- Does the format improve reuse or evaluation?
- Could a table, checklist, or label set work better than free text?
