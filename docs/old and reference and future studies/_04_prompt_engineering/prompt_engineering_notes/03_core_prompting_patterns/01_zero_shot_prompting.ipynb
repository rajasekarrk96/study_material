**Module:** Core Prompting Patterns

**Topic Focus:** Direct prompting without examples.


## Learning Objectives

## Why This Topic Matters

Zero-shot prompting is often the simplest and fastest approach, and many tasks do not need examples if the instruction is clear enough.

Prompt engineering improves when you can identify the exact job a prompt must do, the failure modes it creates, and the levers available to improve it.

## Core Notes

### Zero-shot prompting relies on task clarity

If the task is common and the desired answer format is obvious, a direct prompt may be enough. Simplicity is often a strength because it reduces prompt noise.

### Zero-shot works best when hidden assumptions are low

The fewer unspoken decisions the model has to make, the better zero-shot tends to perform. Ambiguous classification boundaries, niche style rules, and unusual schemas often need stronger scaffolding.

### A strong zero-shot prompt is still structured

Zero-shot does not mean careless. You can still specify audience, constraints, labels, and output format without giving examples.



## Frameworks And Design Patterns

### Direct task template

**Use when:** The task is familiar and examples would add little value.

**How it works:** State the task, define the scope, and constrain the output without demonstrations.

**Watch out for:** If outputs vary too much, you may need few-shot examples.



## Prompt Examples

### Zero-shot extraction

**Scenario:** Simple fact extraction from support messages.

```text
Extract the customer name, order number, reported issue, and requested resolution from the message below. Return the result as a four-column table.
```

**Why this example matters:** It shows a task that is often handled well without examples.



## Common Mistakes And Corrections

### Mistake 1

**Problem:** Using zero-shot for unstable edge-case classification.

**Why it hurts:** Boundary decisions become inconsistent.

**Better move:** Add examples or clearer label definitions.

### Mistake 2

**Problem:** Leaving output shape implicit.

**Why it hurts:** The answer is usable but inconsistent across runs.

**Better move:** State the exact response format.



## Evaluation Checklist

- Can the task be understood without demonstration?
- Are label definitions or format instructions enough on their own?
- Would examples reduce ambiguity significantly?

## Summary

Zero-shot prompting is powerful when tasks are clear, common, and well-bounded by explicit instructions.
