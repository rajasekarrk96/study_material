# Delimiters Structure And Prompt Layout

**Module:** Core Prompting Patterns

**Level:** Intermediate

**Topic Focus:** Using visible structure to separate instructions, data, and examples.


## Learning Objectives

- Use delimiters to separate prompt parts clearly.
- Improve readability for long or multi-part prompts.
- Avoid instruction-data mixing that causes failures.


## Why This Topic Matters

Clear layout reduces prompt ambiguity and makes complex prompts easier for both humans and models to follow.

Prompt engineering improves when you can identify the exact job a prompt must do, the failure modes it creates, and the levers available to improve it.

## Core Notes

### Structure is a reliability tool

Long prompts fail when instructions, examples, and source text blend together. Delimiters and headers separate what the model should do from what it should analyze.

### Readable prompts are easier to debug

A structured layout is not only for the model. It helps you inspect which part of the prompt may be causing confusion or conflict.

### Prompt layout affects downstream maintainability

When prompts become team assets, readable structure allows others to revise them safely without rewriting the entire instruction block.



## Frameworks And Design Patterns

### Instruction / context / output sections

**Use when:** A prompt includes source text, examples, and several rules.

**How it works:** Place the task first, then provide clearly labeled context and a separate output specification.

**Watch out for:** Do not create so many sections that the prompt becomes harder to scan.



## Prompt Examples

### Delimited analysis prompt

**Scenario:** Document review with strong separation.

```text
TASK:
Review the document for missing assumptions and unclear decisions.

DOCUMENT:
<<<
[paste document here]
>>>

OUTPUT:
Return a table with columns: issue, evidence, why it matters, suggested fix.
```

**Why this example matters:** It cleanly separates the instruction from the source material.



## Common Mistakes And Corrections

### Mistake 1

**Problem:** Embedding instructions inside the middle of raw source text.

**Why it hurts:** The model may not reliably distinguish command from content.

**Better move:** Place instructions in distinct sections with clear markers.

### Mistake 2

**Problem:** Using inconsistent delimiter conventions.

**Why it hurts:** Prompt readability and maintenance suffer.

**Better move:** Choose a simple consistent layout pattern.



## Evaluation Checklist

- Can a human quickly identify the task, data, and output format?
- Are source materials clearly bounded?
- Would the prompt stay readable after more context is added?
