## Core Notes

### Hallucination is often a task design issue as well as a model issue

When prompts demand answers without enough evidence, the model may fill gaps with plausible language. Prompts that allow uncertainty and define evidence boundaries can reduce this pressure.

### Uncertainty should be operationalized

A good prompt may instruct the model to say 'not supported by the provided material', 'insufficient evidence', or 'requires external verification' in specific cases. This is more useful than generic cautionary language.

### Evidence discipline improves user trust

When answers cite their basis or clearly separate known facts from inference, users can judge them more intelligently.



## Frameworks And Design Patterns

### Answer / evidence / uncertainty split

**Use when:** The task requires groundedness and honest limits.

**How it works:** Return the answer, the supporting evidence, and any uncertainty separately.

**Watch out for:** If evidence is weak, the answer should remain proportionally cautious.



## Prompt Examples

### Evidence-bounded response prompt

**Scenario:** You want the model to avoid unsupported statements.

```text
Answer only from the supplied materials. For each conclusion, provide one supporting quote or phrase. If the materials do not support a conclusion, say 'insufficient evidence from supplied materials.'
```

**Why this example matters:** It operationalizes evidence and uncertainty handling.



## Common Mistakes And Corrections

### Mistake 1

**Problem:** Demanding certainty where the context does not justify it.

**Why it hurts:** The model is pushed toward confident invention.

**Better move:** Allow uncertainty and define how to express it.

### Mistake 2

**Problem:** Assuming a polished tone implies evidence.

**Why it hurts:** Unsupported content is trusted too easily.

**Better move:** Request visible evidence or support markers.



## Evaluation Checklist

- Does the prompt allow the model to say 'I do not know' in a useful way?
- What evidence should accompany major claims?
- Is the answer clearly bounded by available information?


## Practice Prompts

Use these as writing drills, not as code exercises.

## Summary

Reliability improves when prompts make evidence visible and uncertainty acceptable instead of silently encouraging guesswork.
