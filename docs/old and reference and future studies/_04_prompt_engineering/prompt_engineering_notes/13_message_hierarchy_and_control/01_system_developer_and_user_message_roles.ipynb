## Why This Topic Matters

Developers working with chat APIs need to understand that prompt behavior is shaped by multiple message layers, not only by the latest user request.

Prompt engineering improves when you can identify the exact job a prompt must do, the failure modes it creates, and the levers available to improve it.

## Core Notes

### Message layers serve different jobs

In chat-based systems, not all instructions should live in the same place. System or developer-level messages usually define durable behavior, while user messages provide the current task or request context.

### Permanent behavior should not be repeated everywhere

If safety style, format rules, or product behavior are meant to apply across tasks, they belong in higher-level instructions rather than being restated inconsistently in every user prompt.

### Conversation history is also prompt material

Earlier assistant and user messages shape later outputs. This means message design is partly about controlling what long-lived context stays in the conversation.



## Frameworks And Design Patterns

### Policy at the top, task at the edge

**Use when:** You are building a chat app or assistant workflow.

**How it works:** Keep stable behavioral rules in higher-priority messages and let user prompts focus on the specific task of the moment.

**Watch out for:** If too much task-specific detail is pushed into top-level messages, the system becomes rigid and hard to reuse.



## Prompt Examples

### Layered assistant design

**Scenario:** A developer wants a reusable application prompt structure.

```text
System: You are a reliable documentation assistant. Always distinguish between facts from provided sources and general guidance. Developer: Return concise markdown with clear section titles. User: Summarize the following API spec for a backend engineer and highlight breaking changes.
```

**Why this example matters:** It shows how stable behavior and one-time task instructions can be separated cleanly.



## Common Mistakes And Corrections

### Mistake 1

**Problem:** Putting all instructions into the user prompt.

**Why it hurts:** Reusable behavior becomes inconsistent and harder to maintain.

**Better move:** Move durable rules to higher-level messages and keep user prompts task-focused.

### Mistake 2

**Problem:** Using assistant history without reviewing its influence.

**Why it hurts:** Past outputs can accidentally bias future responses.

**Better move:** Treat conversation history as part of prompt design and prune it when needed.



## Evaluation Checklist

- Which instructions should persist across sessions or features?
- Which instructions belong only to the current task?
- Does the conversation history help the current turn or create drift?
