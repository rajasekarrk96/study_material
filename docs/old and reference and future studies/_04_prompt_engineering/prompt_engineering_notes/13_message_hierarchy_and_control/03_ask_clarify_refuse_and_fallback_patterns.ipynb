# Ask Clarify Refuse And Fallback Patterns

**Module:** Message Hierarchy Instruction Priority And Control

**Level:** Advanced

**Topic Focus:** Controlling what the model should do when the task is ambiguous, unsafe, or underspecified.


## Learning Objectives

- Know when a prompt should encourage clarification.
- Use abstain and fallback behaviors intentionally.
- Avoid forcing the model to guess when uncertainty is high.


## Why This Topic Matters

Developers need explicit patterns for when the model should answer directly, ask a question first, abstain, or provide a constrained fallback.

Prompt engineering improves when you can identify the exact job a prompt must do, the failure modes it creates, and the levers available to improve it.

## Core Notes

### Not every task should be answered immediately

If key requirements are missing, the best behavior may be to ask a concise clarifying question rather than generate a low-quality guess.

### Fallback behavior can preserve usefulness

A system may not be able to complete the user's ideal request, but it can often still provide a safe alternative such as a template, checklist, or explanation of missing inputs.

### Refusal should be specific and helpful

A bare refusal is less useful than one that explains the limitation and, where appropriate, offers a safe next step.



## Frameworks And Design Patterns

### Answer if possible, ask if needed, abstain if unsafe

**Use when:** The task space includes ambiguity, risk, or missing information.

**How it works:** Define conditions for direct answering, clarification, and refusal so the model does not improvise its own decision policy.

**Watch out for:** Too much clarification can make the assistant slow and frustrating for straightforward tasks.



## Prompt Examples

### Clarify-first developer prompt

**Scenario:** A coding assistant gets underspecified tasks.

```text
If the request lacks critical implementation details, ask at most two short clarifying questions before proposing code. If the missing details do not block a safe draft, state assumptions clearly and continue.
```

**Why this example matters:** It gives a practical ambiguity policy suited to developer workflows.



## Common Mistakes And Corrections

### Mistake 1

**Problem:** Always forcing an answer.

**Why it hurts:** The model guesses and creates avoidable errors.

**Better move:** Add explicit clarify or abstain conditions.

### Mistake 2

**Problem:** Refusing without offering any next step.

**Why it hurts:** The assistant becomes safe but not useful.

**Better move:** Provide fallback guidance or input requirements when possible.



## Evaluation Checklist

- What should the assistant do when key information is missing?
- When is a clarifying question better than a draft with assumptions?
- What is the safest useful fallback if the ideal task cannot be completed?
