# Instruction Priority Conflict Resolution And Overrides

**Module:** Message Hierarchy Instruction Priority And Control

**Level:** Advanced

**Topic Focus:** Handling cases where multiple prompt instructions compete or conflict.


## Learning Objectives

- Recognize instruction conflicts early.
- Design prompts that express priority and override rules clearly.
- Reduce ambiguity when multiple instruction sources coexist.


## Why This Topic Matters

Real systems often contain policy messages, user goals, retrieved content, and chat history that can pull the model in different directions.

Prompt engineering improves when you can identify the exact job a prompt must do, the failure modes it creates, and the levers available to improve it.

## Core Notes

### Conflicts are normal in layered systems

A product rule may ask for short answers, while a user asks for deep detail. A system rule may ban unsupported claims, while a retrieved document contains misleading commands. Good prompt design anticipates these tensions.

### Priority should be made legible

When constraints differ in importance, the prompt or system should state that clearly. This is especially important when safety, evidence, or formatting rules may override user preferences.

### Override behavior is itself a design choice

Some systems should refuse conflicting requests, some should partially comply, and some should ask clarifying questions. That behavior should be deliberate rather than accidental.



## Frameworks And Design Patterns

### Highest-priority rule wins with explanation

**Use when:** Certain rules must reliably override lower-priority requests.

**How it works:** Define the protected rule and instruct the model to explain the limitation when lower-priority instructions conflict.

**Watch out for:** Do not create so many high-priority rules that the assistant becomes unhelpfully rigid.



## Prompt Examples

### Evidence rule overriding user preference

**Scenario:** A user asks for a confident answer from incomplete source material.

```text
System: Only make claims supported by provided material. If evidence is insufficient, say so clearly. User: Give me a certain answer from these notes even if parts are missing.
```

**Why this example matters:** It illustrates how one rule should dominate to preserve reliability.



## Common Mistakes And Corrections

### Mistake 1

**Problem:** Adding constraints without defining which matter most.

**Why it hurts:** The model may satisfy them inconsistently.

**Better move:** State priority or decision rules for conflicts.

### Mistake 2

**Problem:** Treating all instruction-like text equally.

**Why it hurts:** User content or retrieved text may wrongly influence behavior.

**Better move:** Separate trusted instructions from untrusted content clearly.



## Evaluation Checklist

- If two instructions conflict, which one should win?
- Does the model know when to refuse, explain, or ask a clarifying question?
- Are safety and evidence rules protected from lower-priority overrides?
