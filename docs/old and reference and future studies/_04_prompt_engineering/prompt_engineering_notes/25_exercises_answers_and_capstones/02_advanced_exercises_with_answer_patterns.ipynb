# Advanced Exercises With Answer Patterns

**Module:** Exercises Answers And Capstone Practice

**Level:** Advanced

**Topic Focus:** More difficult prompt tasks involving evaluation, retrieval, tool use, and workflow design.


## Learning Objectives

- Practice advanced prompt design under realistic constraints.
- Learn what strong answer patterns look like without locking into one exact wording.
- Exercise failure diagnosis and redesign thinking.


## Why This Topic Matters

Advanced learners need exercises that involve realistic ambiguity, system tradeoffs, and structured reasoning rather than only prompt beautification.

Prompt engineering improves when you can identify the exact job a prompt must do, the failure modes it creates, and the levers available to improve it.

## Core Notes

### Advanced exercises should test judgment, not only syntax

A learner should decide when to retrieve, when to ask clarifying questions, how to structure outputs, and how to design evaluation criteria for the prompt they create.

### Answer patterns are better than single perfect answers

For complex prompt tasks, there may be several strong approaches. Teaching answer patterns helps learners reason about tradeoffs instead of memorizing one canonical prompt.



## Frameworks And Design Patterns

### Task prompt plus answer pattern

**Use when:** You want learners to compare their solution with a strong reference shape.

**How it works:** Give the task, then provide a reference outline of what a strong answer should contain.

**Watch out for:** Do not turn answer patterns into rigid formulas that discourage original design choices.



## Prompt Examples

### Advanced RAG exercise

**Scenario:** A learner designs a grounded support prompt.

```text
Design a prompt for a support assistant that must answer only from retrieved docs, expose conflicts, and escalate unresolved issues. Then list the evaluation dimensions you would use to test it.
```

**Why this example matters:** It combines prompting and evaluation design in one realistic task.



## Common Mistakes And Corrections

### Mistake 1

**Problem:** Treating advanced practice as bigger versions of beginner drills.

**Why it hurts:** Learners miss the systems thinking aspect of advanced prompting.

**Better move:** Include ambiguity, failure tradeoffs, and evaluation design in the exercise itself.



## Evaluation Checklist

- Does the exercise require actual tradeoff thinking?
- Are answer patterns specific enough to teach but broad enough to allow variation?
- Does the task connect prompting to evaluation or workflow design?


## Practice Prompts

Use these as writing drills, not as code exercises.

1. Design a tool-calling prompt with missing-parameter handling and a fallback policy.
2. Create an evaluation rubric for a code review assistant and compare two prompt variants against it.


## Summary

Advanced exercises should train prompt judgment under uncertainty, not only the ability to write more elaborate prompt text.

A strong prompt engineer can explain the topic simply, apply it deliberately, and diagnose when it is missing from a workflow.
