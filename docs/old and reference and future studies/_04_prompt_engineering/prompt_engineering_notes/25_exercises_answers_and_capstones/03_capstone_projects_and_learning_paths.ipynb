# Capstone Projects And Learning Paths

**Module:** Exercises Answers And Capstone Practice

**Level:** Advanced to Expert

**Topic Focus:** Larger end-to-end projects that combine multiple prompt engineering concepts into one buildable workflow.


## Learning Objectives

- Apply multiple modules together in one project.
- Practice prompt system design, evals, and iteration end to end.
- Build portfolio-style developer prompt engineering projects.


## Why This Topic Matters

Capstones turn isolated notebook knowledge into integrated practical skill by forcing learners to design complete systems and evaluate them.

Prompt engineering improves when you can identify the exact job a prompt must do, the failure modes it creates, and the levers available to improve it.


## Small Topics Inside This Topic

Study this notebook from the smallest ideas upward:

- Project scoping
- Workflow architecture
- Eval plan
- Failure analysis
- Iteration loop
- Final review


## Core Notes

### Capstones reveal whether concepts transfer

A learner may understand isolated prompt patterns but still struggle to combine retrieval, output contracts, evaluation, safety, and fallback design into one coherent system.

### A good capstone includes both build and critique

Learners should not only design a prompt system but also define how they will test it, what its likely failure modes are, and how they would improve it after initial results.

### Developer capstones should feel real

Projects such as a code review assistant, documentation QA system, incident analysis workflow, or internal support helper are more valuable than abstract prompt-only exercises.



## Frameworks And Design Patterns

### Design -> implement -> evaluate -> iterate

**Use when:** A learner is ready to combine many prompt engineering skills in one project.

**How it works:** Scope a project, design the prompt system, define evals, test it, analyze failures, and revise the workflow.

**Watch out for:** Capstones that are too large may become architecture exercises without enough prompt depth.



## Prompt Examples

### Developer capstone menu

**Scenario:** A learner wants a portfolio project.

```text
Choose one capstone: 1. code review assistant, 2. internal documentation RAG assistant, 3. incident analysis workflow, 4. structured support triage assistant. For the chosen project, define system prompts, task prompts, evaluation plan, failure modes, and iteration strategy.
```

**Why this example matters:** It turns the course into project-oriented practice with explicit system thinking.



## Common Mistakes And Corrections

### Mistake 1

**Problem:** Treating capstones as prompt-writing only.

**Why it hurts:** Learners miss evaluation, ops, and safety thinking.

**Better move:** Require architecture, testing, and failure analysis in the capstone definition.

### Mistake 2

**Problem:** Choosing projects without clear task boundaries.

**Why it hurts:** Work becomes too broad to learn from effectively.

**Better move:** Scope capstones around a focused workflow and evaluation target.



## Evaluation Checklist

- Does the capstone integrate prompting, evaluation, and failure analysis?
- Is the project scoped tightly enough to finish and review?
- Would the final artifact demonstrate real developer prompt engineering skill?


## Practice Prompts

Use these as writing drills, not as code exercises.

1. Pick one capstone and write its system design plus eval plan.
2. List the top five likely failure modes for your chosen capstone and how you would test them.


## Summary

Capstones are where prompt engineering stops being a reading subject and becomes a build-and-review discipline.

A strong prompt engineer can explain the topic simply, apply it deliberately, and diagnose when it is missing from a workflow.
