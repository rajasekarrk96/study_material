# Context Selection And Relevance

**Module:** Prompt Anatomy And Core Building Blocks

**Level:** Intermediate

**Topic Focus:** Choosing the right supporting information and excluding distractions.


## Learning Objectives

- Select context by task relevance.
- Label context so the model can use it effectively.
- Reduce irrelevant background and noisy attachments.


## Why This Topic Matters

Too little context leads to guesswork, while too much irrelevant context degrades signal and wastes space.

Prompt engineering improves when you can identify the exact job a prompt must do, the failure modes it creates, and the levers available to improve it.

## Core Notes

### Context should serve the task, not the user's desire to include everything

Users often paste all available material into a prompt. Expert prompt design asks which pieces actually change the answer quality and removes the rest.

### Labeling matters

When multiple documents or context fragments are present, name them clearly. Source A, policy excerpt, customer email, and metrics table are easier to reason over than an unlabeled text wall.

### Compression is a core prompt skill

You often need to summarize or distill background before the real task prompt. Context engineering is an increasingly important part of prompt engineering.



## Frameworks And Design Patterns

### Need-to-know filtering

**Use when:** You have too much source material.

**How it works:** Include only information that changes the answer, evaluation, or decision-making path.

**Watch out for:** Removing exceptions or caveats can create misleading simplifications.



## Prompt Examples

### Relevant-context extraction step

**Scenario:** Long customer transcript before action extraction.

```text
From the transcript below, extract only the facts needed to determine refund eligibility, contractual exceptions, and promised follow-up actions. Ignore rapport, repetition, and small talk.
```

**Why this example matters:** It shows context narrowing before the main task.



## Common Mistakes And Corrections

### Mistake 1

**Problem:** Passing raw context without telling the model what parts matter.

**Why it hurts:** Important details compete with irrelevant ones.

**Better move:** State the context priorities explicitly.

### Mistake 2

**Problem:** Over-compressing critical nuance.

**Why it hurts:** The model loses conditions, exceptions, or edge cases.

**Better move:** Compress aggressively only after identifying must-keep details.



## Evaluation Checklist

- Does each context block have a clear purpose?
- Are important sources labeled and distinguishable?
- What can be removed with minimal effect on answer quality?


## Practice Prompts

Use these as writing drills, not as code exercises.

1. Take a noisy source document and write a pre-filtering prompt to isolate only task-relevant facts.
2. Design a prompt that compares two labeled documents without confusing their claims.


## Summary

Context selection is a high-leverage skill: better context often beats more context.

A strong prompt engineer can explain the topic simply, apply it deliberately, and diagnose when it is missing from a workflow.
