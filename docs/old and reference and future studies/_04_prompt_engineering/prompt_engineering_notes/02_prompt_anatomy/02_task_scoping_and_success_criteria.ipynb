# Task Scoping And Success Criteria

**Module:** Prompt Anatomy And Core Building Blocks

**Level:** Intermediate

**Topic Focus:** Turning broad asks into bounded, testable prompt tasks.


## Learning Objectives

- Bound a task so it becomes testable.
- Define success criteria inside the prompt or surrounding workflow.
- Separate must-have outputs from nice-to-have extras.


## Why This Topic Matters

A model cannot reliably satisfy a vague objective. Expert prompts define scope, deliverables, and what success means.

Prompt engineering improves when you can identify the exact job a prompt must do, the failure modes it creates, and the levers available to improve it.

## Core Notes

### Broad tasks need operational definitions

Requests such as 'analyze this' or 'improve this' are too open unless you define the analysis lens, target outcome, and completion condition.

### Success criteria improve both generation and evaluation

If the model knows what good looks like, outputs become more aligned. If the evaluator knows what good looks like, revisions become more objective.

### Non-goals prevent wasted effort

A strong prompt may say not to comment on branding, not to invent data, or not to discuss history. Non-goals reduce unhelpful expansion.



## Frameworks And Design Patterns

### Goal -> deliverable -> criteria

**Use when:** A prompt feels too broad or subjective.

**How it works:** State the overall goal, name the concrete output, then specify what makes it successful.

**Watch out for:** If criteria conflict, the model may satisfy them unevenly.



## Prompt Examples

### Scoped document review

**Scenario:** You want a focused review, not generic feedback.

```text
Review this product requirements document only for requirement ambiguity, dependency gaps, and missing acceptance criteria. Do not comment on writing style unless it causes ambiguity. Return findings in priority order.
```

**Why this example matters:** It narrows the model's job and makes output easier to evaluate.



## Common Mistakes And Corrections

### Mistake 1

**Problem:** Combining too many goals into one prompt.

**Why it hurts:** Coverage becomes inconsistent and shallow.

**Better move:** Split the task or prioritize the goals explicitly.

### Mistake 2

**Problem:** Leaving success undefined.

**Why it hurts:** You cannot tell whether the output is truly correct or just plausible.

**Better move:** Name the criteria that matter most.



## Evaluation Checklist

- Can the task be completed in a clearly observable way?
- Are must-have requirements distinguished from optional depth?
- Would a reviewer know how to judge the answer?


## Practice Prompts

Use these as writing drills, not as code exercises.

1. Rewrite 'analyze this startup idea' into a bounded evaluation prompt.
2. Take a large task and split it into two smaller prompts with distinct success criteria.


## Summary

Task scoping turns a prompt from a broad wish into an operational request with visible quality standards.

A strong prompt engineer can explain the topic simply, apply it deliberately, and diagnose when it is missing from a workflow.
