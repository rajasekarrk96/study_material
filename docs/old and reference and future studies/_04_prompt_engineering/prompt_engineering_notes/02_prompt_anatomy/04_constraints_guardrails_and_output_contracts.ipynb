## Why This Topic Matters

Expert prompts often behave well because they specify what must happen, what must not happen, and how outputs should be shaped.

Prompt engineering improves when you can identify the exact job a prompt must do, the failure modes it creates, and the levers available to improve it.

## Core Notes

### Constraints reduce answer space

A model has many possible valid continuations. Constraints reduce that space by setting boundaries around content, form, style, evidence, and allowed assumptions.

### Output contracts stabilize downstream use

If another person or system depends on the output, define the schema or response shape clearly. This is especially important for tables, bullet groups, JSON-like structures, or decision labels.

### Over-constraining can backfire

Too many rules may create contradictions or force unnatural behavior. A good prompt balances clarity and room for the model to perform the task well.



## Frameworks And Design Patterns

### Must / may / must not

**Use when:** You need clear rule hierarchy.

**How it works:** Divide prompt rules into required, optional, and forbidden behaviors.

**Watch out for:** If every rule is critical, priority becomes unclear.



## Prompt Examples

### Decision output contract

**Scenario:** A classification output must be consistent.

```text
Classify each support ticket as Billing, Technical, Account Access, or Other. Return a table with columns: ticket_id, label, confidence, and short evidence phrase. Do not invent labels outside the four allowed classes.
```

**Why this example matters:** It defines the label set and the response schema.



## Common Mistakes And Corrections

### Mistake 1

**Problem:** Using vague constraints like 'be detailed' or 'be smart'.

**Why it hurts:** The model interprets them inconsistently.

**Better move:** Translate vague preferences into observable behaviors.

### Mistake 2

**Problem:** Asking for contradictory outputs.

**Why it hurts:** The model cannot reliably satisfy all conditions.

**Better move:** Rank priorities and remove impossible combinations.



## Evaluation Checklist

- Are the constraints observable in the final output?
- Is the output format specific enough for reuse?
- Could any constraints be merged or prioritized more clearly?
