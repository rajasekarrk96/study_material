## Core Notes

### A template should capture the stable logic of a prompt

If the task repeats across many inputs, convert the common structure into a template with named variables. This keeps the reasoning pattern stable while allowing input variation.

### Template design is more than string substitution

A good template also includes when to use it, what assumptions it makes, expected input shape, output schema, and known failure modes.

### Prompt libraries need curation

A large folder of half-duplicate prompts is not a library. It is noise. Libraries improve when prompts have distinct purposes, documentation, and version history.



## Frameworks And Design Patterns

### Template + notes + examples

**Use when:** A prompt will be reused by you or a team.

**How it works:** Store the prompt pattern, variable descriptions, example inputs, example outputs, and known caveats together.

**Watch out for:** Over-general templates can become vague and weak.



## Prompt Examples

### Template-oriented critique prompt

**Scenario:** Reusable document review across teams.

```text
Template fields: audience_type, review_focus, source_document, output_format, severity_scale. Reusable base instruction: review the document only for the specified focus and return findings using the requested severity scale.
```

**Why this example matters:** It shows how a prompt pattern can be parameterized without losing task intent.



## Common Mistakes And Corrections

### Mistake 1

**Problem:** Treating every prompt as a unique one-off.

**Why it hurts:** You repeat work and lose shared learning.

**Better move:** Promote stable prompt patterns into templates.

### Mistake 2

**Problem:** Creating templates with undefined variables.

**Why it hurts:** Users fill them inconsistently and outputs drift.

**Better move:** Document variable meaning and expected formats.



## Evaluation Checklist

- Can another person use the template correctly without extra explanation?
- Are the variable inputs clearly defined?
- Is the template narrow enough to stay strong but broad enough to reuse?
