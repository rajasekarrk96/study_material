# Prompt Review Checklist

**Module:** Reference Patterns Checklists And Playbooks

**Level:** All Levels

**Topic Focus:** A reusable checklist for inspecting prompt quality before use or revision.


## Learning Objectives

- Create a final prompt review habit.
- Check prompts for clarity, context, structure, and safety.
- Use the checklist during drafting and debugging.


## Why This Topic Matters

A compact checklist helps translate the whole course into a repeatable review habit.

Prompt engineering improves when you can identify the exact job a prompt must do, the failure modes it creates, and the levers available to improve it.

## Core Notes

### A checklist turns principles into practice

Before using or shipping a prompt, run through a short set of questions. This catches basic design mistakes before they become bigger workflow problems.

### Checklists are especially valuable under speed pressure

When teams move fast, common prompt problems reappear: vague tasks, unclear schemas, missing evidence rules, and weak failure handling. A review checklist reduces this slippage.



## Frameworks And Design Patterns

### Clarity -> context -> constraints -> output -> risk

**Use when:** You need a fast but meaningful final review.

**How it works:** Inspect the prompt in five passes, each focused on one quality dimension.

**Watch out for:** A checklist helps review but does not replace testing on real inputs.



## Prompt Examples

### Final review questions

**Scenario:** You are reviewing a prompt before reuse.

```text
Is the task unambiguous? Is the needed context present? Are constraints observable and non-conflicting? Is the output format explicit? Does the prompt encourage groundedness and safe uncertainty?
```

**Why this example matters:** It distills the course into practical review questions.



## Common Mistakes And Corrections

### Mistake 1

**Problem:** Reviewing only wording and ignoring system fit.

**Why it hurts:** The prompt looks clean but still fails in real use.

**Better move:** Check both prompt content and workflow context.



## Evaluation Checklist

- Could a teammate understand and use the prompt correctly?
- What is the likeliest failure mode if this prompt is used at scale?
- Does the prompt need retrieval, tools, or human review to be safe enough?
