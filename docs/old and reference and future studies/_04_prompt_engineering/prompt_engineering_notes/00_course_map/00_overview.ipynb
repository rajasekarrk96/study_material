# Prompt Engineering Course Map

This notebook lists the full curriculum generated by `generate.py`.


## Study Philosophy

- Learn prompt engineering as a system, not as a bag of tricks.
- Move from clarity and structure to evaluation, safety, retrieval, tool use, and prompt operations.
- Reuse the prompt patterns, but always adapt them to task, model, and constraints.
