# Tool Hallucination Misuse And Partial Failure Handling

**Module:** Agent Reliability And Tool Failure Patterns

**Level:** Expert

**Topic Focus:** Preventing agents from inventing tool outcomes, misusing tools, or mishandling incomplete tool results.


## Learning Objectives

- Recognize tool-specific failure modes in agent workflows.
- Prevent hallucinated tool results and unsafe assumptions.
- Design fallback behavior for partial or failed tool actions.


## Why This Topic Matters

Tool-using systems can fail even when the language model is strong, because the agent may call the wrong tool, interpret results badly, or pretend a failed action succeeded.

Prompt engineering improves when you can identify the exact job a prompt must do, the failure modes it creates, and the levers available to improve it.

## Core Notes

### Tool misuse is not the same as bad prose

An agent can produce fluent text while acting incorrectly under the hood. It may select the wrong tool, omit required arguments, or claim an action completed when it did not.

### Partial failure should be first-class behavior

Real tools timeout, return incomplete data, or succeed only partially. Prompts and workflows should define how the agent reports and recovers from these states.

### Confirmation is often worth the extra step

For sensitive actions or fragile workflows, the system may need to restate the planned action and wait for confirmation before executing.



## Frameworks And Design Patterns

### Call, verify, interpret, fallback

**Use when:** Tool actions can fail or return ambiguous results.

**How it works:** Execute the tool, verify what actually happened, interpret the result, and use fallback or escalation logic if the action failed or returned partial data.

**Watch out for:** Do not collapse verification and interpretation into one unchecked step.



## Prompt Examples

### Partial-failure policy

**Scenario:** A tool returns some but not all requested records.

```text
If a tool returns partial data, say what was retrieved successfully, what remains missing, and whether a retry or user clarification is needed. Do not present partial results as a complete answer.
```

**Why this example matters:** It turns a common tool failure into explicit user-visible behavior.



## Common Mistakes And Corrections

### Mistake 1

**Problem:** Assuming tool success unless an error is explicit.

**Why it hurts:** The system may report false completion.

**Better move:** Verify tool outputs and define success criteria clearly.

### Mistake 2

**Problem:** Retrying blindly without a failure policy.

**Why it hurts:** Agents waste steps or loop on the same broken action.

**Better move:** Add retry limits and fallback logic.



## Evaluation Checklist

- Can the system distinguish full success from partial success?
- What happens when a tool returns ambiguous or incomplete results?
- Does the agent ever imply a tool outcome it did not verify?


## Practice Prompts

Use these as writing drills, not as code exercises.

## Summary

Reliable tool use depends on verification, partial-failure handling, and disciplined fallback behavior rather than on trusting tool calls implicitly.

A strong prompt engineer can explain the topic simply, apply it deliberately, and diagnose when it is missing from a workflow.
