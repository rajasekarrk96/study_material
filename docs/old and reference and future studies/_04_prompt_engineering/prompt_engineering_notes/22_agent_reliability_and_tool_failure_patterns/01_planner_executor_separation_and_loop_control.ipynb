# Planner Executor Separation And Loop Control

**Module:** Agent Reliability And Tool Failure Patterns

**Level:** Expert

**Topic Focus:** Structuring agent workflows so planning, acting, and stopping are controlled explicitly.


## Learning Objectives

- Separate planning from acting where reliability benefits from it.
- Define loop limits and stopping rules clearly.
- Reduce drift in multi-step autonomous workflows.


## Why This Topic Matters

Agent-like systems often fail because planning and execution are blended together without enough constraints, causing drift, repeated work, or endless loops.

Prompt engineering improves when you can identify the exact job a prompt must do, the failure modes it creates, and the levers available to improve it.

## Core Notes

### Planner and executor roles solve different problems

A planner proposes what should happen. An executor performs bounded steps. Separating these roles can reduce chaotic agent behavior, especially on complex tasks.

### Loops need explicit termination logic

Without max-step rules, completion criteria, or escalation conditions, agent workflows may continue searching, revising, or calling tools long after value has flattened.

### Autonomy should be bounded by purpose

Agents do not become better by being allowed to do everything. They become more reliable when autonomy is aligned to a clearly defined task boundary.



## Frameworks And Design Patterns

### Plan once, act in bounded steps, stop intentionally

**Use when:** A workflow needs more than one action but should remain controllable.

**How it works:** Have the system plan, execute only the next bounded step, check progress, and stop when criteria are met or limits are reached.

**Watch out for:** Too much separation can create overhead if the task is simple.



## Prompt Examples

### Bounded agent loop policy

**Scenario:** An assistant can search and summarize repeatedly.

```text
Before acting, produce a short plan. Then execute at most one step at a time. After each step, check whether the goal is already satisfied. Stop after three steps unless new evidence clearly justifies continuing.
```

**Why this example matters:** It shows how loop control turns vague autonomy into a governed workflow.



## Common Mistakes And Corrections

### Mistake 1

**Problem:** Letting the same prompt both plan and act indefinitely.

**Why it hurts:** The workflow may drift, repeat, or over-expand.

**Better move:** Separate planning and execution or at least add explicit loop constraints.

### Mistake 2

**Problem:** Failing to define completion criteria.

**Why it hurts:** The agent cannot tell when the job is done.

**Better move:** State what counts as success and when to stop.



## Evaluation Checklist

- Does the workflow need a separate planning step?
- What stops the system from looping unnecessarily?
- Can the agent explain why it stopped or escalated?
