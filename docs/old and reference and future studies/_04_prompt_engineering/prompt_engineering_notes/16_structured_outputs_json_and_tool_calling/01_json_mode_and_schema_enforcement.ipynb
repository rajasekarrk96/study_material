## Why This Topic Matters

Developers often need outputs that are not just readable but reliably parsable and schema-compliant.

Prompt engineering improves when you can identify the exact job a prompt must do, the failure modes it creates, and the levers available to improve it.

## Core Notes

### Machine readability changes what counts as success

A helpful prose answer may still be a workflow failure if the downstream parser expects strict structure. Developers need prompts that prioritize schema compliance as strongly as content quality.

### Schema rules should be explicit and testable

Define field names, allowed values, null behavior, nesting expectations, and any forbidden extras. Small ambiguities create large integration problems.

### Prompting alone may not guarantee schema validity

Where possible, combine careful prompting with stronger structured output mechanisms and validation logic.



## Frameworks And Design Patterns

### Field contract plus failure behavior

**Use when:** The output must be consumed programmatically.

**How it works:** Specify the schema and also define what to do when the source lacks a value or a classification is uncertain.

**Watch out for:** A schema that ignores uncertainty can force bad invented values.



## Prompt Examples

### Strict extraction schema

**Scenario:** A service expects predictable fields from text.

```text
Return only a JSON object with keys customer_name, issue_type, urgency, evidence, and missing_information. If a field cannot be determined, use null and explain the gap inside missing_information. Do not add extra keys.
```

**Why this example matters:** It combines schema constraints with uncertainty handling.



## Common Mistakes And Corrections

### Mistake 1

**Problem:** Describing structure loosely.

**Why it hurts:** Parsers fail or require brittle cleanup.

**Better move:** Define exact fields and output rules.

### Mistake 2

**Problem:** Ignoring null and empty-state behavior.

**Why it hurts:** The model may invent values or vary inconsistently.

**Better move:** State missing-data handling explicitly.



## Evaluation Checklist

- Can the output be parsed reliably without post-hoc cleanup?
- Are field semantics and allowed values clear?
- What should happen when information is missing or uncertain?
