Prompt engineering improves when you can identify the exact job a prompt must do, the failure modes it creates, and the levers available to improve it.

## Core Notes

### Tool calling is decision-making, not just formatting

A model must decide whether a tool is needed, which tool fits, whether the available information is sufficient, and how to map user intent into arguments.

### Missing arguments need explicit behavior

If required parameters are absent, the model may need to ask a question, make a safe assumption, or refuse the tool call. That policy should be designed intentionally.

### Tool outputs need second-stage prompting

After a tool call succeeds, the model still needs guidance on how to summarize or act on the result appropriately.



## Frameworks And Design Patterns

### Need tool? validate inputs -> call -> synthesize

**Use when:** An assistant can either answer directly or act through one or more tools.

**How it works:** Prompt the model to check whether a tool is needed, verify required arguments, then interpret results responsibly.

**Watch out for:** If missing-input behavior is unclear, tool calls become brittle or unsafe.



## Prompt Examples

### Argument-aware tool prompt

**Scenario:** A scheduling assistant may need calendar actions.

```text
Use a calendar tool only if the user requests scheduling, rescheduling, or availability lookup. If date, timezone, or participant details are missing and are required for a correct action, ask one concise follow-up question before calling the tool.
```

**Why this example matters:** It gives a concrete decision and missing-input policy.



## Common Mistakes And Corrections

### Mistake 1

**Problem:** Letting the model call tools without input sufficiency checks.

**Why it hurts:** Actions fail or act on wrong assumptions.

**Better move:** Define when to ask for missing parameters.

### Mistake 2

**Problem:** Treating tool output as the final user answer.

**Why it hurts:** The assistant may return raw data instead of useful interpretation.

**Better move:** Design post-tool synthesis prompts as part of the workflow.



## Evaluation Checklist

- When should the model answer directly instead of using a tool?
- What arguments are required for safe and correct tool use?
- How should tool outputs be transformed for the user?
