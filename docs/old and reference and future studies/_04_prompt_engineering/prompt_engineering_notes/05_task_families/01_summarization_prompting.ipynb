# Summarization Prompting

**Module:** Task Specific Prompting

**Level:** Intermediate

**Topic Focus:** Controlling summary goals, compression ratio, and audience fit.


## Learning Objectives

- Differentiate summary types such as executive, technical, and action-oriented summaries.
- Control what gets compressed and what must remain visible.
- Avoid summaries that are short but unhelpful.


## Why This Topic Matters

Summarization seems simple, but good summaries depend on purpose, audience, and what should be preserved or excluded.

Prompt engineering improves when you can identify the exact job a prompt must do, the failure modes it creates, and the levers available to improve it.

## Core Notes

### A summary is always for someone and for something

If you do not define the audience and purpose, the model usually produces a generic overview. Summaries differ when they support decisions, teaching, memory refresh, or risk review.

### Compression should follow priorities

The prompt should say what information must survive compression. For one task that may be deadlines and blockers; for another it may be claims and evidence; for another it may be themes and contradictions.

### Good summaries often include structure

A powerful summary format may contain a one-line overview, key points, open questions, and action items. Structure often matters more than raw brevity.



## Frameworks And Design Patterns

### Audience + purpose + preservation rules

**Use when:** You want summaries that are more useful than generic compression.

**How it works:** Define who the summary is for, what they need from it, and which information categories must stay visible.

**Watch out for:** If preservation rules are too broad, the summary becomes bloated.



## Prompt Examples

### Executive action summary

**Scenario:** Leadership needs a short decision-oriented version of a long report.

```text
Summarize this report for an executive who has two minutes to read it. Return: 1 sentence overview, 5 critical facts, 3 risks, and 2 decisions that require action this week. Ignore historical detail unless it changes a decision.
```

**Why this example matters:** It aligns summary content with leadership use.



## Common Mistakes And Corrections

### Mistake 1

**Problem:** Asking for a summary without specifying purpose.

**Why it hurts:** The output may be accurate but not useful.

**Better move:** Define the reader and the decision or learning goal.

### Mistake 2

**Problem:** Compressing away uncertainty or caveats.

**Why it hurts:** The summary becomes overconfident and misleading.

**Better move:** Tell the model which caveats must be preserved.



## Evaluation Checklist

- Does the summary serve a real reader and use case?
- Did critical facts survive compression?
- Were non-essential details removed without losing needed nuance?

## Summary

Strong summarization prompts specify the reader, the purpose, and the information that must survive compression.

A strong prompt engineer can explain the topic simply, apply it deliberately, and diagnose when it is missing from a workflow.
