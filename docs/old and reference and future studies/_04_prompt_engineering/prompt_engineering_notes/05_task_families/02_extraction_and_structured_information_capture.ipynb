## Core Notes

### Extraction tasks need schema clarity

The model must know what fields exist, how they should be filled, and what to do if the source does not contain the answer. Otherwise extraction drifts into guessing or inconsistent formatting.

### Grounding improves trust

When possible, ask the model to include the supporting phrase or source location for each extracted item. This makes review easier and discourages unsupported guesses.

### Missing data behavior is part of the task

A prompt should define whether missing data should be returned as blank, null, unknown, not mentioned, or not supported by source. This small detail matters greatly in real workflows.



## Frameworks And Design Patterns

### Field + evidence + missing rule

**Use when:** You need extraction that is auditable and reusable.

**How it works:** For each target field, define the value type, evidence expectation, and missing-data behavior.

**Watch out for:** If fields are underspecified, the model may infer formats inconsistently.



## Prompt Examples

### Invoice extraction prompt

**Scenario:** You want stable data from semi-structured text.

```text
Extract the following fields from the invoice text: vendor_name, invoice_number, invoice_date, due_date, currency, total_amount. For each field, also provide the exact evidence phrase from the text. If a field is missing, return 'not found'.
```

**Why this example matters:** It combines schema, evidence, and missing-data rules.



## Common Mistakes And Corrections

### Mistake 1

**Problem:** Not specifying what to do with absent fields.

**Why it hurts:** The model may invent or infer values.

**Better move:** Set an explicit missing-data policy.

### Mistake 2

**Problem:** Using field names without definitions.

**Why it hurts:** Ambiguous interpretation causes drift.

**Better move:** Define tricky fields and acceptable value forms.



## Evaluation Checklist

- Is every field clearly defined?
- Does the output include evidence or grounding where needed?
- Is the missing-data behavior explicit and stable?
