## Core Notes

### A useful code review assistant needs more than a review prompt

It needs clear review criteria, stable output ordering, attention to regressions and missing tests, and a way to compare performance across different pull requests.

### Findings-first structure improves usefulness

Developers usually want review issues in severity order, tied to evidence and impact, before style commentary or optional suggestions.

### Evaluation should use real review cases

A code review assistant should be tested on examples with known defects, subtle regressions, and benign changes to measure both recall and false positives.



## Frameworks And Design Patterns

### Context -> rubric -> findings -> evidence

**Use when:** You are building or refining a review assistant workflow.

**How it works:** Provide diff context, define the review lens, require severity-ranked findings, and ask for evidence from the change.

**Watch out for:** Overly broad review scopes lead to noisy or shallow comments.



## Prompt Examples

### Review workflow sketch

**Scenario:** A team wants consistent LLM-assisted PR review.

```text
Review the diff for correctness, regressions, edge cases, API contract changes, and missing tests. Return only actionable findings in severity order. For each finding, cite the relevant changed code and explain the likely impact.
```

**Why this example matters:** It demonstrates a practical, high-signal structure for an end-to-end review assistant.



## Common Mistakes And Corrections

### Mistake 1

**Problem:** Treating code review as generic feedback generation.

**Why it hurts:** Outputs become verbose but low value.

**Better move:** Anchor the workflow in a review rubric and actionable findings format.

### Mistake 2

**Problem:** Ignoring benign-change examples in evaluation.

**Why it hurts:** False positives remain hidden.

**Better move:** Include both defect-rich and defect-light test cases.



## Evaluation Checklist

- Does the workflow prioritize real engineering risk over style noise?
- Can findings be traced to changed code and concrete impact?
- How often does the assistant miss important regressions or over-report harmless ones?
