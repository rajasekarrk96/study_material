## Core Notes

### Support assistants fail when they smooth over uncertainty

In support scenarios, it is often better to say what the docs support, where the docs are unclear, and when a human should step in than to guess a confident but wrong answer.

### Grounded answer shape matters

Support replies may need a direct answer, the supporting documentation reference, known limitations, and escalation guidance if the issue is unresolved.

### RAG support assistants need difficult evaluation cases

Use examples with outdated docs, conflicting guidance, missing exceptions, and ambiguous user phrasing to test the real limits of the workflow.



## Frameworks And Design Patterns

### Retrieve -> ground -> answer -> escalate if needed

**Use when:** A support feature answers from internal knowledge or product documentation.

**How it works:** Retrieve relevant content, generate a source-bounded answer, expose uncertainty, and escalate when documentation is incomplete or conflicting.

**Watch out for:** If the support prompt hides uncertainty, users may trust unsupported advice too easily.



## Prompt Examples

### Support answer contract

**Scenario:** An internal support assistant answers product questions.

```text
Answer only from retrieved docs. Return: direct answer, cited evidence, any conflicting guidance found, and whether the issue should be escalated to a human support engineer.
```

**Why this example matters:** It packages groundedness and escalation into one usable answer structure.



## Common Mistakes And Corrections

### Mistake 1

**Problem:** Optimizing for friendly tone while neglecting evidence quality.

**Why it hurts:** The assistant sounds helpful but gives unsupported guidance.

**Better move:** Treat source-boundedness and escalation rules as first-class requirements.

### Mistake 2

**Problem:** Evaluating only easy known-answer support questions.

**Why it hurts:** The workflow looks stronger than it is.

**Better move:** Include ambiguity, contradiction, and outdated-doc cases.



## Evaluation Checklist

- Does the support answer stay grounded to documentation?
- Can the workflow expose missing or conflicting guidance safely?
- When does it escalate instead of over-claiming?

## Summary

A strong RAG support assistant balances helpfulness with explicit grounding, uncertainty handling, and escalation discipline.

A strong prompt engineer can explain the topic simply, apply it deliberately, and diagnose when it is missing from a workflow.
