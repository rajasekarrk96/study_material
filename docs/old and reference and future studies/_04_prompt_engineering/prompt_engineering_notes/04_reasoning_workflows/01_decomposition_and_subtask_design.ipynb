# Decomposition And Subtask Design

**Module:** Reasoning And Multi Step Workflows

**Level:** Intermediate to Advanced

**Topic Focus:** Breaking difficult prompts into smaller, safer operations.


## Learning Objectives

- Learn when to split a task into stages.
- Design subtasks that reduce ambiguity and cognitive load.
- Prevent quality loss from overloaded prompts.


## Why This Topic Matters

Complex tasks often fail because too many cognitive demands are packed into one prompt. Decomposition improves coverage and control.

Prompt engineering improves when you can identify the exact job a prompt must do, the failure modes it creates, and the levers available to improve it.

## Core Notes

### One big prompt is not always one good prompt

When a task requires extraction, interpretation, decision-making, and formatting all at once, the model may skip steps or perform them unevenly. Decomposition makes each step more legible and testable.

### Intermediate outputs are useful design tools

A workflow may first extract facts, then classify them, then produce recommendations. Each intermediate artifact can be inspected or constrained before moving forward.

### Subtasks should have clean ownership

Each step should have one clear job. If a subtask prompt still feels overloaded, decompose further.



## Frameworks And Design Patterns

### Extract -> analyze -> decide -> format

**Use when:** A task mixes raw information handling with judgment and presentation.

**How it works:** Separate information gathering from interpretation and decision from final communication.

**Watch out for:** Too many steps can increase latency and operational complexity.



## Prompt Examples

### Multi-step contract review

**Scenario:** Complex legal-like review without mixing tasks.

```text
Step 1: Extract all obligations, deadlines, penalties, and exception clauses from the contract. Step 2: Based only on the extracted items, identify the top business risks. Step 3: Summarize the risks for a non-lawyer executive.
```

**Why this example matters:** It separates evidence collection from risk communication.



## Common Mistakes And Corrections

### Mistake 1

**Problem:** Splitting tasks without defining handoff outputs.

**Why it hurts:** Later steps inherit messy or inconsistent inputs.

**Better move:** Specify the exact intermediate format.

### Mistake 2

**Problem:** Keeping everything in one prompt for convenience.

**Why it hurts:** The model drops nuance or mixes responsibilities.

**Better move:** Split where the cognitive demands differ significantly.



## Evaluation Checklist

- Which parts of the task require different kinds of reasoning?
- Can intermediate results be checked before final output?
- Would splitting improve explainability or reliability?
