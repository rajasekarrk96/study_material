# Self Critique Reflection And Revision Loops

**Module:** Reasoning And Multi Step Workflows

**Level:** Advanced

**Topic Focus:** Using prompts that ask the model to inspect, critique, or improve its own work.


## Learning Objectives

- Use critique loops deliberately rather than superstitiously.

## Why This Topic Matters

A second pass can reveal omissions, shallow reasoning, or formatting mistakes that the first draft missed.

Prompt engineering improves when you can identify the exact job a prompt must do, the failure modes it creates, and the levers available to improve it.

## Core Notes

### Critique works best when it has a job

A generic request like 'check your answer' is weak. A focused critique such as 'find unsupported claims, missing assumptions, and formatting violations' gives the model a concrete review lens.

### Self-critique is not independent verification

The same model that made the original mistake may miss it again. Critique loops can improve quality, but they should not be mistaken for proof of correctness.

### Revision prompts should preserve what already works

A critique step should identify what to change and what to keep. Otherwise a second pass may fix one problem while damaging clarity or structure elsewhere.



## Frameworks And Design Patterns

### Draft -> critique -> targeted revision

**Use when:** The first answer is useful but incomplete, inconsistent, or risky.

**How it works:** Generate a first draft, review it against explicit criteria, then revise only the flagged weaknesses.

**Watch out for:** Repeated self-revision can drift or overfit to the critique wording.



## Prompt Examples

### Targeted critique prompt

**Scenario:** You want a safer second pass on an analysis memo.

```text
Review the draft above for three things only: unsupported claims, missing counterarguments, and places where the recommendation jumps ahead of the evidence. List the issues first, then provide a revised version.
```

**Why this example matters:** The critique is narrow and actionable.



## Common Mistakes And Corrections

### Mistake 1

**Problem:** Using vague critique instructions.

**Why it hurts:** The second pass repeats surface-level praise or generic changes.

**Better move:** Name the failure modes you want checked.

### Mistake 2

**Problem:** Running endless revision loops.

**Why it hurts:** Outputs become bloated or unstable.

**Better move:** Limit revision rounds and judge against explicit criteria.



## Evaluation Checklist

- Which quality dimension should the critique focus on?
- Would another model, tool, or human be better for verification?
- Did the revision preserve the strengths of the original draft?

## Summary

Self-critique is useful when it is targeted, bounded, and treated as quality improvement rather than ground truth verification.

A strong prompt engineer can explain the topic simply, apply it deliberately, and diagnose when it is missing from a workflow.
