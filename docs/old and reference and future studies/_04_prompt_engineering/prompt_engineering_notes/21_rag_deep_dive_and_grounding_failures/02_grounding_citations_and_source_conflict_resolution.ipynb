Prompt engineering improves when you can identify the exact job a prompt must do, the failure modes it creates, and the levers available to improve it.

## Core Notes

### Grounding requires visible evidence use

If a prompt asks for an answer from retrieved passages, it should also define how evidence appears in the output. Citations, quoted phrases, or passage IDs help reviewers verify the reasoning path.

### Conflicting sources are a feature, not a bug to hide

A mature retrieval workflow should not flatten contradictions into a false consensus. Prompts should allow the model to surface disagreement, compare source strength, and identify missing evidence.

### The model should distinguish support from speculation

Even in grounded systems, the model may infer beyond the retrieved text. Good prompts separate direct support, interpretation, and unresolved uncertainty explicitly.



## Frameworks And Design Patterns

### Answer, cite, compare, qualify

**Use when:** Retrieved evidence is mixed, partial, or from multiple sources.

**How it works:** Ask for the answer, the supporting evidence, any source conflicts, and a qualified conclusion that matches the evidence strength.

**Watch out for:** If the citation format is vague, outputs may look grounded without being verifiable.



## Prompt Examples

### Conflict-aware QA prompt

**Scenario:** Two retrieved documents disagree on a policy exception.

```text
Answer using only the retrieved passages. First cite the passages that support each possible interpretation. Then explain whether the sources agree, conflict, or leave the issue unresolved. If unresolved, state the uncertainty directly instead of choosing a side.
```

**Why this example matters:** It demonstrates how to turn conflicting retrieval into explicit reasoning rather than hidden guessing.



## Common Mistakes And Corrections

### Mistake 1

**Problem:** Treating all retrieved passages as equally authoritative.

**Why it hurts:** Weak or outdated sources can distort the answer.

**Better move:** Prompt for comparison and qualification when evidence quality varies.

### Mistake 2

**Problem:** Using citations as decoration instead of evidence anchors.

**Why it hurts:** The answer appears grounded while still drifting semantically.

**Better move:** Require claim-evidence linkage, not just source mentions.



## Evaluation Checklist

- Can each major claim be traced to a source passage?
- Does the prompt surface contradictions instead of hiding them?
- Is uncertainty expressed in proportion to the evidence quality?
