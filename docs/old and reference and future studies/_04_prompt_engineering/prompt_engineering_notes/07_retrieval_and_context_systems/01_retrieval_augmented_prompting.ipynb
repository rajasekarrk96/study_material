## Why This Topic Matters

When tasks depend on specific source material, retrieval-augmented prompting is often the practical path to accuracy and relevance.

Prompt engineering improves when you can identify the exact job a prompt must do, the failure modes it creates, and the levers available to improve it.

## Core Notes

### Retrieval changes the prompt's job

Without retrieval, the prompt mainly guides generation. With retrieval, the prompt must also tell the model how to use evidence, what to do when passages conflict, and how to admit insufficiency.

### Grounding is stronger when evidence use is explicit

Ask for cited support, quoted fragments, or section references where possible. This encourages the model to bind claims to source material.

### Missing evidence is itself a valid result

A good retrieval prompt allows the model to say the retrieved passages do not answer the question fully. This is often better than a speculative fill-in.



## Frameworks And Design Patterns

### Answer only from provided sources

**Use when:** You want grounded responses and controlled hallucination risk.

**How it works:** Constrain the model to retrieved material, require citations or quotes, and define what to do when evidence is missing.

**Watch out for:** The model can still misread sources, so evaluation remains necessary.



## Prompt Examples

### Grounded QA prompt

**Scenario:** Answering from documentation snippets.

```text
Use only the retrieved passages below to answer the question. First list the relevant passage IDs, then answer in plain language, then note anything the passages do not establish. Do not use outside knowledge.
```

**Why this example matters:** It makes evidence use, answer synthesis, and uncertainty handling explicit.



## Common Mistakes And Corrections

### Mistake 1

**Problem:** Assuming retrieved text automatically prevents hallucination.

**Why it hurts:** The model may still overgeneralize or interpolate beyond the source.

**Better move:** Define a source-bounded answer policy clearly.

### Mistake 2

**Problem:** Stuffing retrieval output into the prompt without telling the model how to use it.

**Why it hurts:** Evidence may be ignored or misweighted.

**Better move:** State the retrieval usage rules explicitly.



## Evaluation Checklist

- Does the prompt define how to use and cite retrieved passages?
- What should happen when evidence is partial or conflicting?
- Is the answer clearly bounded by the sources?

## Summary

Retrieval-augmented prompting is strongest when it treats evidence handling as part of the prompt design, not as an afterthought.

A strong prompt engineer can explain the topic simply, apply it deliberately, and diagnose when it is missing from a workflow.
