## Core Notes

### Generation parameters shape behavior beyond wording

A strong prompt can still behave differently under different sampling settings. Low-variability settings often suit extraction, classification, and strict formatting, while higher variability may help brainstorming or creative writing.

### Temperature and top_p are related but not identical

Both influence token selection diversity, but they do so in different ways. Developers should avoid changing many generation parameters at once without a reasoned test plan.

### Task type should drive parameter choices

The right settings for deterministic routing labels are usually not the right settings for idea generation or exploratory design prompts.



## Frameworks And Design Patterns

### Stability for structured tasks, variety for ideation

**Use when:** You are deciding default generation settings for an application feature.

**How it works:** Use low-variability settings where reproducibility matters and allow broader sampling only where variation is beneficial.

**Watch out for:** Do not assume low variability guarantees correctness; it mainly improves consistency.

## Common Mistakes And Corrections

### Mistake 1

**Problem:** Using high-variability settings for strict schema tasks.

**Why it hurts:** Format drift and inconsistent labels become more likely.

**Better move:** Use lower-variability settings for structured outputs.

### Mistake 2

**Problem:** Tweaking many controls without a test set.

**Why it hurts:** You cannot tell which parameter actually changed behavior.

**Better move:** Adjust one meaningful variable at a time and compare outputs.



## Evaluation Checklist

- Does this task need consistency, diversity, or a balance of both?
- Have parameter changes been tested on representative inputs?
- Is the generation behavior appropriate for downstream consumers?
