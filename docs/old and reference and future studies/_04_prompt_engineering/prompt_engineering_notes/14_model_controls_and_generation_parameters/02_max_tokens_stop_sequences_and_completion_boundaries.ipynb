## Core Notes

### Output budget is part of prompt design

If an answer must include multiple sections or fields, the output token budget should be realistic. Length control is not just a runtime setting; it interacts with task scope and prompt structure.

### Stop behavior affects reliability

In some applications, you want the model to stop at a predictable boundary. This is especially useful when outputs feed a parser or when multiple completions are chained together.

### Short prompts can still produce oversized answers

If the task scope is broad and no stopping behavior is defined, the output may grow beyond what the workflow can handle efficiently.



## Frameworks And Design Patterns

### Scope + budget + stop design

**Use when:** A feature has cost, latency, or parser constraints.

**How it works:** Define what the answer must contain, how long it may be, and where it should stop.

**Watch out for:** Overly tight budgets can cause dropped sections or shallow reasoning.



## Prompt Examples

### Bounded structured output

**Scenario:** A service expects one compact structured answer.

```text
Return exactly three bullets and one closing sentence. Do not add a preface or follow-up questions. Stop after the closing sentence.
```

**Why this example matters:** It illustrates how prompt wording can complement runtime stop behavior and token limits.



## Common Mistakes And Corrections

### Mistake 1

**Problem:** Setting output limits without considering required content.

**Why it hurts:** Answers become clipped or incomplete.

**Better move:** Match budget decisions to the true minimum answer shape.

### Mistake 2

**Problem:** Ignoring stop conditions in chained workflows.

**Why it hurts:** One stage may bleed into the next or produce noisy endings.

**Better move:** Define completion boundaries where multi-step systems need them.



## Evaluation Checklist

- Is the output budget realistic for the requested structure?
- Does the workflow need an explicit stop rule?
- What failure happens if the answer is longer than expected?
